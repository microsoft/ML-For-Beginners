# وقت کی سیریز کی پیش گوئی ARIMA کے ساتھ

اس نوٹ بک میں، ہم یہ دکھائیں گے کہ کیسے:
- وقت کی سیریز کے ڈیٹا کو ARIMA وقت کی سیریز کی پیش گوئی کے ماڈل کے لیے تیار کیا جائے
- ایک سادہ ARIMA ماڈل کو نافذ کریں تاکہ وقت کی سیریز میں اگلے HORIZON مراحل (وقت *t+1* سے *t+HORIZON* تک) کی پیش گوئی کی جا سکے
- ماڈل کا جائزہ لیں
