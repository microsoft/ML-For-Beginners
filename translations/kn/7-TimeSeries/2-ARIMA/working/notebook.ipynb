# ARIMA ಬಳಸಿ ಕಾಲ ಸರಣಿ ಮುನ್ಸೂಚನೆ

ಈ ನೋಟ್ಬುಕ್‌ನಲ್ಲಿ, ನಾವು ಹೇಗೆ ಮಾಡುವುದು ಎಂದು ತೋರಿಸುತ್ತೇವೆ:
- ARIMA ಕಾಲ ಸರಣಿ ಮುನ್ಸೂಚನೆ ಮಾದರಿಗಾಗಿ ಕಾಲ ಸರಣಿ ಡೇಟಾವನ್ನು ತರಬೇತಿಗೆ ಸಿದ್ಧಪಡಿಸುವುದು
- ಸರಳ ARIMA ಮಾದರಿಯನ್ನು ಅನುಷ್ಠಾನಗೊಳಿಸಿ ಮುಂದಿನ HORIZON ಹಂತಗಳನ್ನು ಮುನ್ಸೂಚಿಸುವುದು (ಕಾಲ *t+1* ರಿಂದ *t+HORIZON* ವರೆಗೆ)
- ಮಾದರಿಯನ್ನು ಮೌಲ್ಯಮಾಪನ ಮಾಡುವುದು

ಈ ಉದಾಹರಣೆಯಲ್ಲಿನ ಡೇಟಾ GEFCom2014 ಮುನ್ಸೂಚನೆ ಸ್ಪರ್ಧೆಯಿಂದ ತೆಗೆದುಕೊಳ್ಳಲಾಗಿದೆ<sup>1</sup>. ಇದು 2012 ರಿಂದ 2014 ರವರೆಗೆ 3 ವರ್ಷಗಳ ಗಂಟೆಗಂಟೆ ವಿದ್ಯುತ್ ಲೋಡ್ ಮತ್ತು ತಾಪಮಾನ ಮೌಲ್ಯಗಳನ್ನು ಒಳಗೊಂಡಿದೆ. ಕಾರ್ಯವು ಭವಿಷ್ಯದ ವಿದ್ಯುತ್ ಲೋಡ್ ಮೌಲ್ಯಗಳನ್ನು ಮುನ್ಸೂಚಿಸುವುದಾಗಿದೆ. ಈ ಉದಾಹರಣೆಯಲ್ಲಿ, ನಾವು ಇತಿಹಾಸದ ಲೋಡ್ ಡೇಟಾವನ್ನು ಮಾತ್ರ ಬಳಸಿಕೊಂಡು ಒಂದು ಕಾಲ ಹಂತ ಮುನ್ಸೂಚಿಸುವುದನ್ನು ತೋರಿಸುತ್ತೇವೆ.

<sup>1</sup>ಟಾವ್ ಹಾಂಗ್, ಪಿಯೆರ್ರೆ ಪಿನ್ಸನ್, ಶು ಫ್ಯಾನ್, ಹಮಿದ್ರೆಜಾ ಜರೈಪೌರ್, ಅಲ್ಬೆರ್ಟೋ ಟ್ರೊಕೋಲ್ಲಿ ಮತ್ತು ರಾಬ್ ಜೆ. ಹಿಂಡ್ಮನ್, "ಪ್ರೊಬಬಿಲಿಸ್ಟಿಕ್ ಎನರ್ಜಿ ಫೋರ್ಕಾಸ್ಟಿಂಗ್: ಗ್ಲೋಬಲ್ ಎನರ್ಜಿ ಫೋರ್ಕಾಸ್ಟಿಂಗ್ ಸ್ಪರ್ಧೆ 2014 ಮತ್ತು ನಂತರ", ಇಂಟರ್‌ನ್ಯಾಷನಲ್ ಜರ್ನಲ್ ಆಫ್ ಫೋರ್ಕಾಸ್ಟಿಂಗ್, ವಾಲ್ಯೂಮ್ 32, ಸಂಖ್ಯೆ 3, ಪುಟಗಳು 896-913, ಜುಲೈ-ಸೆಪ್ಟೆಂಬರ್, 2016.
