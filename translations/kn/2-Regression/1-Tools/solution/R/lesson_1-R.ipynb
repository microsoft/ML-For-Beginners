‡≤à‡≤ó ‡≤®‡≤Æ‡≤ó‡≥Ü ‡≤°‡≥á‡≤ü‡≤æ ‡≤á‡≤¶‡≥ç‡≤¶‡≥Å, ‡≤à ‡≤µ‡≥ç‡≤Ø‡≤æ‡≤Ø‡≤æ‡≤Æ‡≤ï‡≥ç‡≤ï‡≤æ‡≤ó‡≤ø ‡≤í‡≤Ç‡≤¶‡≥Å ‡≤µ‡≥à‡≤∂‡≤ø‡≤∑‡≥ç‡≤ü‡≥ç‡≤Ø (`bmi`) ‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤ó‡≥Å‡≤∞‡≤ø‡≤Ø‡≤æ‡≤ó‡≤ø‡≤∏‡≥ã‡≤£. ‡≤á‡≤¶‡≤ï‡≥ç‡≤ï‡≤æ‡≤ó‡≤ø ‡≤®‡≤æ‡≤µ‡≥Å ‡≤¨‡≥á‡≤ï‡≤æ‡≤¶ ‡≤ï‡≤æ‡≤≤‡≤Æ‡≥ç‚Äå‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å ‡≤Ü‡≤Ø‡≥ç‡≤ï‡≥Ü ‡≤Æ‡≤æ‡≤°‡≤¨‡≥á‡≤ï‡≤æ‡≤ó‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü. ‡≤π‡≤æ‡≤ó‡≤æ‡≤¶‡≤∞‡≥Ü, ‡≤®‡≤æ‡≤µ‡≥Å ‡≤á‡≤¶‡≤®‡≥ç‡≤®‡≥Å ‡≤π‡≥á‡≤ó‡≥Ü ‡≤Æ‡≤æ‡≤°‡≤¨‡≤π‡≥Å‡≤¶‡≥Å?

- ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø **‡≤Æ‡≥ã‡≤°‡≥ç** ‡≤∏‡≤æ‡≤Æ‡≤æ‡≤®‡≥ç‡≤Ø ‡≤Ü‡≤Ø‡≥ç‡≤ï‡≥Ü‡≤ó‡≤≥‡≥Å regression ‡≤Æ‡≤§‡≥ç‡≤§‡≥Å classification ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤í‡≤≥‡≤ó‡≥ä‡≤Ç‡≤°‡≤ø‡≤¶‡≥Ü; ‡≤ï‡≥Ü‡≤≤‡≤µ‡≥Å ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø ‡≤™‡≥ç‡≤∞‡≤ï‡≤æ‡≤∞‡≤ó‡≤≥‡≥Å ‡≤á‡≤µ‡≥Å‡≤ó‡≤≥‡≤≤‡≥ç‡≤≤‡≤ø ‡≤Ø‡≤æ‡≤µ‡≥Å‡≤¶‡≤æ‡≤¶‡≤∞‡≥Ç ‡≤í‡≤Ç‡≤¶‡≥Å ‡≤Ö‡≤•‡≤µ‡≤æ ‡≤é‡≤∞‡≤°‡≤®‡≥ç‡≤®‡≥Ç ‡≤¨‡≥Ü‡≤Ç‡≤¨‡≤≤‡≤ø‡≤∏‡≥Å‡≤§‡≥ç‡≤§‡≤µ‡≥Ü.

- ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø **‡≤é‡≤Ç‡≤ú‡≤ø‡≤®‡≥ç** ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤π‡≥ä‡≤Ç‡≤¶‡≤ø‡≤∏‡≤≤‡≥Å ‡≤¨‡≤≥‡≤∏‡≤≤‡≤æ‡≤ó‡≥Å‡≤µ ‡≤ó‡≤£‡≤®‡≤æ‡≤§‡≥ç‡≤Æ‡≤ï ‡≤∏‡≤æ‡≤ß‡≤®‡≤µ‡≤æ‡≤ó‡≤ø‡≤¶‡≥Ü. ‡≤∏‡≤æ‡≤Æ‡≤æ‡≤®‡≥ç‡≤Ø‡≤µ‡≤æ‡≤ó‡≤ø ‡≤á‡≤µ‡≥Å R ‡≤™‡≥ç‡≤Ø‡≤æ‡≤ï‡≥á‡≤ú‡≥ç‚Äå‡≤ó‡≤≥‡≥Å, ‡≤â‡≤¶‡≤æ‡≤π‡≤∞‡≤£‡≥Ü‡≤ó‡≥Ü **`"lm"`** ‡≤Ö‡≤•‡≤µ‡≤æ **`"ranger"`**

‡≤à ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø ‡≤Æ‡≤æ‡≤π‡≤ø‡≤§‡≤ø‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø ‡≤®‡≤ø‡≤∞‡≥ç‡≤¶‡≤ø‡≤∑‡≥ç‡≤ü‡≤™‡≤°‡≤ø‡≤∏‡≥Å‡≤µ‡≤ø‡≤ï‡≥Ü‡≤Ø‡≤≤‡≥ç‡≤≤‡≤ø ‡≤∏‡≥Ü‡≤∞‡≥Ü‡≤π‡≤ø‡≤°‡≤ø‡≤Ø‡≤≤‡≤æ‡≤ó‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü, ‡≤Ü‡≤¶‡≥ç‡≤¶‡≤∞‡≤ø‡≤Ç‡≤¶ ‡≤í‡≤Ç‡≤¶‡≤®‡≥ç‡≤®‡≥Å ‡≤®‡≤ø‡≤∞‡≥ç‡≤Æ‡≤ø‡≤∏‡≥ã‡≤£!


```r
# Build a linear model specification
lm_spec <- 
  # Type
  linear_reg() %>% 
  # Engine
  set_engine("lm") %>% 
  # Mode
  set_mode("regression")


# Print the model specification
lm_spec
```

‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø‡≤Ø‡≤®‡≥ç‡≤®‡≥Å *‡≤®‡≤ø‡≤∞‡≥ç‡≤¶‡≤ø‡≤∑‡≥ç‡≤ü‡≤™‡≤°‡≤ø‡≤∏‡≤ø‡≤¶* ‡≤®‡≤Ç‡≤§‡≤∞, ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤∏‡≤æ‡≤Æ‡≤æ‡≤®‡≥ç‡≤Ø‡≤µ‡≤æ‡≤ó‡≤ø ‡≤∏‡≥Ç‡≤§‡≥ç‡≤∞ ‡≤Æ‡≤§‡≥ç‡≤§‡≥Å ‡≤ï‡≥Ü‡≤≤‡≤µ‡≥Å ‡≤°‡≥á‡≤ü‡≤æ‡≤µ‡≤®‡≥ç‡≤®‡≥Å ‡≤¨‡≤≥‡≤∏‡≤ø [`fit()`](https://parsnip.tidymodels.org/reference/fit.html) ‡≤´‡≤Ç‡≤ï‡≥ç‡≤∑‡≤®‡≥ç ‡≤¨‡≤≥‡≤∏‡≤ø `‡≤Ö‡≤Ç‡≤¶‡≤æ‡≤ú‡≥Å` ‡≤Ö‡≤•‡≤µ‡≤æ `‡≤§‡≤∞‡≤¨‡≥á‡≤§‡≤ø` ‡≤Æ‡≤æ‡≤°‡≤¨‡≤π‡≥Å‡≤¶‡≥Å.

`y ~ .` ‡≤Ö‡≤Ç‡≤¶‡≤∞‡≥Ü ‡≤®‡≤æ‡≤µ‡≥Å `y` ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤≠‡≤µ‡≤ø‡≤∑‡≥ç‡≤Ø‡≤µ‡≤æ‡≤£‡≤ø ‡≤™‡≥ç‡≤∞‡≤Æ‡≤æ‡≤£/‡≤≤‡≤ï‡≥ç‡≤∑‡≥ç‡≤Ø‡≤µ‡≤æ‡≤ó‡≤ø ‡≤π‡≥ä‡≤Ç‡≤¶‡≤ø‡≤∏‡≥Å‡≤µ‡≥Ü‡≤µ‡≥Å, ‡≤é‡≤≤‡≥ç‡≤≤‡≤æ ‡≤≠‡≤µ‡≤ø‡≤∑‡≥ç‡≤Ø‡≤µ‡≤æ‡≤£‡≤ø‡≤ï‡≤æ‡≤∞‡≤∞‡≥Å/‡≤µ‡≥à‡≤∂‡≤ø‡≤∑‡≥ç‡≤ü‡≥ç‡≤Ø‡≤ó‡≤≥‡≤ø‡≤Ç‡≤¶ ‡≤µ‡≤ø‡≤µ‡≤∞‡≤ø‡≤∏‡≤≤‡≤æ‡≤ó‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü ‡≤Ö‡≤Ç‡≤¶‡≤∞‡≥Ü, `.` (‡≤à ‡≤™‡≥ç‡≤∞‡≤ï‡≤∞‡≤£‡≤¶‡≤≤‡≥ç‡≤≤‡≤ø, ‡≤®‡≤Æ‡≤ó‡≥Ü ‡≤í‡≤Ç‡≤¶‡≥á ‡≤≠‡≤µ‡≤ø‡≤∑‡≥ç‡≤Ø‡≤µ‡≤æ‡≤£‡≤ø‡≤ï‡≤æ‡≤∞‡≤®‡≥Å ‡≤á‡≤¶‡≥Ü: `bmi`)


```r
# Build a linear model specification
lm_spec <- linear_reg() %>% 
  set_engine("lm") %>%
  set_mode("regression")


# Train a linear regression model
lm_mod <- lm_spec %>% 
  fit(y ~ ., data = diabetes_train)

# Print the model
lm_mod
```

‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø ‡≤î‡≤ü‡≥ç‚Äå‡≤™‡≥Å‡≤ü‡≥ç‚Äå‡≤®‡≤ø‡≤Ç‡≤¶, ‡≤®‡≤æ‡≤µ‡≥Å ‡≤§‡≤∞‡≤¨‡≥á‡≤§‡≤ø ‡≤∏‡≤Æ‡≤Ø‡≤¶‡≤≤‡≥ç‡≤≤‡≤ø ‡≤ï‡≤≤‡≤ø‡≤§ ‡≤∏‡≤π‡≤ó‡≥Å‡≤£‡≤æ‡≤Ç‡≤ï‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å ‡≤®‡≥ã‡≤°‡≤¨‡≤π‡≥Å‡≤¶‡≥Å. ‡≤Ö‡≤µ‡≥Å‡≤ó‡≤≥‡≥Å ‡≤®‡≤ø‡≤ú‡≤µ‡≤æ‡≤¶ ‡≤Æ‡≤§‡≥ç‡≤§‡≥Å ‡≤≠‡≤µ‡≤ø‡≤∑‡≥ç‡≤Ø‡≤µ‡≤æ‡≤£‡≤ø ‡≤Æ‡≤æ‡≤°‡≤≤‡≤æ‡≤¶ ‡≤ö‡≤∞‡≤ó‡≤≥ ‡≤®‡≤°‡≥Å‡≤µ‡≥Ü ‡≤ï‡≤®‡≤ø‡≤∑‡≥ç‡≤† ‡≤í‡≤ü‡≥ç‡≤ü‡≥Å ‡≤¶‡≥ã‡≤∑‡≤µ‡≤®‡≥ç‡≤®‡≥Å ‡≤®‡≥Ä‡≤°‡≥Å‡≤µ ‡≤â‡≤§‡≥ç‡≤§‡≤Æ ‡≤π‡≥ä‡≤Ç‡≤¶‡≤æ‡≤£‡≤ø‡≤ï‡≥Ü‡≤Ø ‡≤∞‡≥á‡≤ñ‡≥Ü‡≤Ø ‡≤∏‡≤π‡≤ó‡≥Å‡≤£‡≤æ‡≤Ç‡≤ï‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å ‡≤™‡≥ç‡≤∞‡≤§‡≤ø‡≤®‡≤ø‡≤ß‡≤ø‡≤∏‡≥Å‡≤§‡≥ç‡≤§‡≤µ‡≥Ü.
<br>

## 5. ‡≤™‡≤∞‡≥Ä‡≤ï‡≥ç‡≤∑‡≤æ ‡≤∏‡≥Ü‡≤ü‡≥ç‚Äå‡≤®‡≤≤‡≥ç‡≤≤‡≤ø ‡≤≠‡≤µ‡≤ø‡≤∑‡≥ç‡≤Ø‡≤µ‡≤æ‡≤£‡≤ø‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å ‡≤Æ‡≤æ‡≤°‡≤ø

‡≤®‡≤æ‡≤µ‡≥Å ‡≤à‡≤ó‡≤æ‡≤ó‡≤≤‡≥á ‡≤í‡≤Ç‡≤¶‡≥Å ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤§‡≤∞‡≤¨‡≥á‡≤§‡≤ø‡≤Æ‡≤æ‡≤°‡≤ø‡≤∞‡≥Å‡≤µ‡≥Å‡≤¶‡≤∞‡≤ø‡≤Ç‡≤¶, ‡≤Ö‡≤¶‡≤®‡≥ç‡≤®‡≥Å [parsnip::predict()](https://parsnip.tidymodels.org/reference/predict.model_fit.html) ‡≤¨‡≤≥‡≤∏‡≤ø ‡≤™‡≤∞‡≥Ä‡≤ï‡≥ç‡≤∑‡≤æ ‡≤°‡≥á‡≤ü‡≤æ‡≤∏‡≥Ü‡≤ü‡≥ç‚Äå‡≤ó‡≥Ü ‡≤∞‡≥ã‡≤ó ‡≤™‡≥ç‡≤∞‡≤ó‡≤§‡≤ø‡≤Ø‡≤®‡≥ç‡≤®‡≥Å y ‡≤≠‡≤µ‡≤ø‡≤∑‡≥ç‡≤Ø‡≤µ‡≤æ‡≤£‡≤ø ‡≤Æ‡≤æ‡≤°‡≤≤‡≥Å ‡≤¨‡≤≥‡≤∏‡≤¨‡≤π‡≥Å‡≤¶‡≥Å. ‡≤á‡≤¶‡≤®‡≥ç‡≤®‡≥Å ‡≤°‡≥á‡≤ü‡≤æ ‡≤ó‡≥Å‡≤Ç‡≤™‡≥Å‡≤ó‡≤≥ ‡≤®‡≤°‡≥Å‡≤µ‡≥Ü ‡≤∞‡≥á‡≤ñ‡≥Ü‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤¨‡≤ø‡≤°‡≤ø‡≤∏‡≤≤‡≥Å ‡≤¨‡≤≥‡≤∏‡≤≤‡≤æ‡≤ó‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü.


```r
# Make predictions for the test set
predictions <- lm_mod %>% 
  predict(new_data = diabetes_test)

# Print out some of the predictions
predictions %>% 
  slice(1:5)
```

‡≤µ‡≤æ‡≤π‡≥ç! üíÉüï∫ ‡≤®‡≤æ‡≤µ‡≥Å ‡≤á‡≤§‡≥ç‡≤§‡≥Ä‡≤ö‡≥Ü‡≤ó‡≥Ü ‡≤í‡≤Ç‡≤¶‡≥Å ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤§‡≤∞‡≤¨‡≥á‡≤§‡≥Å‡≤ó‡≥ä‡≤Ç‡≤°‡≥Å ‡≤Ö‡≤¶‡≤®‡≥ç‡≤®‡≥Å ‡≤≠‡≤µ‡≤ø‡≤∑‡≥ç‡≤Ø‡≤µ‡≤æ‡≤£‡≤ø ‡≤Æ‡≤æ‡≤°‡≤≤‡≥Å ‡≤¨‡≤≥‡≤∏‡≤ø‡≤¶‡≥ç‡≤¶‡≥á‡≤µ‡≥Ü!

‡≤≠‡≤µ‡≤ø‡≤∑‡≥ç‡≤Ø‡≤µ‡≤æ‡≤£‡≤ø ‡≤Æ‡≤æ‡≤°‡≥Å‡≤µ‡≤æ‡≤ó, tidymodels ‡≤∏‡≤Ç‡≤™‡≥ç‡≤∞‡≤¶‡≤æ‡≤Ø‡≤µ‡≥Å ‡≤Ø‡≤æ‡≤µ‡≤æ‡≤ó‡≤≤‡≥Ç ‡≤´‡≤≤‡≤ø‡≤§‡≤æ‡≤Ç‡≤∂‡≤ó‡≤≥ ‡≤§‡≤¨‡≤≤‡≥ç/‡≤°‡≥á‡≤ü‡≤æ ‡≤´‡≥ç‡≤∞‡≥á‡≤Æ‡≥ç ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤Æ‡≤æ‡≤®‡≤ï‡≥É‡≤§ ‡≤ï‡≤æ‡≤≤‡≤Æ‡≥ç ‡≤π‡≥Ü‡≤∏‡≤∞‡≥Å‡≤ó‡≤≥‡≥ä‡≤Ç‡≤¶‡≤ø‡≤ó‡≥Ü ‡≤â‡≤§‡≥ç‡≤™‡≤æ‡≤¶‡≤ø‡≤∏‡≥Å‡≤µ‡≥Å‡≤¶‡≤æ‡≤ó‡≤ø‡≤¶‡≥Ü. ‡≤á‡≤¶‡≤∞‡≤ø‡≤Ç‡≤¶ ‡≤Æ‡≥Ç‡≤≤ ‡≤°‡≥á‡≤ü‡≤æ ‡≤Æ‡≤§‡≥ç‡≤§‡≥Å ‡≤≠‡≤µ‡≤ø‡≤∑‡≥ç‡≤Ø‡≤µ‡≤æ‡≤£‡≤ø‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å subsequent ‡≤ï‡≤æ‡≤∞‡≥ç‡≤Ø‡≤æ‡≤ö‡≤∞‡≤£‡≥Ü‡≤ó‡≤≥‡≤ø‡≤ó‡≥Ü, ‡≤â‡≤¶‡≤æ‡≤π‡≤∞‡≤£‡≥Ü‡≤ó‡≥Ü ‡≤™‡≥ç‡≤≤‡≤æ‡≤ü‡≤ø‡≤Ç‡≤ó‡≥ç‚Äå‡≤ó‡≥Ü, ‡≤¨‡≤≥‡≤∏‡≤¨‡≤π‡≥Å‡≤¶‡≤æ‡≤¶ ‡≤∏‡≥ç‡≤µ‡≤∞‡≥Ç‡≤™‡≤¶‡≤≤‡≥ç‡≤≤‡≤ø ‡≤∏‡≤Ç‡≤Ø‡≥ã‡≤ú‡≤ø‡≤∏‡≥Å‡≤µ‡≥Å‡≤¶‡≥Å ‡≤∏‡≥Å‡≤≤‡≤≠‡≤µ‡≤æ‡≤ó‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü.

`dplyr::bind_cols()` ‡≤¨‡≤π‡≥Å ‡≤°‡≥á‡≤ü‡≤æ ‡≤´‡≥ç‡≤∞‡≥á‡≤Æ‡≥ç‚Äå‡≤ó‡≤≥ ‡≤ï‡≤æ‡≤≤‡≤Æ‡≥ç‚Äå‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å ‡≤™‡≤∞‡≤ø‡≤£‡≤æ‡≤Æ‡≤ï‡≤æ‡≤∞‡≤ø‡≤Ø‡≤æ‡≤ó‡≤ø ‡≤¨‡≤Ç‡≤ß‡≤ø‡≤∏‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü.


```r
# Combine the predictions and the original test set
results <- diabetes_test %>% 
  bind_cols(predictions)


results %>% 
  slice(1:5)
```

## 6. ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø ‡≤´‡≤≤‡≤ø‡≤§‡≤æ‡≤Ç‡≤∂‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å ‡≤ö‡≤ø‡≤§‡≥ç‡≤∞‡≤ø‡≤∏‡≤ø
