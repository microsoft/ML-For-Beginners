```r
# Load the core Tidyverse packages
library(tidyverse)
library(lubridate)

# Import the pumpkins data
pumpkins <- read_csv(file = "https://raw.githubusercontent.com/microsoft/ML-For-Beginners/main/2-Regression/data/US-pumpkins.csv")


# Get a glimpse and dimensions of the data
glimpse(pumpkins)


# Print the first 50 rows of the data set
pumpkins %>% 
  slice_head(n = 5)
```

```r
# Return column names
pumpkins %>% 
  names()
```

ü§î ‡≤®‡≤æ‡≤µ‡≥Å ‡≤á‡≤®‡≥ç‡≤®‡≥Ç ‡≤â‡≤§‡≥ç‡≤§‡≤Æ ‡≤Æ‡≤æ‡≤°‡≤¨‡≤π‡≥Å‡≤¶‡≥Å. ‡≤à ‡≤ï‡≤æ‡≤≤‡≤Æ‡≥ç ‡≤π‡≥Ü‡≤∏‡≤∞‡≥Å‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å `janitor::clean_names` ‡≤¨‡≤≥‡≤∏‡≤ø [snake_case](https://en.wikipedia.org/wiki/Snake_case) ‡≤∏‡≤Ç‡≤™‡≥ç‡≤∞‡≤¶‡≤æ‡≤Ø‡≤ï‡≥ç‡≤ï‡≥Ü ‡≤™‡≤∞‡≤ø‡≤µ‡≤∞‡≥ç‡≤§‡≤ø‡≤∏‡≤ø `friendR` ‡≤Ü‡≤ó‡≤ø‡≤∏‡≥ã‡≤£. ‡≤à ‡≤´‡≤Ç‡≤ï‡≥ç‡≤∑‡≤®‡≥ç ‡≤¨‡≤ó‡≥ç‡≤ó‡≥Ü ‡≤á‡≤®‡≥ç‡≤®‡≤∑‡≥ç‡≤ü‡≥Å ‡≤§‡≤ø‡≤≥‡≤ø‡≤Ø‡≤≤‡≥Å: `?clean_names`


```r
# Clean names to the snake_case convention
pumpkins <- pumpkins %>% 
  clean_names(case = "snake")

# Return column names
pumpkins %>% 
  names()
```

‡≤¨‡≤π‡≤≥ tidyR üßπ! ‡≤à‡≤ó, ‡≤π‡≤ø‡≤Ç‡≤¶‡≤ø‡≤® ‡≤™‡≤æ‡≤†‡≤¶‡≤≤‡≥ç‡≤≤‡≤ø ‡≤á‡≤¶‡≥ç‡≤¶‡≤Ç‡≤§‡≥Ü `dplyr` ‡≤¨‡≤≥‡≤∏‡≤ø ‡≤°‡≥á‡≤ü‡≤æ‡≤¶‡≥ä‡≤Ç‡≤¶‡≤ø‡≤ó‡≥Ü ‡≤®‡≥É‡≤§‡≥ç‡≤Ø ‡≤Æ‡≤æ‡≤°‡≥ã‡≤£! üíÉ


```r
# Select desired columns
pumpkins <- pumpkins %>% 
  select(variety, city_name, package, low_price, high_price, date)



# Extract the month from the dates to a new column
pumpkins <- pumpkins %>%
  mutate(date = mdy(date),
         month = month(date)) %>% 
  select(-date)



# Create a new column for average Price
pumpkins <- pumpkins %>% 
  mutate(price = (low_price + high_price)/2)


# Retain only pumpkins with the string "bushel"
new_pumpkins <- pumpkins %>% 
  filter(str_detect(string = package, pattern = "bushel"))


# Normalize the pricing so that you show the pricing per bushel, not per 1 1/9 or 1/2 bushel
new_pumpkins <- new_pumpkins %>% 
  mutate(price = case_when(
    str_detect(package, "1 1/9") ~ price/(1.1),
    str_detect(package, "1/2") ~ price*2,
    TRUE ~ price))

# Relocate column positions
new_pumpkins <- new_pumpkins %>% 
  relocate(month, .before = variety)


# Display the first 5 rows
new_pumpkins %>% 
  slice_head(n = 5)
```

```r
set.seed(2056)
# Split the data into training and test sets
pumpkins_split <- new_pumpkins %>% 
  initial_split(prop = 0.8)


# Extract training and test data
pumpkins_train <- training(pumpkins_split)
pumpkins_test <- testing(pumpkins_split)



# Create a recipe for preprocessing the data
lm_pumpkins_recipe <- recipe(price ~ package, data = pumpkins_train) %>% 
  step_integer(all_predictors(), zero_based = TRUE)



# Create a linear model specification
lm_spec <- linear_reg() %>% 
  set_engine("lm") %>% 
  set_mode("regression")
```

Tidymodels ‡≤®‡≤≤‡≥ç‡≤≤‡≤ø, ‡≤à ‡≤Ö‡≤®‡≥Å‡≤ï‡≥Ç‡≤≤‡≤ï‡≤∞ ‡≤µ‡≤∏‡≥ç‡≤§‡≥Å‡≤µ‡≤®‡≥ç‡≤®‡≥Å [`workflow`](https://workflows.tidymodels.org/) ‡≤é‡≤Ç‡≤¶‡≥Å ‡≤ï‡≤∞‡≥Ü‡≤Ø‡≤≤‡≤æ‡≤ó‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü ‡≤Æ‡≤§‡≥ç‡≤§‡≥Å ‡≤á‡≤¶‡≥Å ‡≤®‡≤ø‡≤Æ‡≥ç‡≤Æ ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø ‡≤ò‡≤ü‡≤ï‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å ‡≤Ö‡≤®‡≥Å‡≤ï‡≥Ç‡≤≤‡≤ï‡≤∞‡≤µ‡≤æ‡≤ó‡≤ø ‡≤π‡≤ø‡≤°‡≤ø‡≤¶‡≤ø‡≤°‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü! Python ‡≤®‡≤≤‡≥ç‡≤≤‡≤ø ‡≤á‡≤¶‡≤®‡≥ç‡≤®‡≥Å *pipelines* ‡≤é‡≤Ç‡≤¶‡≥Å ‡≤ï‡≤∞‡≥Ü‡≤Ø‡≥Å‡≤§‡≥ç‡≤§‡≥á‡≤µ‡≥Ü.

‡≤π‡≥Ä‡≤ó‡≤æ‡≤ó‡≤ø ‡≤é‡≤≤‡≥ç‡≤≤‡≤µ‡≤®‡≥ç‡≤®‡≥Ç workflow ‡≤ó‡≥Ü ‡≤¨‡≤Ç‡≤°‡≤≤‡≥ç ‡≤Æ‡≤æ‡≤°‡≥ã‡≤£!üì¶


```r
# Hold modelling components in a workflow
lm_wf <- workflow() %>% 
  add_recipe(lm_pumpkins_recipe) %>% 
  add_model(lm_spec)

# Print out the workflow
lm_wf
```

üëå ‡≤ú‡≥ä‡≤§‡≥Ü‡≤ó‡≥Ü, ‡≤í‡≤Ç‡≤¶‡≥Å ‡≤µ‡≤∞‡≥ç‡≤ï‡≥ç‚Äå‡≤´‡≥ç‡≤≤‡≥ã‡≤µ‡≤®‡≥ç‡≤®‡≥Å ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤§‡≤∞‡≤¨‡≥á‡≤§‡≥Å‡≤ó‡≥ä‡≤≥‡≤ø‡≤∏‡≥Å‡≤µ‡≤Ç‡≤§‡≥Ü ‡≤Ö‡≤•‡≤µ‡≤æ ‡≤π‡≥ä‡≤Ç‡≤¶‡≤ø‡≤∏‡≥Å‡≤µ‡≤Ç‡≤§‡≥Ü ‡≤Æ‡≤æ‡≤°‡≤¨‡≤π‡≥Å‡≤¶‡≥Å.


```r
# Train the model
lm_wf_fit <- lm_wf %>% 
  fit(data = pumpkins_train)

# Print the model coefficients learned 
lm_wf_fit
```

‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø ‡≤î‡≤ü‡≥ç‚Äå‡≤™‡≥Å‡≤ü‡≥ç‚Äå‡≤®‡≤ø‡≤Ç‡≤¶, ‡≤®‡≤æ‡≤µ‡≥Å ‡≤§‡≤∞‡≤¨‡≥á‡≤§‡≤ø ‡≤∏‡≤Æ‡≤Ø‡≤¶‡≤≤‡≥ç‡≤≤‡≤ø ‡≤ï‡≤≤‡≤ø‡≤§ ‡≤∏‡≤π‡≤ó‡≥Å‡≤£‡≤æ‡≤Ç‡≤ï‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å ‡≤®‡≥ã‡≤°‡≤¨‡≤π‡≥Å‡≤¶‡≥Å. ‡≤Ö‡≤µ‡≥Å‡≤ó‡≤≥‡≥Å ‡≤®‡≤ø‡≤ú‡≤µ‡≤æ‡≤¶ ‡≤Æ‡≤§‡≥ç‡≤§‡≥Å ‡≤≠‡≤µ‡≤ø‡≤∑‡≥ç‡≤Ø‡≤µ‡≤æ‡≤£‡≤ø ‡≤Æ‡≤æ‡≤°‡≤≤‡≤æ‡≤¶ ‡≤ö‡≤∞‡≤ó‡≤≥ ‡≤®‡≤°‡≥Å‡≤µ‡≥Ü ‡≤í‡≤ü‡≥ç‡≤ü‡≥Å ‡≤§‡≤™‡≥ç‡≤™‡≥Å ‡≤ï‡≤°‡≤ø‡≤Æ‡≥Ü ‡≤Ü‡≤ó‡≥Å‡≤µ ‡≤Ö‡≤§‡≥ç‡≤Ø‡≥Å‡≤§‡≥ç‡≤§‡≤Æ ‡≤∏‡≤∞‡≤≥‡≤∞‡≥á‡≤ñ‡≥Ü‡≤Ø ‡≤∏‡≤π‡≤ó‡≥Å‡≤£‡≤æ‡≤Ç‡≤ï‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å ‡≤™‡≥ç‡≤∞‡≤§‡≤ø‡≤®‡≤ø‡≤ß‡≤ø‡≤∏‡≥Å‡≤§‡≥ç‡≤§‡≤µ‡≥Ü.


#### ‡≤™‡≤∞‡≥Ä‡≤ï‡≥ç‡≤∑‡≤æ ‡≤∏‡≥Ü‡≤ü‡≥ç ‡≤¨‡≤≥‡≤∏‡≤ø ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø ‡≤ï‡≤æ‡≤∞‡≥ç‡≤Ø‡≤ï‡≥ç‡≤∑‡≤Æ‡≤§‡≥Ü‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤Æ‡≥å‡≤≤‡≥ç‡≤Ø‡≤Æ‡≤æ‡≤™‡≤® ‡≤Æ‡≤æ‡≤°‡≤ø

‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø ‡≤π‡≥á‡≤ó‡≥Ü ‡≤ï‡≤æ‡≤∞‡≥ç‡≤Ø‡≤®‡≤ø‡≤∞‡≥ç‡≤µ‡≤π‡≤ø‡≤∏‡≤ø‡≤§‡≥Å ‡≤é‡≤Ç‡≤¶‡≥Å ‡≤®‡≥ã‡≤°‡≤≤‡≥Å ‡≤∏‡≤Æ‡≤Ø‡≤µ‡≤æ‡≤ó‡≤ø‡≤¶‡≥Ü üìè! ‡≤®‡≤æ‡≤µ‡≥Å ‡≤á‡≤¶‡≤®‡≥ç‡≤®‡≥Å ‡≤π‡≥á‡≤ó‡≥Ü ‡≤Æ‡≤æ‡≤°‡≥Å‡≤§‡≥ç‡≤§‡≥á‡≤µ‡≥Ü?

‡≤à‡≤ó ‡≤®‡≤æ‡≤µ‡≥Å ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤§‡≤∞‡≤¨‡≥á‡≤§‡≤ø‡≤Æ‡≤æ‡≤°‡≤ø‡≤¶‡≥ç‡≤¶‡≤∞‡≤ø‡≤Ç‡≤¶, `parsnip::predict()` ‡≤¨‡≤≥‡≤∏‡≤ø test_set ‡≤ó‡≥Ü ‡≤≠‡≤µ‡≤ø‡≤∑‡≥ç‡≤Ø‡≤µ‡≤æ‡≤£‡≤ø‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å ‡≤Æ‡≤æ‡≤°‡≤¨‡≤π‡≥Å‡≤¶‡≥Å. ‡≤®‡≤Ç‡≤§‡≤∞ ‡≤®‡≤æ‡≤µ‡≥Å ‡≤à ‡≤≠‡≤µ‡≤ø‡≤∑‡≥ç‡≤Ø‡≤µ‡≤æ‡≤£‡≤ø‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å ‡≤®‡≤ø‡≤ú‡≤µ‡≤æ‡≤¶ ‡≤≤‡≥á‡≤¨‡≤≤‡≥ç ‡≤Æ‡≥å‡≤≤‡≥ç‡≤Ø‡≤ó‡≤≥‡≥ä‡≤Ç‡≤¶‡≤ø‡≤ó‡≥Ü ‡≤π‡≥ã‡≤≤‡≤ø‡≤∏‡≤ø ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø ‡≤é‡≤∑‡≥ç‡≤ü‡≥Å ‡≤ö‡≥Ü‡≤®‡≥ç‡≤®‡≤æ‡≤ó‡≤ø (‡≤Ö‡≤•‡≤µ‡≤æ ‡≤á‡≤≤‡≥ç‡≤≤!) ‡≤ï‡≤æ‡≤∞‡≥ç‡≤Ø‡≤®‡≤ø‡≤∞‡≥ç‡≤µ‡≤π‡≤ø‡≤∏‡≥Å‡≤§‡≥ç‡≤§‡≤ø‡≤¶‡≥Ü ‡≤é‡≤Ç‡≤¶‡≥Å ‡≤Æ‡≥å‡≤≤‡≥ç‡≤Ø‡≤Æ‡≤æ‡≤™‡≤® ‡≤Æ‡≤æ‡≤°‡≤¨‡≤π‡≥Å‡≤¶‡≥Å.

‡≤Æ‡≥ä‡≤¶‡≤≤‡≥Å ‡≤™‡≤∞‡≥Ä‡≤ï‡≥ç‡≤∑‡≤æ ‡≤∏‡≥Ü‡≤ü‡≥ç‚Äå‡≤ó‡≥Ü ‡≤≠‡≤µ‡≤ø‡≤∑‡≥ç‡≤Ø‡≤µ‡≤æ‡≤£‡≤ø‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å ‡≤Æ‡≤æ‡≤°‡≥ã‡≤£ ‡≤Æ‡≤§‡≥ç‡≤§‡≥Å ‡≤®‡≤Ç‡≤§‡≤∞ ‡≤Ü ‡≤ï‡≤æ‡≤≤‡≤Æ‡≥ç‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å ‡≤™‡≤∞‡≥Ä‡≤ï‡≥ç‡≤∑‡≤æ ‡≤∏‡≥Ü‡≤ü‡≥ç‚Äå‡≤ó‡≥Ü ‡≤ú‡≥ã‡≤°‡≤ø‡≤∏‡≥ã‡≤£.


```r
# Make predictions for the test set
predictions <- lm_wf_fit %>% 
  predict(new_data = pumpkins_test)


# Bind predictions to the test set
lm_results <- pumpkins_test %>% 
  select(c(package, price)) %>% 
  bind_cols(predictions)


# Print the first ten rows of the tibble
lm_results %>% 
  slice_head(n = 10)
```

‡≤π‡≥å‡≤¶‡≥Å, ‡≤®‡≥Ä‡≤µ‡≥Å ‡≤á‡≤§‡≥ç‡≤§‡≥Ä‡≤ö‡≥Ü‡≤ó‡≥Ü ‡≤í‡≤Ç‡≤¶‡≥Å ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤§‡≤∞‡≤¨‡≥á‡≤§‡≥Å‡≤ó‡≥ä‡≤Ç‡≤°‡≥Å ‡≤Ö‡≤¶‡≤®‡≥ç‡≤®‡≥Å ‡≤≠‡≤µ‡≤ø‡≤∑‡≥ç‡≤Ø‡≤µ‡≤æ‡≤£‡≤ø ‡≤Æ‡≤æ‡≤°‡≤≤‡≥Å ‡≤¨‡≤≥‡≤∏‡≤ø‡≤¶‡≥ç‡≤¶‡≥Ä‡≤∞‡≤ø!üîÆ ‡≤á‡≤¶‡≥Å ‡≤é‡≤∑‡≥ç‡≤ü‡≥Å ‡≤â‡≤§‡≥ç‡≤§‡≤Æ‡≤µ‡≤æ‡≤ó‡≤ø‡≤¶‡≥Ü ‡≤é‡≤Ç‡≤¶‡≥Å ‡≤®‡≥ã‡≤°‡≥ã‡≤£, ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø‡≤Ø ‡≤ï‡≤æ‡≤∞‡≥ç‡≤Ø‡≤ï‡≥ç‡≤∑‡≤Æ‡≤§‡≥Ü‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤Æ‡≥å‡≤≤‡≥ç‡≤Ø‡≤Æ‡≤æ‡≤™‡≤® ‡≤Æ‡≤æ‡≤°‡≥ã‡≤£!

Tidymodels ‡≤®‡≤≤‡≥ç‡≤≤‡≤ø, ‡≤®‡≤æ‡≤µ‡≥Å ‡≤á‡≤¶‡≤®‡≥ç‡≤®‡≥Å `yardstick::metrics()` ‡≤¨‡≤≥‡≤∏‡≤ø ‡≤Æ‡≤æ‡≤°‡≥Å‡≤§‡≥ç‡≤§‡≥á‡≤µ‡≥Ü! ‡≤∞‡≥á‡≤ñ‡≥Ä‡≤Ø ‡≤∞‡≤ø‡≤ó‡≥ç‡≤∞‡≥Ü‡≤∂‡≤®‡≥ç‚Äå‡≤ó‡≤æ‡≤ó‡≤ø, ‡≤ï‡≥Ü‡≤≥‡≤ó‡≤ø‡≤® ‡≤Æ‡≥Ü‡≤ü‡≥ç‡≤∞‡≤ø‡≤ï‡≥ç‡≤∏‡≥ç‚Äå‡≤ó‡≤≥ ‡≤Æ‡≥á‡≤≤‡≥Ü ‡≤ó‡≤Æ‡≤®‡≤π‡≤∞‡≤ø‡≤∏‡≥ã‡≤£:

-   `Root Mean Square Error (RMSE)`: [MSE](https://en.wikipedia.org/wiki/Mean_squared_error) ‡≤Ø ‡≤ö‡≤¶‡≤∞ ‡≤Æ‡≥Ç‡≤≤. ‡≤á‡≤¶‡≥Å ‡≤≤‡≥á‡≤¨‡≤≤‡≥ç‚Äå‡≤® (‡≤à ‡≤™‡≥ç‡≤∞‡≤ï‡≤∞‡≤£‡≤¶‡≤≤‡≥ç‡≤≤‡≤ø, ‡≤ï‡≤Ç‡≤¨‡≤≥‡≤ø‡≤Ø ‡≤¨‡≥Ü‡≤≤‡≥Ü) ‡≤Ö‡≤¶‡≥á ‡≤ò‡≤ü‡≤ï‡≤¶‡≤≤‡≥ç‡≤≤‡≤ø ‡≤™‡≤∞‡≤Æ‡≤æ‡≤£‡≥Å ‡≤Æ‡≥Ü‡≤ü‡≥ç‡≤∞‡≤ø‡≤ï‡≥ç ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤®‡≥Ä‡≤°‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü. ‡≤Æ‡≥å‡≤≤‡≥ç‡≤Ø‡≤µ‡≥Å ‡≤ï‡≤°‡≤ø‡≤Æ‡≥Ü ‡≤á‡≤¶‡≥ç‡≤¶‡≤∞‡≥Ü, ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø ‡≤â‡≤§‡≥ç‡≤§‡≤Æ‡≤µ‡≤æ‡≤ó‡≤ø‡≤¶‡≥Ü (‡≤∏‡≤∞‡≤≥‡≤µ‡≤æ‡≤ó‡≤ø ‡≤π‡≥á‡≤≥‡≥Å‡≤µ‡≥Å‡≤¶‡≤æ‡≤¶‡≤∞‡≥Ü, ‡≤á‡≤¶‡≥Å ‡≤≠‡≤µ‡≤ø‡≤∑‡≥ç‡≤Ø‡≤µ‡≤æ‡≤£‡≤ø‡≤ó‡≤≥‡≥Å ‡≤§‡≤™‡≥ç‡≤™‡≤æ‡≤ó‡≤ø‡≤∞‡≥Å‡≤µ ‡≤∏‡≤∞‡≤æ‡≤∏‡≤∞‡≤ø ‡≤¨‡≥Ü‡≤≤‡≥Ü‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤™‡≥ç‡≤∞‡≤§‡≤ø‡≤®‡≤ø‡≤ß‡≤ø‡≤∏‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü!)

-   `Coefficient of Determination (‡≤∏‡≤æ‡≤Æ‡≤æ‡≤®‡≥ç‡≤Ø‡≤µ‡≤æ‡≤ó‡≤ø R-squared ‡≤Ö‡≤•‡≤µ‡≤æ R2 ‡≤é‡≤Ç‡≤¶‡≥Å ‡≤ï‡≤∞‡≥Ü‡≤Ø‡≤≤‡≥ç‡≤™‡≤°‡≥Å‡≤µ‡≥Å‡≤¶‡≥Å)`: ‡≤á‡≤¶‡≥Å ‡≤∏‡≤æ‡≤™‡≥á‡≤ï‡≥ç‡≤∑ ‡≤Æ‡≥Ü‡≤ü‡≥ç‡≤∞‡≤ø‡≤ï‡≥ç ‡≤Ü‡≤ó‡≤ø‡≤¶‡≥ç‡≤¶‡≥Å, ‡≤Æ‡≥å‡≤≤‡≥ç‡≤Ø‡≤µ‡≥Å ‡≤π‡≥Ü‡≤ö‡≥ç‡≤ö‡≥Å ‡≤á‡≤¶‡≥ç‡≤¶‡≤∞‡≥Ü, ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø‡≤Ø ‡≤π‡≥ä‡≤Ç‡≤¶‡≤æ‡≤£‡≤ø‡≤ï‡≥Ü ‡≤â‡≤§‡≥ç‡≤§‡≤Æ‡≤µ‡≤æ‡≤ó‡≤ø‡≤¶‡≥Ü. ‡≤Æ‡≥Ç‡≤≤‡≤§‡≤É, ‡≤à ‡≤Æ‡≥Ü‡≤ü‡≥ç‡≤∞‡≤ø‡≤ï‡≥ç ‡≤≠‡≤µ‡≤ø‡≤∑‡≥ç‡≤Ø‡≤µ‡≤æ‡≤£‡≤ø ‡≤Æ‡≤æ‡≤°‡≤ø‡≤¶ ‡≤Æ‡≤§‡≥ç‡≤§‡≥Å ‡≤®‡≤ø‡≤ú‡≤µ‡≤æ‡≤¶ ‡≤≤‡≥á‡≤¨‡≤≤‡≥ç ‡≤Æ‡≥å‡≤≤‡≥ç‡≤Ø‡≤ó‡≤≥ ‡≤®‡≤°‡≥Å‡≤µ‡≤ø‡≤® ‡≤µ‡≥ç‡≤Ø‡≤§‡≥ç‡≤Ø‡≤æ‡≤∏‡≤µ‡≤®‡≥ç‡≤®‡≥Å ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø ‡≤é‡≤∑‡≥ç‡≤ü‡≥Å ‡≤µ‡≤ø‡≤µ‡≤∞‡≤ø‡≤∏‡≤¨‡≤π‡≥Å‡≤¶‡≥Å ‡≤é‡≤Ç‡≤¨‡≥Å‡≤¶‡≤®‡≥ç‡≤®‡≥Å ‡≤™‡≥ç‡≤∞‡≤§‡≤ø‡≤®‡≤ø‡≤ß‡≤ø‡≤∏‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü.


```r
# Evaluate performance of linear regression
metrics(data = lm_results,
        truth = price,
        estimate = .pred)
```

‡≤Æ‡≤æ‡≤°‡≥Ü‡≤≤‡≥ç ‡≤ï‡≤æ‡≤∞‡≥ç‡≤Ø‡≤ï‡≥ç‡≤∑‡≤Æ‡≤§‡≥Ü ‡≤á‡≤≤‡≥ç‡≤≤‡≤ø‡≤ó‡≥Ü ‡≤π‡≥ã‡≤ó‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü. ‡≤™‡≥ç‡≤Ø‡≤æ‡≤ï‡≥á‡≤ú‡≥ç ‡≤Æ‡≤§‡≥ç‡≤§‡≥Å ‡≤¨‡≥Ü‡≤≤‡≥Ü‡≤Ø scatter ‡≤™‡≥ç‡≤≤‡≤æ‡≤ü‡≥ç ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤¶‡≥É‡≤∂‡≥ç‡≤Ø‡≥Ä‡≤ï‡≤∞‡≤ø‡≤∏‡≥Å‡≤µ ‡≤Æ‡≥Ç‡≤≤‡≤ï ‡≤®‡≤æ‡≤µ‡≥Å ‡≤â‡≤§‡≥ç‡≤§‡≤Æ ‡≤∏‡≥Ç‡≤ö‡≤®‡≥Ü‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤™‡≤°‡≥Ü‡≤Ø‡≤¨‡≤π‡≥Å‡≤¶‡≥á ‡≤é‡≤Ç‡≤¶‡≥Å ‡≤®‡≥ã‡≤°‡≥ã‡≤£, ‡≤®‡≤Ç‡≤§‡≤∞ ‡≤Æ‡≤æ‡≤°‡≤≤‡≤æ‡≤¶ ‡≤≠‡≤µ‡≤ø‡≤∑‡≥ç‡≤Ø‡≤µ‡≤æ‡≤£‡≤ø‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å ‡≤¨‡≤≥‡≤∏‡≤ø‡≤ï‡≥ä‡≤Ç‡≤°‡≥Å ‡≤â‡≤§‡≥ç‡≤§‡≤Æ ‡≤π‡≥ä‡≤Ç‡≤¶‡≤æ‡≤£‡≤ø‡≤ï‡≥Ü‡≤Ø ‡≤∞‡≥á‡≤ñ‡≥Ü‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤ì‡≤µ‡≤∞‡≥ç‚Äå‡≤≤‡≥Ü ‡≤Æ‡≤æ‡≤°‡≥ã‡≤£.

‡≤á‡≤¶‡≤ï‡≥ç‡≤ï‡≥Ü ‡≤Ö‡≤∞‡≥ç‡≤•, ‡≤®‡≤æ‡≤µ‡≥Å ‡≤™‡≥ç‡≤Ø‡≤æ‡≤ï‡≥á‡≤ú‡≥ç ‡≤ï‡≤æ‡≤≤‡≤Æ‡≥ç ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤é‡≤®‡≥ç‚Äå‡≤ï‡≥ã‡≤°‡≥ç ‡≤Æ‡≤æ‡≤°‡≤≤‡≥Å ‡≤ü‡≥Ü‡≤∏‡≥ç‡≤ü‡≥ç ‡≤∏‡≥Ü‡≤ü‡≥ç ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤§‡≤Ø‡≤æ‡≤∞‡≤ø‡≤∏‡≤ø ‡≤¨‡≥á‡≤ï‡≥ç ‡≤Æ‡≤æ‡≤°‡≤¨‡≥á‡≤ï‡≤æ‡≤ó‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü, ‡≤®‡≤Ç‡≤§‡≤∞ ‡≤á‡≤¶‡≤®‡≥ç‡≤®‡≥Å ‡≤®‡≤Æ‡≥ç‡≤Æ ‡≤Æ‡≤æ‡≤°‡≥Ü‡≤≤‡≥ç ‡≤Æ‡≤æ‡≤°‡≤ø‡≤¶ ‡≤≠‡≤µ‡≤ø‡≤∑‡≥ç‡≤Ø‡≤µ‡≤æ‡≤£‡≤ø‡≤ó‡≤≥‡≥ä‡≤Ç‡≤¶‡≤ø‡≤ó‡≥Ü ‡≤¨‡≤æ‡≤Ç‡≤ß‡≤ø‡≤∏‡≤¨‡≥á‡≤ï‡≥Å.
