```r
# Load the core tidyverse and make it available in your current R session
library(tidyverse)

# Import the data into a tibble
df <- read_csv(file = "https://raw.githubusercontent.com/microsoft/ML-For-Beginners/main/5-Clustering/data/nigerian-songs.csv", show_col_types = FALSE)

# Narrow down to top 3 popular genres
nigerian_songs <- df %>% 
  # Concentrate on top 3 genres
  filter(artist_top_genre %in% c("afro dancehall", "afropop","nigerian pop")) %>% 
  # Remove unclassified observations
  filter(popularity != 0)



# Visualize popular genres using bar plots
theme_set(theme_light())
nigerian_songs %>%
  count(artist_top_genre) %>%
  ggplot(mapping = aes(x = artist_top_genre, y = n,
                       fill = artist_top_genre)) +
  geom_col(alpha = 0.8) +
  paletteer::scale_fill_paletteer_d("ggsci::category10_d3") +
  ggtitle("Top genres") +
  theme(plot.title = element_text(hjust = 0.5))
```

‡≤®‡≤æ‡≤µ‡≥Å ‡≤™‡≥ç‡≤∞‡≤§‡≤ø ‡≤∏‡≤Ç‡≤ñ‡≥ç‡≤Ø‡≤æ‡≤§‡≥ç‡≤Æ‡≤ï ‡≤≤‡≤ï‡≥ç‡≤∑‡≤£‡≤ó‡≤≥‡≤ø‡≤ó‡≤æ‡≤ó‡≤ø ‡≤¨‡≤æ‡≤ï‡≥ç‡≤∏‡≥ç‚Äå‡≤™‡≥ç‡≤≤‡≤æ‡≤ü‡≥ç ‡≤Æ‡≤æ‡≤°‡≤≤‡≤ø‡≤¶‡≥ç‡≤¶‡≥á‡≤µ‡≥Ü ‡≤Æ‡≤§‡≥ç‡≤§‡≥Å ‡≤≤‡≥Ç‡≤™‡≥ç‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å ‡≤¨‡≤≥‡≤∏‡≥Å‡≤µ‡≥Å‡≤¶‡≤®‡≥ç‡≤®‡≥Å ‡≤§‡≤™‡≥ç‡≤™‡≤ø‡≤∏‡≤≤‡≥Å ‡≤¨‡≤Ø‡≤∏‡≥Å‡≤§‡≥ç‡≤§‡≥á‡≤µ‡≥Ü, ‡≤Ü‡≤¶‡≥ç‡≤¶‡≤∞‡≤ø‡≤Ç‡≤¶ ‡≤®‡≤Æ‡≥ç‡≤Æ ‡≤°‡≥á‡≤ü‡≤æ‡≤µ‡≤®‡≥ç‡≤®‡≥Å *‡≤¶‡≥Ä‡≤∞‡≥ç‡≤ò* ‡≤∏‡≥ç‡≤µ‡≤∞‡≥Ç‡≤™‡≤ï‡≥ç‡≤ï‡≥Ü ‡≤Æ‡≤∞‡≥Å‡≤∞‡≥Ç‡≤™‡≤ó‡≥ä‡≤≥‡≤ø‡≤∏‡≥ã‡≤£, ‡≤á‡≤¶‡≥Å ‡≤®‡≤Æ‡≤ó‡≥Ü `facets` - ‡≤™‡≥ç‡≤∞‡≤§‡≤ø ‡≤â‡≤™‡≤∏‡≤Æ‡≥Ç‡≤π‡≤µ‡≤®‡≥ç‡≤®‡≥Å ‡≤™‡≥ç‡≤∞‡≤¶‡≤∞‡≥ç‡≤∂‡≤ø‡≤∏‡≥Å‡≤µ ‡≤â‡≤™‡≤ö‡≤ø‡≤§‡≥ç‡≤∞‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å ‡≤¨‡≤≥‡≤∏‡≤≤‡≥Å ‡≤Ö‡≤µ‡≤ï‡≤æ‡≤∂ ‡≤®‡≥Ä‡≤°‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü.

```r
# Make a box plot
df_numeric_long %>% 
  ggplot(mapping = aes(x = feature_names, y = values, fill = feature_names)) +
  geom_boxplot() +
  facet_wrap(~ feature_names, ncol = 4, scales = "free") +
  theme(legend.position = "none")
```

## 3. R ‡≤®‡≤≤‡≥ç‡≤≤‡≤ø k-means ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≤ø‡≤Ç‡≤ó‡≥ç ‡≤≤‡≥Ü‡≤ï‡≥ç‡≤ï‡≤π‡≤æ‡≤ï‡≥Å‡≤µ‡≥Å‡≤¶‡≥Å

‡≤®‡≤æ‡≤µ‡≥Å R ‡≤®‡≤≤‡≥ç‡≤≤‡≤ø ‡≤®‡≤ø‡≤∞‡≥ç‡≤Æ‡≤ø‡≤§ `kmeans` ‡≤´‡≤Ç‡≤ï‡≥ç‡≤∑‡≤®‡≥ç ‡≤¨‡≤≥‡≤∏‡≤ø k-means ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤≤‡≥Ü‡≤ï‡≥ç‡≤ï‡≤π‡≤æ‡≤ï‡≤¨‡≤π‡≥Å‡≤¶‡≥Å, `help("kmeans()")` ‡≤®‡≥ã‡≤°‡≤ø. `kmeans()` ‡≤´‡≤Ç‡≤ï‡≥ç‡≤∑‡≤®‡≥ç ‡≤§‡≤®‡≥ç‡≤® ‡≤™‡≥ç‡≤∞‡≤æ‡≤•‡≤Æ‡≤ø‡≤ï ‡≤Ü‡≤∞‡≥ç‡≤ó‡≥ç‡≤Ø‡≥Å‡≤Æ‡≥Ü‡≤Ç‡≤ü‡≥ç ‡≤Ü‡≤ó‡≤ø ‡≤é‡≤≤‡≥ç‡≤≤‡≤æ ‡≤∏‡≤Ç‡≤ñ‡≥ç‡≤Ø‡≤æ‡≤§‡≥ç‡≤Æ‡≤ï ‡≤ï‡≤æ‡≤≤‡≤Æ‡≥ç‚Äå‡≤ó‡≤≥‡≤ø‡≤∞‡≥Å‡≤µ ‡≤°‡≥á‡≤ü‡≤æ ‡≤´‡≥ç‡≤∞‡≥á‡≤Æ‡≥ç ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤∏‡≥ç‡≤µ‡≥Ä‡≤ï‡≤∞‡≤ø‡≤∏‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü.

k-means ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≤ø‡≤Ç‡≤ó‡≥ç ‡≤¨‡≤≥‡≤∏‡≥Å‡≤µ ‡≤Æ‡≥ä‡≤¶‡≤≤ ‡≤π‡≤Ç‡≤§‡≤µ‡≥Å ‡≤Ö‡≤Ç‡≤§‡≤ø‡≤Æ ‡≤™‡≤∞‡≤ø‡≤π‡≤æ‡≤∞‡≤¶‡≤≤‡≥ç‡≤≤‡≤ø ‡≤∞‡≤ö‡≤ø‡≤∏‡≤≤‡≤æ‡≤ó‡≥Å‡≤µ ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç‚Äå‡≤ó‡≤≥ ‡≤∏‡≤Ç‡≤ñ‡≥ç‡≤Ø‡≥Ü (k) ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤®‡≤ø‡≤∞‡≥ç‡≤¶‡≤ø‡≤∑‡≥ç‡≤ü‡≤™‡≤°‡≤ø‡≤∏‡≥Å‡≤µ‡≥Å‡≤¶‡≥Å. ‡≤®‡≤æ‡≤µ‡≥Å ‡≤°‡≥á‡≤ü‡≤æ‡≤∏‡≥Ü‡≤ü‡≥ç‚Äå‡≤®‡≤ø‡≤Ç‡≤¶ 3 ‡≤π‡≤æ‡≤°‡≥Å ‡≤∂‡≥à‡≤≤‡≤ø‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å ‡≤π‡≥ä‡≤∞‡≤§‡≥Ü‡≤ó‡≥Ü‡≤Ø‡≤≤‡≤æ‡≤ó‡≤ø‡≤¶‡≥Ü ‡≤é‡≤Ç‡≤¶‡≥Å ‡≤§‡≤ø‡≤≥‡≤ø‡≤¶‡≤ø‡≤¶‡≥ç‡≤¶‡≥á‡≤µ‡≥Ü, ‡≤Ü‡≤¶‡≥ç‡≤¶‡≤∞‡≤ø‡≤Ç‡≤¶ 3 ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤™‡≥ç‡≤∞‡≤Ø‡≤§‡≥ç‡≤®‡≤ø‡≤∏‡≥ã‡≤£:


```r
set.seed(2056)
# Kmeans clustering for 3 clusters
kclust <- kmeans(
  df_numeric_select,
  # Specify the number of clusters
  centers = 3,
  # How many random initial configurations
  nstart = 25
)

# Display clustering object
kclust
```

kmeans ‡≤µ‡≤∏‡≥ç‡≤§‡≥Å‡≤µ‡≤ø‡≤®‡≤≤‡≥ç‡≤≤‡≤ø ‡≤π‡≤≤‡≤µ‡≤æ‡≤∞‡≥Å ‡≤Æ‡≤æ‡≤π‡≤ø‡≤§‡≤ø‡≤Ø ‡≤§‡≥Å‡≤Ç‡≤°‡≥Å‡≤ó‡≤≥‡≥Å ‡≤á‡≤µ‡≥Ü, ‡≤Ö‡≤µ‡≥Å‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å `help("kmeans()")` ‡≤®‡≤≤‡≥ç‡≤≤‡≤ø ‡≤ö‡≥Ü‡≤®‡≥ç‡≤®‡≤æ‡≤ó‡≤ø ‡≤µ‡≤ø‡≤µ‡≤∞‡≤ø‡≤∏‡≤≤‡≤æ‡≤ó‡≤ø‡≤¶‡≥Ü. ‡≤à‡≤ó, ‡≤ï‡≥Ü‡≤≤‡≤µ‡≥Å ‡≤µ‡≤ø‡≤∑‡≤Ø‡≤ó‡≤≥ ‡≤Æ‡≥á‡≤≤‡≥Ü ‡≤ó‡≤Æ‡≤®‡≤π‡≤∞‡≤ø‡≤∏‡≥ã‡≤£. ‡≤°‡≥á‡≤ü‡≤æ 65, 110, 111 ‡≤ó‡≤æ‡≤§‡≥ç‡≤∞‡≤¶ 3 ‡≤ó‡≥Å‡≤Ç‡≤™‡≥Å‡≤ó‡≤≥‡≤æ‡≤ó‡≤ø ‡≤ó‡≥Å‡≤Ç‡≤™‡≥Å‡≤¨‡≤¶‡≥ç‡≤ß‡≤ó‡≥ä‡≤Ç‡≤°‡≤ø‡≤¶‡≥Ü ‡≤é‡≤Ç‡≤¶‡≥Å ‡≤®‡≤æ‡≤µ‡≥Å ‡≤®‡≥ã‡≤°‡≥Å‡≤§‡≥ç‡≤§‡≥á‡≤µ‡≥Ü. ‡≤î‡≤ü‡≥ç‚Äå‡≤™‡≥Å‡≤ü‡≥ç‚Äå‡≤®‡≤≤‡≥ç‡≤≤‡≤ø 5 ‡≤ö‡≤∞‡≤ó‡≤≥‡≤æ‡≤¶ 3 ‡≤ó‡≥Å‡≤Ç‡≤™‡≥Å‡≤ó‡≤≥ ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç ‡≤ï‡≥á‡≤Ç‡≤¶‡≥ç‡≤∞‡≤ó‡≤≥‡≥Å (‡≤∏‡≤∞‡≤æ‡≤∏‡≤∞‡≤ø) ‡≤ï‡≥Ç‡≤° ‡≤∏‡≥á‡≤∞‡≤ø‡≤µ‡≥Ü.

‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≤ø‡≤Ç‡≤ó‡≥ç ‡≤µ‡≥Ü‡≤ï‡≥ç‡≤ü‡≤∞‡≥ç ‡≤™‡≥ç‡≤∞‡≤§‡≤ø ‡≤µ‡≥Ä‡≤ï‡≥ç‡≤∑‡≤£‡≥Ü‡≤Ø ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç ‡≤®‡≤ø‡≤Ø‡≥ã‡≤ú‡≤®‡≥Ü ‡≤Ü‡≤ó‡≤ø‡≤¶‡≥Ü. ‡≤Æ‡≥Ç‡≤≤ ‡≤°‡≥á‡≤ü‡≤æ ‡≤∏‡≥Ü‡≤ü‡≥ç‚Äå‡≤ó‡≥Ü ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç ‡≤®‡≤ø‡≤Ø‡≥ã‡≤ú‡≤®‡≥Ü‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤∏‡≥á‡≤∞‡≤ø‡≤∏‡≤≤‡≥Å `augment` ‡≤´‡≤Ç‡≤ï‡≥ç‡≤∑‡≤®‡≥ç ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤¨‡≤≥‡≤∏‡≥ã‡≤£.


```r
# Add predicted cluster assignment to data set
augment(kclust, df_numeric_select) %>% 
  relocate(.cluster) %>% 
  slice_head(n = 10)
```

### **Silhouette score**

[Silhouette ‡≤µ‡≤ø‡≤∂‡≥ç‡≤≤‡≥á‡≤∑‡≤£‡≥Ü](https://en.wikipedia.org/wiki/Silhouette_(clustering)) ‡≤´‡≤≤‡≤ø‡≤§‡≤æ‡≤Ç‡≤∂ ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç‚Äå‡≤ó‡≤≥ ‡≤®‡≤°‡≥Å‡≤µ‡≤ø‡≤® ‡≤µ‡≤ø‡≤≠‡≤ú‡≤®‡≥Ü ‡≤¶‡≥Ç‡≤∞‡≤µ‡≤®‡≥ç‡≤®‡≥Å ‡≤Ö‡≤ß‡≥ç‡≤Ø‡≤Ø‡≤® ‡≤Æ‡≤æ‡≤°‡≤≤‡≥Å ‡≤¨‡≤≥‡≤∏‡≤¨‡≤π‡≥Å‡≤¶‡≥Å. ‡≤à ‡≤∏‡≥ç‡≤ï‡≥ã‡≤∞‡≥ç -1 ‡≤∞‡≤ø‡≤Ç‡≤¶ 1 ‡≤∞‡≤µ‡≤∞‡≥Ü‡≤ó‡≥Ü ‡≤¨‡≤¶‡≤≤‡≤æ‡≤ó‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü, ‡≤Æ‡≤§‡≥ç‡≤§‡≥Å ‡≤∏‡≥ç‡≤ï‡≥ã‡≤∞‡≥ç 1 ‡≤ó‡≥Ü ‡≤π‡≤§‡≥ç‡≤§‡≤ø‡≤∞ ‡≤á‡≤¶‡≥ç‡≤¶‡≤∞‡≥Ü, ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç ‡≤ó‡≤ü‡≥ç‡≤ü‡≤ø‡≤Ø‡≤æ‡≤ó‡≤ø‡≤¶‡≥ç‡≤¶‡≥Å ‡≤á‡≤§‡≤∞ ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç‚Äå‡≤ó‡≤≥‡≤ø‡≤Ç‡≤¶ ‡≤ö‡≥Ü‡≤®‡≥ç‡≤®‡≤æ‡≤ó‡≤ø ‡≤µ‡≤ø‡≤≠‡≤ú‡≤ø‡≤∏‡≤≤‡≤æ‡≤ó‡≤ø‡≤¶‡≥Ü. 0 ‡≤ó‡≥Ü ‡≤π‡≤§‡≥ç‡≤§‡≤ø‡≤∞‡≤¶ ‡≤Æ‡≥å‡≤≤‡≥ç‡≤Ø‡≤µ‡≥Å ‡≤π‡≤§‡≥ç‡≤§‡≤ø‡≤∞‡≤¶ ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç‚Äå‡≤ó‡≤≥ ‡≤®‡≤ø‡≤∞‡≥ç‡≤ß‡≤æ‡≤∞ ‡≤ó‡≤°‡≤ø‡≤¨‡≤ø‡≤°‡≤ø‡≤Ø ‡≤¨‡≤≥‡≤ø ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø‡≤ó‡≤≥‡≥ä‡≤Ç‡≤¶‡≤ø‡≤ó‡≥Ü ‡≤í‡≤§‡≥ç‡≤§‡≤°‡≤¶ ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç‚Äå‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å ‡≤™‡≥ç‡≤∞‡≤§‡≤ø‡≤®‡≤ø‡≤ß‡≤ø‡≤∏‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü.[‡≤Æ‡≥Ç‡≤≤](https://dzone.com/articles/kmeans-silhouette-score-explained-with-python-exam).

‡≤∏‡≤∞‡≤æ‡≤∏‡≤∞‡≤ø silhouette ‡≤µ‡≤ø‡≤ß‡≤æ‡≤®‡≤µ‡≥Å ‡≤µ‡≤ø‡≤≠‡≤ø‡≤®‡≥ç‡≤® *k* ‡≤Æ‡≥å‡≤≤‡≥ç‡≤Ø‡≤ó‡≤≥‡≤ø‡≤ó‡≤æ‡≤ó‡≤ø ‡≤µ‡≥Ä‡≤ï‡≥ç‡≤∑‡≤£‡≥Ü‡≤ó‡≤≥ ‡≤∏‡≤∞‡≤æ‡≤∏‡≤∞‡≤ø silhouette ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤≤‡≥Ü‡≤ï‡≥ç‡≤ï‡≤π‡≤æ‡≤ï‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü. ‡≤â‡≤®‡≥ç‡≤®‡≤§ ‡≤∏‡≤∞‡≤æ‡≤∏‡≤∞‡≤ø silhouette ‡≤∏‡≥ç‡≤ï‡≥ã‡≤∞‡≥ç ‡≤â‡≤§‡≥ç‡≤§‡≤Æ ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≤ø‡≤Ç‡≤ó‡≥ç ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤∏‡≥Ç‡≤ö‡≤ø‡≤∏‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü.

‡≤∏‡≤∞‡≤æ‡≤∏‡≤∞‡≤ø silhouette ‡≤Ö‡≤ó‡≤≤‡≤µ‡≤®‡≥ç‡≤®‡≥Å ‡≤≤‡≥Ü‡≤ï‡≥ç‡≤ï‡≤π‡≤æ‡≤ï‡≤≤‡≥Å ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç ‡≤™‡≥ç‡≤Ø‡≤æ‡≤ï‡≥á‡≤ú‡≤ø‡≤® `silhouette` ‡≤´‡≤Ç‡≤ï‡≥ç‡≤∑‡≤®‡≥ç ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤¨‡≤≥‡≤∏‡≤¨‡≤π‡≥Å‡≤¶‡≥Å.

> silhouette ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤Ø‡≤æ‡≤µ‡≥Å‡≤¶‡≥á [‡≤¶‡≥Ç‡≤∞](https://en.wikipedia.org/wiki/Distance "Distance") ‡≤Æ‡≤æ‡≤™‡≤ï‡≤¶‡≤ø‡≤Ç‡≤¶ ‡≤≤‡≥Ü‡≤ï‡≥ç‡≤ï‡≤π‡≤æ‡≤ï‡≤¨‡≤π‡≥Å‡≤¶‡≥Å, ‡≤â‡≤¶‡≤æ‡≤π‡≤∞‡≤£‡≥Ü‡≤ó‡≥Ü ‡≤®‡≤æ‡≤µ‡≥Å [‡≤π‡≤ø‡≤Ç‡≤¶‡≤ø‡≤® ‡≤™‡≤æ‡≤†‡≤¶‡≤≤‡≥ç‡≤≤‡≤ø](https://github.com/microsoft/ML-For-Beginners/blob/main/5-Clustering/1-Visualize/solution/R/lesson_14-R.ipynb) ‡≤ö‡≤∞‡≥ç‡≤ö‡≤ø‡≤∏‡≤ø‡≤¶‡≤Ç‡≤§‡≥Ü [‡≤Ø‡≥Ç‡≤ï‡≥ç‡≤≤‡≤ø‡≤°‡≤ø‡≤Ø‡≤®‡≥ç ‡≤¶‡≥Ç‡≤∞](https://en.wikipedia.org/wiki/Euclidean_distance "Euclidean distance") ‡≤Ö‡≤•‡≤µ‡≤æ [‡≤Æ‡≥ç‡≤Ø‡≤æ‡≤®‡≥ç‡≤π‡≥ç‡≤Ø‡≤æ‡≤ü‡≤®‡≥ç ‡≤¶‡≥Ç‡≤∞](https://en.wikipedia.org/wiki/Manhattan_distance "Manhattan distance").


```r
# Load cluster package
library(cluster)

# Compute average silhouette score
ss <- silhouette(kclust$cluster,
                 # Compute euclidean distance
                 dist = dist(df_numeric_select))
mean(ss[, 3])
```

‡≤®‡≤æ‡≤µ‡≥Å ‡≤™‡≥ç‡≤∞‡≤Ø‡≤§‡≥ç‡≤®‡≤ø‡≤∏‡≤¨‡≤π‡≥Å‡≤¶‡≤æ‡≤¶ ‡≤í‡≤Ç‡≤¶‡≥Å ‡≤µ‡≤ø‡≤ß‡≤æ‡≤®‡≤µ‡≥Ü‡≤Ç‡≤¶‡≤∞‡≥Ü ‡≤°‡≥á‡≤ü‡≤æ ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤¨‡≤≥‡≤∏‡≤ø‡≤ï‡≥ä‡≤Ç‡≤°‡≥Å ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç‚Äå‡≤ó‡≤≥ ‡≤∏‡≤Ç‡≤ñ‡≥ç‡≤Ø‡≥Ü‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤ï‡≥ç‡≤∞‡≤Æ‡≥á‡≤£ ‡≤π‡≥Ü‡≤ö‡≥ç‡≤ö‡≤ø‡≤∏‡≥Å‡≤µ ‡≤Æ‡≥Ç‡≤≤‡≤ï (‡≤â‡≤¶‡≤æ: 1-10 ‡≤∞‡≤µ‡≤∞‡≥Ü‡≤ó‡≥Ü) `‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≤ø‡≤Ç‡≤ó‡≥ç ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø‡≤ó‡≤≥ ‡≤∏‡≤∞‡≤£‡≤ø‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤∞‡≤ö‡≤ø‡≤∏‡≥Å‡≤µ‡≥Å‡≤¶‡≥Å` ‡≤Æ‡≤§‡≥ç‡≤§‡≥Å **‡≤∏‡≤ø‡≤≤‡≥ç‡≤π‡≥å‡≤ü‡≥ç ‡≤∏‡≥ç‡≤ï‡≥ã‡≤∞‡≥ç** ‡≤Æ‡≥Å‡≤Ç‡≤§‡≤æ‡≤¶ ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≤ø‡≤Ç‡≤ó‡≥ç ‡≤Æ‡≥Ü‡≤ü‡≥ç‡≤∞‡≤ø‡≤ï‡≥ç‡≤∏‡≥ç‚Äå‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å ‡≤Æ‡≥å‡≤≤‡≥ç‡≤Ø‡≤Æ‡≤æ‡≤™‡≤® ‡≤Æ‡≤æ‡≤°‡≥Å‡≤µ‡≥Å‡≤¶‡≥Å.

‡≤µ‡≤ø‡≤≠‡≤ø‡≤®‡≥ç‡≤® *k* ‡≤Æ‡≥å‡≤≤‡≥ç‡≤Ø‡≤ó‡≤≥‡≤ø‡≤ó‡≤æ‡≤ó‡≤ø ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≤ø‡≤Ç‡≤ó‡≥ç ‡≤Ü‡≤≤‡≥ç‡≤ó‡≥ã‡≤∞‡≤ø‡≤¶‡≤Æ‡≥ç ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤ó‡≤£‡≤®‡≥Ü ‡≤Æ‡≤æ‡≤°‡≤ø ‡≤Æ‡≤§‡≥ç‡≤§‡≥Å **‡≤µ‡≤ø‡≤•‡≤ø‡≤®‡≥ç ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç ‡≤∏‡≤Æ‡≥ç ‡≤Ü‡≤´‡≥ç ‡≤∏‡≥ç‡≤ï‡≥ç‡≤µ‡≥á‡≤∞‡≥ç‡≤∏‡≥ç** (WCSS) ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤Æ‡≥å‡≤≤‡≥ç‡≤Ø‡≤Æ‡≤æ‡≤™‡≤® ‡≤Æ‡≤æ‡≤°‡≥Å‡≤µ ‡≤Æ‡≥Ç‡≤≤‡≤ï ‡≤Ö‡≤§‡≥ç‡≤Ø‡≥Å‡≤§‡≥ç‡≤§‡≤Æ ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç ‡≤∏‡≤Ç‡≤ñ‡≥ç‡≤Ø‡≥Ü‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤®‡≤ø‡≤∞‡≥ç‡≤ß‡≤∞‡≤ø‡≤∏‡≥ã‡≤£. ‡≤í‡≤ü‡≥ç‡≤ü‡≥Å ‡≤µ‡≤ø‡≤•‡≤ø‡≤®‡≥ç-‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç ‡≤∏‡≤Æ‡≥ç ‡≤Ü‡≤´‡≥ç ‡≤∏‡≥ç‡≤ï‡≥ç‡≤µ‡≥á‡≤∞‡≥ç‡≤∏‡≥ç (WCSS) ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≤ø‡≤Ç‡≤ó‡≥ç‚Äå‡≤® ‡≤∏‡≤Ç‡≤ï‡≥Å‡≤ö‡≤ø‡≤§‡≤§‡≥Ü‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤Ö‡≤≥‡≥Ü‡≤Ø‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü ‡≤Æ‡≤§‡≥ç‡≤§‡≥Å ‡≤®‡≤æ‡≤µ‡≥Å ‡≤Ö‡≤¶‡≤®‡≥ç‡≤®‡≥Å ‡≤∏‡≤æ‡≤ß‡≥ç‡≤Ø‡≤µ‡≤æ‡≤¶‡≤∑‡≥ç‡≤ü‡≥Å ‡≤ï‡≤°‡≤ø‡≤Æ‡≥Ü ‡≤á‡≤∞‡≤ø‡≤∏‡≤≤‡≥Å ‡≤¨‡≤Ø‡≤∏‡≥Å‡≤§‡≥ç‡≤§‡≥á‡≤µ‡≥Ü, ‡≤ï‡≤°‡≤ø‡≤Æ‡≥Ü ‡≤Æ‡≥å‡≤≤‡≥ç‡≤Ø‡≤ó‡≤≥‡≥Å ‡≤Ö‡≤Ç‡≤¶‡≤∞‡≥Ü ‡≤°‡≥á‡≤ü‡≤æ ‡≤™‡≤æ‡≤Ø‡≤ø‡≤Ç‡≤ü‡≥ç‚Äå‡≤ó‡≤≥‡≥Å ‡≤π‡≤§‡≥ç‡≤§‡≤ø‡≤∞‡≤µ‡≤ø‡≤∞‡≥Å‡≤µ‡≥Å‡≤¶‡≤®‡≥ç‡≤®‡≥Å ‡≤∏‡≥Ç‡≤ö‡≤ø‡≤∏‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü.

1 ‡≤∞‡≤ø‡≤Ç‡≤¶ 10 ‡≤∞‡≤µ‡≤∞‡≥Ü‡≤ó‡≥Ü `k` ‡≤® ‡≤µ‡≤ø‡≤≠‡≤ø‡≤®‡≥ç‡≤® ‡≤Ü‡≤Ø‡≥ç‡≤ï‡≥Ü‡≤ó‡≤≥ ‡≤™‡≤∞‡≤ø‡≤£‡≤æ‡≤Æ‡≤µ‡≤®‡≥ç‡≤®‡≥Å ‡≤à ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≤ø‡≤Ç‡≤ó‡≥ç ‡≤Æ‡≥á‡≤≤‡≥Ü ‡≤Ö‡≤®‡≥ç‡≤µ‡≥á‡≤∑‡≤ø‡≤∏‡≥ã‡≤£.


```r
# Create a series of clustering models
kclusts <- tibble(k = 1:10) %>% 
  # Perform kmeans clustering for 1,2,3 ... ,10 clusters
  mutate(model = map(k, ~ kmeans(df_numeric_select, centers = .x, nstart = 25)),
  # Farm out clustering metrics eg WCSS
         glanced = map(model, ~ glance(.x))) %>% 
  unnest(cols = glanced)
  

# View clustering rsulsts
kclusts
```

‡≤à‡≤ó ‡≤®‡≤æ‡≤µ‡≥Å ‡≤™‡≥ç‡≤∞‡≤§‡≤ø ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≤ø‡≤Ç‡≤ó‡≥ç ‡≤Ü‡≤≤‡≥ç‡≤ó‡≥ã‡≤∞‡≤ø‡≤¶‡≤Æ‡≥ç‚Äå‡≤ó‡≥Ü ‡≤ï‡≥á‡≤Ç‡≤¶‡≥ç‡≤∞ *k* ‡≤á‡≤∞‡≥Å‡≤µ ‡≤í‡≤ü‡≥ç‡≤ü‡≥Å ‡≤í‡≤≥‡≤ó‡≤ø‡≤®-‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç ‡≤∏‡≤Æ‡≥ç‡≤Æ‡≤ø‡≤∂‡≥ç‡≤∞‡≤£‡≤ó‡≤≥ ‡≤Æ‡≥ä‡≤§‡≥ç‡≤§ (tot.withinss) ‡≤π‡≥ä‡≤Ç‡≤¶‡≤ø‡≤¶‡≥ç‡≤¶‡≥á‡≤µ‡≥Ü, ‡≤®‡≤æ‡≤µ‡≥Å [‡≤é‡≤≤‡≥ç‡≤¨‡≥ã ‡≤µ‡≤ø‡≤ß‡≤æ‡≤®](https://en.wikipedia.org/wiki/Elbow_method_(clustering)) ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤¨‡≤≥‡≤∏‡≤ø‡≤ï‡≥ä‡≤Ç‡≤°‡≥Å ‡≤Ö‡≤§‡≥ç‡≤Ø‡≥Å‡≤§‡≥ç‡≤§‡≤Æ ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç ‡≤∏‡≤Ç‡≤ñ‡≥ç‡≤Ø‡≥Ü‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤ï‡≤Ç‡≤°‡≥Å‡≤π‡≤ø‡≤°‡≤ø‡≤Ø‡≥Å‡≤§‡≥ç‡≤§‡≥á‡≤µ‡≥Ü. ‡≤à ‡≤µ‡≤ø‡≤ß‡≤æ‡≤®‡≤µ‡≥Å ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç ‡≤∏‡≤Ç‡≤ñ‡≥ç‡≤Ø‡≥Ü‡≤Ø ‡≤ï‡≤æ‡≤∞‡≥ç‡≤Ø‡≤µ‡≤æ‡≤ó‡≤ø WCSS ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤ö‡≤ø‡≤§‡≥ç‡≤∞‡≤ø‡≤∏‡≥Å‡≤µ‡≥Å‡≤¶‡≤®‡≥ç‡≤®‡≥Å ‡≤í‡≤≥‡≤ó‡≥ä‡≤Ç‡≤°‡≤ø‡≤¶‡≥ç‡≤¶‡≥Å, ‡≤¨‡≤≥‡≤ï‡≥Ü ‡≤Æ‡≤æ‡≤°‡≤≤‡≥Å ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç ‡≤∏‡≤Ç‡≤ñ‡≥ç‡≤Ø‡≥Ü‡≤Ø‡≤æ‡≤ó‡≤ø [‡≤µ‡≤ï‡≥ç‡≤∞‡≤¶ ‡≤é‡≤≤‡≥ç‡≤¨‡≥ã](https://en.wikipedia.org/wiki/Elbow_of_the_curve "Elbow of the curve") ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤Ü‡≤Ø‡≥ç‡≤ï‡≥Ü‡≤Æ‡≤æ‡≤°‡≥Å‡≤µ‡≥Å‡≤¶‡≥Å.


```r
set.seed(2056)
# Use elbow method to determine optimum number of clusters
kclusts %>% 
  ggplot(mapping = aes(x = k, y = tot.withinss)) +
  geom_line(size = 1.2, alpha = 0.8, color = "#FF7F0EFF") +
  geom_point(size = 2, color = "#FF7F0EFF")
```

‡≤ö‡≤ø‡≤§‡≥ç‡≤∞‡≤µ‡≥Å ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç‚Äå‡≤ó‡≤≥ ‡≤∏‡≤Ç‡≤ñ‡≥ç‡≤Ø‡≥Ü ‡≤í‡≤Ç‡≤¶‡≥Å‡≤∞‡≤ø‡≤Ç‡≤¶ ‡≤é‡≤∞‡≤°‡≥Å‡≤ï‡≥ç‡≤ï‡≥Ü ‡≤π‡≥Ü‡≤ö‡≥ç‡≤ö‡≤æ‡≤¶‡≤Ç‡≤§‡≥Ü WCSS ‡≤®‡≤≤‡≥ç‡≤≤‡≤ø ‡≤¶‡≥ä‡≤°‡≥ç‡≤° ‡≤ï‡≤°‡≤ø‡≤§‡≤µ‡≤®‡≥ç‡≤®‡≥Å ‡≤§‡≥ã‡≤∞‡≤ø‡≤∏‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü (‡≤π‡≥Ä‡≤ó‡≤æ‡≤ó‡≤ø ‡≤π‡≥Ü‡≤ö‡≥ç‡≤ö‡≥Å *‡≤¶‡≥É‡≤¢‡≤§‡≥Ü*), ‡≤Æ‡≤§‡≥ç‡≤§‡≥Å ‡≤é‡≤∞‡≤°‡≥Å‡≤∞‡≤ø‡≤Ç‡≤¶ ‡≤Æ‡≥Ç‡≤∞‡≥Å ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç‚Äå‡≤ó‡≤≥‡≤ø‡≤ó‡≥Ü ‡≤Æ‡≤§‡≥ç‡≤§‡≤∑‡≥ç‡≤ü‡≥Å ‡≤ó‡≤Æ‡≤®‡≤æ‡≤∞‡≥ç‡≤π ‡≤ï‡≤°‡≤ø‡≤§‡≤µ‡≤®‡≥ç‡≤®‡≥Å ‡≤§‡≥ã‡≤∞‡≤ø‡≤∏‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü. ‡≤Ö‡≤¶‡≤æ‡≤¶ ‡≤Æ‡≥á‡≤≤‡≥Ü, ‡≤ï‡≤°‡≤ø‡≤§‡≤µ‡≥Å ‡≤ï‡≤°‡≤ø‡≤Æ‡≥Ü ‡≤∏‡≥ç‡≤™‡≤∑‡≥ç‡≤ü‡≤µ‡≤æ‡≤ó‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü, ‡≤´‡≤≤‡≤µ‡≤æ‡≤ó‡≤ø ‡≤ö‡≤æ‡≤∞‡≥ç‡≤ü‡≥ç‚Äå‡≤®‡≤≤‡≥ç‡≤≤‡≤ø ‡≤∏‡≥Å‡≤Æ‡≤æ‡≤∞‡≥Å ‡≤Æ‡≥Ç‡≤∞‡≥Å ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç‚Äå‡≤ó‡≤≥ ‡≤¨‡≤≥‡≤ø `‡≤é‡≤≤‡≥ç‡≤¨‡≥ã` üí™ ‡≤ï‡≤æ‡≤£‡≤ø‡≤∏‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü. ‡≤á‡≤¶‡≥Å ‡≤°‡≥á‡≤ü‡≤æ ‡≤™‡≤æ‡≤Ø‡≤ø‡≤Ç‡≤ü‡≥ç‚Äå‡≤ó‡≤≥ ‡≤é‡≤∞‡≤°‡≥Å ‡≤∞‡≤ø‡≤Ç‡≤¶ ‡≤Æ‡≥Ç‡≤∞‡≥Å ‡≤∏‡≤Æ‡≤∞‡≥ç‡≤™‡≤ï‡≤µ‡≤æ‡≤ó‡≤ø ‡≤µ‡≤ø‡≤≠‡≤ú‡≤ø‡≤§ ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç‚Äå‡≤ó‡≤≥‡≤ø‡≤∞‡≥Å‡≤µ ‡≤â‡≤§‡≥ç‡≤§‡≤Æ ‡≤∏‡≥Ç‡≤ö‡≤®‡≥Ü.

‡≤®‡≤æ‡≤µ‡≥Å ‡≤à‡≤ó ‡≤Æ‡≥Å‡≤Ç‡≤¶‡≥Å‡≤µ‡≤∞‡≤ø‡≤¶‡≥Å `k = 3` ‡≤á‡≤∞‡≥Å‡≤µ ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≤ø‡≤Ç‡≤ó‡≥ç ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤π‡≥ä‡≤∞‡≤§‡≥Ü‡≤ó‡≥Ü‡≤Ø‡≤¨‡≤π‡≥Å‡≤¶‡≥Å:

> `pull()`: ‡≤í‡≤Ç‡≤¶‡≥Å ‡≤ï‡≤æ‡≤≤‡≤Æ‡≥ç ‡≤Ö‡≤®‡≥ç‡≤®‡≥Å ‡≤π‡≥ä‡≤∞‡≤§‡≥Ü‡≤ó‡≥Ü‡≤Ø‡≤≤‡≥Å ‡≤¨‡≤≥‡≤∏‡≤≤‡≤æ‡≤ó‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü
>
> `pluck()`: ‡≤™‡≤ü‡≥ç‡≤ü‡≤ø‡≤ó‡≤≥‡≤Ç‡≤§‡≤π ‡≤°‡≥á‡≤ü‡≤æ ‡≤∞‡≤ö‡≤®‡≥Ü‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å ‡≤∏‡≥Ç‡≤ö‡≤ø‡≤∏‡≤≤‡≥Å ‡≤¨‡≤≥‡≤∏‡≤≤‡≤æ‡≤ó‡≥Å‡≤§‡≥ç‡≤§‡≤¶‡≥Ü


```r
# Extract k = 3 clustering
final_kmeans <- kclusts %>% 
  filter(k == 3) %>% 
  pull(model) %>% 
  pluck(1)


final_kmeans
```

‡≤∂‡≥ç‡≤∞‡≥á‡≤∑‡≥ç‡≤†! ‡≤®‡≤æ‡≤µ‡≥Å ‡≤™‡≤°‡≥Ü‡≤¶ ‡≤ï‡≥ç‡≤≤‡≤∏‡≥ç‡≤ü‡≤∞‡≥ç‚Äå‡≤ó‡≤≥‡≤®‡≥ç‡≤®‡≥Å ‡≤¶‡≥É‡≤∂‡≥ç‡≤Ø‡≥Ä‡≤ï‡≤∞‡≤ø‡≤∏‡≥ã‡≤£. `plotly` ‡≤¨‡≤≥‡≤∏‡≤ø ‡≤∏‡≥ç‡≤µ‡≤≤‡≥ç‡≤™ ‡≤á‡≤Ç‡≤ü‡≤∞‡≤æ‡≤ï‡≥ç‡≤ü‡≤ø‡≤µ‡≤ø‡≤ü‡≤ø ‡≤¨‡≥á‡≤ï‡≥Ü?


```r
# Add predicted cluster assignment to data set
results <-  augment(final_kmeans, df_numeric_select) %>% 
  bind_cols(df_numeric %>% select(artist_top_genre)) 

# Plot cluster assignments
clust_plt <- results %>% 
  ggplot(mapping = aes(x = popularity, y = danceability, color = .cluster, shape = artist_top_genre)) +
  geom_point(size = 2, alpha = 0.8) +
  paletteer::scale_color_paletteer_d("ggthemes::Tableau_10")

ggplotly(clust_plt)
```

‡≤®‡≤æ‡≤µ‡≥Å ‡≤Æ‡≤æ‡≤¶‡≤∞‡≤ø‡≤Ø ‡≤®‡≤ø‡≤ñ‡≤∞‡≤§‡≥Ü‡≤Ø‡≤®‡≥ç‡≤®‡≥Å ‡≤®‡≥ã‡≤°‡≥ã‡≤£.


```r
# Assign genres to predefined integers
label_count <- results %>% 
  group_by(artist_top_genre) %>% 
  mutate(id = cur_group_id()) %>% 
  ungroup() %>% 
  summarise(correct_labels = sum(.cluster == id))


# Print results  
cat("Result:", label_count$correct_labels, "out of", nrow(results), "samples were correctly labeled.")

cat("\nAccuracy score:", label_count$correct_labels/nrow(results))
```
