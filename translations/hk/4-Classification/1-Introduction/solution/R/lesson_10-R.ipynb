## **ç·´ç¿’ - ä½¿ç”¨ dplyr æ¢ç´¢ä¸åŒèœç³»çš„ä¸»è¦é£Ÿæ**

ç¾åœ¨ä½ å¯ä»¥æ·±å…¥ç ”ç©¶æ•¸æ“šï¼Œäº†è§£æ¯ç¨®èœç³»çš„å…¸å‹é£Ÿæã€‚ä½ éœ€è¦æ¸…ç†ä¸€äº›é‡è¤‡çš„æ•¸æ“šï¼Œé€™äº›æ•¸æ“šå¯èƒ½æœƒåœ¨èœç³»ä¹‹é–“é€ æˆæ··æ·†ï¼Œå› æ­¤è®“æˆ‘å€‘ä¾†äº†è§£é€™å€‹å•é¡Œã€‚

åœ¨ R ä¸­å‰µå»ºä¸€å€‹åç‚º `create_ingredient()` çš„å‡½æ•¸ï¼Œè©²å‡½æ•¸æœƒè¿”å›ä¸€å€‹é£Ÿæçš„æ•¸æ“šæ¡†ã€‚é€™å€‹å‡½æ•¸å°‡é¦–å…ˆåˆªé™¤ä¸€å€‹ç„¡ç”¨çš„åˆ—ï¼Œç„¶å¾Œæ ¹æ“šé£Ÿæçš„æ•¸é‡é€²è¡Œæ’åºã€‚

R ä¸­å‡½æ•¸çš„åŸºæœ¬çµæ§‹å¦‚ä¸‹ï¼š

å¯ä»¥åœ¨[é€™è£¡](https://skirmer.github.io/presentations/functions_with_r.html#1)æ‰¾åˆ°ä¸€å€‹ç°¡æ½”çš„ R å‡½æ•¸å…¥é–€ä»‹ç´¹ã€‚

è®“æˆ‘å€‘ç›´æ¥é–‹å§‹å§ï¼æˆ‘å€‘å°‡ä½¿ç”¨ [dplyr å‹•è©](https://dplyr.tidyverse.org/)ï¼Œé€™äº›å‹•è©æˆ‘å€‘åœ¨ä¹‹å‰çš„èª²ç¨‹ä¸­å·²ç¶“å­¸éã€‚ä»¥ä¸‹æ˜¯å›é¡§ï¼š

-   `dplyr::select()`: å¹«åŠ©ä½ é¸æ“‡è¦ä¿ç•™æˆ–æ’é™¤çš„**åˆ—**ã€‚

-   `dplyr::pivot_longer()`: å¹«åŠ©ä½ å°‡æ•¸æ“šâ€œæ‹‰é•·â€ï¼Œå¢åŠ è¡Œæ•¸ä¸¦æ¸›å°‘åˆ—æ•¸ã€‚

-   `dplyr::group_by()` å’Œ `dplyr::summarise()`: å¹«åŠ©ä½ æ‰¾åˆ°ä¸åŒçµ„çš„çµ±è¨ˆæ‘˜è¦ï¼Œä¸¦å°‡å®ƒå€‘æ”¾å…¥ä¸€å€‹æ•´é½Šçš„è¡¨æ ¼ä¸­ã€‚

-   `dplyr::filter()`: å‰µå»ºä¸€å€‹åƒ…åŒ…å«æ»¿è¶³æ¢ä»¶çš„è¡Œçš„æ•¸æ“šå­é›†ã€‚

-   `dplyr::mutate()`: å¹«åŠ©ä½ å‰µå»ºæˆ–ä¿®æ”¹åˆ—ã€‚

æŸ¥çœ‹é€™å€‹ç”± Allison Horst è£½ä½œçš„[å……æ»¿è—è¡“æ„Ÿçš„ learnr æ•™ç¨‹](https://allisonhorst.shinyapps.io/dplyr-learnr/#section-welcome)ï¼Œå®ƒä»‹ç´¹äº†ä¸€äº›åœ¨ dplyr *(Tidyverse çš„ä¸€éƒ¨åˆ†)* ä¸­éå¸¸æœ‰ç”¨çš„æ•¸æ“šæ•´ç†å‡½æ•¸ã€‚


```r
# Creates a functions that returns the top ingredients by class

create_ingredient <- function(df){
  
  # Drop the id column which is the first colum
  ingredient_df = df %>% select(-1) %>% 
  # Transpose data to a long format
    pivot_longer(!cuisine, names_to = "ingredients", values_to = "count") %>% 
  # Find the top most ingredients for a particular cuisine
    group_by(ingredients) %>% 
    summarise(n_instances = sum(count)) %>% 
    filter(n_instances != 0) %>% 
  # Arrange by descending order
    arrange(desc(n_instances)) %>% 
    mutate(ingredients = factor(ingredients) %>% fct_inorder())
  
  
  return(ingredient_df)
} # End of function
```

ç¾åœ¨æˆ‘å€‘å¯ä»¥ä½¿ç”¨é€™å€‹å‡½æ•¸ä¾†äº†è§£æŒ‰èœç³»åˆ†é¡çš„åå¤§æœ€å—æ­¡è¿é£Ÿæã€‚è®“æˆ‘å€‘ç”¨ `thai_df` è©¦è©¦çœ‹å§ã€‚


```r
# Call create_ingredient and display popular ingredients
thai_ingredient_df <- create_ingredient(df = thai_df)

thai_ingredient_df %>% 
  slice_head(n = 10)
```

åœ¨ä¸Šä¸€ç¯€ä¸­ï¼Œæˆ‘å€‘ä½¿ç”¨äº† `geom_col()`ï¼Œç¾åœ¨è®“æˆ‘å€‘çœ‹çœ‹å¦‚ä½•ä½¿ç”¨ `geom_bar` ä¾†è£½ä½œæŸ±ç‹€åœ–ã€‚ä½¿ç”¨ `?geom_bar` äº†è§£æ›´å¤šè³‡è¨Šã€‚


```r
# Make a bar chart for popular thai cuisines
thai_ingredient_df %>% 
  slice_head(n = 10) %>% 
  ggplot(aes(x = n_instances, y = ingredients)) +
  geom_bar(stat = "identity", width = 0.5, fill = "steelblue") +
  xlab("") + ylab("")
```

```r
# Get popular ingredients for Japanese cuisines and make bar chart
create_ingredient(df = japanese_df) %>% 
  slice_head(n = 10) %>%
  ggplot(aes(x = n_instances, y = ingredients)) +
  geom_bar(stat = "identity", width = 0.5, fill = "darkorange", alpha = 0.8) +
  xlab("") + ylab("")
```

æ—¢ç„¶é€™ç¯€èª²æ˜¯é—œæ–¼çƒ¹é£ªï¼Œæˆ‘å€‘éœ€è¦å°‡ `recipes` æ”¾åˆ°åˆé©çš„èƒŒæ™¯ä¸­ã€‚

Tidymodels æä¾›äº†å¦ä¸€å€‹éå¸¸æ–¹ä¾¿çš„å¥—ä»¶ï¼š`recipes`â€”â€”ä¸€å€‹ç”¨æ–¼é è™•ç†æ•¸æ“šçš„å¥—ä»¶ã€‚

å¦‚ä½ æ‰€è¦‹ï¼Œèœå¼çš„æ•¸é‡åˆ†ä½ˆéå¸¸ä¸å‡è¡¡ã€‚éŸ“åœ‹èœçš„æ•¸é‡å¹¾ä¹æ˜¯æ³°åœ‹èœçš„ä¸‰å€ã€‚ä¸å‡è¡¡çš„æ•¸æ“šé€šå¸¸æœƒå°æ¨¡å‹çš„è¡¨ç¾ç”¢ç”Ÿè² é¢å½±éŸ¿ã€‚è©¦æƒ³ä¸€å€‹äºŒå…ƒåˆ†é¡å•é¡Œï¼Œå¦‚æœä½ çš„å¤§éƒ¨åˆ†æ•¸æ“šéƒ½å±¬æ–¼æŸä¸€å€‹é¡åˆ¥ï¼Œæ©Ÿå™¨å­¸ç¿’æ¨¡å‹å°±æœƒæ›´é »ç¹åœ°é æ¸¬è©²é¡åˆ¥ï¼Œåƒ…åƒ…å› ç‚ºå®ƒæœ‰æ›´å¤šçš„æ•¸æ“šå¯ä¾›å­¸ç¿’ã€‚å¹³è¡¡æ•¸æ“šå¯ä»¥ä¿®æ­£ä»»ä½•åæ–œçš„æ•¸æ“šï¼Œå¹«åŠ©æ¶ˆé™¤é€™ç¨®ä¸å‡è¡¡ã€‚è¨±å¤šæ¨¡å‹åœ¨è§€æ¸¬æ•¸é‡ç›¸ç­‰æ™‚è¡¨ç¾æœ€ä½³ï¼Œå› æ­¤åœ¨è™•ç†ä¸å‡è¡¡æ•¸æ“šæ™‚å¾€å¾€æœƒé‡åˆ°å›°é›£ã€‚

è™•ç†ä¸å‡è¡¡æ•¸æ“šé›†ä¸»è¦æœ‰å…©ç¨®æ–¹æ³•ï¼š

-   ç‚ºå°‘æ•¸é¡åˆ¥æ·»åŠ è§€æ¸¬å€¼ï¼š`éæ¡æ¨£`ï¼Œä¾‹å¦‚ä½¿ç”¨ SMOTE æ¼”ç®—æ³•

-   å¾å¤šæ•¸é¡åˆ¥ç§»é™¤è§€æ¸¬å€¼ï¼š`æ¬ æ¡æ¨£`

ç¾åœ¨è®“æˆ‘å€‘æ¼”ç¤ºå¦‚ä½•ä½¿ç”¨ `recipe` ä¾†è™•ç†ä¸å‡è¡¡æ•¸æ“šé›†ã€‚`recipe` å¯ä»¥è¢«è¦–ç‚ºä¸€å€‹è—åœ–ï¼Œæè¿°äº†æ‡‰è©²å°æ•¸æ“šé›†æ‡‰ç”¨å“ªäº›æ­¥é©Ÿï¼Œä»¥ä½¿å…¶æº–å‚™å¥½é€²è¡Œæ•¸æ“šåˆ†æã€‚


```r
# Load themis package for dealing with imbalanced data
library(themis)

# Create a recipe for preprocessing data
cuisines_recipe <- recipe(cuisine ~ ., data = df_select) %>% 
  step_smote(cuisine)

cuisines_recipe
```

è®“æˆ‘å€‘é€æ­¥äº†è§£é è™•ç†çš„æ­¥é©Ÿã€‚

-   ä½¿ç”¨å…¬å¼èª¿ç”¨ `recipe()` æ™‚ï¼Œæœƒæ ¹æ“š `df_select` æ•¸æ“šä½œç‚ºåƒè€ƒï¼Œå‘Šè¨´ recipe è®Šæ•¸çš„*è§’è‰²*ã€‚ä¾‹å¦‚ï¼Œ`cuisine` åˆ—è¢«åˆ†é…äº† `outcome` è§’è‰²ï¼Œè€Œå…¶ä»–åˆ—å‰‡è¢«åˆ†é…äº† `predictor` è§’è‰²ã€‚

-   [`step_smote(cuisine)`](https://themis.tidymodels.org/reference/step_smote.html) å‰µå»ºäº†ä¸€å€‹ recipe æ­¥é©Ÿçš„*è¦ç¯„*ï¼Œä½¿ç”¨é€™äº›æ¡ˆä¾‹çš„æœ€è¿‘é„°å±…åˆæˆç”Ÿæˆå°‘æ•¸é¡åˆ¥çš„æ–°æ¨£æœ¬ã€‚

ç¾åœ¨ï¼Œå¦‚æœæˆ‘å€‘æƒ³æŸ¥çœ‹é è™•ç†å¾Œçš„æ•¸æ“šï¼Œå°±éœ€è¦ä½¿ç”¨ [**`prep()`**](https://recipes.tidymodels.org/reference/prep.html) å’Œ [**`bake()`**](https://recipes.tidymodels.org/reference/bake.html) ä¾†è™•ç†æˆ‘å€‘çš„ recipeã€‚

`prep()`ï¼šå¾è¨“ç·´é›†ä¼°ç®—æ‰€éœ€çš„åƒæ•¸ï¼Œé€™äº›åƒæ•¸å¯ä»¥ç¨å¾Œæ‡‰ç”¨åˆ°å…¶ä»–æ•¸æ“šé›†ã€‚

`bake()`ï¼šä½¿ç”¨å·²æº–å‚™å¥½çš„ recipeï¼Œä¸¦å°‡æ“ä½œæ‡‰ç”¨åˆ°ä»»ä½•æ•¸æ“šé›†ã€‚


```r
# Prep and bake the recipe
preprocessed_df <- cuisines_recipe %>% 
  prep() %>% 
  bake(new_data = NULL) %>% 
  relocate(cuisine)

# Display data
preprocessed_df %>% 
  slice_head(n = 5)

# Quick summary stats
preprocessed_df %>% 
  introduce()
```
