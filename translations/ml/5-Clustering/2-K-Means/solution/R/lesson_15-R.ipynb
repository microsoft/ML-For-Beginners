## 1. ‡¥°‡¥æ‡¥±‡µç‡¥±‡¥Ø‡µÅ‡¥Æ‡¥æ‡¥Ø‡¥ø ‡¥í‡¥∞‡µÅ ‡¥®‡µÉ‡¥§‡µç‡¥§‡¥Ç: ‡¥è‡¥±‡µç‡¥±‡¥µ‡µÅ‡¥Ç ‡¥ú‡¥®‡¥™‡µç‡¥∞‡¥ø‡¥Ø‡¥Æ‡¥æ‡¥Ø 3 ‡¥∏‡¥Ç‡¥ó‡µÄ‡¥§ ‡¥∂‡µà‡¥≤‡¥ø‡¥ï‡µæ‡¥ï‡µç‡¥ï‡µç ‡¥ï‡µÅ‡¥±‡¥Ø‡µç‡¥ï‡µç‡¥ï‡µÅ‡¥ï

‡¥á‡¥§‡µç ‡¥Æ‡µÅ‡¥Æ‡µç‡¥™‡¥§‡µç‡¥§‡µÜ ‡¥™‡¥æ‡¥†‡¥§‡µç‡¥§‡¥ø‡µΩ ‡¥û‡¥ô‡µç‡¥ô‡µæ ‡¥ö‡µÜ‡¥Ø‡µç‡¥§‡¥§‡¥ø‡¥®‡µç‡¥±‡µÜ ‡¥í‡¥∞‡µÅ ‡¥∏‡¥Ç‡¥ó‡µç‡¥∞‡¥π‡¥Æ‡¥æ‡¥£‡µç. ‡¥®‡¥Æ‡µÅ‡¥ï‡µç‡¥ï‡µç ‡¥ö‡¥ø‡¥≤ ‡¥°‡¥æ‡¥±‡µç‡¥± ‡¥ï‡¥∑‡¥£‡¥ô‡µç‡¥ô‡¥≥‡¥æ‡¥ï‡µç‡¥ï‡¥ø ‡¥µ‡¥ø‡¥∂‡¥ï‡¥≤‡¥®‡¥Ç ‡¥ö‡µÜ‡¥Ø‡µç‡¥Ø‡¥æ‡¥Ç!


```r
# Load the core tidyverse and make it available in your current R session
library(tidyverse)

# Import the data into a tibble
df <- read_csv(file = "https://raw.githubusercontent.com/microsoft/ML-For-Beginners/main/5-Clustering/data/nigerian-songs.csv", show_col_types = FALSE)

# Narrow down to top 3 popular genres
nigerian_songs <- df %>% 
  # Concentrate on top 3 genres
  filter(artist_top_genre %in% c("afro dancehall", "afropop","nigerian pop")) %>% 
  # Remove unclassified observations
  filter(popularity != 0)



# Visualize popular genres using bar plots
theme_set(theme_light())
nigerian_songs %>%
  count(artist_top_genre) %>%
  ggplot(mapping = aes(x = artist_top_genre, y = n,
                       fill = artist_top_genre)) +
  geom_col(alpha = 0.8) +
  paletteer::scale_fill_paletteer_d("ggsci::category10_d3") +
  ggtitle("Top genres") +
  theme(plot.title = element_text(hjust = 0.5))
```

‡¥§‡¥ø‡¥∞‡¥û‡µç‡¥û‡µÜ‡¥ü‡µÅ‡¥™‡µç‡¥™‡µç ‡¥∏‡¥π‡¥æ‡¥Ø‡¥ø‡¥Ø‡¥æ‡¥Ø `where` ‡¥á‡¥§‡¥ø‡¥®‡µÜ ‡¥é‡¥≥‡µÅ‡¥™‡µç‡¥™‡¥Æ‡¥æ‡¥ï‡µç‡¥ï‡µÅ‡¥®‡µç‡¥®‡¥§‡µç ‡¥é‡¥ô‡µç‡¥ô‡¥®‡µÜ ‡¥ï‡¥æ‡¥£‡µÇ üíÅ? ‡¥á‡¥§‡µç‡¥§‡¥∞‡¥§‡µç‡¥§‡¥ø‡¥≤‡µÅ‡¥≥‡µç‡¥≥ ‡¥Æ‡¥±‡µç‡¥±‡µç ‡¥´‡¥Ç‡¥ó‡µç‡¥∑‡¥®‡µÅ‡¥ï‡µæ [‡¥á‡¥µ‡¥ø‡¥ü‡µÜ](https://tidyselect.r-lib.org/) ‡¥™‡¥∞‡¥ø‡¥∂‡µã‡¥ß‡¥ø‡¥ï‡µç‡¥ï‡µÅ‡¥ï.

## 3. R-‡µΩ k-means ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡¥ø‡¥Ç‡¥ó‡µç ‡¥ï‡¥£‡¥ï‡µç‡¥ï‡¥æ‡¥ï‡µç‡¥ï‡µΩ

‡¥®‡¥æ‡¥Ç R-‡µΩ ‡¥â‡µæ‡¥™‡µç‡¥™‡µÜ‡¥ü‡µÅ‡¥§‡µç‡¥§‡¥ø‡¥Ø‡¥ø‡¥∞‡¥ø‡¥ï‡µç‡¥ï‡µÅ‡¥®‡µç‡¥® `kmeans` ‡¥´‡¥Ç‡¥ó‡µç‡¥∑‡µª ‡¥â‡¥™‡¥Ø‡µã‡¥ó‡¥ø‡¥ö‡µç‡¥ö‡µç k-means ‡¥ï‡¥£‡¥ï‡µç‡¥ï‡¥æ‡¥ï‡µç‡¥ï‡¥æ‡¥Ç, `help("kmeans()")` ‡¥ï‡¥æ‡¥£‡µÅ‡¥ï. `kmeans()` ‡¥´‡¥Ç‡¥ó‡µç‡¥∑‡µª ‡¥™‡µç‡¥∞‡¥ß‡¥æ‡¥®‡¥Æ‡¥æ‡¥Ø‡µÅ‡¥Ç ‡¥é‡¥≤‡µç‡¥≤‡¥æ ‡¥∏‡¥Ç‡¥ñ‡µç‡¥Ø‡¥æ‡¥§‡µç‡¥Æ‡¥ï ‡¥ï‡µã‡¥≥‡¥ô‡µç‡¥ô‡¥≥‡µÅ‡¥≥‡µç‡¥≥ ‡¥í‡¥∞‡µÅ ‡¥°‡¥æ‡¥±‡µç‡¥±‡¥æ ‡¥´‡µç‡¥∞‡µÜ‡¥Ø‡¥ø‡¥Ç ‡¥∏‡µç‡¥µ‡µÄ‡¥ï‡¥∞‡¥ø‡¥ï‡µç‡¥ï‡µÅ‡¥®‡µç‡¥®‡µÅ.

k-means ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡¥ø‡¥Ç‡¥ó‡µç ‡¥â‡¥™‡¥Ø‡µã‡¥ó‡¥ø‡¥ï‡µç‡¥ï‡µÅ‡¥Æ‡µç‡¥™‡µã‡µæ ‡¥Ü‡¥¶‡µç‡¥Ø‡¥§‡µç‡¥§‡µÜ ‡¥ò‡¥ü‡µç‡¥ü‡¥Ç ‡¥Ö‡¥µ‡¥∏‡¥æ‡¥® ‡¥™‡¥∞‡¥ø‡¥π‡¥æ‡¥∞‡¥§‡µç‡¥§‡¥ø‡µΩ ‡¥∏‡µÉ‡¥∑‡µç‡¥ü‡¥ø‡¥ï‡µç‡¥ï‡¥™‡µç‡¥™‡µÜ‡¥ü‡µÅ‡¥®‡µç‡¥® ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡µÅ‡¥ï‡¥≥‡µÅ‡¥ü‡µÜ ‡¥é‡¥£‡µç‡¥£‡¥Ç (k) ‡¥®‡¥ø‡µº‡¥¶‡µç‡¥¶‡µá‡¥∂‡¥ø‡¥ï‡µç‡¥ï‡µÅ‡¥ï‡¥Ø‡¥æ‡¥£‡µç. ‡¥°‡¥æ‡¥±‡µç‡¥±‡¥æ‡¥∏‡µÜ‡¥±‡µç‡¥±‡¥ø‡µΩ ‡¥®‡¥ø‡¥®‡µç‡¥®‡µç ‡¥®‡¥æ‡¥Ç 3 ‡¥ó‡¥æ‡¥® ‡¥∂‡µà‡¥≤‡¥ø‡¥ï‡µæ ‡¥µ‡µá‡µº‡¥§‡¥ø‡¥∞‡¥ø‡¥ö‡µç‡¥ö‡µÜ‡¥ü‡µÅ‡¥§‡µç‡¥§‡¥§‡¥æ‡¥Ø‡¥ø ‡¥Ö‡¥±‡¥ø‡¥Ø‡¥æ‡¥Ç, ‡¥Ö‡¥§‡¥ø‡¥®‡¥æ‡µΩ 3 ‡¥™‡¥∞‡µÄ‡¥ï‡µç‡¥∑‡¥ø‡¥ï‡µç‡¥ï‡¥æ‡¥Ç:


```r
set.seed(2056)
# Kmeans clustering for 3 clusters
kclust <- kmeans(
  df_numeric_select,
  # Specify the number of clusters
  centers = 3,
  # How many random initial configurations
  nstart = 25
)

# Display clustering object
kclust
```

kmeans ‡¥í‡¥¨‡µç‡¥ú‡¥ï‡µç‡¥±‡µç‡¥±‡¥ø‡µΩ `help("kmeans()")` ‡µΩ ‡¥®‡¥®‡µç‡¥®‡¥æ‡¥Ø‡¥ø ‡¥µ‡¥ø‡¥∂‡¥¶‡µÄ‡¥ï‡¥∞‡¥ø‡¥ö‡µç‡¥ö‡¥ø‡¥∞‡¥ø‡¥ï‡µç‡¥ï‡µÅ‡¥®‡µç‡¥® ‡¥®‡¥ø‡¥∞‡¥µ‡¥ß‡¥ø ‡¥µ‡¥ø‡¥µ‡¥∞‡¥ô‡µç‡¥ô‡µæ ‡¥Ö‡¥ü‡¥ô‡µç‡¥ô‡¥ø‡¥Ø‡¥ø‡¥∞‡¥ø‡¥ï‡µç‡¥ï‡µÅ‡¥®‡µç‡¥®‡µÅ. ‡¥á‡¥™‡µç‡¥™‡µã‡µæ, ‡¥ö‡¥ø‡¥≤‡¥§‡¥ø‡µΩ ‡¥∂‡µç‡¥∞‡¥¶‡µç‡¥ß ‡¥ï‡µá‡¥®‡µç‡¥¶‡µç‡¥∞‡µÄ‡¥ï‡¥∞‡¥ø‡¥ï‡µç‡¥ï‡¥æ‡¥Ç. ‡¥°‡¥æ‡¥±‡µç‡¥± 65, 110, 111 ‡¥é‡¥®‡µç‡¥® ‡¥µ‡¥≤‡µÅ‡¥™‡µç‡¥™‡¥Æ‡µÅ‡¥≥‡µç‡¥≥ 3 ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡µÅ‡¥ï‡¥≥‡¥æ‡¥Ø‡¥ø ‡¥ó‡µç‡¥∞‡µÇ‡¥™‡µç‡¥™‡µÅ‡¥ö‡µÜ‡¥Ø‡µç‡¥§‡¥ø‡¥ü‡µç‡¥ü‡µÅ‡¥£‡µç‡¥ü‡µÜ‡¥®‡µç‡¥®‡µç ‡¥ï‡¥æ‡¥£‡¥æ‡¥Ç. ‡¥î‡¥ü‡µç‡¥ü‡µç‡¥™‡µÅ‡¥ü‡µç‡¥ü‡¥ø‡µΩ 5 ‡¥µ‡µç‡¥Ø‡¥§‡µç‡¥Ø‡¥∏‡µç‡¥§ ‡¥µ‡µá‡¥∞‡¥ø‡¥Ø‡¥¨‡¥ø‡¥≥‡µÅ‡¥ï‡¥≥‡¥ø‡¥≤‡¥æ‡¥Ø‡¥ø 3 ‡¥ó‡µç‡¥∞‡µÇ‡¥™‡µç‡¥™‡µÅ‡¥ï‡¥≥‡µÅ‡¥ü‡µÜ ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡µº ‡¥∏‡µÜ‡¥®‡µç‡¥±‡¥±‡µÅ‡¥ï‡¥≥‡µÅ‡¥Ç (‡¥Æ‡µÄ‡¥®‡µÅ‡¥ï‡¥≥‡µÅ‡¥Ç) ‡¥â‡µæ‡¥™‡µç‡¥™‡µÜ‡¥ü‡µÅ‡¥®‡µç‡¥®‡µÅ.

‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡¥ø‡¥Ç‡¥ó‡µç ‡¥µ‡µÜ‡¥ï‡µç‡¥ü‡µº ‡¥ì‡¥∞‡µã ‡¥®‡¥ø‡¥∞‡µÄ‡¥ï‡µç‡¥∑‡¥£‡¥§‡µç‡¥§‡¥ø‡¥®‡µç‡¥±‡µÜ‡¥Ø‡µÅ‡¥Ç ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡µº ‡¥®‡¥ø‡¥Ø‡µã‡¥ó‡¥Æ‡¥æ‡¥£‡µç. ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡µº ‡¥®‡¥ø‡¥Ø‡µã‡¥ó‡¥Ç ‡¥Æ‡µó‡¥≤‡¥ø‡¥ï ‡¥°‡¥æ‡¥±‡µç‡¥±‡¥æ ‡¥∏‡µÜ‡¥±‡µç‡¥±‡¥ø‡µΩ ‡¥ö‡µá‡µº‡¥ï‡µç‡¥ï‡¥æ‡µª `augment` ‡¥´‡¥Ç‡¥ó‡µç‡¥∑‡µª ‡¥â‡¥™‡¥Ø‡µã‡¥ó‡¥ø‡¥ï‡µç‡¥ï‡¥æ‡¥Ç.


```r
# Add predicted cluster assignment to data set
augment(kclust, df_numeric_select) %>% 
  relocate(.cluster) %>% 
  slice_head(n = 10)
```

Perfect, ‡¥®‡¥æ‡¥Ç ‡¥á‡¥™‡µç‡¥™‡µã‡µæ ‡¥®‡¥Æ‡µç‡¥Æ‡µÅ‡¥ü‡µÜ ‡¥°‡¥æ‡¥±‡µç‡¥±‡¥æ ‡¥∏‡µÜ‡¥±‡µç‡¥±‡µç 3 ‡¥ó‡µç‡¥∞‡µÇ‡¥™‡µç‡¥™‡µÅ‡¥ï‡¥≥‡¥æ‡¥Ø‡¥ø ‡¥µ‡¥ø‡¥≠‡¥ú‡¥ø‡¥ö‡µç‡¥ö‡µÅ. ‡¥Ö‡¥§‡¥ø‡¥®‡¥æ‡µΩ, ‡¥®‡¥Æ‡µç‡¥Æ‡µÅ‡¥ü‡µÜ ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡¥ø‡¥Ç‡¥ó‡µç ‡¥é‡¥§‡µç‡¥∞‡¥§‡µç‡¥§‡µã‡¥≥‡¥Ç ‡¥®‡¥≤‡µç‡¥≤‡¥§‡¥æ‡¥£‡µç ü§∑? ‡¥®‡¥Æ‡µÅ‡¥ï‡µç‡¥ï‡µç `Silhouette score` ‡¥®‡µã‡¥ï‡µç‡¥ï‡¥æ‡¥Ç

### **Silhouette score**

[Silhouette analysis](https://en.wikipedia.org/wiki/Silhouette_(clustering)) ‡¥â‡¥™‡¥Ø‡µã‡¥ó‡¥ø‡¥ö‡µç‡¥ö‡µç ‡¥´‡¥≤‡¥Æ‡¥æ‡¥Ø ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡µÅ‡¥ï‡µæ ‡¥§‡¥Æ‡µç‡¥Æ‡¥ø‡¥≤‡µÅ‡¥≥‡µç‡¥≥ ‡¥µ‡µá‡µº‡¥§‡¥ø‡¥∞‡¥ø‡¥µ‡µç ‡¥Ö‡¥≥‡¥ï‡µç‡¥ï‡¥æ‡µª ‡¥ï‡¥¥‡¥ø‡¥Ø‡µÅ‡¥Ç. ‡¥à ‡¥∏‡µç‡¥ï‡µã‡µº -1 ‡¥Æ‡µÅ‡¥§‡µΩ 1 ‡¥µ‡¥∞‡µÜ ‡¥Æ‡¥æ‡¥±‡µÅ‡¥®‡µç‡¥®‡µÅ, ‡¥∏‡µç‡¥ï‡µã‡µº 1-‡¥®‡µç ‡¥Ö‡¥ü‡µÅ‡¥§‡µç‡¥§‡µç ‡¥Ü‡¥£‡µÜ‡¥ô‡µç‡¥ï‡¥ø‡µΩ, ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡µº ‡¥∏‡¥æ‡¥®‡µç‡¥¶‡µç‡¥∞‡¥µ‡µÅ‡¥Ç ‡¥Æ‡¥±‡µç‡¥±‡µç ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡µÅ‡¥ï‡¥≥‡¥ø‡µΩ ‡¥®‡¥ø‡¥®‡µç‡¥®‡µç ‡¥®‡¥®‡µç‡¥®‡¥æ‡¥Ø‡¥ø ‡¥µ‡µá‡µº‡¥§‡¥ø‡¥∞‡¥ø‡¥ö‡µç‡¥ö‡µÅ‡¥Æ‡¥æ‡¥£‡µç. 0-‡¥®‡µç ‡¥Ö‡¥ü‡µÅ‡¥§‡µç‡¥§‡µÅ‡¥≥‡µç‡¥≥ ‡¥Æ‡µÇ‡¥≤‡µç‡¥Ø‡¥Ç ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡µÅ‡¥ï‡µæ ‡¥í‡¥§‡µÅ‡¥ï‡µç‡¥ï‡¥ø‡¥Ø‡¥ø‡¥∞‡¥ø‡¥ï‡µç‡¥ï‡µÅ‡¥®‡µç‡¥®‡¥§‡µÅ‡¥Ç, ‡¥∏‡¥æ‡¥Æ‡µç‡¥™‡¥ø‡¥≥‡µÅ‡¥ï‡µæ ‡¥Ö‡¥ü‡µÅ‡¥§‡µç‡¥§‡µÅ‡¥≥‡µç‡¥≥ ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡µÅ‡¥ï‡¥≥‡µÅ‡¥ü‡µÜ ‡¥§‡µÄ‡¥∞‡µÅ‡¥Æ‡¥æ‡¥® ‡¥Ö‡¥§‡¥ø‡¥∞‡¥ø‡¥®‡µã‡¥ü‡µç ‡¥µ‡¥≥‡¥∞‡µÜ ‡¥Ö‡¥ü‡µÅ‡¥§‡µç‡¥§‡µÅ‡¥≥‡µç‡¥≥‡¥§‡µÅ‡¥Æ‡¥æ‡¥£‡µç ‡¥∏‡µÇ‡¥ö‡¥ø‡¥™‡µç‡¥™‡¥ø‡¥ï‡µç‡¥ï‡µÅ‡¥®‡µç‡¥®‡¥§‡µç.[source](https://dzone.com/articles/kmeans-silhouette-score-explained-with-python-exam).

‡¥∏‡¥æ‡¥ß‡¥æ‡¥∞‡¥£ silhouette ‡¥∞‡µÄ‡¥§‡¥ø ‡¥µ‡µç‡¥Ø‡¥§‡µç‡¥Ø‡¥∏‡µç‡¥§ *k* ‡¥Æ‡µÇ‡¥≤‡µç‡¥Ø‡¥ô‡µç‡¥ô‡µæ‡¥ï‡µç‡¥ï‡µç ‡¥®‡¥ø‡¥∞‡µÄ‡¥ï‡µç‡¥∑‡¥£‡¥ô‡µç‡¥ô‡¥≥‡µÅ‡¥ü‡µÜ ‡¥∂‡¥∞‡¥æ‡¥∂‡¥∞‡¥ø silhouette ‡¥ï‡¥£‡¥ï‡µç‡¥ï‡¥æ‡¥ï‡µç‡¥ï‡µÅ‡¥®‡µç‡¥®‡µÅ. ‡¥â‡¥Ø‡µº‡¥®‡µç‡¥® ‡¥∂‡¥∞‡¥æ‡¥∂‡¥∞‡¥ø silhouette ‡¥∏‡µç‡¥ï‡µã‡µº ‡¥®‡¥≤‡µç‡¥≤ ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡¥ø‡¥Ç‡¥ó‡µç ‡¥∏‡µÇ‡¥ö‡¥ø‡¥™‡µç‡¥™‡¥ø‡¥ï‡µç‡¥ï‡µÅ‡¥®‡µç‡¥®‡µÅ.

`silhouette` ‡¥´‡¥Ç‡¥ó‡µç‡¥∑‡µª ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡µº ‡¥™‡¥æ‡¥ï‡µç‡¥ï‡µá‡¥ú‡¥ø‡µΩ ‡¥∂‡¥∞‡¥æ‡¥∂‡¥∞‡¥ø silhouette ‡¥µ‡µÄ‡¥§‡¥ø ‡¥ï‡¥£‡¥ï‡µç‡¥ï‡¥æ‡¥ï‡µç‡¥ï‡¥æ‡µª ‡¥â‡¥™‡¥Ø‡µã‡¥ó‡¥ø‡¥ï‡µç‡¥ï‡µÅ‡¥®‡µç‡¥®‡µÅ.

> silhouette ‡¥è‡¥§‡µÜ‡¥ô‡µç‡¥ï‡¥ø‡¥≤‡µÅ‡¥Ç [distance](https://en.wikipedia.org/wiki/Distance "Distance") ‡¥Æ‡µÜ‡¥ü‡µç‡¥∞‡¥ø‡¥ï‡µç ‡¥â‡¥™‡¥Ø‡µã‡¥ó‡¥ø‡¥ö‡µç‡¥ö‡µç ‡¥ï‡¥£‡¥ï‡µç‡¥ï‡¥æ‡¥ï‡µç‡¥ï‡¥æ‡¥Ç, ‡¥â‡¥¶‡¥æ‡¥π‡¥∞‡¥£‡¥§‡µç‡¥§‡¥ø‡¥®‡µç [Euclidean distance](https://en.wikipedia.org/wiki/Euclidean_distance "Euclidean distance") ‡¥Ö‡¥≤‡µç‡¥≤‡µÜ‡¥ô‡µç‡¥ï‡¥ø‡µΩ [Manhattan distance](https://en.wikipedia.org/wiki/Manhattan_distance "Manhattan distance") ‡¥é‡¥®‡µç‡¥®‡¥µ, ‡¥û‡¥ô‡µç‡¥ô‡µæ [‡¥Æ‡µÅ‡µª‡¥™‡¥§‡µç‡¥§‡µÜ ‡¥™‡¥æ‡¥†‡¥§‡µç‡¥§‡¥ø‡µΩ](https://github.com/microsoft/ML-For-Beginners/blob/main/5-Clustering/1-Visualize/solution/R/lesson_14-R.ipynb) ‡¥ö‡µº‡¥ö‡µç‡¥ö ‡¥ö‡µÜ‡¥Ø‡µç‡¥§‡¥§‡µÅ‡¥™‡µã‡¥≤‡µÜ.


```r
# Load cluster package
library(cluster)

# Compute average silhouette score
ss <- silhouette(kclust$cluster,
                 # Compute euclidean distance
                 dist = dist(df_numeric_select))
mean(ss[, 3])
```

‡¥®‡¥Æ‡µç‡¥Æ‡µÅ‡¥ü‡µÜ ‡¥∏‡µç‡¥ï‡µã‡µº **.549** ‡¥Ü‡¥£‡µç, ‡¥Ö‡¥§‡¥æ‡¥Ø‡¥§‡µç ‡¥Æ‡¥ß‡µç‡¥Ø‡¥§‡µç‡¥§‡¥ø‡µΩ ‡¥§‡¥®‡µç‡¥®‡µÜ. ‡¥á‡¥§‡µç ‡¥∏‡µÇ‡¥ö‡¥ø‡¥™‡µç‡¥™‡¥ø‡¥ï‡µç‡¥ï‡µÅ‡¥®‡µç‡¥®‡¥§‡µç ‡¥®‡¥Æ‡µç‡¥Æ‡µÅ‡¥ü‡µÜ ‡¥°‡¥æ‡¥±‡µç‡¥± ‡¥à ‡¥§‡¥∞‡¥§‡µç‡¥§‡¥ø‡¥≤‡µÅ‡¥≥‡µç‡¥≥ ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡¥ø‡¥Ç‡¥ó‡¥ø‡¥®‡µç ‡¥™‡µç‡¥∞‡¥§‡µç‡¥Ø‡µá‡¥ï‡¥ø‡¥ö‡µç‡¥ö‡µç ‡¥Ö‡¥®‡µÅ‡¥Ø‡µã‡¥ú‡µç‡¥Ø‡¥Æ‡¥æ‡¥Ø‡¥§‡¥≤‡µç‡¥≤ ‡¥é‡¥®‡µç‡¥®‡¥§‡¥æ‡¥£‡µç. ‡¥®‡¥Æ‡µÅ‡¥ï‡µç‡¥ï‡µç ‡¥à ‡¥∏‡¥Ç‡¥∂‡¥Ø‡¥Ç ‡¥¶‡µÉ‡¥∂‡µç‡¥Ø‡¥Æ‡¥æ‡¥Ø‡¥ø ‡¥∏‡µç‡¥•‡¥ø‡¥∞‡µÄ‡¥ï‡¥∞‡¥ø‡¥ï‡µç‡¥ï‡¥æ‡¥Æ‡µã ‡¥é‡¥®‡µç‡¥®‡µç ‡¥®‡µã‡¥ï‡µç‡¥ï‡¥æ‡¥Ç. [factoextra ‡¥™‡¥æ‡¥ï‡µç‡¥ï‡µá‡¥ú‡µç](https://rpkgs.datanovia.com/factoextra/index.html) ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡¥ø‡¥Ç‡¥ó‡µç ‡¥¶‡µÉ‡¥∂‡µç‡¥Ø‡¥µ‡¥§‡µç‡¥ï‡¥∞‡¥ø‡¥ï‡µç‡¥ï‡¥æ‡µª (`fviz_cluster()`) ‡¥´‡¥Ç‡¥ó‡µç‡¥∑‡¥®‡µÅ‡¥ï‡µæ ‡¥®‡µΩ‡¥ï‡µÅ‡¥®‡µç‡¥®‡µÅ.

## 4. ‡¥è‡¥±‡µç‡¥±‡¥µ‡µÅ‡¥Ç ‡¥Ö‡¥®‡µÅ‡¥Ø‡µã‡¥ú‡µç‡¥Ø‡¥Æ‡¥æ‡¥Ø ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡µÅ‡¥ï‡µæ ‡¥®‡¥ø‡µº‡¥£‡µç‡¥£‡¥Ø‡¥ø‡¥ï‡µç‡¥ï‡µΩ

K-‡¥Æ‡µÄ‡µª‡¥∏‡µç ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡¥ø‡¥Ç‡¥ó‡¥ø‡µΩ ‡¥∏‡¥æ‡¥ß‡¥æ‡¥∞‡¥£‡¥Ø‡¥æ‡¥Ø‡¥ø ‡¥â‡¥Ø‡¥∞‡µÅ‡¥®‡µç‡¥® ‡¥í‡¥∞‡µÅ ‡¥Ö‡¥ü‡¥ø‡¥∏‡µç‡¥•‡¥æ‡¥® ‡¥ö‡µã‡¥¶‡µç‡¥Ø‡¥Æ‡¥æ‡¥£‡µç - ‡¥Ö‡¥±‡¥ø‡¥Ø‡¥™‡µç‡¥™‡µÜ‡¥ü‡¥æ‡¥§‡µç‡¥§ ‡¥ï‡µç‡¥≤‡¥æ‡¥∏‡µç ‡¥≤‡µá‡¥¨‡¥≤‡µÅ‡¥ï‡µæ ‡¥á‡¥≤‡µç‡¥≤‡¥æ‡¥§‡µÜ, ‡¥®‡¥ø‡¥ô‡µç‡¥ô‡¥≥‡µÅ‡¥ü‡µÜ ‡¥°‡¥æ‡¥±‡µç‡¥± ‡¥é‡¥§‡µç‡¥∞ ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡µÅ‡¥ï‡¥≥‡¥æ‡¥Ø‡¥ø ‡¥µ‡µá‡µº‡¥§‡¥ø‡¥∞‡¥ø‡¥ï‡µç‡¥ï‡¥£‡¥Æ‡µÜ‡¥®‡µç‡¥®‡µç ‡¥®‡¥ø‡¥ô‡µç‡¥ô‡µæ ‡¥é‡¥ô‡µç‡¥ô‡¥®‡µÜ ‡¥Ö‡¥±‡¥ø‡¥Ø‡µÅ‡¥Ç?

‡¥®‡¥æ‡¥Ç ‡¥ï‡¥£‡µç‡¥ü‡µÜ‡¥§‡µç‡¥§‡¥æ‡µª ‡¥∂‡µç‡¥∞‡¥Æ‡¥ø‡¥ï‡µç‡¥ï‡¥æ‡¥µ‡µÅ‡¥®‡µç‡¥® ‡¥í‡¥∞‡µÅ ‡¥Æ‡¥æ‡µº‡¥ó‡¥Ç ‡¥°‡¥æ‡¥±‡µç‡¥± ‡¥∏‡¥æ‡¥Æ‡µç‡¥™‡¥ø‡µæ ‡¥â‡¥™‡¥Ø‡µã‡¥ó‡¥ø‡¥ö‡µç‡¥ö‡µç ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡µÅ‡¥ï‡¥≥‡µÅ‡¥ü‡µÜ ‡¥é‡¥£‡µç‡¥£‡¥Ç ‡¥ï‡µç‡¥∞‡¥Æ‡¥Æ‡¥æ‡¥Ø‡¥ø ‡¥µ‡µº‡¥¶‡µç‡¥ß‡¥ø‡¥™‡µç‡¥™‡¥ø‡¥ö‡µç‡¥ö‡µç (‡¥â‡¥¶‡¥æ: 1-10 ‡¥µ‡¥∞‡µÜ) `‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡¥ø‡¥Ç‡¥ó‡µç ‡¥Æ‡µã‡¥°‡¥≤‡µÅ‡¥ï‡¥≥‡µÅ‡¥ü‡µÜ ‡¥í‡¥∞‡µÅ ‡¥™‡¥∞‡¥Æ‡µç‡¥™‡¥∞ ‡¥∏‡µÉ‡¥∑‡µç‡¥ü‡¥ø‡¥ï‡µç‡¥ï‡µÅ‡¥ï` ‡¥é‡¥®‡µç‡¥®‡¥§‡¥æ‡¥£‡µç, ‡¥ï‡µÇ‡¥ü‡¥æ‡¥§‡µÜ **‡¥∏‡¥ø‡¥≤‡¥π‡µç‡¥µ‡µÜ‡¥±‡µç‡¥±‡µç ‡¥∏‡µç‡¥ï‡µã‡µº** ‡¥™‡µã‡¥≤‡µÅ‡¥≥‡µç‡¥≥ ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡¥ø‡¥Ç‡¥ó‡µç ‡¥Æ‡µÜ‡¥ü‡µç‡¥∞‡¥ø‡¥ï‡µç‡¥ï‡µÅ‡¥ï‡µæ ‡¥µ‡¥ø‡¥≤‡¥Ø‡¥ø‡¥∞‡µÅ‡¥§‡µç‡¥§‡µÅ‡¥ï.

‡¥µ‡¥ø‡¥µ‡¥ø‡¥ß *k* ‡¥Æ‡µÇ‡¥≤‡µç‡¥Ø‡¥ô‡µç‡¥ô‡µæ‡¥ï‡µç‡¥ï‡µç ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡¥ø‡¥Ç‡¥ó‡µç ‡¥Ü‡µΩ‡¥ó‡µã‡¥∞‡¥ø‡¥§‡¥Ç ‡¥ï‡¥£‡¥ï‡µç‡¥ï‡¥æ‡¥ï‡µç‡¥ï‡¥ø **Within Cluster Sum of Squares** (WCSS) ‡¥µ‡¥ø‡¥≤‡¥Ø‡¥ø‡¥∞‡µÅ‡¥§‡µç‡¥§‡¥ø ‡¥è‡¥±‡µç‡¥±‡¥µ‡µÅ‡¥Ç ‡¥Ö‡¥®‡µÅ‡¥Ø‡µã‡¥ú‡µç‡¥Ø‡¥Æ‡¥æ‡¥Ø ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡µÅ‡¥ï‡¥≥‡µÅ‡¥ü‡µÜ ‡¥é‡¥£‡µç‡¥£‡¥Ç ‡¥®‡¥Æ‡µÅ‡¥ï‡µç‡¥ï‡µç ‡¥®‡¥ø‡µº‡¥£‡µç‡¥£‡¥Ø‡¥ø‡¥ï‡µç‡¥ï‡¥æ‡¥Ç. ‡¥Æ‡µä‡¥§‡µç‡¥§‡¥Ç ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡µº ‡¥â‡¥≥‡µç‡¥≥‡¥ø‡¥≤‡µÜ ‡¥∏‡µç‡¥ï‡µç‡¥µ‡¥Ø‡µº ‡¥∏‡¥Æ‡¥æ‡¥π‡¥æ‡¥∞‡¥Ç (WCSS) ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡¥ø‡¥Ç‡¥ó‡¥ø‡¥®‡µç‡¥±‡µÜ ‡¥∏‡¥æ‡¥®‡µç‡¥¶‡µç‡¥∞‡¥§ ‡¥Ö‡¥≥‡¥ï‡µç‡¥ï‡µÅ‡¥®‡µç‡¥®‡µÅ, ‡¥á‡¥§‡µç œåœÉŒø ‡¥ö‡µÜ‡¥±‡µÅ‡¥§‡¥æ‡¥Ø‡¥ø‡¥∞‡¥ø‡¥ï‡µç‡¥ï‡µÅ‡¥Ç, ‡¥°‡¥æ‡¥±‡µç‡¥± ‡¥™‡µã‡¥Ø‡¥ø‡¥®‡µç‡¥±‡µÅ‡¥ï‡µæ ‡¥§‡¥Æ‡µç‡¥Æ‡¥ø‡¥≤‡µÅ‡¥≥‡µç‡¥≥ ‡¥¶‡µÇ‡¥∞‡¥Ç ‡¥ï‡µÅ‡¥±‡¥µ‡¥æ‡¥Ø‡¥ø‡¥∞‡¥ø‡¥ï‡µç‡¥ï‡µÅ‡¥Ç ‡¥é‡¥®‡µç‡¥®‡µç ‡¥Ö‡µº‡¥§‡µç‡¥•‡¥Ç.

1 ‡¥Æ‡µÅ‡¥§‡µΩ 10 ‡¥µ‡¥∞‡µÜ ‡¥µ‡µç‡¥Ø‡¥§‡µç‡¥Ø‡¥∏‡µç‡¥§ `k` ‡¥§‡¥ø‡¥∞‡¥û‡µç‡¥û‡µÜ‡¥ü‡µÅ‡¥™‡µç‡¥™‡µÅ‡¥ï‡¥≥‡µÅ‡¥ü‡µÜ ‡¥à ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡¥ø‡¥Ç‡¥ó‡¥ø‡µΩ ‡¥â‡¥≥‡µç‡¥≥ ‡¥´‡¥≤‡¥ô‡µç‡¥ô‡µæ ‡¥®‡¥Æ‡µÅ‡¥ï‡µç‡¥ï‡µç ‡¥™‡¥∞‡¥ø‡¥∂‡µã‡¥ß‡¥ø‡¥ï‡µç‡¥ï‡¥æ‡¥Ç.


```r
# Create a series of clustering models
kclusts <- tibble(k = 1:10) %>% 
  # Perform kmeans clustering for 1,2,3 ... ,10 clusters
  mutate(model = map(k, ~ kmeans(df_numeric_select, centers = .x, nstart = 25)),
  # Farm out clustering metrics eg WCSS
         glanced = map(model, ~ glance(.x))) %>% 
  unnest(cols = glanced)
  

# View clustering rsulsts
kclusts
```

‡¥á‡¥™‡µç‡¥™‡µã‡µæ ‡¥ì‡¥∞‡µã ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡¥ø‡¥Ç‡¥ó‡µç ‡¥Ü‡µΩ‡¥ó‡µã‡¥±‡¥ø‡¥§‡¥§‡µç‡¥§‡¥ø‡¥®‡µÅ‡¥Ç ‡¥∏‡µÜ‡¥®‡µç‡¥±‡µº *k* ‡¥â‡¥≥‡µç‡¥≥ ‡¥Æ‡µä‡¥§‡µç‡¥§‡¥Ç ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡µº-‡¥Ö‡¥ï‡¥§‡µç‡¥§‡µç ‡¥∏‡µç‡¥ï‡µç‡¥µ‡¥Ø‡µº‡¥∏‡¥ø‡¥®‡µç‡¥±‡µÜ ‡¥∏‡¥Ç‡¥ñ‡µç‡¥Ø (tot.withinss) ‡¥≤‡¥≠‡¥ø‡¥ö‡µç‡¥ö‡¥§‡¥ø‡¥®‡µÅ‡¥∂‡µá‡¥∑‡¥Ç, ‡¥è‡¥±‡µç‡¥±‡¥µ‡µÅ‡¥Ç ‡¥Ö‡¥®‡µÅ‡¥Ø‡µã‡¥ú‡µç‡¥Ø‡¥Æ‡¥æ‡¥Ø ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡µÅ‡¥ï‡¥≥‡µÅ‡¥ü‡µÜ ‡¥é‡¥£‡µç‡¥£‡¥Ç ‡¥ï‡¥£‡µç‡¥ü‡µÜ‡¥§‡µç‡¥§‡¥æ‡µª [‡¥é‡µΩ‡¥¨‡µã ‡¥Æ‡µÜ‡¥§‡µç‡¥§‡¥°‡µç](https://en.wikipedia.org/wiki/Elbow_method_(clustering)) ‡¥â‡¥™‡¥Ø‡µã‡¥ó‡¥ø‡¥ï‡µç‡¥ï‡µÅ‡¥®‡µç‡¥®‡µÅ. ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡µÅ‡¥ï‡¥≥‡µÅ‡¥ü‡µÜ ‡¥é‡¥£‡µç‡¥£‡¥§‡µç‡¥§‡¥ø‡¥®‡µç‡¥±‡µÜ ‡¥´‡¥Ç‡¥ó‡µç‡¥∑‡¥®‡¥æ‡¥Ø‡¥ø WCSS ‡¥™‡µç‡¥≤‡µã‡¥ü‡µç‡¥ü‡µç ‡¥ö‡µÜ‡¥Ø‡µç‡¥Ø‡µÅ‡¥ï‡¥Ø‡µÅ‡¥Ç, ‡¥â‡¥™‡¥Ø‡µã‡¥ó‡¥ø‡¥ï‡µç‡¥ï‡µá‡¥£‡µç‡¥ü ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡µÅ‡¥ï‡¥≥‡µÅ‡¥ü‡µÜ ‡¥é‡¥£‡µç‡¥£‡¥Æ‡¥æ‡¥Ø‡¥ø [‡¥µ‡¥ï‡µç‡¥∞‡¥§‡µç‡¥§‡¥ø‡¥®‡µç‡¥±‡µÜ ‡¥é‡µΩ‡¥¨‡µã](https://en.wikipedia.org/wiki/Elbow_of_the_curve "Elbow of the curve") ‡¥§‡¥ø‡¥∞‡¥û‡µç‡¥û‡µÜ‡¥ü‡µÅ‡¥ï‡µç‡¥ï‡µÅ‡¥ï‡¥Ø‡µÅ‡¥Ç ‡¥ö‡µÜ‡¥Ø‡µç‡¥Ø‡µÅ‡¥®‡µç‡¥®‡¥§‡¥æ‡¥£‡µç ‡¥à ‡¥∞‡µÄ‡¥§‡¥ø.


```r
set.seed(2056)
# Use elbow method to determine optimum number of clusters
kclusts %>% 
  ggplot(mapping = aes(x = k, y = tot.withinss)) +
  geom_line(size = 1.2, alpha = 0.8, color = "#FF7F0EFF") +
  geom_point(size = 2, color = "#FF7F0EFF")
```

The plot shows a large reduction in WCSS (so greater *tightness*) as the number of clusters increases from one to two, and a further noticeable reduction from two to three clusters. After that, the reduction is less pronounced, resulting in an `elbow` üí™in the chart at around three clusters. This is a good indication that there are two to three reasonably well separated clusters of data points.

We can now go ahead and extract the clustering model where `k = 3`:

> `pull()`: used to extract a single column
>
> `pluck()`: used to index data structures such as lists


```r
# Extract k = 3 clustering
final_kmeans <- kclusts %>% 
  filter(k == 3) %>% 
  pull(model) %>% 
  pluck(1)


final_kmeans
```

‡¥∂‡µç‡¥∞‡µá‡¥∑‡µç‡¥†‡¥Ç! ‡¥®‡¥Æ‡µÅ‡¥ï‡µç‡¥ï‡µç ‡¥≤‡¥≠‡¥ø‡¥ö‡µç‡¥ö ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡µÅ‡¥ï‡µæ ‡¥¶‡µÉ‡¥∂‡µç‡¥Ø‡¥µ‡¥§‡µç‡¥ï‡¥∞‡¥ø‡¥ï‡µç‡¥ï‡¥æ‡¥Ç. `plotly` ‡¥â‡¥™‡¥Ø‡µã‡¥ó‡¥ø‡¥ö‡µç‡¥ö‡µç ‡¥ö‡¥ø‡¥≤ ‡¥á‡¥®‡µç‡¥±‡¥±‡¥æ‡¥ï‡µç‡¥ü‡¥ø‡¥µ‡¥ø‡¥±‡µç‡¥±‡¥ø ‡¥µ‡µá‡¥£‡µã?


```r
# Add predicted cluster assignment to data set
results <-  augment(final_kmeans, df_numeric_select) %>% 
  bind_cols(df_numeric %>% select(artist_top_genre)) 

# Plot cluster assignments
clust_plt <- results %>% 
  ggplot(mapping = aes(x = popularity, y = danceability, color = .cluster, shape = artist_top_genre)) +
  geom_point(size = 2, alpha = 0.8) +
  paletteer::scale_color_paletteer_d("ggthemes::Tableau_10")

ggplotly(clust_plt)
```

‡¥™‡µç‡¥∞‡¥§‡µç‡¥Ø‡µá‡¥ï ‡¥®‡¥ø‡¥±‡¥ô‡µç‡¥ô‡¥≥‡¥ø‡µΩ ‡¥™‡µç‡¥∞‡¥§‡¥ø‡¥®‡¥ø‡¥ß‡µÄ‡¥ï‡¥∞‡¥ø‡¥ö‡µç‡¥ö‡¥ø‡¥∞‡¥ø‡¥ï‡µç‡¥ï‡µÅ‡¥®‡µç‡¥® ‡¥ì‡¥∞‡µã ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡µÅ‡¥Ç ‡¥µ‡µç‡¥Ø‡¥§‡µç‡¥Ø‡¥∏‡µç‡¥§ ‡¥Ü‡¥ï‡µÉ‡¥§‡¥ø‡¥ï‡¥≥‡¥ø‡µΩ ‡¥™‡µç‡¥∞‡¥§‡¥ø‡¥®‡¥ø‡¥ß‡µÄ‡¥ï‡¥∞‡¥ø‡¥ö‡µç‡¥ö‡¥ø‡¥∞‡¥ø‡¥ï‡µç‡¥ï‡µÅ‡¥®‡µç‡¥® ‡¥µ‡µç‡¥Ø‡¥§‡µç‡¥Ø‡¥∏‡µç‡¥§ ‡¥ú‡¥æ‡¥®‡¥±‡µÅ‡¥ï‡µæ ‡¥â‡¥£‡µç‡¥ü‡¥æ‡¥ï‡µÅ‡¥Æ‡µÜ‡¥®‡µç‡¥®‡µç ‡¥®‡¥Æ‡µç‡¥Æ‡µæ ‡¥™‡µç‡¥∞‡¥§‡µÄ‡¥ï‡µç‡¥∑‡¥ø‡¥ö‡µç‡¥ö‡¥ø‡¥∞‡µÅ‡¥®‡µç‡¥®‡µã.

‡¥Æ‡µã‡¥°‡¥≤‡¥ø‡¥®‡µç‡¥±‡µÜ ‡¥ï‡µÉ‡¥§‡µç‡¥Ø‡¥§ ‡¥®‡µã‡¥ï‡µç‡¥ï‡¥æ‡¥Ç.


```r
# Assign genres to predefined integers
label_count <- results %>% 
  group_by(artist_top_genre) %>% 
  mutate(id = cur_group_id()) %>% 
  ungroup() %>% 
  summarise(correct_labels = sum(.cluster == id))


# Print results  
cat("Result:", label_count$correct_labels, "out of", nrow(results), "samples were correctly labeled.")

cat("\nAccuracy score:", label_count$correct_labels/nrow(results))
```

‡¥à ‡¥Æ‡µã‡¥°‡¥≤‡¥ø‡¥®‡µç‡¥±‡µÜ ‡¥ï‡µÉ‡¥§‡µç‡¥Ø‡¥§ ‡¥Æ‡µã‡¥∂‡¥Æ‡¥≤‡µç‡¥≤, ‡¥™‡¥ï‡µç‡¥∑‡µá ‡¥Æ‡¥ø‡¥ï‡¥ö‡µç‡¥ö‡¥§‡µÅ‡¥Æ‡¥≤‡µç‡¥≤. ‡¥°‡¥æ‡¥±‡µç‡¥± K-Means ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡¥ø‡¥Ç‡¥ó‡¥ø‡¥®‡µç ‡¥Ö‡¥®‡µÅ‡¥Ø‡µã‡¥ú‡µç‡¥Ø‡¥Æ‡¥≤‡µç‡¥≤‡¥æ‡¥Ø‡¥ø‡¥∞‡¥ø‡¥ï‡µç‡¥ï‡¥æ‡¥Ç. ‡¥à ‡¥°‡¥æ‡¥±‡µç‡¥± ‡¥µ‡¥≥‡¥∞‡µÜ ‡¥Ö‡¥∏‡¥Æ‡¥§‡µÅ‡¥≤‡¥ø‡¥§‡¥Æ‡¥æ‡¥£‡µç, correlations ‡¥µ‡¥≥‡¥∞‡µÜ ‡¥ï‡µÅ‡¥±‡¥µ‡¥æ‡¥£‡µç, ‡¥ï‡µÇ‡¥ü‡¥æ‡¥§‡µÜ ‡¥ï‡µã‡¥≥‡¥Ç ‡¥Æ‡µÇ‡¥≤‡µç‡¥Ø‡¥ô‡µç‡¥ô‡µæ‡¥ï‡µç‡¥ï‡¥ø‡¥ü‡¥Ø‡¥ø‡µΩ ‡¥µ‡µç‡¥Ø‡¥§‡µç‡¥Ø‡¥æ‡¥∏‡¥Ç ‡¥µ‡¥≥‡¥∞‡µÜ ‡¥ï‡µÇ‡¥ü‡µÅ‡¥§‡¥≤‡¥æ‡¥£‡µç, ‡¥Ö‡¥§‡¥ø‡¥®‡¥æ‡µΩ ‡¥®‡¥≤‡µç‡¥≤ ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡¥ø‡¥Ç‡¥ó‡µç ‡¥∏‡¥æ‡¥ß‡µç‡¥Ø‡¥Æ‡¥≤‡µç‡¥≤. ‡¥µ‡¥æ‡¥∏‡µç‡¥§‡¥µ‡¥§‡µç‡¥§‡¥ø‡µΩ, ‡¥∞‡µÇ‡¥™‡¥™‡µç‡¥™‡µÜ‡¥ü‡µÅ‡¥®‡µç‡¥® ‡¥ï‡µç‡¥≤‡¥∏‡µç‡¥±‡µç‡¥±‡¥±‡µÅ‡¥ï‡µæ ‡¥Æ‡µÅ‡¥ï‡¥≥‡¥ø‡µΩ ‡¥®‡¥æ‡¥Ç ‡¥®‡¥ø‡µº‡¥µ‡¥ö‡¥ø‡¥ö‡µç‡¥ö ‡¥Æ‡µÇ‡¥®‡µç‡¥®‡µç ‡¥ú‡¥æ‡¥®‡µº ‡¥µ‡¥ø‡¥≠‡¥æ‡¥ó‡¥ô‡µç‡¥ô‡µæ ‡¥Æ‡µÇ‡¥≤‡¥Ç ‡¥∂‡¥ï‡µç‡¥§‡¥Æ‡¥æ‡¥Ø‡¥ø ‡¥∏‡µç‡¥µ‡¥æ‡¥ß‡µÄ‡¥®‡¥ø‡¥ï‡µç‡¥ï‡¥™‡µç‡¥™‡µÜ‡¥ü‡µç‡¥ü‡¥§‡µã ‡¥µ‡¥ï‡µç‡¥∞‡¥Æ‡¥æ‡¥Ø‡¥§‡µã ‡¥Ü‡¥Ø‡¥ø‡¥∞‡¥ø‡¥ï‡µç‡¥ï‡¥æ‡¥Ç.
