## 3. เดชเดฐเดฟเดถเตเดฒเดจเดตเตเด เดชเดฐเดฟเดถเตเดงเดจเดพ เดกเดพเดฑเตเดฑเดฏเตเด

เดชเดฐเดฟเดถเตเดฒเดฟเดค เดชเดเดจเดคเตเดคเดฟเตฝ เดกเดพเดฑเตเดฑเดฏเต เดฐเดฃเตเดเต เดเดชเดธเดฎเตเดนเดเตเดเดณเดพเดฏเดฟ *เดชเดฟเดฐเดฟเดเตเดเตเดตเดฟเดเตเด* เดเดจเตเดจเดคเต เดธเดพเดงเดพเดฐเดฃ เดชเตเดฐเดเตเดฐเดฟเดฏเดฏเดพเดฃเต; เดฎเตเดกเตฝ เดชเดฐเดฟเดถเตเดฒเดฟเดชเตเดชเดฟเดเตเดเดพเตป เดเดชเดฏเตเดเดฟเดเตเดเตเดจเตเดจ (เดธเดพเดงเดพเดฐเดฃเดฏเดพเดฏเดฟ เดตเดฒเดฟเดฏ) เดเดฐเต เดธเตเดฑเตเดฑเต, เดฎเตเดกเดฒเดฟเดจเตเดฑเต เดชเตเดฐเดเดเดจเด เดเดพเดฃเดพเตป เดเดชเดฏเตเดเดฟเดเตเดเตเดจเตเดจ เดเตเดฑเดฟเดฏ "เดนเตเตพเดกเต-เดฌเดพเดเตเดเต" เดธเตเดฑเตเดฑเต.

เดเดชเตเดชเตเตพ เดกเดพเดฑเตเดฑ เดคเดฏเตเดฏเดพเดฑเดพเดฏเดคเดฟเดจเดพเตฝ, เด เดกเดพเดฑเตเดฑเดพเดธเตเดฑเตเดฑเดฟเดฒเต เดธเดเดเตเดฏเดเดณเดฟเตฝ เดฒเดเดฟเดเตเดเตฝ เดธเตเดชเตเดฒเดฟเดฑเตเดฑเต เดจเดฟเตผเดฃเตเดฃเดฏเดฟเดเตเดเดพเตป เดฏเดจเตเดคเตเดฐเด เดธเดนเดพเดฏเดฟเดเตเดเตเดฎเต เดเดจเตเดจเต เดจเตเดเตเดเดพเด. เดกเดพเดฑเตเดฑ เดชเดฟเดฐเดฟเดเตเดเดพเตป *เดเดเตเดเดจเต* เดเดจเตเดจ เดตเดฟเดตเดฐเดเตเดเตพ เดเตพเดเตเดเตเดณเตเดณเตเดจเตเดจ เดเดฐเต เดเดฌเตเดเดเตเดฑเตเดฑเต เดธเตเดทเตเดเดฟเดเตเดเดพเตป, Tidymodels เดซเตเดฐเตเดฏเดฟเดเดตเตผเดเดฟเดจเตเดฑเต เดญเดพเดเดฎเดพเดฏ [rsample](https://tidymodels.github.io/rsample/) เดชเดพเดเตเดเตเดเต เดเดชเดฏเตเดเดฟเดเตเดเดพเด, เดคเตเดเตผเดจเตเดจเต เดธเตเดทเตเดเดฟเดเตเด เดชเดฐเดฟเดถเตเดฒเดจเดตเตเด เดชเดฐเดฟเดถเตเดงเดจเดพ เดธเตเดฑเตเดฑเตเดเดณเตเด เดเดเตเดเตเดเดพเตป เดฐเดฃเตเดเต rsample เดซเดเดเตเดทเดจเตเดเตพ เดเดชเดฏเตเดเดฟเดเตเดเดพเด:


```r
set.seed(2056)
# Split 67% of the data for training and the rest for tesing
diabetes_split <- diabetes_select %>% 
  initial_split(prop = 0.67)

# Extract the resulting train and test sets
diabetes_train <- training(diabetes_split)
diabetes_test <- testing(diabetes_split)

# Print the first 3 rows of the training set
diabetes_train %>% 
  slice(1:10)
```

เดเดชเตเดชเตเตพ เดจเดพเด เดจเดฎเตเดฎเตเดเต เดฎเตเดกเตฝ เดเตเดฐเตเดฏเดฟเตป เดเตเดฏเตเดฏเดพเตป เดคเดฏเตเดฏเดพเดฑเดพเดฃเต!

Tidymodels-เตฝ, เดจเดฟเดเตเดเตพ `parsnip()` เดเดชเดฏเตเดเดฟเดเตเดเต เดฎเตเดกเดฒเตเดเตพ เดจเดฟเตผเดตเดเดฟเดเตเดเตเดฎเตเดชเตเตพ เดฎเตเดจเตเดจเต เดเดถเดฏเดเตเดเตพ เดตเตเดฏเดเตเดคเดฎเดพเดเตเดเตเดจเตเดจเต:

- เดฎเตเดกเตฝ **เดคเดฐเด** เดฒเตเดจเดฟเดฏเตผ เดฑเตเดเตเดฐเดทเตป, เดฒเตเดเดฟเดธเตเดฑเตเดฑเดฟเดเต เดฑเตเดเตเดฐเดทเตป, เดกเดฟเดธเดฟเดทเตป เดเตเดฐเต เดฎเตเดกเดฒเตเดเตพ เดคเตเดเดเตเดเดฟเดฏเดตเดฏเต เดตเตเดฏเดคเตเดฏเดธเตเดคเดฎเดพเดเตเดเตเดจเตเดจเต.

- เดฎเตเดกเตฝ **เดฎเตเดกเต** เดฑเตเดเตเดฐเดทเดจเตเด เดเตเดฒเดพเดธเดฟเดซเดฟเดเตเดเตเดทเดจเตเด เดชเตเดฒเตเดณเตเดณ เดชเตเดคเตเดตเดพเดฏ เดเดชเตเดทเดจเตเดเตพ เดเตพเดเตเดเตเดณเตเดณเตเดจเตเดจเต; เดเดฟเดฒ เดฎเตเดกเตฝ เดคเดฐเด เดเดตเดฏเดฟเตฝ เดเดคเตเดเตเดเดฟเดฒเตเด เดเดจเตเดจเต เดชเดฟเดจเตเดคเตเดฃเดฏเตเดเตเดเตเด, เดเดฟเดฒเดคเต เดเดฐเต เดฎเตเดกเดฟเตฝ เดฎเดพเดคเตเดฐเดฎเต เดเดฃเตเดเดพเดเต.

- เดฎเตเดกเตฝ **เดเดเตเดเดฟเตป** เดฎเตเดกเตฝ เดซเดฟเดฑเตเดฑเต เดเตเดฏเตเดฏเดพเตป เดเดชเดฏเตเดเดฟเดเตเดเตเดจเตเดจ เดเดเดชเตเดฏเตเดเตเดเตเดทเดฃเตฝ เดเดชเดเดฐเดฃเด เดเดฃเต. เดธเดพเดงเดพเดฐเดฃเดฏเดพเดฏเดฟ เดเดต R เดชเดพเดเตเดเตเดเตเดเตพ เดเดฃเต, เดเดฆเดพเดนเดฐเดฃเดคเตเดคเดฟเดจเต **`"lm"`** เดเดฒเตเดฒเตเดเตเดเดฟเตฝ **`"ranger"`**

เด เดฎเตเดกเดฒเดฟเดเดเต เดตเดฟเดตเดฐเดเตเดเตพ เดเดฐเต เดฎเตเดกเตฝ เดธเตเดชเตเดธเดฟเดซเดฟเดเตเดเตเดทเดจเดฟเตฝ เดชเดเตผเดคเตเดคเดชเตเดชเตเดเตเดจเตเดจเต, เดเดคเดฟเดจเดพเตฝ เดจเดฎเตเดเตเดเต เดเดฐเต เดฎเตเดกเตฝ เดจเดฟเตผเดฎเตเดฎเดฟเดเตเดเดพเด!


```r
# Build a linear model specification
lm_spec <- 
  # Type
  linear_reg() %>% 
  # Engine
  set_engine("lm") %>% 
  # Mode
  set_mode("regression")


# Print the model specification
lm_spec
```

เดเดฐเต เดฎเตเดกเตฝ *เดจเดฟเตผเดตเดเดฟเดเตเดเดถเตเดทเด*, เดฎเตเดกเตฝ `estimated` เดเดฒเตเดฒเตเดเตเดเดฟเตฝ `trained` เดเตเดฏเตเดฏเดพเตป เดธเดพเดงเดฟเดเตเดเตเด, เดธเดพเดงเดพเดฐเดฃเดฏเดพเดฏเดฟ เดเดฐเต เดซเตเตผเดฎเตเดฒเดฏเตเด เดเดฟเดฒ เดกเดพเดฑเตเดฑเดฏเตเด เดเดชเดฏเตเดเดฟเดเตเดเต [`fit()`](https://parsnip.tidymodels.org/reference/fit.html) เดซเดเดเตเดทเตป เดเดชเดฏเตเดเดฟเดเตเดเต.

`y ~ .` เดเดจเตเดจเดคเต `y` เดจเดพเด เดชเตเดฐเดตเดเดฟเดเตเดเดชเตเดชเตเดเตเดจเตเดจ เดเดณเดตเดพเดฏเดฟ/เดฒเดเตเดทเตเดฏเดฎเดพเดฏเดฟ เดซเดฟเดฑเตเดฑเต เดเตเดฏเตเดฏเตเดฎเตเดจเตเดจเต เดเตผเดคเตเดฅเด, เดเดฒเตเดฒเดพ เดชเตเดฐเดตเดเดเดเตเดเดณเดพเตฝ/เดซเตเดเตเดเดฑเตเดเดณเดพเตฝ เดตเดฟเดถเดฆเตเดเดฐเดฟเดเตเดเดชเตเดชเตเดเตเดจเตเดจเต, เดเดฅเดตเดพ `.` (เด เดเตเดธเดฟเตฝ, เดจเดฎเตเดเตเดเต เดเดฐเต เดชเตเดฐเดตเดเดเตป เดฎเดพเดคเตเดฐเดฎเต เดเดณเตเดณเต: `bmi`)


```r
# Build a linear model specification
lm_spec <- linear_reg() %>% 
  set_engine("lm") %>%
  set_mode("regression")


# Train a linear regression model
lm_mod <- lm_spec %>% 
  fit(y ~ ., data = diabetes_train)

# Print the model
lm_mod
```

From the model output, we can see the coefficients learned during training. They represent the coefficients of the line of best fit that gives us the lowest overall error between the actual and predicted variable.
<br>

## 5. Make predictions on the test set

Now that we've trained a model, we can use it to predict the disease progression y for the test dataset using [parsnip::predict()](https://parsnip.tidymodels.org/reference/predict.model_fit.html). This will be used to draw the line between data groups.


```r
# Make predictions for the test set
predictions <- lm_mod %>% 
  predict(new_data = diabetes_test)

# Print out some of the predictions
predictions %>% 
  slice(1:5)
```

`dplyr::bind_cols()` เดชเดฒ เดกเดพเดฑเตเดฑเดพ เดซเตเดฐเตเดฏเดฟเดฎเตเดเดณเตเด เดเดพเดฐเตเดฏเดเตเดทเดฎเดฎเดพเดฏเดฟ เดเตเดณเด เดเดเดฟเดธเตเดฅเดพเดจเดคเตเดคเดฟเตฝ เดฌเดจเตเดงเดฟเดชเตเดชเดฟเดเตเดเตเดจเตเดจเต.


```r
# Combine the predictions and the original test set
results <- diabetes_test %>% 
  bind_cols(predictions)


results %>% 
  slice(1:5)
```
