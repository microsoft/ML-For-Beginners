## เดเดญเตเดฏเดพเดธเด - เดกเดพเดฑเตเดฑ เดถเตเดฆเตเดงเดฎเดพเดเตเดเตเด

เดเดฆเตเดฏเด, เดกเดพเดฑเตเดฑ เดเตเดฑเดเตเดเต เดถเตเดฆเตเดงเดฎเดพเดเตเดเตเด, เดจเตพ เดฎเตเดฒเตเดฏเดเตเดเตพ เดเดดเดฟเดตเดพเดเตเดเดฟ เดเดฟเดฒ เดเตเดณเดเตเดเตพ เดฎเดพเดคเตเดฐเด เดคเดฟเดฐเดเตเดเตเดเตเดเตเดเตเด:

```r
# Load the core tidyverse packages
library(tidyverse)

# Import the data and clean column names
pumpkins <- read_csv(file = "https://raw.githubusercontent.com/microsoft/ML-For-Beginners/main/2-Regression/data/US-pumpkins.csv") %>% 
  clean_names()

# Select desired columns
pumpkins_select <- pumpkins %>% 
  select(c(city_name, package, variety, origin, item_size, color)) 

# Drop rows containing missing values and encode color as factor (category)
pumpkins_select <- pumpkins_select %>% 
  drop_na() %>% 
  mutate(color = factor(color))

# View the first few rows
pumpkins_select %>% 
  slice_head(n = 5)
```

```r
pumpkins_select %>% 
  glimpse()
```

### Visualization - categorical plot
เดเดชเตเดชเตเตพ เดจเดฟเดเตเดเตพ เดชเดเดชเตเดเดฟเตป เดกเดพเดฑเตเดฑ เดตเตเดฃเตเดเตเด เดฒเตเดกเต เดเตเดฏเตเดคเต เดเดฟเดฒ เดตเตเดฐเดฟเดฏเดฌเดฟเดณเตเดเตพ เดเตพเดชเตเดชเตเดเตเดจเตเดจ เดเดฐเต เดกเดพเดฑเตเดฑเดพเดธเตเดฑเตเดฑเต เดธเดเดฐเดเตเดทเดฟเดเตเดเดพเตป เดเตเดฒเตเตป เดเตเดฏเตเดคเดฟเดเตเดเตเดฃเตเดเต, เดเดคเดฟเตฝ Color เดเตพเดชเตเดชเตเดเตเดจเตเดจเต. เดจเตเดเตเดเตเดฌเตเดเตเดเดฟเตฝ ggplot เดฒเตเดฌเตเดฐเดฑเดฟ เดเดชเดฏเตเดเดฟเดเตเดเต เดกเดพเดฑเตเดฑเดพเดซเตเดฐเตเดฏเดฟเด เดฆเตเดถเตเดฏเดตเตฝเดเตเดเดฐเดฟเดเตเดเดพเด.

ggplot เดฒเตเดฌเตเดฐเดฑเดฟ เดจเดฟเดเตเดเดณเตเดเต เดกเดพเดฑเตเดฑ เดฆเตเดถเตเดฏเดตเตฝเดเตเดเดฐเดฟเดเตเดเดพเตป เดเดฟเดฒ เดจเดฒเตเดฒ เดฎเดพเตผเดเดเตเดเตพ เดจเตฝเดเตเดจเตเดจเต. เดเดฆเดพเดนเดฐเดฃเดคเตเดคเดฟเดจเต, เดเดฐเต Variety, Color เดเดจเตเดจเดฟเดตเดฏเตเดเต เดกเดพเดฑเตเดฑเดฏเตเดเต เดตเดฟเดคเดฐเดฃเดเตเดเตพ categorical plot-เตฝ เดคเดพเดฐเดคเดฎเตเดฏเด เดเตเดฏเตเดฏเดพเด.

1. geombar เดซเดเดเตเดทเตป เดเดชเดฏเตเดเดฟเดเตเดเต, เดจเดฎเตเดฎเตเดเต เดชเดเดชเตเดเดฟเตป เดกเดพเดฑเตเดฑ เดเดชเดฏเตเดเดฟเดเตเดเต, เดเดฐเต เดชเดเดชเตเดเดฟเตป เดตเดฟเดญเดพเดเดคเตเดคเดฟเดจเตเด (เดเดฑเดเตเดเต เดเดฒเตเดฒเตเดเตเดเดฟเตฝ เดตเตเดฑเตเดฑเต) เดเดณเตผ เดฎเดพเดชเตเดชเดฟเดเดเต เดตเตเดฏเดเตเดคเดฎเดพเดเตเดเดฟเดฏเดพเดฃเต เดเดคเตเดคเดฐเดฎเตเดฐเต เดชเตเดฒเตเดเตเดเต เดธเตเดทเตเดเดฟเดเตเดเตเด:

เดเดฟเดกเดฟเดฎเตเดกเตฝเดธเต เดฎเดฑเตเดฑเตเดฐเต เดจเดฒเตเดฒ เดชเดพเดเตเดเตเดเต เดจเตฝเดเตเดจเตเดจเต: [recipes](https://recipes.tidymodels.org/) - เดกเดพเดฑเตเดฑ เดชเตเดฐเตเดชเตเดฐเตเดธเดธเตเดธเดฟเดเดเดฟเดจเตเดณเตเดณ เดเดฐเต เดชเดพเดเตเดเตเดเต. เดเดฒเตเดฒเดพ เดชเตเดฐเดกเดฟเดเตเดเตผ เดเตเดณเดเตเดเดณเตเด เดธเดเดเตเดฏเดเดณเดพเดฏเดฟ เดเตปเดเตเดกเต เดเตเดฏเตเดฏเดชเตเดชเตเดเดฃเดฎเตเดจเตเดจเต เดจเดฟเตผเดฆเตเดฆเตเดถเดฟเดเตเดเตเดจเตเดจ เดเดฐเต `recipe` เดจเดพเด เดจเดฟเตผเดตเดเดฟเดเตเดเตเด, เดเดคเดฟเดจเต `prep` เดเตเดฏเตเดคเต เดเดตเดถเตเดฏเดฎเดพเดฏ เดเดณเดตเตเดเดณเตเด เดธเตเดฅเดฟเดคเดฟเดตเดฟเดตเดฐเดเตเดเดณเตเด เดเดฃเดเตเดเดพเดเตเดเตเด, เดเดเตเดตเดฟเตฝ `bake` เดเดชเดฏเตเดเดฟเดเตเดเต เดชเตเดคเดฟเดฏ เดกเดพเดฑเตเดฑเดฏเดฟเตฝ เดเดฃเดเตเดเตเดเตพ เดชเตเดฐเดฏเตเดเดฟเดเตเดเตเด.

> เดธเดพเดงเดพเดฐเดฃเดฏเดพเดฏเดฟ, recipes เดฎเตเดกเดฒเดฟเดเดเดฟเดจเตเดณเตเดณ เดชเตเดฐเตเดชเตเดฐเตเดธเดธเตเดธเดฑเดพเดฏเดฟ เดเดชเดฏเตเดเดฟเดเตเดเตเดจเตเดจเต, เดเดคเดฟเตฝ เดกเดพเดฑเตเดฑ เดธเตเดฑเตเดฑเดฟเตฝ เดเดคเต เดเดเตเดเดเตเดเตพ เดชเตเดฐเดฏเตเดเดฟเดเตเดเดฃเดฎเตเดจเตเดจเต เดจเดฟเตผเดตเดเดฟเดเตเดเตเดจเตเดจเต, เดฎเตเดกเดฒเดฟเดเดเดฟเดจเต เดคเดฏเตเดฏเดพเดฑเดพเดเตเดเดพเตป. เด เดธเดพเดนเดเดฐเตเดฏเดคเตเดคเดฟเตฝ `workflow()` เดเดชเดฏเตเดเดฟเดเตเดเตเดจเตเดจเดคเต **เดฎเดฟเดเดเตเดเดคเดพเดฃเต**, เดชเดเดฐเด `prep` เดเด `bake` เดเด เดเดชเดฏเตเดเดฟเดเตเดเต เดฎเดพเดจเตเดตเดฒเดพเดฏเดฟ เดเดฐเต recipe เดเดฃเดเตเดเดพเดเตเดเตเดจเตเดจเดคเดฟเดจเต. เดจเดพเด เดเดคเตเดฒเตเดฒเดพเด เดเดเตป เดเดพเดฃเตเด.
>
> เดเดจเตเดจเดพเตฝ เดเดชเตเดชเตเตพ, เดจเดพเด recipes + prep + bake เดเดชเดฏเตเดเดฟเดเตเดเต เดกเดพเดฑเตเดฑเดพ เดตเดฟเดถเดเดฒเดจเดคเตเดคเดฟเดจเต เดคเดฏเตเดฏเดพเดฑเดพเดเตเดเดพเตป เดเดคเต เดเดเตเดเดเตเดเตพ เดชเตเดฐเดฏเตเดเดฟเดเตเดเดฃเดฎเตเดจเตเดจเต เดจเดฟเตผเดฆเตเดฆเตเดถเดฟเดเตเดเตเดเดฏเตเด, เด เดเดเตเดเดเตเดเตพ เดชเตเดฐเดฏเตเดเดฟเดเตเด เดชเตเดฐเตเดชเตเดฐเตเดธเดธเตเดธเตเดเตเดฏเตเดค เดกเดพเดฑเตเดฑ เดเดเตเดเตเดเตเดเดฏเตเด เดเตเดฏเตเดฏเตเดจเตเดจเต.


```r
# Preprocess and extract data to allow some data analysis
baked_pumpkins <- recipe(color ~ ., data = pumpkins_select) %>%
  # Define ordering for item_size column
  step_mutate(item_size = ordered(item_size, levels = c('sml', 'med', 'med-lge', 'lge', 'xlge', 'jbo', 'exjbo'))) %>%
  # Convert factors to numbers using the order defined above (Ordinal encoding)
  step_integer(item_size, zero_based = F) %>%
  # Encode all other predictors using one hot encoding
  step_dummy(all_nominal(), -all_outcomes(), one_hot = TRUE) %>%
  prep(data = pumpkin_select) %>%
  bake(new_data = NULL)

# Display the first few rows of preprocessed data
baked_pumpkins %>% 
  slice_head(n = 5)
```

Tidymodels-เตฝ เดฒเตเดเดฟเดธเตเดฑเตเดฑเดฟเดเต เดฑเตเดเตเดฐเดทเตป เดฎเตเดกเตฝ เดตเตเดฏเดเตเดคเดฎเดพเดเตเดเดพเดจเตเดณเตเดณ เดจเดฟเดฐเดตเดงเดฟ เดฎเดพเตผเดเตเดเดเตเดเดณเตเดฃเตเดเต. `?logistic_reg()` เดเดพเดฃเตเด. เดเดชเตเดชเตเตพ, เดจเดพเด เดกเดฟเดซเตเตพเดเตเดเต `stats::glm()` เดเดเตเดเดฟเตป เดตเดดเดฟ เดเดฐเต เดฒเตเดเดฟเดธเตเดฑเตเดฑเดฟเดเต เดฑเตเดเตเดฐเดทเตป เดฎเตเดกเตฝ เดตเตเดฏเดเตเดคเดฎเดพเดเตเดเตเด.


```r
# Create a recipe that specifies preprocessing steps for modelling
pumpkins_recipe <- recipe(color ~ ., data = pumpkins_train) %>% 
  step_mutate(item_size = ordered(item_size, levels = c('sml', 'med', 'med-lge', 'lge', 'xlge', 'jbo', 'exjbo'))) %>%
  step_integer(item_size, zero_based = F) %>%  
  step_dummy(all_nominal(), -all_outcomes(), one_hot = TRUE)

# Create a logistic model specification
log_reg <- logistic_reg() %>% 
  set_engine("glm") %>% 
  set_mode("classification")
```

[positive predictive value](https://en.wikipedia.org/wiki/Positive_predictive_value "Positive predictive value") เดเดจเตเดจเตเด เดตเดฟเดณเดฟเดเตเดเตเดจเตเดจเต.

๐ Specificity: `TN/(TN + FP)` เดฏเดฅเดพเตผเดคเตเดฅ เดจเตเดเดฑเตเดฑเตเดตเต เดธเดพเดฎเตเดชเดฟเดณเตเดเดณเดฟเตฝ เดจเดฟเดจเตเดจเตเดณเตเดณ เดจเตเดเดฑเตเดฑเตเดตเต เดซเดฒเดเตเดเดณเตเดเต เดเดจเตเดชเดพเดคเด.

๐ Accuracy: `TP + TN/(TP + TN + FP + FN)` เดเดฐเต เดธเดพเดฎเตเดชเดฟเดณเดฟเดจเดพเดฏเดฟ เดถเดฐเดฟเดฏเดพเดฏเดฟ เดชเตเดฐเดตเดเดฟเดเตเด เดฒเตเดฌเดฒเตเดเดณเตเดเต เดถเดคเดฎเดพเดจเด.

๐ F Measure: Precision เดเด Recall เดเด เดคเดฎเตเดฎเดฟเดฒเตเดณเตเดณ เดญเดพเดฐเดฟเดค เดถเดฐเดพเดถเดฐเดฟ, เดเดฑเตเดฑเดตเตเด เดจเดฒเตเดฒเดคเต 1, เดเดฑเตเดฑเดตเตเด เดฎเตเดถเด 0.

เดเดชเตเดชเตเตพ เด เดฎเตเดเตเดฐเดฟเดเตเดเตเดเตพ เดเดฃเดเตเดเดพเดเตเดเดพเด!


```r
# Combine metric functions and calculate them all at once
eval_metrics <- metric_set(ppv, recall, spec, f_meas, accuracy)
eval_metrics(data = results, truth = color, estimate = .pred_class)
```
