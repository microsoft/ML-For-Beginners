เดกเดพเดฑเตเดฑเดพ เดธเดฏเดจเตเดฑเดฟเดธเตเดฑเตเดฑเตเดเตพ เดจเตเดฐเดฟเดเตเดฃเตเด เดเดฑเตเดฑเดตเตเด เดธเดพเดงเดพเดฐเดฃ เดชเตเดฐเดถเตเดจเดเตเดเดณเดฟเตฝ เดเดจเตเดจเดพเดฃเต เดเดชเตเตผเดฃเตเดฃเดฎเดพเดฏ เดเดฒเตเดฒเตเดเตเดเดฟเตฝ เดจเดทเตเดเดฎเดพเดฏ เดกเดพเดฑเตเดฑ. R เดจเดทเตเดเดฎเดพเดฏ เดเดฒเตเดฒเตเดเตเดเดฟเตฝ เดเดฑเดฟเดฏเดชเตเดชเตเดเดพเดคเตเดค เดฎเตเดฒเตเดฏเดเตเดเดณเต เดชเตเดฐเดคเตเดฏเตเด เดธเตเดจเตเดฑเดฟเดจเตฝ เดฎเตเดฒเตเดฏเดฎเดพเดฏ `NA` (Not Available) เดเดชเดฏเตเดเดฟเดเตเดเต เดชเตเดฐเดคเดฟเดจเดฟเดงเตเดเดฐเดฟเดเตเดเตเดจเตเดจเต.

<br>
-   เดเดฐเต เดจเตเดฐเดฟเดเตเดเตเดณเตเดณ เดฎเดพเตผเดเด เดฌเตเดธเต R เดซเดเดเตเดทเตป `anyNA` เดเดชเดฏเตเดเดฟเดเตเดเตเดเดฏเดพเดฃเต, เดเดคเต เดฒเดเดฟเดเตเดเตฝ เดเดฌเตเดเดเตเดฑเตเดฑเตเดเตพ เดเดฏ `TRUE` เดเดฒเตเดฒเตเดเตเดเดฟเตฝ `FALSE` เดคเดฟเดฐเดฟเดเต เดจเตฝเดเตเดจเตเดจเต.

-   เดฎเดฑเตเดฑเตเดฐเต เดฎเดพเตผเดเด `is.na()` เดซเดเดเตเดทเตป เดเดชเดฏเตเดเดฟเดเตเดเตเดเดฏเดพเดฃเต, เดเดคเต เดเดคเตเดเตเดเต เดตเตเดฏเดเตเดคเดฟเดเดค เดเตเดณเด เดเดเดเดเตเดเตพ เดจเดทเตเดเดชเตเดชเตเดเตเดเดฟเดเตเดเตเดฃเตเดเตเดจเตเดจเต เดฒเดเดฟเดเตเดเตฝ `TRUE` เดเดฏเดฟ เดธเตเดเดฟเดชเตเดชเดฟเดเตเดเตเดจเตเดจเต.

เดจเดฟเดเตเดเดณเตเดเต เดกเดพเดฑเตเดฑเดพ เดซเตเดฐเตเดฏเดฟเด เดเดชเดฏเตเดเดฟเดเตเดเดพเตป เดเดณเตเดชเตเดชเดฎเดพเดเตเดเดพเตป, เดเดคเดฟเดฒเต เดเดฟเดฒ เดเตเดณเดเตเดเตพ เดเดดเดฟเดตเดพเดเตเดเตเด, `select()` เดเดชเดฏเตเดเดฟเดเตเดเต, เดจเดฟเดเตเดเตพเดเตเดเต เดเดตเดถเตเดฏเดฎเดพเดฏ เดเตเดณเดเตเดเตพ เดฎเดพเดคเตเดฐเด เดธเตเดเตเดทเดฟเดเตเดเตเด.

เดเดฆเดพเดนเดฐเดฃเดคเตเดคเดฟเดจเต, เด เดเดญเตเดฏเดพเดธเดคเตเดคเดฟเตฝ, เดจเดฎเตเดฎเตเดเต เดตเดฟเดถเดเดฒเดจเด `Package`, `Low Price`, `High Price` เฎฎเฎฑเฏเฎฑเฏเฎฎเฏ `Date` เดเดจเตเดจ เดเตเดณเดเตเดเตพ เดเตพเดชเตเดชเตเดเตเด. เด เดเตเดณเดเตเดเตพ เดคเดฟเดฐเดเตเดเตเดเตเดเตเดเดพเด.

#### dplyr::mutate()

`mutate()` เดเดจเตเดจเดคเต `dplyr` เดชเดพเดเตเดเตเดเดฟเดฒเตเดณเตเดณ เดเดฐเต เดซเดเดเตเดทเดจเดพเดฃเต, เดเดคเต เดจเดฟเดฒเดตเดฟเดฒเตเดณเตเดณ เดเตเดณเดเตเดเตพ เดจเดฟเดฒเดจเดฟเตผเดคเตเดคเดฟเดเตเดเตเดฃเตเดเต เดชเตเดคเดฟเดฏ เดเตเดณเดเตเดเตพ เดธเตเดทเตเดเดฟเดเตเดเตเดเดฏเต เดฎเดพเดฑเตเดฑเด เดตเดฐเตเดคเตเดคเตเดเดฏเต เดเตเดฏเตเดฏเดพเตป เดธเดนเดพเดฏเดฟเดเตเดเตเดจเตเดจเต.

mutate เดจเตเดฑเต เดชเตเดคเตเดตเดพเดฏ เดเดเดจ:

`data %>%   mutate(new_column_name = what_it_contains)`

`Date` เดเตเดณเด เดเดชเดฏเตเดเดฟเดเตเดเต `mutate` เดชเดฐเตเดเตเดทเดฟเดเตเดเดพเด, เดคเดพเดดเต เดชเดฑเดฏเตเดจเตเดจ เดชเตเดฐเดตเตผเดคเตเดคเดจเดเตเดเตพ เดเตเดฏเตเดฏเตเด:

1.  เดคเตเดฏเดคเดฟเดเตพ (เดเดชเตเดชเตเตพ character เดคเดฐเด) เดฎเดพเดธ เดซเตเตผเดฎเดพเดฑเตเดฑเดฟเดฒเตเดเตเดเต เดฎเดพเดฑเตเดฑเตเด (เดเดต US เดคเตเดฏเดคเดฟเดเดณเดพเดฃเต, เดเดคเดฟเดจเดพเตฝ เดซเตเตผเดฎเดพเดฑเตเดฑเต `MM/DD/YYYY` เดเดฃเต).

2.  เดคเตเดฏเดคเดฟเดเดณเดฟเตฝ เดจเดฟเดจเตเดจเต เดฎเดพเดธเด เดชเตเดคเดฟเดฏ เดเตเดณเดคเตเดคเดฟเดฒเตเดเตเดเต เดเดเตเดเตเดเตเด.

R เตฝ, [lubridate](https://lubridate.tidyverse.org/) เดชเดพเดเตเดเตเดเต Date-time เดกเดพเดฑเตเดฑ เดเตเดเดพเดฐเตเดฏเด เดเตเดฏเตเดฏเดพเตป เดเดณเตเดชเตเดชเดฎเดพเดเตเดเตเดจเตเดจเต. เดเดคเดฟเดจเดพเตฝ, `dplyr::mutate()`, `lubridate::mdy()`, `lubridate::month()` เดเดชเดฏเตเดเดฟเดเตเดเต เดฎเตเดเดณเดฟเตฝ เดชเดฑเดเตเด เดฒเดเตเดทเตเดฏเดเตเดเตพ เดเดเตเดเดจเต เดจเตเดเดพเดฎเตเดจเตเดจเต เดจเตเดเตเดเดพเด. เดชเดฟเดจเตเดจเตเดเต เดชเตเดฐเดตเตผเดคเตเดคเดจเดเตเดเดณเดฟเตฝ Date เดเตเดณเด เดเดตเดถเตเดฏเดฎเดฟเดฒเตเดฒเดพเดคเตเดคเดคเดฟเดจเดพเตฝ เดเดคเต เดเดดเดฟเดตเดพเดเตเดเดพเด.


```r
# Load lubridate
library(lubridate)

pumpkins <- pumpkins %>% 
  # Convert the Date column to a date object
  mutate(Date = mdy(Date)) %>% 
  # Extract month from Date
  mutate(Month = month(Date)) %>% 
  # Drop Date column
  select(-Date)

# View the first few rows
pumpkins %>% 
  slice_head(n = 7)
```

[`dplyr::filter()`](https://dplyr.tidyverse.org/reference/filter.html): เดจเดฟเดเตเดเดณเตเดเต เดจเดฟเดฌเดจเตเดงเดจเดเตพ เดชเดพเดฒเดฟเดเตเดเตเดจเตเดจ **เดชเดเตเดเตเดเตพ** เดฎเดพเดคเตเดฐเด เดเตพเดเตเดเตเดณเตเดณเตเดจเตเดจ เดกเดพเดฑเตเดฑเดฏเตเดเต เดเดฐเต เดเดชเดธเดฎเตเดนเด เดธเตเดทเตเดเดฟเดเตเดเตเดจเตเดจเต, เด เดเตเดธเดฟเตฝ, `Package` เดเตเดณเดคเตเดคเดฟเตฝ *bushel* เดเดจเตเดจ เดธเตเดเตเดฐเดฟเดเดเต เดเดณเตเดณ เดชเดเดชเตเดเดฟเดจเตเดเตพ.

[stringr::str_detect()](https://stringr.tidyverse.org/reference/str_detect.html): เดเดฐเต เดธเตเดเตเดฐเดฟเดเดเดฟเตฝ เดเดฐเต เดชเดพเดฑเตเดฑเตเตบ เดเดณเตเดณเดคเต เดเดฒเตเดฒเดพเดคเตเดฏเต เดเดฃเตเดเตเดคเตเดคเตเดจเตเดจเต.

[`stringr`](https://github.com/tidyverse/stringr) เดชเดพเดเตเดเตเดเต เดธเดพเดงเดพเดฐเดฃ เดธเตเดเตเดฐเดฟเดเดเต เดชเตเดฐเดตเตผเดคเตเดคเดจเดเตเดเตพเดเตเดเต เดฒเดณเดฟเดคเดฎเดพเดฏ เดซเดเดเตเดทเดจเตเดเตพ เดจเตฝเดเตเดจเตเดจเต.


```r
# Retain only pumpkins with "bushel"
new_pumpkins <- pumpkins %>% 
       filter(str_detect(Package, "bushel"))

# Get the dimensions of the new data
dim(new_pumpkins)

# View a few rows of the new data
new_pumpkins %>% 
  slice_head(n = 5)
```

#### dplyr::case_when()

**เดเดพเดคเตเดคเดฟเดฐเดฟเดเตเดเตเด! เดเตเดฏเตเดฏเตเดฃเตเด เดฎเดฑเตเดฑเตเดฐเต เดเดพเดฐเตเดฏเด เดเดฃเตเดเต**

เดฌเดทเตเตฝ เดคเตเด เดเดฐเต เดตเดฐเดฟเดฏเดฟเดฒเตเด เดตเตเดฏเดคเตเดฏเดธเตเดคเดฎเดพเดฃเตเดจเตเดจเต เดจเดฟเดเตเดเตพ เดถเตเดฐเดฆเตเดงเดฟเดเตเดเต? 1 1/9 เดเดฒเตเดฒเตเดเตเดเดฟเตฝ 1/2 เดฌเดทเตเตฝเดเตเดเต เดชเดเดฐเด เดฌเดทเตเตฝเดเตเดเต เดคเตเด เดเดพเดฃเดฟเดเตเดเดพเตป เดตเดฟเดฒ เดธเดพเดงเดพเดฐเดฃเดฎเดพเดเตเดเตเดฃเตเดเดคเตเดฃเตเดเต. เดเดคเต เดธเดพเดงเดพเดฐเดฃเดฎเดพเดเตเดเดพเตป เดเตเดฑเดเตเดเต เดเดฃเดฟเดคเด เดเตเดฏเตเดฏเตเดฃเตเด เดธเดฎเดฏเด.

เดจเดพเด เดเดฟเดฒ เดจเดฟเดฌเดจเตเดงเดจเดเดณเตเดเต เดเดเดฟเดธเตเดฅเดพเดจเดคเตเดคเดฟเตฝ เดตเดฟเดฒ เดเตเดณเด *เดฎเตเดฏเตเดเตเดเตเดฑเตเดฑเต* เดเตเดฏเตเดฏเดพเตป [`case_when()`](https://dplyr.tidyverse.org/reference/case_when.html) เดซเดเดเตเดทเตป เดเดชเดฏเตเดเดฟเดเตเดเตเด. `case_when` เดจเดฟเดเตเดเตพเดเตเดเต เดชเดฒ `if_else()` เดชเตเดฐเดธเตเดคเดพเดตเดจเดเดณเตเด เดตเตเดเตเดเดฑเตเดธเต เดเตเดฏเตเดฏเดพเตป เดเดจเตเดตเดฆเดฟเดเตเดเตเดจเตเดจเต.


```r
# Convert the price if the Package contains fractional bushel values
new_pumpkins <- new_pumpkins %>% 
  mutate(Price = case_when(
    str_detect(Package, "1 1/9") ~ Price/(1 + 1/9),
    str_detect(Package, "1/2") ~ Price/(1/2),
    TRUE ~ Price))

# View the first few rows of the data
new_pumpkins %>% 
  slice_head(n = 30)
```

R เดเตเดฐเดพเดซเตเดเตพ เดธเตเดทเตเดเดฟเดเตเดเดพเตป เดจเดฟเดฐเดตเดงเดฟ เดธเดฟเดธเตเดฑเตเดฑเดเตเดเตพ เดจเตฝเดเตเดจเตเดจเต, เดชเดเตเดทเต [`ggplot2`](https://ggplot2.tidyverse.org/index.html) เดเดฑเตเดฑเดตเตเด เดธเตเดจเตเดฆเดฐเดตเตเด เดฌเดนเตเดฎเตเดเดตเตเดฎเดพเดฃเต. `ggplot2` เดจเดฟเดเตเดเตพเดเตเดเต **เดธเตเดตเดคเดจเตเดคเตเดฐ เดเดเดเดเตเดเตพ เดธเดเดฏเตเดเดฟเดชเตเดชเดฟเดเตเดเต** เดเตเดฐเดพเดซเตเดเตพ เดฐเตเดชเดเตฝเดชเตเดชเดจ เดเตเดฏเตเดฏเดพเตป เดเดจเตเดตเดฆเดฟเดเตเดเตเดจเตเดจเต.

Price, Month เดเตเดณเดเตเดเดณเดฟเดฒเตเดเตเดเตเดณเตเดณ เดเดฐเต เดฒเดณเดฟเดคเดฎเดพเดฏ เดธเตเดเดพเดฑเตเดฑเตผ เดชเตเดฒเตเดเตเดเดฟเตฝ เดจเดฟเดจเตเดจเต เดคเตเดเดเตเดเดพเด.

เดเดคเดฟเดจเดพเดฏเดฟ, เดจเดพเด [`ggplot()`](https://ggplot2.tidyverse.org/reference/ggplot.html) เดเดชเดฏเตเดเดฟเดเตเดเต เดเดฐเต เดกเดพเดฑเตเดฑเดพเดธเตเดฑเตเดฑเต, เดเดธเตเดฑเตเดฑเดฑเตเดฑเดฟเดเต เดฎเดพเดชเตเดชเดฟเดเดเต ( [`aes()`](https://ggplot2.tidyverse.org/reference/aes.html) เดเดชเดฏเตเดเดฟเดเตเดเต) เดจเตฝเดเตเดเดฏเตเด, เดธเตเดเดพเดฑเตเดฑเตผ เดชเตเดฒเตเดเตเดเตเดเตพเดเตเดเดพเดฏเดฟ [`geom_point()`](https://ggplot2.tidyverse.org/reference/geom_point.html) เดชเตเดฒเตเดณเตเดณ เดฒเตเดฏเดฑเตเดเตพ เดเตเตผเดเตเดเตเดเดฏเตเด เดเตเดฏเตเดฏเตเด.


```r
# Set a theme for the plots
theme_set(theme_light())

# Create a scatter plot
p <- ggplot(data = new_pumpkins, aes(x = Price, y = Month))
p + geom_point()
```

### **เดจเดพเด เดเดคเต เดเดเตเดเดจเต เดชเตเดฐเดฏเตเดเดจเดชเตเดชเตเดเตเดคเตเดคเดพเด?**

เดเดพเตผเดเตเดเตเดเดณเดฟเตฝ เดชเตเดฐเดฏเตเดเดจเดเดฐเดฎเดพเดฏ เดกเดพเดฑเตเดฑ เดชเตเดฐเดฆเตผเดถเดฟเดชเตเดชเดฟเดเตเดเดพเตป, เดธเดพเดงเดพเดฐเดฃเดฏเดพเดฏเดฟ เดกเดพเดฑเตเดฑเดฏเต เดเดคเตเดเตเดเดฟเดฒเตเด เดตเดฟเดงเดคเตเดคเดฟเตฝ เดเตเดฐเตเดชเตเดชเต เดเตเดฏเตเดฏเตเดฃเตเดเดคเตเดฃเตเดเต. เดเดฆเดพเดนเดฐเดฃเดคเตเดคเดฟเดจเต, เดจเดฎเตเดฎเตเดเต เดเตเดธเดฟเตฝ, เดเดฐเต เดฎเดพเดธเดคเตเดคเดฟเดจเตเด เดชเดเดชเตเดเดฟเตปเดธเดฟเดจเตเดฑเต เดถเดฐเดพเดถเดฐเดฟ เดตเดฟเดฒ เดเดฃเตเดเตเดคเตเดคเตเดจเตเดจเดคเต เดจเดฎเตเดฎเตเดเต เดกเดพเดฑเตเดฑเดฏเดฟเดฒเต เดเดเดฟเดธเตเดฅเดฟเดค เดชเดพเดฑเตเดฑเตเดฃเตเดเตพเดเตเดเต เดเตเดเตเดคเตฝๆดๅฏเดเตเดเตพ เดจเตฝเดเตเด. เดเดคเต เดจเดฎเตเดฎเต เดฎเดฑเตเดฑเตเดฐเต **dplyr** เดซเตเดฒเตเดฌเตเดฏเดฟเดฒเตเดเตเดเต เดจเดฏเดฟเดเตเดเตเดจเตเดจเต:

#### `dplyr::group_by() %>% summarize()`

R-เตฝ เดเตเดฐเตเดชเตเดชเต เดเตเดฏเตเดค เดเดเตเดฐเดฟเดเตเดทเตป เดเดณเตเดชเตเดชเดคเตเดคเดฟเตฝ เดเดฃเดเตเดเดพเดเตเดเดพเตป เดเดดเดฟเดฏเตเด

`dplyr::group_by() %>% summarize()`

-   `dplyr::group_by()` เดชเตเตผเดฃเตเดฃ เดกเดพเดฑเตเดฑเดพเดธเตเดฑเตเดฑเดฟเตฝ เดจเดฟเดจเตเดจเตเดณเตเดณ เดตเดฟเดถเดเดฒเดจ เดฏเตเดฃเดฟเดฑเตเดฑเต เดเดฐเต เดเตเดฐเตเดชเตเดชเตเดเดณเดฟเดฒเตเดเตเดเต เดฎเดพเดฑเตเดฑเตเดจเตเดจเต, เดเดฆเดพเดนเดฐเดฃเดคเตเดคเดฟเดจเต เดเดฐเต เดฎเดพเดธเด.

-   `dplyr::summarize()` เดจเดฟเดเตเดเตพ เดจเดฟเตผเดฆเตเดฆเตเดถเดฟเดเตเด เดธเดเดเตเดฐเดน เดธเดพเดเตเดเตเดคเดฟเดเดตเดฟเดฆเตเดฏเดเตพเดเตเดเดพเดฏเดฟ เดเดฐเต เดเตเดฐเตเดชเตเดชเดฟเดเดเต เดตเตเดฐเดฟเดฏเดฌเดฟเดณเดฟเดจเตเด เดเดฐเต เดเตเดณเดตเตเด เดเดฐเต เดธเดเดเตเดฐเดน เดธเดพเดเตเดเตเดคเดฟเดเดตเดฟเดฆเตเดฏเดฏเตเดเตเดเตเด เดเดฐเต เดเตเดณเดตเตเด เดเดณเตเดณ เดชเตเดคเดฟเดฏ เดกเดพเดฑเตเดฑเดพ เดซเตเดฐเตเดฏเดฟเด เดธเตเดทเตเดเดฟเดเตเดเตเดจเตเดจเต.

เดเดฆเดพเดนเดฐเดฃเดคเตเดคเดฟเดจเต, `dplyr::group_by() %>% summarize()` เดเดชเดฏเตเดเดฟเดเตเดเต เดชเดเดชเตเดเดฟเตปเดธเดฟเดจเต **เดฎเดพเดธเด** เดเตเดณเด เดเดเดฟเดธเตเดฅเดพเดจเดฎเดพเดเตเดเดฟ เดเตเดฐเตเดชเตเดชเดพเดเตเดเดฟ, เดเดฐเต เดฎเดพเดธเดคเตเดคเดฟเดจเตเด **เดถเดฐเดพเดถเดฐเดฟ เดตเดฟเดฒ** เดเดฃเตเดเตเดคเตเดคเดพเด.


```r
# Find the average price of pumpkins per month
new_pumpkins %>%
  group_by(Month) %>% 
  summarise(mean_price = mean(Price))
```

เดธเดเดเตเดทเดฟเดชเตเดคเด!โจ

เดฎเดพเดธเดเตเดเตพ เดชเตเดฒเตเดณเตเดณ เดตเตผเดเตเดเตเดฏ เดธเดตเดฟเดถเตเดทเดคเดเตพ เดฌเดพเตผ เดชเตเดฒเตเดเตเดเต ๐ เดเดชเดฏเตเดเดฟเดเตเดเต เดฎเดฟเดเดเตเด เดฐเตเดคเดฟเดฏเดฟเตฝ เดชเตเดฐเดคเดฟเดจเดฟเดงเตเดเดฐเดฟเดเตเดเดพเด. เดฌเดพเตผ เดเดพเตผเดเตเดเตเดเตพเดเตเดเต เดเดคเตเดคเดฐเดตเดพเดฆเดฟเดฏเดพเดฏ เดฒเตเดฏเดฑเตเดเตพ `geom_bar()` เดเด `geom_col()` เดเด เดเดฃเต. เดเตเดเตเดคเตฝ เดเดฑเดฟเดฏเดพเตป `?geom_bar` เดเดพเดฃเตเด.
