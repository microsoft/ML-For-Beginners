เดเตเดเตเดคเตฝ tidyR ๐งน! เดเดชเตเดชเตเตพ, เดฎเตเตปเดชเดคเตเดคเต เดชเดพเดเดคเตเดคเดฟเตฝเดชเตเดฒเต `dplyr` เดเดชเดฏเตเดเดฟเดเตเดเต เดกเดพเดฑเตเดฑเดฏเตเดฎเดพเดฏเดฟ เดเดฐเต เดจเตเดคเตเดคเด! ๐


```r
# Select desired columns
pumpkins <- pumpkins %>% 
  select(variety, city_name, package, low_price, high_price, date)



# Extract the month from the dates to a new column
pumpkins <- pumpkins %>%
  mutate(date = mdy(date),
         month = month(date)) %>% 
  select(-date)



# Create a new column for average Price
pumpkins <- pumpkins %>% 
  mutate(price = (low_price + high_price)/2)


# Retain only pumpkins with the string "bushel"
new_pumpkins <- pumpkins %>% 
  filter(str_detect(string = package, pattern = "bushel"))


# Normalize the pricing so that you show the pricing per bushel, not per 1 1/9 or 1/2 bushel
new_pumpkins <- new_pumpkins %>% 
  mutate(price = case_when(
    str_detect(package, "1 1/9") ~ price/(1.1),
    str_detect(package, "1/2") ~ price*2,
    TRUE ~ price))

# Relocate column positions
new_pumpkins <- new_pumpkins %>% 
  relocate(month, .before = variety)


# Display the first 5 rows
new_pumpkins %>% 
  slice_head(n = 5)
```

เดเดคเต เดเตเดฏเตเดฏเดพเตป, เดเดฆเตเดฏเด เดกเดพเดฑเตเดฑ 80% เดเตเดฐเตเดฏเดฟเดจเดฟเดเดเดฟเดจเตเด 20% เดเตเดธเตเดฑเตเดฑเต เดธเตเดฑเตเดฑเดฟเดจเตเด เดตเดฟเดญเดเดฟเดเตเดเต, เดชเดฟเดจเตเดจเตเดเต predictor เดเตเดณเด เดเดจเตเดฑเดฟเดเดฑเตเดเดณเตเดเต เดธเตเดฑเตเดฑเดพเดฏเดฟ เดเตปเดเตเดกเต เดเตเดฏเตเดฏเดพเดจเตเดณเตเดณ เดเดฐเต เดฑเตเดธเดฟเดชเตเดชเดฟ เดจเดฟเตผเดตเดเดฟเดเตเดเต, เดฎเตเดกเตฝ เดธเตเดชเตเดธเดฟเดซเดฟเดเตเดเตเดทเตป เดจเดฟเตผเดฎเตเดฎเดฟเดเตเดเตเด. เดฑเตเดธเดฟเดชเตเดชเดฟ เดชเตเดฐเตเดชเตเดชเต เดเตเดฏเตเดคเต เดฌเตเดเตเดเต เดเตเดฏเตเดฏเดฟเดฒเตเดฒ, เดเดพเดฐเดฃเด เดเดคเต เดกเดพเดฑเตเดฑ เดชเตเดฐเดคเตเดเตเดทเดฟเดเตเดเดคเตเดชเตเดฒเต เดชเตเดฐเตเดธเดธเต เดเตเดฏเตเดฏเตเด เดเดจเตเดจเต เดจเดฎเตเดเตเดเต เดเดคเดฟเดจเดเด เดเดฑเดฟเดฏเดพเด.


```r
set.seed(2056)
# Split the data into training and test sets
pumpkins_split <- new_pumpkins %>% 
  initial_split(prop = 0.8)


# Extract training and test data
pumpkins_train <- training(pumpkins_split)
pumpkins_test <- testing(pumpkins_split)



# Create a recipe for preprocessing the data
lm_pumpkins_recipe <- recipe(price ~ package, data = pumpkins_train) %>% 
  step_integer(all_predictors(), zero_based = TRUE)



# Create a linear model specification
lm_spec <- linear_reg() %>% 
  set_engine("lm") %>% 
  set_mode("regression")
```

Tidymodels-เตฝ, เด เดธเตเดเดฐเตเดฏเดชเตเดฐเดฆเดฎเดพเดฏ เดเดฌเตเดเดเตเดฑเตเดฑเต [`workflow`](https://workflows.tidymodels.org/) เดเดจเตเดจเต เดตเดฟเดณเดฟเดเตเดเตเดจเตเดจเต, เดเดคเต เดจเดฟเดเตเดเดณเตเดเต เดฎเตเดกเดฒเดฟเดเดเต เดเดเดเดเตเดเตพ เดธเตเดเดฐเตเดฏเดชเตเดฐเดฆเดฎเดพเดฏเดฟ เดเตเดตเดถเด เดตเดฏเตเดเตเดเตเดจเตเดจเต! Python-เตฝ เดเดคเดฟเดจเต *pipelines* เดเดจเตเดจเต เดตเดฟเดณเดฟเดเตเดเตเด.

เดเดชเตเดชเตเตพ เดจเดฎเตเดเตเดเต เดเดฒเตเดฒเดพเด workflow-เดฏเดฟเดฒเตเดเตเดเต เดฌเดฃเตเดเดฟเตฝ เดเตเดฏเตเดฏเดพเด!๐ฆ


```r
# Hold modelling components in a workflow
lm_wf <- workflow() %>% 
  add_recipe(lm_pumpkins_recipe) %>% 
  add_model(lm_spec)

# Print out the workflow
lm_wf
```

๐ เดเตเดเดพเดคเต, เดเดฐเต เดตเตผเดเตเดเตโเดซเตเดฒเต เดเดฐเต เดฎเตเดกเดฒเดฟเดจเต เดชเตเดฒเต เดคเดจเตเดจเต เดซเดฟเดฑเตเดฑเต/เดเตเดฐเตเดฏเดฟเตป เดเตเดฏเตเดฏเดพเตป เดเดดเดฟเดฏเตเด.


```r
# Train the model
lm_wf_fit <- lm_wf %>% 
  fit(data = pumpkins_train)

# Print the model coefficients learned 
lm_wf_fit
```

เดฎเตเดกเตฝ เดเดเตเดเตเดชเตเดเตเดเดฟเตฝ เดจเดฟเดจเตเดจเต, เดชเดฐเดฟเดถเตเดฒเดจเดคเตเดคเดฟเดจเดฟเดเต เดชเดเดฟเดเตเด เดเตเดซเดฟเดทเตเดฏเดจเตเดฑเตเดเตพ เดเดพเดฃเดพเด. เดเดต เดฏเดฅเดพเตผเดคเตเดฅเดตเตเด เดชเตเดฐเดตเดเดฟเดเตเดเดตเตเดฎเดพเดฏ เดตเตเดฏเดคเตเดฏเดพเดธเด เดเดฑเตเดฑเดตเตเด เดเตเดฑเดตเตเดณเตเดณ เดฎเดฟเดเดเตเด เดซเดฟเดฑเตเดฑเต เดตเดฐเดฟเดฏเตเดเต เดเตเดซเดฟเดทเตเดฏเดจเตเดฑเตเดเดณเดพเดฃเต.

เดฎเตเดกเตฝ เดเดเตเดเดจเต เดชเตเดฐเดตเตผเดคเตเดคเดฟเดเตเดเต เดเดจเตเดจเต เดเดพเดฃเดพเดจเตเดณเตเดณ เดธเดฎเดฏเด ๐! เดเดคเต เดเดเตเดเดจเต เดเตเดฏเตเดฏเดพเด?

เดเดชเตเดชเตเตพ เดฎเตเดกเตฝ เดชเดฐเดฟเดถเตเดฒเดฟเดชเตเดชเดฟเดเตเดเดคเดฟเดจเดพเตฝ, `parsnip::predict()` เดเดชเดฏเตเดเดฟเดเตเดเต test_set เดจเตเดณเตเดณ เดชเตเดฐเดตเดเดจเดเตเดเตพ เดจเดเดคเตเดคเดพเด. เดชเดฟเดจเตเดจเตเดเต เด เดชเตเดฐเดตเดเดจเดเตเดเดณเต เดฏเดฅเดพเตผเดคเตเดฅ เดฒเตเดฌเตฝ เดฎเตเดฒเตเดฏเดเตเดเดณเตเดฎเดพเดฏเดฟ เดคเดพเดฐเดคเดฎเตเดฏเด เดเตเดฏเตเดคเต เดฎเตเดกเตฝ เดเดคเตเดฐเดคเตเดคเตเดณเด (เดเดฒเตเดฒเตเดเตเดเดฟเตฝ เดเดคเตเดฐเดคเตเดคเตเดณเด เดเดฒเตเดฒ!) เดชเตเดฐเดตเตผเดคเตเดคเดฟเดเตเดเตเดจเตเดจเตเดตเตเดจเตเดจเต เดตเดฟเดฒเดฏเดฟเดฐเตเดคเตเดคเดพเด.

เดเดฆเตเดฏเด เดเตเดธเตเดฑเตเดฑเต เดธเตเดฑเตเดฑเดฟเดจเตเดณเตเดณ เดชเตเดฐเดตเดเดจเดเตเดเตพ เดจเดเดคเตเดคเดพเด, เดชเดฟเดจเตเดจเตเดเต เด เดเตเดณเดเตเดเตพ เดเตเดธเตเดฑเตเดฑเต เดธเตเดฑเตเดฑเดฟเดจเตเดเต เดเตเตผเดเตเดเดพเด.


```r
# Make predictions for the test set
predictions <- lm_wf_fit %>% 
  predict(new_data = pumpkins_test)


# Bind predictions to the test set
lm_results <- pumpkins_test %>% 
  select(c(package, price)) %>% 
  bind_cols(predictions)


# Print the first ten rows of the tibble
lm_results %>% 
  slice_head(n = 10)
```

เดเดคเต, เดจเดฟเดเตเดเตพ เดเดฐเต เดฎเตเดกเตฝ เดชเดฐเดฟเดถเตเดฒเดฟเดชเตเดชเดฟเดเตเดเต เดเดคเต เดชเตเดฐเดตเดเดจเดเตเดเตพ เดจเดเดคเตเดคเดพเตป เดเดชเดฏเตเดเดฟเดเตเดเต!๐ฎ เดเดคเต เดเดคเตเดฐเดคเตเดคเตเดณเด เดจเดฒเตเดฒเดคเดพเดฃเตเดจเตเดจเต เดจเตเดเตเดเดพเด, เดฎเตเดกเดฒเดฟเดจเตเดฑเต เดชเตเดฐเดเดเดจเด เดตเดฟเดฒเดฏเดฟเดฐเตเดคเตเดคเดพเด!

Tidymodels-เตฝ, เดเดคเต `yardstick::metrics()` เดเดชเดฏเตเดเดฟเดเตเดเต เดเตเดฏเตเดฏเดพเด! เดฒเตเดจเดฟเดฏเตผ เดฑเตเดเตเดฐเดทเดจเดฟเดจเดพเดฏเดฟ, เดคเดพเดดเตเดชเตเดชเดฑเดฏเตเดจเตเดจ เดฎเตเดเตเดฐเดฟเดเตเดเตเดธเตเดเดณเดฟเตฝ เดถเตเดฐเดฆเตเดง เดเตเดจเตเดฆเตเดฐเตเดเดฐเดฟเดเตเดเดพเด:

-   `Root Mean Square Error (RMSE)`: [MSE](https://en.wikipedia.org/wiki/Mean_squared_error)เดฏเตเดเต เดเดคเตเดฐเดถเตเดฐเดฎเตเดฒเตเดฏเด. เดเดคเต เดฒเตเดฌเดฒเดฟเดจเตเดฑเต (เด เดเตเดธเดฟเตฝ, เดชเดเดชเตเดเดฟเดจเตเดฑเต เดตเดฟเดฒ) เดธเดฎเดพเดจ เดฏเตเดฃเดฟเดฑเตเดฑเดฟเดฒเตเดณเตเดณ เดเดฐเต เดเดฌเตเดธเดฒเตเดฏเตเดเตเดเต เดฎเตเดเตเดฐเดฟเดเต เดจเตฝเดเตเดจเตเดจเต. เดฎเตเดฒเตเดฏเด เดเตเดฑเตเดคเดพเดฏเดฟเดฐเดฟเดเตเดเตเด, เดฎเตเดกเตฝ เดจเดฒเตเดฒเดคเดพเดฃเตเดจเตเดจเต เดธเตเดเดฟเดชเตเดชเดฟเดเตเดเตเดจเตเดจเต (เดธเดพเดงเดพเดฐเดฃ เดเตผเดคเตเดฅเดคเตเดคเดฟเตฝ, เดชเตเดฐเดตเดเดจเดเตเดเตพ เดเดคเตเดฐเดคเตเดคเตเดณเด เดถเดฐเดฟเดฏเดฒเตเดฒ เดเดจเตเดจ เดถเดฐเดพเดถเดฐเดฟ เดตเดฟเดฒ เดชเตเดฐเดคเดฟเดจเดฟเดงเตเดเดฐเดฟเดเตเดเตเดจเตเดจเต!)

-   เดตเตผเดเตเดเตโเดซเตเดฒเต เดซเดฟเดฑเตเดฑเต เดเตเดฏเตเดคเต เดฎเตเดกเตฝ เดธเตเดทเตเดเดฟเดเตเดเตเด

-   เดเตเดธเตเดฑเตเดฑเต เดกเดพเดฑเตเดฑเดฏเดฟเตฝ เดฎเตเดกเดฒเดฟเดจเตเดฑเต เดชเตเดฐเดเดเดจเด เดตเดฟเดฒเดฏเดฟเดฐเตเดคเตเดคเตเด

เดจเดฎเตเดเตเดเต เดเดเตป เดคเตเดเดเตเดเดพเด!


```r
# Specify a recipe
poly_pumpkins_recipe <-
  recipe(price ~ package, data = pumpkins_train) %>%
  step_integer(all_predictors(), zero_based = TRUE) %>% 
  step_poly(all_predictors(), degree = 4)


# Create a model specification
poly_spec <- linear_reg() %>% 
  set_engine("lm") %>% 
  set_mode("regression")


# Bundle recipe and model spec into a workflow
poly_wf <- workflow() %>% 
  add_recipe(poly_pumpkins_recipe) %>% 
  add_model(poly_spec)


# Create a model
poly_wf_fit <- poly_wf %>% 
  fit(data = pumpkins_train)


# Print learned model coefficients
poly_wf_fit

  
```
