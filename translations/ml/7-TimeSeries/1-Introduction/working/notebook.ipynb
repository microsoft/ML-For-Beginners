# ഡാറ്റ സജ്ജീകരണം

ഈ നോട്ട്‌ബുക്കിൽ, ഞങ്ങൾ കാണിക്കുന്നു എങ്ങനെ:

ഈ മോഡ്യൂളിനായി ടൈം സീരീസ് ഡാറ്റ സജ്ജീകരിക്കാം
ഡാറ്റ ദൃശ്യവൽക്കരിക്കാം
ഈ ഉദാഹരണത്തിലെ ഡാറ്റ GEFCom2014 ഫോറ്കാസ്റ്റിംഗ് മത്സരം1-ൽ നിന്നാണ് എടുത്തത്. ഇത് 2012 മുതൽ 2014 വരെ 3 വർഷത്തെ മണിക്കൂറു അടിസ്ഥാനത്തിലുള്ള വൈദ്യുതി ലോഡ് மற்றும் താപനില മൂല്യങ്ങൾ ഉൾക്കൊള്ളുന്നു.

1Tao Hong, Pierre Pinson, Shu Fan, Hamidreza Zareipour, Alberto Troccoli and Rob J. Hyndman, "Probabilistic energy forecasting: Global Energy Forecasting Competition 2014 and beyond", International Journal of Forecasting, vol.32, no.3, pp 896-913, July-September, 2016.
