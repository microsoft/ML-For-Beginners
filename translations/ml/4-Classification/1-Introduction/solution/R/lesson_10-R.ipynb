เดจเตเดฐเต เดคเตเดเดเตเดเดพเด! เดจเดพเด เดฎเตเดฎเตเดชเดคเตเดคเต เดชเดพเดเดเตเดเดณเดฟเตฝ เดชเดเดฟเดเตเด [dplyr เดเตเดฐเดฟเดฏเดพเดชเดฆเดเตเดเตพ](https://dplyr.tidyverse.org/) เดเดชเดฏเตเดเดฟเดเตเดเตเด. เดเดฐเต เดธเดเดเตเดทเตเดชเดฎเดพเดฏเดฟ:

-   `dplyr::pivot_longer()`: เดกเดพเดฑเตเดฑ "เดจเตเดณเดฎเดพเดเตเดเดพเตป" เดธเดนเดพเดฏเดฟเดเตเดเตเดจเตเดจเต, เดตเดฐเดฟเดเดณเตเดเต เดเดฃเตเดฃเด เดตเตผเดฆเตเดงเดฟเดชเตเดชเดฟเดเตเดเต เดเตเดณเดเตเดเดณเตเดเต เดเดฃเตเดฃเด เดเตเดฑเดฏเตเดเตเดเตเดจเตเดจเต.

-   `dplyr::group_by()` and `dplyr::summarise()`: เดตเตเดฏเดคเตเดฏเดธเตเดค เดเตเดฐเตเดชเตเดชเตเดเตพเดเตเดเดพเดฏเดฟ เดธเดเดเตเดฐเดน เดธเตเดฅเดฟเดคเดฟเดตเดฟเดตเดฐเดเตเดเตพ เดเดฃเตเดเตเดคเตเดคเดพเดจเตเด เดเดต เดเดฐเต เดจเดฒเตเดฒ เดชเดเตเดเดฟเดเดฏเดฟเดฒเดพเดเตเดเดพเดจเตเด เดธเดนเดพเดฏเดฟเดเตเดเตเดจเตเดจเต.

-   `dplyr::filter()`: เดจเดฟเดเตเดเดณเตเดเต เดจเดฟเดฌเดจเตเดงเดจเดเตพ เดชเดพเดฒเดฟเดเตเดเตเดจเตเดจ เดตเดฐเดฟเดเตพ เดฎเดพเดคเตเดฐเด เดเตพเดเตเดเตเดณเตเดณเตเดจเตเดจ เดกเดพเดฑเตเดฑเดฏเตเดเต เดเดชเดธเดฎเตเดนเด เดธเตเดทเตเดเดฟเดเตเดเตเดจเตเดจเต.

-   `dplyr::mutate()`: เดเตเดณเดเตเดเตพ เดธเตเดทเตเดเดฟเดเตเดเดพเดจเตเด เดฎเดพเดฑเตเดฑเดพเดจเตเด เดธเดนเดพเดฏเดฟเดเตเดเตเดจเตเดจเต.

Allison Horst เดเดฐเตเดเตเดเดฟเดฏ เด [*เดเดฒเดพเดชเดฐเดฎเดพเดฏ* learnr เดเตเดฏเตเดเตเดเตเดฑเดฟเดฏเตฝ](https://allisonhorst.shinyapps.io/dplyr-learnr/#section-welcome) เดจเตเดเตเดเต, dplyr-เตฝ เดเดฟเดฒ เดเดชเดเดพเดฐเดชเตเดฐเดฆเดฎเดพเดฏ เดกเดพเดฑเตเดฑ เดเตเดเดพเดฐเตเดฏเด เดเตเดฏเตเดฏเตฝ เดซเดเดเตเดทเดจเตเดเตพ เดชเดฐเดฟเดเดฏเดชเตเดชเตเดเตเดคเตเดคเตเดจเตเดจเต *(Tidyverse-เดจเตเดฑเต เดญเดพเดเด)*


```r
# Creates a functions that returns the top ingredients by class

create_ingredient <- function(df){
  
  # Drop the id column which is the first colum
  ingredient_df = df %>% select(-1) %>% 
  # Transpose data to a long format
    pivot_longer(!cuisine, names_to = "ingredients", values_to = "count") %>% 
  # Find the top most ingredients for a particular cuisine
    group_by(ingredients) %>% 
    summarise(n_instances = sum(count)) %>% 
    filter(n_instances != 0) %>% 
  # Arrange by descending order
    arrange(desc(n_instances)) %>% 
    mutate(ingredients = factor(ingredients) %>% fct_inorder())
  
  
  return(ingredient_df)
} # End of function
```

เดเดชเตเดชเตเตพ เดจเดพเด เด เดซเดเดเตเดทเตป เดเดชเดฏเตเดเดฟเดเตเดเต เดเดฐเต เดเตเดฏเตเดธเดฟเดจเดฟเดฏเตเดเตเดฏเตเด เดเดฑเตเดฑเดตเตเด เดเดจเดชเตเดฐเดฟเดฏเดฎเดพเดฏ เดชเดคเตเดคเต เดเดเดเดเตเดเดณเตเดเต เดเดฐเต เดเดถเดฏเด เดจเตเดเดพเด. `thai_df` เดเดชเดฏเตเดเดฟเดเตเดเต เดเดคเต เดชเดฐเตเดเตเดทเดฟเดเตเดเดพเด.

เดเดธเดฎเดคเตเดตเดฎเตเดณเตเดณ เดกเดพเดฑเตเดฑ เดธเตเดฑเตเดฑเตเดเดณเต เดเตเดเดพเดฐเตเดฏเด เดเตเดฏเตเดฏเดพเดจเตเดณเตเดณ เดชเตเดฐเดงเดพเดจเดฎเดพเดฏ เดฐเดฃเตเดเต เดฎเดพเตผเดเตเดเดเตเดเตพ เดเดฃเตเดเต:

- เดญเตเดฐเดฟเดชเดเตเดท เดเตเดฒเดพเดธเตเดธเดฟเตฝ เดจเดฟเดจเตเดจเตเดณเตเดณ เดจเดฟเดฐเตเดเตเดทเดฃเดเตเดเตพ เดจเตเดเตเดเด เดเตเดฏเตเดฏเตฝ: `Under-sampling`

เดเดชเตเดชเตเตพ `recipe` เดเดชเดฏเตเดเดฟเดเตเดเต เดเดธเดฎเดคเตเดตเดฎเตเดณเตเดณ เดกเดพเดฑเตเดฑ เดธเตเดฑเตเดฑเตเดเดณเต เดเดเตเดเดจเต เดเตเดเดพเดฐเตเดฏเด เดเตเดฏเตเดฏเดพเดฎเตเดจเตเดจเต เดเดพเดฃเดฟเดเตเดเดพเด. เดเดฐเต recipe เดเดฐเต เดฌเตเดฒเตเดชเตเดฐเดฟเดจเตเดฑเต เดชเตเดฒเต เดเดฐเตเดคเดพเด, เดเดคเต เดเดฐเต เดกเดพเดฑเตเดฑ เดธเตเดฑเตเดฑเดฟเตฝ เดเดคเต เดเดเตเดเดเตเดเตพ เดชเตเดฐเดฏเตเดเดฟเดเตเดเดฃเด เดเดจเตเดจเต เดตเดฟเดตเดฐเดฟเดเตเดเตเดจเตเดจเต, เดกเดพเดฑเตเดฑ เดตเดฟเดถเดเดฒเดจเดคเตเดคเดฟเดจเต เดคเดฏเตเดฏเดพเดฑเดพเดเตเดเดพเตป.

-   [`step_smote(cuisine)`](https://themis.tidymodels.org/reference/step_smote.html) เดเดฐเต เดฑเตเดธเดฟเดชเตเดชเดฟ เดเดเตเดเดคเตเดคเดฟเดจเตเดฑเต *เดธเตเดชเตเดธเดฟเดซเดฟเดเตเดเตเดทเตป* เดธเตเดทเตเดเดฟเดเตเดเตเดจเตเดจเต, เดเดคเต เดจเตเดฏเตเดจเดชเดเตเดท เดเตเดฒเดพเดธเตเดธเดฟเดจเตเดฑเต เดชเตเดคเดฟเดฏ เดเดฆเดพเดนเดฐเดฃเดเตเดเตพ เดธเดฟเดจเตเดคเดฑเตเดฑเดฟเดเตเดเดพเดฏเดฟ เด เดเตเดธเตเดเดณเตเดเต เดเดเตเดคเตเดคเตเดณเตเดณ เดเดฏเตฝเดเตเดเดพเดฐเต เดเดชเดฏเตเดเดฟเดเตเดเต เดธเตเดทเตเดเดฟเดเตเดเตเดจเตเดจเต.

เดเดชเตเดชเตเตพ, เดจเดฎเตเดเตเดเต เดชเตเดฐเตเดชเตเดฐเตเดธเดธเตเดธเตเดเตเดฏเตเดค เดกเดพเดฑเตเดฑ เดเดพเดฃเดฃเดฎเตเดเตเดเดฟเตฝ, เดจเดฎเตเดเตเดเต [**`prep()`**](https://recipes.tidymodels.org/reference/prep.html)เดฏเตเด [**`bake()`**](https://recipes.tidymodels.org/reference/bake.html)เดฏเตเด เดจเดฎเตเดฎเตเดเต เดฑเตเดธเดฟเดชเตเดชเดฟเดฏเดฟเตฝ เดเตเดฏเตเดฏเตเดฃเตเดเดคเตเดฃเตเดเต.

`prep()`: เดชเดฐเดฟเดถเตเดฒเดจ เดธเตเดฑเตเดฑเดฟเตฝ เดจเดฟเดจเตเดจเตเดณเตเดณ เดเดตเดถเตเดฏเดฎเดพเดฏ เดชเดพเดฐเดพเดฎเตเดฑเตเดฑเดฑเตเดเตพ เดเดฃเดเตเดเดพเดเตเดเตเดจเตเดจเต, เดชเดฟเดจเตเดจเตเดเต เดฎเดฑเตเดฑเต เดกเดพเดฑเตเดฑเดพ เดธเตเดฑเตเดฑเตเดเดณเดฟเตฝ เดชเตเดฐเดฏเตเดเดฟเดเตเดเดพเดตเตเดจเตเดจเดคเดพเดฃเต.

`bake()`: เดชเตเดฐเตเดชเต เดเตเดฏเตเดค เดฑเตเดธเดฟเดชเตเดชเดฟ เดเดเตเดคเตเดคเต เดชเตเดฐเดตเตผเดคเตเดคเดจเดเตเดเตพ เดเดคเตเดเตเดเดฟเดฒเตเด เดกเดพเดฑเตเดฑเดพ เดธเตเดฑเตเดฑเดฟเตฝ เดชเตเดฐเดฏเตเดเดฟเดเตเดเตเดจเตเดจเต.


```r
# Prep and bake the recipe
preprocessed_df <- cuisines_recipe %>% 
  prep() %>% 
  bake(new_data = NULL) %>% 
  relocate(cuisine)

# Display data
preprocessed_df %>% 
  slice_head(n = 5)

# Quick summary stats
preprocessed_df %>% 
  introduce()
```

เดเดชเตเดชเตเตพ เดจเดฎเตเดเตเดเต เดจเดฎเตเดฎเตเดเต เดญเดเตเดทเดฃเดถเตเดฒเดฟเดเดณเตเดเต เดตเดฟเดคเดฐเดฃเดตเตเด เดเดตเดฏเต เดเดธเดฎเดคเตเดฒเตเดฏ เดกเดพเดฑเตเดฑเดฏเตเดฎเดพเดฏเดฟ เดคเดพเดฐเดคเดฎเตเดฏเด เดเตเดฏเตเดฏเดฒเตเด เดชเดฐเดฟเดถเตเดงเดฟเดเตเดเดพเด.


```r
# Distribution of cuisines
new_label_count <- preprocessed_df %>% 
  count(cuisine) %>% 
  arrange(desc(n))

list(new_label_count = new_label_count,
     old_label_count = old_label_count)
```

> เดธเดพเดงเดพเดฐเดฃเดฏเดพเดฏเดฟ, เดเดฐเต เดฑเตเดธเดฟเดชเตเดชเดฟ เดฎเตเดกเดฒเดฟเดเดเดฟเดจเดพเดฏเดฟ เดชเตเดฐเตเดชเตเดฐเตเดธเดธเดฑเดพเดฏเดพเดฃเต เดเดชเดฏเตเดเดฟเดเตเดเตเดจเตเดจเดคเต, เดเดคเดพเดฏเดคเต เดฎเตเดกเดฒเดฟเดเดเดฟเดจเดพเดฏเดฟ เดกเดพเดฑเตเดฑ เดธเตเดฑเตเดฑเดฟเตฝ เดเดคเต เดเดเตเดเดเตเดเตพ เดชเตเดฐเดฏเตเดเดฟเดเตเดเดฃเดฎเตเดจเตเดจเต เดจเดฟเตผเดตเดเดฟเดเตเดเตเดจเตเดจเต. เดเดคเตเดคเดฐเดคเตเดคเดฟเตฝ, เดเดฐเต `workflow()` เดธเดพเดงเดพเดฐเดฃเดฏเดพเดฏเดฟ เดเดชเดฏเตเดเดฟเดเตเดเตเดจเตเดจเต (เดฎเตเตปเดชเดคเตเดคเต เดชเดพเดเดเตเดเดณเดฟเตฝ เดจเดพเด เดเดคเดฟเดจเดเด เดเดฃเตเดเดคเตเดชเตเดฒเต) เดฑเตเดธเดฟเดชเตเดชเดฟ เดฎเดพเดจเตเดตเดฒเดพเดฏเดฟ เดเดฃเดเตเดเดพเดเตเดเตเดจเตเดจเดคเดฟเดจเต เดชเดเดฐเด
>
> เดเดคเดฟเดจเดพเตฝ, tidymodels เดเดชเดฏเตเดเดฟเดเตเดเตเดฎเตเดชเตเตพ เดธเดพเดงเดพเดฐเดฃเดฏเดพเดฏเดฟ **`prep()`** เดเด **`bake()`** เดเด เดฑเตเดธเดฟเดชเตเดชเดฟเดเตพเดเตเดเต เดเดตเดถเตเดฏเดฎเดฟเดฒเตเดฒ, เดชเดเตเดทเต เดจเดฎเตเดฎเตเดเต เดเตเดธเดฟเตฝ เดชเตเดฒเต เดฑเตเดธเดฟเดชเตเดชเดฟเดเตพ เดจเดฟเดเตเดเตพ เดชเตเดฐเดคเตเดเตเดทเดฟเดเตเดเตเดจเตเดจเดคเตเดชเตเดฒเต เดชเตเดฐเดตเตผเดคเตเดคเดฟเดเตเดเตเดจเตเดจเตเดฃเตเดเตเดจเตเดจเต เดธเตเดฅเดฟเดฐเตเดเดฐเดฟเดเตเดเดพเตป เดเดต เดเดชเดเดพเดฐเดชเตเดฐเดฆเดฎเดพเดฏ เดซเดเดเตเดทเดจเตเดเดณเดพเดฃเต.
>
> เดจเดฟเดเตเดเตพ **`bake()`** เดเตเดฏเตเดคเดชเตเดชเตเตพ เดชเตเดฐเตเดชเตเดฐเตเดชเตเดชเตเดเตเดฏเตเดค เดฑเตเดธเดฟเดชเตเดชเดฟ **`new_data = NULL`** เดเดชเดฏเตเดเดฟเดเตเดเต, เดจเดฟเดเตเดเตพ เดฑเตเดธเดฟเดชเตเดชเดฟ เดจเดฟเตผเดตเดเดฟเดเตเดเตเดฎเตเดชเตเตพ เดจเตฝเดเดฟเดฏ เดกเดพเดฑเตเดฑ เดคเดจเตเดจเต เดคเดฟเดฐเดฟเดเต เดฒเดญเดฟเดเตเดเตเด, เดเดจเตเดจเดพเตฝ เดชเตเดฐเตเดชเตเดฐเตเดธเดธเดฟเดเดเต เดเดเตเดเดเตเดเตพ เดเดเดจเตเดจเตเดชเตเดฏเดฟเดฐเดฟเดเตเดเตเด.
