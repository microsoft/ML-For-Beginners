## 2. เดกเดพเดฑเตเดฑ เดตเดฟเดญเดเดฟเดเตเดเต เดเดธเดฎเดคเตเดฒเตเดฏเดฎเดพเดฏ เดกเดพเดฑเตเดฑ เดธเตเดฑเตเดฑเดฟเดจเต เดเตเดเดพเดฐเตเดฏเด เดเตเดฏเตเดฏเตเด.

-   เดตเตเดฏเดคเตเดฏเดธเตเดค เดเตเดฏเตเดธเดฟเดจเดฟเดเตพ เดคเดฎเตเดฎเดฟเตฝ เดเดถเดฏเดเตเดเตเดดเดชเตเดชเด เดธเตเดทเตเดเดฟเดเตเดเตเดจเตเดจ เดเดฑเตเดฑเดตเตเด เดชเตเดคเตเดตเดพเดฏ เดเดเดเดเตเดเตพ `dplyr::select()` เดเดชเดฏเตเดเดฟเดเตเดเต เดเดดเดฟเดตเดพเดเตเดเตเด.

-   เดฎเตเดกเดฒเดฟเดเดเดฟเดจเต เดคเดฏเตเดฏเดพเดฑเดพเดเตเดเดพเตป เดกเดพเดฑเตเดฑ เดชเตเดฐเตเดชเตเดฐเตเดธเดธเต เดเตเดฏเตเดฏเตเดจเตเดจ เดเดฐเต `recipe` เดเดชเดฏเตเดเดฟเดเตเดเตเด, เดเดคเดฟเตฝ `over-sampling` เดเตฝเดเตเดฐเดฟเดคเด เดชเตเดฐเดฏเตเดเดฟเดเตเดเตเด.

เดฎเตเตปเดชเดคเตเดคเต เดชเดพเดเดคเตเดคเดฟเตฝ เดเดคเต เดจเตเดเตเดเดฟเดฏเดคเดฟเดจเดพเตฝ เดเดคเต เดเดณเตเดชเตเดชเด เดเดฏเดฟเดฐเดฟเดเตเดเดฃเด ๐ฅณ!


```r
# Load the core Tidyverse and Tidymodels packages
library(tidyverse)
library(tidymodels)

# Load the original cuisines data
df <- read_csv(file = "https://raw.githubusercontent.com/microsoft/ML-For-Beginners/main/4-Classification/data/cuisines.csv")

# Drop id column, rice, garlic and ginger from our original data set
df_select <- df %>% 
  select(-c(1, rice, garlic, ginger)) %>%
  # Encode cuisine column as categorical
  mutate(cuisine = factor(cuisine))


# Create data split specification
set.seed(2056)
cuisines_split <- initial_split(data = df_select,
                                strata = cuisine,
                                prop = 0.7)

# Extract the data in each split
cuisines_train <- training(cuisines_split)
cuisines_test <- testing(cuisines_split)

# Display distribution of cuisines in the training set
cuisines_train %>% 
  count(cuisine) %>% 
  arrange(desc(n))
```

เดธเดชเตเดชเตเตผเดเตเดเต-เดตเตเดเตเดเตผ เดเตเดฒเดธเตเดฑเตเดฑเดฑเดฟเดเดเต (SVC) ML เดธเดพเดเตเดเตเดคเดฟเดเดตเดฟเดฆเตเดฏเดเดณเดพเดฏ เดธเดชเตเดชเตเตผเดเตเดเต-เดตเตเดเตเดเตผ เดฎเตเดทเตเดจเตเดเดณเตเดเต เดเตเดเตเดเดฌเดคเตเดคเดฟเดฒเต เดเดฐเต เดถเดพเดเดฏเดพเดฃเต. SVC-เดฏเดฟเตฝ, เดนเตเดชเตเดชเตผเดชเตเดฒเตเดฏเดฟเตป เดชเดฐเดฟเดถเตเดฒเดจ เดจเดฟเดฐเตเดเตเดทเดฃเดเตเดเดณเตเดเต `เดฎเดฟเดเตเดเดตเดพเดฑเตเด` เดถเดฐเดฟเดฏเดพเดฏเดฟ เดตเตเตผเดคเดฟเดฐเดฟเดเตเดเดพเตป เดคเดฟเดฐเดเตเดเตเดเตเดเตเดเดชเตเดชเตเดเตเดจเตเดจเต, เดชเดเตเดทเต เดเดฟเดฒ เดจเดฟเดฐเตเดเตเดทเดฃเดเตเดเตพ `เดคเตเดฑเตเดฑเดพเดฏเดฟ เดตเตผเดเตเดเตเดเดฐเดฟเดเตเดเดชเตเดชเตเดเดพเด`. เดเดฟเดฒ เดชเตเดฏเดฟเดจเตเดฑเตเดเตพ เดคเตเดฑเตเดฑเดพเดฏ เดตเดถเดคเตเดคเต เดเดฃเตเดเดพเดเดพเตป เดเดจเตเดตเดฆเดฟเดเตเดเตเดจเตเดจเดคเดฟเดฒเตเดเต, SVM เดเดเตเดเตเดฒเดฏเดฑเตเดเดณเตเดเต เดเตเดเตเดคเตฝ เดฆเตเดขเดฎเดพเดเตเดจเตเดจเต, เดเดคเดฟเดจเดพเตฝ เดชเตเดคเดฟเดฏ เดกเดพเดฑเตเดฑเดฏเดฟเดฒเตเดเตเดเต เดฎเดฟเดเดเตเด เดชเตเดคเตเดตเดพเดฏ เดชเตเดฐเดฏเตเดเด เดธเดพเดงเตเดฏเดฎเดพเดเตเดจเตเดจเต. เด เดฒเดเดเดจเด เดจเดฟเดฏเดจเตเดคเตเดฐเดฟเดเตเดเตเดจเตเดจ เดชเดพเดฐเดพเดฎเตเดฑเตเดฑเตผ `cost` เดเดจเตเดจเดฑเดฟเดฏเดชเตเดชเตเดเตเดจเตเดจเต, เดเดคเดฟเดจเต เดกเดฟเดซเตเตพเดเตเดเต เดฎเตเดฒเตเดฏเด 1 เดเดฃเต (`help("svm_poly")` เดเดพเดฃเตเด).

เดชเตเดณเดฟเดจเตเดฎเดฟเดฏเตฝ SVM เดฎเตเดกเดฒเดฟเตฝ `degree = 1` เดธเดเตเดเดฎเดพเดเตเดเดฟ เดเดฐเต เดฒเตเดจเดฟเดฏเตผ SVC เดธเตเดทเตเดเดฟเดเตเดเดพเด.


```r
# Make a linear SVC specification
svc_linear_spec <- svm_poly(degree = 1) %>% 
  set_engine("kernlab") %>% 
  set_mode("classification")

# Bundle specification and recipe into a worklow
svc_linear_wf <- workflow() %>% 
  add_recipe(cuisines_recipe) %>% 
  add_model(svc_linear_spec)

# Print out workflow
svc_linear_wf
```

เดเดชเตเดชเตเตพ เดจเดพเด เดชเตเดฐเตเดชเตเดฐเตเดธเดธเตเดธเดฟเดเดเต เดเดเตเดเดเตเดเดณเตเด เดฎเตเดกเตฝ เดธเตเดชเตเดธเดฟเดซเดฟเดเตเดเตเดทเดจเตเด เดเดฐเต *workflow* เดเดฏเดฟ เดชเดเตผเดคเตเดคเดฟเดฏเดคเดฟเดจเตเดถเตเดทเด, เดจเดพเด linear SVC เดเตเดฐเตเดฏเดฟเตป เดเตเดฏเตเดคเต เดซเดฒเดเตเดเตพ เดตเดฟเดฒเดฏเดฟเดฐเตเดคเตเดคเดพเตป เดฎเตเดจเตเดจเตเดเตเดเต เดชเตเดเดพเด. เดชเตเดฐเดเดเดจ เดฎเตเดเตเดฐเดฟเดเตโเดธเตเดเตพเดเตเดเดพเดฏเดฟ, เดจเดพเด เดตเดฟเดฒเดฏเดฟเดฐเตเดคเตเดคเตเดจเตเดจ เดเดฐเต เดฎเตเดเตเดฐเดฟเดเต เดธเตเดฑเตเดฑเต เดธเตเดทเตเดเดฟเดเตเดเดพเด: `accuracy`, `sensitivity`, `Positive Predicted Value` เดเดจเตเดจเดฟเดตเดฏเตเด `F Measure` เดเด.

> `augment()` เดจเตฝเดเดฟเดฏ เดกเดพเดฑเตเดฑเดฏเดฟเตฝ เดชเตเดฐเดตเดเดจเดเตเดเตพเดเตเดเดพเดฏเดฟ เดเตเดณเด(เดเตพ) เดเตเตผเดเตเดเตเด.


```r
# Train a linear SVC model
svc_linear_fit <- svc_linear_wf %>% 
  fit(data = cuisines_train)

# Create a metric set
eval_metrics <- metric_set(ppv, sens, accuracy, f_meas)


# Make predictions and Evaluate model performance
svc_linear_fit %>% 
  augment(new_data = cuisines_test) %>% 
  eval_metrics(truth = cuisine, estimate = .pred_class)
```

#### Support Vector Machine

เดธเดชเตเดชเตเตผเดเตเดเต เดตเตเดเตเดเตผ เดฎเตเดทเตเตป (SVM) เดเตเดฒเดพเดธเตเดเตพเดเตเดเดฟเดเดฏเดฟเดฒเต เดจเตเตบ-เดฒเตเดจเดฟเดฏเตผ เดฌเตเดฃเตเดเดฑเดฟ เดธเตเดตเตเดเดฐเดฟเดเตเดเดพเตป เดธเดชเตเดชเตเตผเดเตเดเต เดตเตเดเตเดเตผ เดเตเดฒเดพเดธเดฟเดซเดฏเดฑเดฟเดจเตเดฑเต เดตเดฟเดชเตเดฒเตเดเดฐเดฃเดฎเดพเดฃเต. เดธเดพเดฐเดพเดเดถเดคเตเดคเดฟเตฝ, SVMเดเตพ เดเตเดฒเดพเดธเตเดเตพเดเตเดเดฟเดเดฏเดฟเดฒเต เดจเตเตบเดฒเตเดจเดฟเดฏเตผ เดฌเดจเตเดงเดเตเดเตพเดเตเดเดพเดฏเดฟ เดซเตเดเตเดเตผ เดธเตเดชเตเดธเต เดตเดฒเตเดคเดพเดเตเดเดพเตป *kernel trick* เดเดชเดฏเตเดเดฟเดเตเดเตเดจเตเดจเต. SVMเดเตพ เดเดชเดฏเตเดเดฟเดเตเดเตเดจเตเดจ เดเดฐเต เดชเตเดฐเดถเดธเตเดคเดตเตเด เดเดคเตเดฏเดจเตเดคเด เดฒเดตเดเดฟเดเดตเตเดฎเดพเดฏ kernel เดซเดเดเตเดทเตป เดเดฃเต *Radial basis function.* เดจเดฎเตเดฎเตเดเต เดกเดพเดฑเตเดฑเดฏเดฟเตฝ เดเดคเต เดเดเตเดเดจเต เดชเตเดฐเดตเตผเดคเตเดคเดฟเดเตเดเตเด เดเดจเตเดจเต เดจเตเดเตเดเดพเด.


```r
set.seed(2056)

# Make an RBF SVM specification
svm_rbf_spec <- svm_rbf() %>% 
  set_engine("kernlab") %>% 
  set_mode("classification")

# Bundle specification and recipe into a worklow
svm_rbf_wf <- workflow() %>% 
  add_recipe(cuisines_recipe) %>% 
  add_model(svm_rbf_spec)


# Train an RBF model
svm_rbf_fit <- svm_rbf_wf %>% 
  fit(data = cuisines_train)


# Make predictions and Evaluate model performance
svm_rbf_fit %>% 
  augment(new_data = cuisines_test) %>% 
  eval_metrics(truth = cuisine, estimate = .pred_class)
```

*K*-เดเดฑเตเดฑเดตเตเด เดเดเตเดคเตเดคเตเดณเตเดณ เดจเตเดฌเตผ (KNN) เดเดฐเต เดเตฝเดเตเดฐเดฟเดคเดฎเดพเดฃเต, เดเดคเดฟเตฝ เดเดฐเต เดจเดฟเดฐเตเดเตเดทเดฃเดตเตเด เดฎเดฑเตเดฑเต เดจเดฟเดฐเตเดเตเดทเดฃเดเตเดเดณเตเดเตเดณเตเดณ เดเดคเดฟเดจเตเดฑเต *เดธเดพเดฆเตเดถเตเดฏเดคเตเดคเดฟเดจเตเดฑเต* เดเดเดฟเดธเตเดฅเดพเดจเดคเตเดคเดฟเตฝ เดชเตเดฐเดตเดเดฟเดเตเดเดชเตเดชเตเดเตเดจเตเดจเต.

เดจเดฎเตเดฎเตเดเต เดกเดพเดฑเตเดฑเดฏเดฟเตฝ เดเดจเตเดจเต เดซเดฟเดฑเตเดฑเต เดเตเดฏเตเดฏเดพเด.


```r
# Make a KNN specification
knn_spec <- nearest_neighbor() %>% 
  set_engine("kknn") %>% 
  set_mode("classification")

# Bundle recipe and model specification into a workflow
knn_wf <- workflow() %>% 
  add_recipe(cuisines_recipe) %>% 
  add_model(knn_spec)

# Train a boosted tree model
knn_wf_fit <- knn_wf %>% 
  fit(data = cuisines_train)


# Make predictions and Evaluate model performance
knn_wf_fit %>% 
  augment(new_data = cuisines_test) %>% 
  eval_metrics(truth = cuisine, estimate = .pred_class)
```

```r
# Make a random forest specification
rf_spec <- rand_forest() %>% 
  set_engine("ranger") %>% 
  set_mode("classification")

# Bundle recipe and model specification into a workflow
rf_wf <- workflow() %>% 
  add_recipe(cuisines_recipe) %>% 
  add_model(rf_spec)

# Train a random forest model
rf_wf_fit <- rf_wf %>% 
  fit(data = cuisines_train)


# Make predictions and Evaluate model performance
rf_wf_fit %>% 
  augment(new_data = cuisines_test) %>% 
  eval_metrics(truth = cuisine, estimate = .pred_class)
```

## 4. เดเดงเดฟเดเด - เดชเดฒ เดฎเตเดกเดฒเตเดเดณเตเด เดคเดพเดฐเดคเดฎเตเดฏเด เดเตเดฏเตเดฏเตฝ

เด เดฒเดพเดฌเดฟเตฝ เดจเดพเด เดตเดณเดฐเต เดฎเตเดกเดฒเตเดเตพ เดซเดฟเดฑเตเดฑเต เดเตเดฏเตเดคเดฟเดเตเดเตเดฃเตเดเต ๐. เดตเตเดฏเดคเตเดฏเดธเตเดค เดธเตเดฑเตเดฑเตเดเดณเตเดเต เดชเตเดฐเตเดชเตเดฐเตเดธเดธเดฑเตเดเดณเตเด/เดเดฅเดตเดพ เดฎเตเดกเตฝ เดธเตเดชเตเดธเดฟเดซเดฟเดเตเดเตเดทเดจเตเดเดณเตเด เดเดชเดฏเตเดเดฟเดเตเดเต เดจเดฟเดฐเดตเดงเดฟ เดตเตผเดเตเดเตโเดซเตเดฒเตเดเตพ เดธเตเดทเตเดเดฟเดเตเดเต เดเดฐเตเดจเตเดจเดฟเดจเตเดฑเตเดฏเตเด เดชเตเดฐเดเดเดจ เดฎเตเดเตเดฐเดฟเดเตเดเตเดเตพ เดเดฑเตเดฑเดคเตเดคเดตเดฃ เดเดฃเดเตเดเดพเดเตเดเตเดจเตเดจเดคเต เดเตเดทเตเดฃเดเดฐเดฎเต เดฌเตเดฆเตเดงเดฟเดฎเตเดเตเดเตเดณเตเดณเดคเต เดเดเดพเด.

เดเตเดฐเตเดฏเดฟเดจเดฟเดเดเต เดธเตเดฑเตเดฑเดฟเตฝ เดตเตผเดเตเดเตโเดซเตเดฒเตเดเดณเตเดเต เดเดฐเต เดฒเดฟเดธเตเดฑเตเดฑเต เดซเดฟเดฑเตเดฑเต เดเตเดฏเตเดคเต เดเตเดธเตเดฑเตเดฑเต เดธเตเดฑเตเดฑเดฟเดจเตเดฑเต เดเดเดฟเดธเตเดฅเดพเดจเดคเตเดคเดฟเตฝ เดชเตเดฐเดเดเดจ เดฎเตเดเตเดฐเดฟเดเตเดเตเดเตพ เดคเดฟเดฐเดฟเดเต เดจเตฝเดเตเดจเตเดจ เดเดฐเต เดซเดเดเตเดทเตป เดธเตเดทเตเดเดฟเดเตเดเต เดเดคเต เดชเดฐเดฟเดนเดฐเดฟเดเตเดเดพเดฎเต เดเดจเตเดจเต เดจเตเดเตเดเดพเด. เดฒเดฟเดธเตเดฑเตเดฑเดฟเดฒเต เดเดฐเต เดเดเดเดคเตเดคเดฟเดฒเตเด เดซเดเดเตเดทเดจเตเดเตพ เดชเตเดฐเดฏเตเดเดฟเดเตเดเดพเตป [purrr](https://purrr.tidyverse.org/) เดชเดพเดเตเดเตเดเดฟเตฝ เดจเดฟเดจเตเดจเตเดณเตเดณ `map()` เดเด `map_dfr()` เดเด เดเดชเดฏเตเดเดฟเดเตเดเดพเด.

> [`map()`](https://purrr.tidyverse.org/reference/map.html) เดซเดเดเตเดทเดจเตเดเตพ เดเตเดกเต เดเตเดเตเดคเตฝ เดธเดเดเตเดทเดฟเดชเตเดคเดตเตเด เดตเดพเดฏเดฟเดเตเดเดพเตป เดเดณเตเดชเตเดชเดตเตเดฎเดพเดเตเดเตเดจเตเดจ เดจเดฟเดฐเดตเดงเดฟ for เดฒเตเดชเตเดชเตเดเตพ เดฎเดพเดฑเตเดฑเดพเตป เดเดจเตเดตเดฆเดฟเดเตเดเตเดจเตเดจเต. [`map()`](https://purrr.tidyverse.org/reference/map.html) เดซเดเดเตเดทเดจเตเดเตพเดเตเดเตเดฑเดฟเดเตเดเต เดชเดเดฟเดเตเดเดพเตป เดเดฑเตเดฑเดตเตเด เดจเดฒเตเดฒ เดธเตเดฅเดฒเด R for data science-เดฒเต [iteration chapter](http://r4ds.had.co.nz/iteration.html) เดเดฃเต.


```r
set.seed(2056)

# Create a metric set
eval_metrics <- metric_set(ppv, sens, accuracy, f_meas)

# Define a function that returns performance metrics
compare_models <- function(workflow_list, train_set, test_set){
  
  suppressWarnings(
    # Fit each model to the train_set
    map(workflow_list, fit, data = train_set) %>% 
    # Make predictions on the test set
      map_dfr(augment, new_data = test_set, .id = "model") %>%
    # Select desired columns
      select(model, cuisine, .pred_class) %>% 
    # Evaluate model performance
      group_by(model) %>% 
      eval_metrics(truth = cuisine, estimate = .pred_class) %>% 
      ungroup()
  )
  
} # End of function
```

เดจเดฎเตเดเตเดเต เดจเดฎเตเดฎเตเดเต เดซเดเดเตเดทเตป เดตเดฟเดณเดฟเดเตเดเต เดฎเตเดกเดฒเตเดเตพเดเตเดเดฟเดเดฏเดฟเดฒเต เดเตเดคเตเดฏเดค เดคเดพเดฐเดคเดฎเตเดฏเด เดเตเดฏเตเดฏเดพเด.


```r
# Make a list of workflows
workflow_list <- list(
  "svc" = svc_linear_wf,
  "svm" = svm_rbf_wf,
  "knn" = knn_wf,
  "random_forest" = rf_wf,
  "xgboost" = boost_wf)

# Call the function
set.seed(2056)
perf_metrics <- compare_models(workflow_list = workflow_list, train_set = cuisines_train, test_set = cuisines_test)

# Print out performance metrics
perf_metrics %>% 
  group_by(.metric) %>% 
  arrange(desc(.estimate)) %>% 
  slice_head(n=7)

# Compare accuracy
perf_metrics %>% 
  filter(.metric == "accuracy") %>% 
  arrange(desc(.estimate))
```

## **๐เดเดฒเดเตเดเต**
