‡∞á‡∞¶‡∞ø ‡∞Æ‡±Å‡∞®‡±Å‡∞™‡∞ü‡∞ø ‡∞™‡∞æ‡∞†‡∞Ç‡∞≤‡±ã ‡∞Æ‡∞®‡∞Ç ‡∞ö‡±á‡∞∏‡∞ø‡∞®‡∞¶‡∞æ‡∞®‡∞ø‡∞ï‡∞ø ‡∞í‡∞ï ‡∞∏‡∞æ‡∞∞‡∞æ‡∞Ç‡∞∂‡∞Ç. ‡∞Æ‡∞®‡∞Ç ‡∞ï‡±ä‡∞Ç‡∞§ ‡∞°‡±á‡∞ü‡∞æ‡∞®‡±Å ‡∞ï‡∞ü‡±ç ‡∞ö‡±á‡∞∏‡∞ø, ‡∞µ‡∞ø‡∞∂‡±ç‡∞≤‡±á‡∞∑‡∞ø‡∞¶‡±ç‡∞¶‡∞æ‡∞Ç!


```r
# Load the core tidyverse and make it available in your current R session
library(tidyverse)

# Import the data into a tibble
df <- read_csv(file = "https://raw.githubusercontent.com/microsoft/ML-For-Beginners/main/5-Clustering/data/nigerian-songs.csv", show_col_types = FALSE)

# Narrow down to top 3 popular genres
nigerian_songs <- df %>% 
  # Concentrate on top 3 genres
  filter(artist_top_genre %in% c("afro dancehall", "afropop","nigerian pop")) %>% 
  # Remove unclassified observations
  filter(popularity != 0)



# Visualize popular genres using bar plots
theme_set(theme_light())
nigerian_songs %>%
  count(artist_top_genre) %>%
  ggplot(mapping = aes(x = artist_top_genre, y = n,
                       fill = artist_top_genre)) +
  geom_col(alpha = 0.8) +
  paletteer::scale_fill_paletteer_d("ggsci::category10_d3") +
  ggtitle("Top genres") +
  theme(plot.title = element_text(hjust = 0.5))
```

‡∞é‡∞≤‡∞æ ‡∞∏‡±Ü‡∞≤‡∞ï‡±ç‡∞∑‡∞®‡±ç ‡∞π‡±Ü‡∞≤‡±ç‡∞™‡∞∞‡±ç `where` ‡∞¶‡±Ä‡∞®‡±ç‡∞®‡∞ø ‡∞∏‡±Å‡∞≤‡∞≠‡∞Ç ‡∞ö‡±á‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡±ã ‡∞ö‡±Ç‡∞°‡∞Ç‡∞°‡∞ø üíÅ? ‡∞á‡∞≤‡∞æ‡∞Ç‡∞ü‡∞ø ‡∞á‡∞§‡∞∞ ‡∞´‡∞Ç‡∞ï‡±ç‡∞∑‡∞®‡±ç‡∞≤‡∞®‡±Å [‡∞á‡∞ï‡±ç‡∞ï‡∞°](https://tidyselect.r-lib.org/) ‡∞Ö‡∞®‡±ç‡∞µ‡±á‡∞∑‡∞ø‡∞Ç‡∞ö‡∞Ç‡∞°‡∞ø.

‡∞™‡±ç‡∞∞‡∞§‡∞ø ‡∞∏‡∞Ç‡∞ñ‡±ç‡∞Ø‡∞æ ‡∞≤‡∞ï‡±ç‡∞∑‡∞£‡∞æ‡∞®‡∞ø‡∞ï‡∞ø ‡∞¨‡∞æ‡∞ï‡±ç‡∞∏‡±ç‚Äå‡∞™‡±ç‡∞≤‡∞æ‡∞ü‡±ç ‡∞§‡∞Ø‡∞æ‡∞∞‡±Å ‡∞ö‡±á‡∞Ø‡∞¨‡±ã‡∞§‡±Å‡∞®‡±ç‡∞®‡∞æ‡∞Æ‡±Å ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å ‡∞≤‡±Ç‡∞™‡±ç‚Äå‡∞≤‡∞®‡±Å ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞ï‡±Å‡∞Ç‡∞°‡∞æ ‡∞â‡∞Ç‡∞°‡∞æ‡∞≤‡∞®‡±Å‡∞ï‡±Å‡∞Ç‡∞ü‡±Å‡∞®‡±ç‡∞®‡∞æ‡∞Æ‡±Å, ‡∞ï‡∞æ‡∞¨‡∞ü‡±ç‡∞ü‡∞ø ‡∞Æ‡∞® ‡∞°‡±á‡∞ü‡∞æ‡∞®‡±Å *‡∞¶‡±Ä‡∞∞‡±ç‡∞ò‡∞Æ‡±à‡∞®* ‡∞´‡∞æ‡∞∞‡±ç‡∞Æ‡∞æ‡∞ü‡±ç‚Äå‡∞≤‡±ã ‡∞™‡±Å‡∞®‡∞É‡∞∞‡±Ç‡∞™‡∞ï‡∞≤‡±ç‡∞™‡∞® ‡∞ö‡±á‡∞¶‡±ç‡∞¶‡∞æ‡∞Ç, ‡∞á‡∞¶‡∞ø ‡∞Æ‡∞®‡∞ï‡±Å `facets` - ‡∞™‡±ç‡∞∞‡∞§‡∞ø ‡∞â‡∞™‡∞∏‡∞Æ‡∞ø‡∞§‡∞ø ‡∞°‡±á‡∞ü‡∞æ‡∞®‡±Å ‡∞™‡±ç‡∞∞‡∞¶‡∞∞‡±ç‡∞∂‡∞ø‡∞Ç‡∞ö‡±á ‡∞â‡∞™‡∞ó‡±ç‡∞∞‡∞æ‡∞´‡±ç‚Äå‡∞≤‡∞®‡±Å ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡±Å‡∞ï‡±Å‡∞®‡±á ‡∞Ö‡∞µ‡∞ï‡∞æ‡∞∂‡∞Ç ‡∞á‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø.


```r
# Pivot data from wide to long
df_numeric_long <- df_numeric %>% 
  pivot_longer(!artist_top_genre, names_to = "feature_names", values_to = "values") 

# Print out data
df_numeric_long %>% 
  slice_head(n = 15)
```

‡∞á‡∞Ç‡∞ï‡∞æ ‡∞ö‡∞æ‡∞≤‡∞æ ‡∞™‡±ä‡∞°‡∞µ‡±Å‡∞ó‡∞æ! ‡∞á‡∞™‡±ç‡∞™‡±Å‡∞°‡±Å ‡∞ï‡±ä‡∞Ç‡∞§ `ggplots` ‡∞∏‡∞Æ‡∞Ø‡∞Ç! ‡∞ï‡∞æ‡∞¨‡∞ü‡±ç‡∞ü‡∞ø ‡∞è `geom` ‡∞®‡±Å ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞¨‡±ã‡∞§‡±Å‡∞®‡±ç‡∞®‡∞æ‡∞Ç?


```r
# Make a box plot
df_numeric_long %>% 
  ggplot(mapping = aes(x = feature_names, y = values, fill = feature_names)) +
  geom_boxplot() +
  facet_wrap(~ feature_names, ncol = 4, scales = "free") +
  theme(legend.position = "none")
```

‡∞á‡∞™‡±ç‡∞™‡±Å‡∞°‡±Å ‡∞Æ‡∞®‡∞Ç ‡∞à ‡∞°‡±á‡∞ü‡∞æ ‡∞ï‡±ä‡∞Ç‡∞ö‡±Ü‡∞Ç ‡∞∂‡∞¨‡±ç‡∞¶‡∞Ç‡∞ó‡∞æ ‡∞â‡∞Ç‡∞¶‡∞®‡∞ø ‡∞ö‡±Ç‡∞°‡∞µ‡∞ö‡±ç‡∞ö‡±Å: ‡∞™‡±ç‡∞∞‡∞§‡∞ø ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‚Äå‡∞®‡±Å ‡∞¨‡∞æ‡∞ï‡±ç‡∞∏‡±ç‚Äå‡∞™‡±ç‡∞≤‡∞æ‡∞ü‡±ç‚Äå‡∞ó‡∞æ ‡∞™‡∞∞‡∞ø‡∞∂‡±Ä‡∞≤‡∞ø‡∞Ç‡∞ö‡∞°‡∞Ç ‡∞¶‡±ç‡∞µ‡∞æ‡∞∞‡∞æ, ‡∞Æ‡±Ä‡∞∞‡±Å ‡∞Ö‡∞µ‡±Å‡∞ü‡±ç‡∞≤‡∞Ø‡∞∞‡±ç‡∞≤‡∞®‡±Å ‡∞ö‡±Ç‡∞°‡∞µ‡∞ö‡±ç‡∞ö‡±Å. ‡∞Æ‡±Ä‡∞∞‡±Å ‡∞°‡±á‡∞ü‡∞æ‡∞∏‡±Ü‡∞ü‡±ç‚Äå‡∞®‡±Å ‡∞ó‡∞Æ‡∞®‡∞ø‡∞Ç‡∞ö‡∞ø ‡∞à ‡∞Ö‡∞µ‡±Å‡∞ü‡±ç‡∞≤‡∞Ø‡∞∞‡±ç‡∞≤‡∞®‡±Å ‡∞§‡±ä‡∞≤‡∞ó‡∞ø‡∞Ç‡∞ö‡∞µ‡∞ö‡±ç‡∞ö‡±Å, ‡∞ï‡∞æ‡∞®‡±Ä ‡∞Ö‡∞¶‡∞ø ‡∞°‡±á‡∞ü‡∞æ‡∞®‡±Å ‡∞ö‡∞æ‡∞≤‡∞æ ‡∞§‡∞ï‡±ç‡∞ï‡±Å‡∞µ‡∞ó‡∞æ ‡∞ö‡±á‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø.

‡∞™‡±ç‡∞∞‡∞∏‡±ç‡∞§‡±Å‡∞§‡∞Ç, ‡∞Æ‡∞®‡∞Ç ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡∞ø‡∞Ç‡∞ó‡±ç ‡∞µ‡±ç‡∞Ø‡∞æ‡∞Ø‡∞æ‡∞Æ‡∞Ç ‡∞ï‡±ã‡∞∏‡∞Ç ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞¨‡±ã‡∞Ø‡±á ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‚Äå‡∞≤‡∞®‡±Å ‡∞é‡∞Ç‡∞ö‡±Å‡∞ï‡±Å‡∞Ç‡∞¶‡∞æ‡∞Ç. ‡∞∏‡∞Æ‡∞æ‡∞® ‡∞™‡∞∞‡∞ø‡∞ß‡±Å‡∞≤‡±Å‡∞®‡±ç‡∞® ‡∞∏‡∞Ç‡∞ñ‡±ç‡∞Ø‡∞æ‡∞§‡±ç‡∞Æ‡∞ï ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‚Äå‡∞≤‡∞®‡±Å ‡∞é‡∞Ç‡∞ö‡±Å‡∞ï‡±Å‡∞Ç‡∞¶‡∞æ‡∞Ç. ‡∞Æ‡∞®‡∞Ç `artist_top_genre` ‡∞®‡±Å ‡∞∏‡∞Ç‡∞ñ‡±ç‡∞Ø‡∞æ‡∞§‡±ç‡∞Æ‡∞ï‡∞Ç‡∞ó‡∞æ ‡∞é‡∞®‡±ç‚Äå‡∞ï‡±ã‡∞°‡±ç ‡∞ö‡±á‡∞Ø‡∞µ‡∞ö‡±ç‡∞ö‡±Å ‡∞ï‡∞æ‡∞®‡±Ä ‡∞á‡∞™‡±ç‡∞™‡∞ü‡∞ø‡∞ï‡±Ä ‡∞¶‡∞æ‡∞®‡±ç‡∞®‡∞ø ‡∞µ‡∞¶‡∞ø‡∞≤‡±á‡∞∏‡±ç‡∞§‡∞æ‡∞Æ‡±Å.


```r
# Select variables with similar ranges
df_numeric_select <- df_numeric %>% 
  select(popularity, danceability, acousticness, loudness, energy) 

# Normalize data
# df_numeric_select <- scale(df_numeric_select)
```

## 3. R ‡∞≤‡±ã k-means ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡∞ø‡∞Ç‡∞ó‡±ç ‡∞≤‡±Ü‡∞ï‡±ç‡∞ï‡∞ø‡∞Ç‡∞ö‡∞°‡∞Ç

‡∞Æ‡∞®‡∞Ç R ‡∞≤‡±ã ‡∞¨‡∞ø‡∞≤‡±ç‡∞ü‡±ç-‡∞á‡∞®‡±ç `kmeans` ‡∞´‡∞Ç‡∞ï‡±ç‡∞∑‡∞®‡±ç ‡∞§‡±ã k-means ‡∞≤‡±Ü‡∞ï‡±ç‡∞ï‡∞ø‡∞Ç‡∞ö‡∞µ‡∞ö‡±ç‡∞ö‡±Å, ‡∞ö‡±Ç‡∞°‡∞Ç‡∞°‡∞ø `help("kmeans()")`. `kmeans()` ‡∞´‡∞Ç‡∞ï‡±ç‡∞∑‡∞®‡±ç ‡∞™‡±ç‡∞∞‡∞æ‡∞•‡∞Æ‡∞ø‡∞ï ‡∞Ü‡∞∞‡±ç‡∞ó‡±ç‡∞Ø‡±Å‡∞Æ‡±Ü‡∞Ç‡∞ü‡±ç ‡∞ó‡∞æ ‡∞Ö‡∞®‡±ç‡∞®‡∞ø ‡∞®‡±ç‡∞Ø‡±Ç‡∞Æ‡∞∞‡∞ø‡∞ï‡±ç ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‡∞∏‡±ç ‡∞â‡∞®‡±ç‡∞® ‡∞°‡±á‡∞ü‡∞æ ‡∞´‡±ç‡∞∞‡±á‡∞Æ‡±ç ‡∞®‡±Å ‡∞Ö‡∞Ç‡∞ó‡±Ä‡∞ï‡∞∞‡∞ø‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø.

k-means ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡∞ø‡∞Ç‡∞ó‡±ç ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡±á ‡∞Æ‡±ä‡∞¶‡∞ü‡∞ø ‡∞¶‡∞∂ ‡∞ö‡∞ø‡∞µ‡∞∞‡∞ø ‡∞™‡∞∞‡∞ø‡∞∑‡±ç‡∞ï‡∞æ‡∞∞‡∞Ç‡∞≤‡±ã ‡∞â‡∞§‡±ç‡∞™‡∞§‡±ç‡∞§‡∞ø ‡∞ö‡±á‡∞Ø‡∞¨‡∞°‡±á ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç‡∞≤ ‡∞∏‡∞Ç‡∞ñ‡±ç‡∞Ø (k) ‡∞®‡±Å ‡∞®‡∞ø‡∞∞‡±ç‡∞¶‡±á‡∞∂‡∞ø‡∞Ç‡∞ö‡∞°‡∞Ç. ‡∞Æ‡∞®‡∞Ç ‡∞°‡±á‡∞ü‡∞æ‡∞∏‡±Ü‡∞ü‡±ç ‡∞®‡±Å‡∞Ç‡∞°‡∞ø 3 ‡∞™‡∞æ‡∞ü ‡∞ú‡∞æ‡∞®‡∞∞‡±ç‡∞≤‡∞®‡±Å ‡∞§‡±Ä‡∞∏‡±Å‡∞ï‡±Å‡∞®‡±ç‡∞®‡∞æ‡∞Æ‡∞®‡∞ø ‡∞§‡±Ü‡∞≤‡±Å‡∞∏‡±Å, ‡∞ï‡∞æ‡∞¨‡∞ü‡±ç‡∞ü‡∞ø 3 ‡∞®‡±Å ‡∞™‡±ç‡∞∞‡∞Ø‡∞§‡±ç‡∞®‡∞ø‡∞¶‡±ç‡∞¶‡∞æ‡∞Ç:


```r
set.seed(2056)
# Kmeans clustering for 3 clusters
kclust <- kmeans(
  df_numeric_select,
  # Specify the number of clusters
  centers = 3,
  # How many random initial configurations
  nstart = 25
)

# Display clustering object
kclust
```

kmeans ‡∞Ü‡∞¨‡±ç‡∞ú‡±Ü‡∞ï‡±ç‡∞ü‡±ç‚Äå‡∞≤‡±ã `help("kmeans()")` ‡∞≤‡±ã ‡∞¨‡∞æ‡∞ó‡∞æ ‡∞µ‡∞ø‡∞µ‡∞∞‡∞ø‡∞Ç‡∞ö‡∞¨‡∞°‡∞ø‡∞® ‡∞Ö‡∞®‡±á‡∞ï ‡∞∏‡∞Æ‡∞æ‡∞ö‡∞æ‡∞∞‡∞Ç ‡∞≠‡∞æ‡∞ó‡∞æ‡∞≤‡±Å ‡∞â‡∞Ç‡∞ü‡∞æ‡∞Ø‡∞ø. ‡∞á‡∞™‡±ç‡∞™‡∞ü‡∞ø‡∞ï‡±Ä, ‡∞ï‡±ä‡∞®‡±ç‡∞®‡∞ø ‡∞µ‡∞ø‡∞∑‡∞Ø‡∞æ‡∞≤‡∞™‡±à ‡∞¶‡±É‡∞∑‡±ç‡∞ü‡∞ø ‡∞™‡±Ü‡∞ü‡±ç‡∞ü‡±Å‡∞ï‡±Å‡∞Ç‡∞¶‡∞æ‡∞Ç. ‡∞°‡±á‡∞ü‡∞æ 65, 110, 111 ‡∞™‡∞∞‡∞ø‡∞Æ‡∞æ‡∞£‡∞æ‡∞≤ 3 ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç‡∞≤‡±Å‡∞ó‡∞æ ‡∞µ‡∞ø‡∞≠‡∞ú‡∞ø‡∞Ç‡∞ö‡∞¨‡∞°‡∞ø‡∞Ç‡∞¶‡∞®‡∞ø ‡∞Æ‡∞®‡∞Ç ‡∞ö‡±Ç‡∞∏‡±ç‡∞§‡±Å‡∞®‡±ç‡∞®‡∞æ‡∞Æ‡±Å. ‡∞Ö‡∞µ‡±Å‡∞ü‡±ç‡∞™‡±Å‡∞ü‡±ç‚Äå‡∞≤‡±ã 5 ‡∞µ‡±á‡∞∞‡∞ø‡∞Ø‡∞¨‡±Å‡∞≤‡±ç‡∞∏‡±ç ‡∞Ö‡∞Ç‡∞§‡∞ü‡∞æ 3 ‡∞ó‡±ç‡∞∞‡±Ç‡∞™‡±Å‡∞≤ ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç ‡∞∏‡±Ü‡∞Ç‡∞ü‡∞∞‡±ç‡∞≤‡±Å (‡∞∏‡∞ó‡∞ü‡±Å‡∞≤‡±Å) ‡∞ï‡±Ç‡∞°‡∞æ ‡∞â‡∞Ç‡∞ü‡∞æ‡∞Ø‡∞ø.

‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡∞ø‡∞Ç‡∞ó‡±ç ‡∞µ‡±Ü‡∞ï‡±ç‡∞ü‡∞∞‡±ç ‡∞Ö‡∞®‡±á‡∞¶‡∞ø ‡∞™‡±ç‡∞∞‡∞§‡∞ø ‡∞™‡∞∞‡∞ø‡∞∂‡±Ä‡∞≤‡∞®‡∞ï‡±Å ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç ‡∞ï‡±á‡∞ü‡∞æ‡∞Ø‡∞ø‡∞Ç‡∞™‡±Å. ‡∞Ö‡∞∏‡∞≤‡±Å ‡∞°‡±á‡∞ü‡∞æ ‡∞∏‡±Ü‡∞ü‡±ç‚Äå‡∞ï‡±Å ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç ‡∞ï‡±á‡∞ü‡∞æ‡∞Ø‡∞ø‡∞Ç‡∞™‡±Å‡∞®‡±Å ‡∞ú‡±ã‡∞°‡∞ø‡∞Ç‡∞ö‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø `augment` ‡∞´‡∞Ç‡∞ï‡±ç‡∞∑‡∞®‡±ç‚Äå‡∞®‡±Å ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞¶‡±ç‡∞¶‡∞æ‡∞Ç.


```r
# Add predicted cluster assignment to data set
augment(kclust, df_numeric_select) %>% 
  relocate(.cluster) %>% 
  slice_head(n = 10)
```

‡∞∏‡∞∞‡±á, ‡∞Æ‡∞®‡∞Ç ‡∞á‡∞™‡±ç‡∞™‡±Å‡∞°‡±Å ‡∞Æ‡∞® ‡∞°‡±á‡∞ü‡∞æ ‡∞∏‡±Ü‡∞ü‡±ç‚Äå‡∞®‡±Å 3 ‡∞ó‡±ç‡∞∞‡±Ç‡∞™‡±Å‡∞≤ ‡∞∏‡±Ü‡∞ü‡±ç‚Äå‡∞ó‡∞æ ‡∞µ‡∞ø‡∞≠‡∞ú‡∞ø‡∞Ç‡∞ö‡∞æ‡∞Æ‡±Å. ‡∞ï‡∞æ‡∞¨‡∞ü‡±ç‡∞ü‡∞ø, ‡∞Æ‡∞® ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡∞ø‡∞Ç‡∞ó‡±ç ‡∞é‡∞Ç‡∞§ ‡∞Æ‡∞Ç‡∞ö‡∞ø‡∞¶‡±ã ‡∞ö‡±Ç‡∞¶‡±ç‡∞¶‡∞æ‡∞Ç ü§∑? ‡∞Æ‡∞®‡∞Ç `Silhouette score` ‡∞®‡∞ø ‡∞ö‡±Ç‡∞¶‡±ç‡∞¶‡∞æ‡∞Ç

### **Silhouette score**

[Silhouette ‡∞µ‡∞ø‡∞∂‡±ç‡∞≤‡±á‡∞∑‡∞£](https://en.wikipedia.org/wiki/Silhouette_(clustering)) ‡∞´‡∞≤‡∞ø‡∞§‡∞Ç‡∞ó‡∞æ ‡∞µ‡∞ö‡±ç‡∞ö‡∞ø‡∞® ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç‡∞≤ ‡∞Æ‡∞ß‡±ç‡∞Ø ‡∞µ‡∞ø‡∞°‡∞ø‡∞™‡±ã‡∞µ‡∞°‡∞Ç ‡∞¶‡±Ç‡∞∞‡∞æ‡∞®‡±ç‡∞®‡∞ø ‡∞Ö‡∞ß‡±ç‡∞Ø‡∞Ø‡∞®‡∞Ç ‡∞ö‡±á‡∞Ø‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞µ‡∞ö‡±ç‡∞ö‡±Å. ‡∞à ‡∞∏‡±ç‡∞ï‡±ã‡∞∞‡±Å -1 ‡∞®‡±Å‡∞Ç‡∞°‡∞ø 1 ‡∞µ‡∞∞‡∞ï‡±Å ‡∞Æ‡∞æ‡∞∞‡±Å‡∞§‡±Å‡∞Ç‡∞¶‡∞ø, ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å ‡∞∏‡±ç‡∞ï‡±ã‡∞∞‡±Å 1‡∞ï‡∞ø ‡∞¶‡∞ó‡±ç‡∞ó‡∞∞‡∞ó‡∞æ ‡∞â‡∞Ç‡∞ü‡±á, ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç ‡∞∏‡∞æ‡∞Ç‡∞¶‡±ç‡∞∞‡∞Ç‡∞ó‡∞æ ‡∞â‡∞Ç‡∞ü‡±Å‡∞Ç‡∞¶‡∞ø ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å ‡∞á‡∞§‡∞∞ ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç‡∞≤ ‡∞®‡±Å‡∞Ç‡∞°‡∞ø ‡∞¨‡∞æ‡∞ó‡∞æ ‡∞µ‡±á‡∞∞‡±Å‡∞ó‡∞æ ‡∞â‡∞Ç‡∞ü‡±Å‡∞Ç‡∞¶‡∞ø. 0‡∞ï‡∞ø ‡∞¶‡∞ó‡±ç‡∞ó‡∞∞‡∞ó‡∞æ ‡∞â‡∞®‡±ç‡∞® ‡∞µ‡∞ø‡∞≤‡±Å‡∞µ ‡∞∏‡∞Æ‡±Ä‡∞™ ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç‡∞≤‡∞§‡±ã decision boundary‡∞ï‡∞ø ‡∞ö‡∞æ‡∞≤‡∞æ ‡∞¶‡∞ó‡±ç‡∞ó‡∞∞‡∞ó‡∞æ ‡∞â‡∞®‡±ç‡∞® ‡∞®‡∞Æ‡±Ç‡∞®‡∞æ‡∞≤‡∞®‡±Å ‡∞∏‡±Ç‡∞ö‡∞ø‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø.[source](https://dzone.com/articles/kmeans-silhouette-score-explained-with-python-exam).

‡∞∏‡∞ó‡∞ü‡±Å silhouette ‡∞™‡∞¶‡±ç‡∞ß‡∞§‡∞ø ‡∞µ‡∞ø‡∞µ‡∞ø‡∞ß *k* ‡∞µ‡∞ø‡∞≤‡±Å‡∞µ‡∞≤ ‡∞ï‡±ã‡∞∏‡∞Ç ‡∞™‡∞∞‡∞ø‡∞∂‡±Ä‡∞≤‡∞®‡∞≤ ‡∞∏‡∞ó‡∞ü‡±Å silhouette‡∞®‡∞ø ‡∞≤‡±Ü‡∞ï‡±ç‡∞ï‡∞ø‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø. ‡∞í‡∞ï ‡∞Ö‡∞ß‡∞ø‡∞ï ‡∞∏‡∞ó‡∞ü‡±Å silhouette ‡∞∏‡±ç‡∞ï‡±ã‡∞∞‡±Å ‡∞Æ‡∞Ç‡∞ö‡∞ø ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡∞ø‡∞Ç‡∞ó‡±ç‚Äå‡∞®‡±Å ‡∞∏‡±Ç‡∞ö‡∞ø‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø.

‡∞∏‡∞ó‡∞ü‡±Å silhouette ‡∞µ‡±Ü‡∞°‡∞≤‡±ç‡∞™‡±Å‡∞®‡±Å ‡∞≤‡±Ü‡∞ï‡±ç‡∞ï‡∞ø‡∞Ç‡∞ö‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø cluster ‡∞™‡±ç‡∞Ø‡∞æ‡∞ï‡±á‡∞ú‡±Ä‡∞≤‡±ã‡∞®‡∞ø `silhouette` ‡∞´‡∞Ç‡∞ï‡±ç‡∞∑‡∞®‡±ç ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞µ‡∞ö‡±ç‡∞ö‡±Å.

> silhouette‡∞®‡∞ø ‡∞è‡∞¶‡±à‡∞®‡∞æ [distance](https://en.wikipedia.org/wiki/Distance "Distance") ‡∞Æ‡±Ü‡∞ü‡±ç‡∞∞‡∞ø‡∞ï‡±ç‚Äå‡∞§‡±ã ‡∞≤‡±Ü‡∞ï‡±ç‡∞ï‡∞ø‡∞Ç‡∞ö‡∞µ‡∞ö‡±ç‡∞ö‡±Å, ‡∞â‡∞¶‡∞æ‡∞π‡∞∞‡∞£‡∞ï‡±Å ‡∞Æ‡∞®‡∞Ç [‡∞Æ‡±Å‡∞®‡±Å‡∞™‡∞ü‡∞ø ‡∞™‡∞æ‡∞†‡∞Ç‡∞≤‡±ã](https://github.com/microsoft/ML-For-Beginners/blob/main/5-Clustering/1-Visualize/solution/R/lesson_14-R.ipynb) ‡∞ö‡∞∞‡±ç‡∞ö‡∞ø‡∞Ç‡∞ö‡∞ø‡∞® [Euclidean distance](https://en.wikipedia.org/wiki/Euclidean_distance "Euclidean distance") ‡∞≤‡±á‡∞¶‡∞æ [Manhattan distance](https://en.wikipedia.org/wiki/Manhattan_distance "Manhattan distance") ‡∞µ‡∞Ç‡∞ü‡∞ø.


```r
# Load cluster package
library(cluster)

# Compute average silhouette score
ss <- silhouette(kclust$cluster,
                 # Compute euclidean distance
                 dist = dist(df_numeric_select))
mean(ss[, 3])
```

‡∞Æ‡∞æ ‡∞∏‡±ç‡∞ï‡±ã‡∞∞‡±Å **.549** ‡∞â‡∞Ç‡∞¶‡∞ø, ‡∞ï‡∞æ‡∞¨‡∞ü‡±ç‡∞ü‡∞ø ‡∞Æ‡∞ß‡±ç‡∞Ø‡∞≤‡±ã‡∞®‡±á ‡∞â‡∞Ç‡∞¶‡∞ø. ‡∞á‡∞¶‡∞ø ‡∞Æ‡∞æ ‡∞°‡±á‡∞ü‡∞æ ‡∞à ‡∞∞‡∞ï‡∞Æ‡±à‡∞® ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡∞ø‡∞Ç‡∞ó‡±ç‚Äå‡∞ï‡±Å ‡∞™‡±ç‡∞∞‡∞§‡±ç‡∞Ø‡±á‡∞ï‡∞Ç‡∞ó‡∞æ ‡∞Ö‡∞®‡±Å‡∞ï‡±Ç‡∞≤‡∞Ç‡∞ó‡∞æ ‡∞≤‡±á‡∞®‡∞ü‡±ç‡∞ü‡±Å‡∞ó‡∞æ ‡∞∏‡±Ç‡∞ö‡∞ø‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø. ‡∞Æ‡∞®‡∞Ç ‡∞à ‡∞Ö‡∞®‡±Å‡∞Æ‡∞æ‡∞®‡∞æ‡∞®‡±ç‡∞®‡∞ø ‡∞¶‡±É‡∞∂‡±ç‡∞Ø‡∞∞‡±Ç‡∞™‡∞Ç‡∞≤‡±ã ‡∞®‡∞ø‡∞∞‡±ç‡∞ß‡∞æ‡∞∞‡∞ø‡∞Ç‡∞ö‡∞ó‡∞≤‡∞Æ‡∞æ ‡∞ö‡±Ç‡∞¶‡±ç‡∞¶‡∞æ‡∞Ç. [factoextra ‡∞™‡±ç‡∞Ø‡∞æ‡∞ï‡±á‡∞ú‡±ç](https://rpkgs.datanovia.com/factoextra/index.html) ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡∞ø‡∞Ç‡∞ó‡±ç‚Äå‡∞®‡±Å ‡∞¶‡±É‡∞∂‡±ç‡∞Ø‡±Ä‡∞ï‡∞∞‡∞ø‡∞Ç‡∞ö‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø (`fviz_cluster()`) ‡∞´‡∞Ç‡∞ï‡±ç‡∞∑‡∞®‡±ç‡∞≤‡∞®‡±Å ‡∞Ö‡∞Ç‡∞¶‡∞ø‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø.


```r
library(factoextra)

# Visualize clustering results
fviz_cluster(kclust, df_numeric_select)
```

## 4. ‡∞â‡∞§‡±ç‡∞§‡∞Æ ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç‡∞≤‡∞®‡±Å ‡∞®‡∞ø‡∞∞‡±ç‡∞£‡∞Ø‡∞ø‡∞Ç‡∞ö‡∞°‡∞Ç

K-Means ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡∞ø‡∞Ç‡∞ó‡±ç‚Äå‡∞≤‡±ã ‡∞§‡∞∞‡∞ö‡±Å‡∞ó‡∞æ ‡∞é‡∞¶‡±Å‡∞∞‡∞Ø‡±ç‡∞Ø‡±á ‡∞í‡∞ï ‡∞™‡±ç‡∞∞‡∞æ‡∞•‡∞Æ‡∞ø‡∞ï ‡∞™‡±ç‡∞∞‡∞∂‡±ç‡∞® ‡∞á‡∞¶‡∞ø - ‡∞§‡±Ü‡∞≤‡∞ø‡∞Ø‡∞®‡∞ø ‡∞ï‡±ç‡∞≤‡∞æ‡∞∏‡±ç ‡∞≤‡±á‡∞¨‡±Å‡∞≤‡±ç‡∞∏‡±ç ‡∞≤‡±á‡∞ï‡±Å‡∞Ç‡∞°‡∞æ, ‡∞Æ‡±Ä‡∞∞‡±Å ‡∞Æ‡±Ä ‡∞°‡±á‡∞ü‡∞æ‡∞®‡±Å ‡∞é‡∞®‡±ç‡∞®‡∞ø ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç‡∞≤‡±Å‡∞ó‡∞æ ‡∞µ‡∞ø‡∞°‡∞ó‡±ä‡∞ü‡±ç‡∞ü‡∞æ‡∞≤‡±ã ‡∞é‡∞≤‡∞æ ‡∞§‡±Ü‡∞≤‡±Å‡∞∏‡±Å‡∞ï‡±Å‡∞Ç‡∞ü‡∞æ‡∞∞‡±Å?

‡∞Æ‡∞®‡∞Ç ‡∞§‡±Ü‡∞≤‡±Å‡∞∏‡±Å‡∞ï‡±ã‡∞µ‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø ‡∞™‡±ç‡∞∞‡∞Ø‡∞§‡±ç‡∞®‡∞ø‡∞Ç‡∞ö‡∞ó‡∞≤ ‡∞í‡∞ï ‡∞Æ‡∞æ‡∞∞‡±ç‡∞ó‡∞Ç ‡∞°‡±á‡∞ü‡∞æ ‡∞®‡∞Æ‡±Ç‡∞®‡∞æ‡∞®‡±Å ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞ø `‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç‡∞≤ ‡∞∏‡∞Ç‡∞ñ‡±ç‡∞Ø ‡∞™‡±Ü‡∞∞‡±Å‡∞ó‡±Å‡∞§‡±Ç` (‡∞â‡∞¶‡∞æ: 1-10 ‡∞µ‡∞∞‡∞ï‡±Å) ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡∞ø‡∞Ç‡∞ó‡±ç ‡∞Æ‡±ã‡∞°‡∞≤‡±ç‡∞∏‡±ç ‡∞∏‡∞ø‡∞∞‡±Ä‡∞∏‡±ç‚Äå‡∞®‡±Å ‡∞∏‡±É‡∞∑‡±ç‡∞ü‡∞ø‡∞Ç‡∞ö‡∞°‡∞Ç, ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å **‡∞∏‡∞ø‡∞≤‡±ç‡∞π‡±å‡∞ü‡±ç ‡∞∏‡±ç‡∞ï‡±ã‡∞∞‡±ç** ‡∞µ‡∞Ç‡∞ü‡∞ø ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡∞ø‡∞Ç‡∞ó‡±ç ‡∞Æ‡±Ü‡∞ü‡±ç‡∞∞‡∞ø‡∞ï‡±ç‡∞∏‡±ç‚Äå‡∞®‡±Å ‡∞Æ‡±Ç‡∞≤‡±ç‡∞Ø‡∞æ‡∞Ç‡∞ï‡∞®‡∞Ç ‡∞ö‡±á‡∞Ø‡∞°‡∞Ç.

‡∞µ‡∞ø‡∞≠‡∞ø‡∞®‡±ç‡∞® *k* ‡∞µ‡∞ø‡∞≤‡±Å‡∞µ‡∞≤ ‡∞ï‡±ã‡∞∏‡∞Ç ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡∞ø‡∞Ç‡∞ó‡±ç ‡∞Ö‡∞≤‡±ç‡∞ó‡±ã‡∞∞‡∞ø‡∞•‡∞Æ‡±ç‚Äå‡∞®‡±Å ‡∞ó‡∞£‡∞ø‡∞Ç‡∞ö‡∞ø, **Within Cluster Sum of Squares** (WCSS) ‡∞®‡±Å ‡∞Æ‡±Ç‡∞≤‡±ç‡∞Ø‡∞æ‡∞Ç‡∞ï‡∞®‡∞Ç ‡∞ö‡±á‡∞Ø‡∞°‡∞Ç ‡∞¶‡±ç‡∞µ‡∞æ‡∞∞‡∞æ ‡∞â‡∞§‡±ç‡∞§‡∞Æ ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç‡∞≤ ‡∞∏‡∞Ç‡∞ñ‡±ç‡∞Ø‡∞®‡±Å ‡∞®‡∞ø‡∞∞‡±ç‡∞£‡∞Ø‡∞ø‡∞¶‡±ç‡∞¶‡∞æ‡∞Ç. ‡∞Æ‡±ä‡∞§‡±ç‡∞§‡∞Ç ‡∞µ‡∞ø‡∞§‡∞ø‡∞®‡±ç-‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç ‡∞∏‡∞Æ‡±ç ‡∞Ü‡∞´‡±ç ‡∞∏‡±ç‡∞ï‡±ç‡∞µ‡±á‡∞∞‡±ç (WCSS) ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡∞ø‡∞Ç‡∞ó‡±ç ‡∞Ø‡±ä‡∞ï‡±ç‡∞ï ‡∞∏‡∞®‡±ç‡∞®‡∞ø‡∞π‡∞ø‡∞§‡∞§‡∞®‡±Å ‡∞ï‡±ä‡∞≤‡±Å‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å ‡∞Æ‡∞®‡∞Ç ‡∞¶‡±Ä‡∞®‡±ç‡∞®‡∞ø œåœÉŒø ‡∞§‡∞ï‡±ç‡∞ï‡±Å‡∞µ‡∞ó‡∞æ ‡∞â‡∞Ç‡∞°‡∞æ‡∞≤‡∞®‡∞ø ‡∞ï‡±ã‡∞∞‡±Å‡∞ï‡±Å‡∞Ç‡∞ü‡∞æ‡∞Æ‡±Å, ‡∞§‡∞ï‡±ç‡∞ï‡±Å‡∞µ ‡∞µ‡∞ø‡∞≤‡±Å‡∞µ‡∞≤‡±Å ‡∞°‡±á‡∞ü‡∞æ ‡∞™‡∞æ‡∞Ø‡∞ø‡∞Ç‡∞ü‡±ç‡∞≤‡±Å ‡∞¶‡∞ó‡±ç‡∞ó‡∞∞‡∞ó‡∞æ ‡∞â‡∞®‡±ç‡∞®‡∞ü‡±ç‡∞≤‡±Å ‡∞∏‡±Ç‡∞ö‡∞ø‡∞∏‡±ç‡∞§‡∞æ‡∞Ø‡∞ø.

1 ‡∞®‡±Å‡∞Ç‡∞°‡∞ø 10 ‡∞µ‡∞∞‡∞ï‡±Å `k` ‡∞Ø‡±ä‡∞ï‡±ç‡∞ï ‡∞µ‡∞ø‡∞µ‡∞ø‡∞ß ‡∞é‡∞Ç‡∞™‡∞ø‡∞ï‡∞≤ ‡∞™‡±ç‡∞∞‡∞≠‡∞æ‡∞µ‡∞æ‡∞®‡±ç‡∞®‡∞ø ‡∞à ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡∞ø‡∞Ç‡∞ó‡±ç‚Äå‡∞™‡±à ‡∞™‡∞∞‡∞ø‡∞∂‡±Ä‡∞≤‡∞ø‡∞¶‡±ç‡∞¶‡∞æ‡∞Ç.


```r
# Create a series of clustering models
kclusts <- tibble(k = 1:10) %>% 
  # Perform kmeans clustering for 1,2,3 ... ,10 clusters
  mutate(model = map(k, ~ kmeans(df_numeric_select, centers = .x, nstart = 25)),
  # Farm out clustering metrics eg WCSS
         glanced = map(model, ~ glance(.x))) %>% 
  unnest(cols = glanced)
  

# View clustering rsulsts
kclusts
```

‡∞á‡∞™‡±ç‡∞™‡±Å‡∞°‡±Å ‡∞Æ‡∞®‡∞ï‡±Å ‡∞™‡±ç‡∞∞‡∞§‡∞ø ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡∞ø‡∞Ç‡∞ó‡±ç ‡∞Ö‡∞≤‡±ç‡∞ó‡±ã‡∞∞‡∞ø‡∞•‡∞Ç ‡∞ï‡±ã‡∞∏‡∞Ç ‡∞∏‡±Ü‡∞Ç‡∞ü‡∞∞‡±ç *k* ‡∞§‡±ã ‡∞Æ‡±ä‡∞§‡±ç‡∞§‡∞Ç ‡∞µ‡∞ø‡∞§‡∞ø‡∞®‡±ç-‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç ‡∞∏‡∞Æ‡±ç-‡∞Ü‡∞´‡±ç-‡∞∏‡±ç‡∞ï‡±ç‡∞µ‡±á‡∞∞‡±ç‡∞≤ (tot.withinss) ‡∞â‡∞Ç‡∞¶‡∞ø, ‡∞Æ‡∞®‡∞Ç [‡∞é‡∞≤‡±ç‡∞¨‡±ã ‡∞™‡∞¶‡±ç‡∞ß‡∞§‡∞ø](https://en.wikipedia.org/wiki/Elbow_method_(clustering)) ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞ø ‡∞Ü‡∞™‡±ç‡∞ü‡∞ø‡∞Æ‡∞≤‡±ç ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç‡∞≤ ‡∞∏‡∞Ç‡∞ñ‡±ç‡∞Ø‡∞®‡±Å ‡∞ï‡∞®‡±Å‡∞ó‡±ä‡∞Ç‡∞ü‡∞æ‡∞Æ‡±Å. ‡∞à ‡∞™‡∞¶‡±ç‡∞ß‡∞§‡∞ø ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç‡∞≤ ‡∞∏‡∞Ç‡∞ñ‡±ç‡∞Ø ‡∞Ø‡±ä‡∞ï‡±ç‡∞ï ‡∞´‡∞Ç‡∞ï‡±ç‡∞∑‡∞®‡±ç‚Äå‡∞ó‡∞æ WCSS ‡∞®‡±Å ‡∞™‡±ç‡∞≤‡∞æ‡∞ü‡±ç ‡∞ö‡±á‡∞Ø‡∞°‡∞Ç ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å ‡∞µ‡∞æ‡∞°‡∞æ‡∞≤‡±ç‡∞∏‡∞ø‡∞® ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç‡∞≤ ‡∞∏‡∞Ç‡∞ñ‡±ç‡∞Ø‡∞ó‡∞æ [‡∞µ‡∞ï‡±ç‡∞∞ ‡∞∞‡±á‡∞ñ ‡∞Ø‡±ä‡∞ï‡±ç‡∞ï ‡∞é‡∞≤‡±ç‡∞¨‡±ã](https://en.wikipedia.org/wiki/Elbow_of_the_curve "Elbow of the curve") ‡∞®‡±Å ‡∞é‡∞Ç‡∞ö‡±Å‡∞ï‡±ã‡∞µ‡∞°‡∞Ç ‡∞ï‡∞≤‡∞ø‡∞ó‡∞ø ‡∞â‡∞Ç‡∞ü‡±Å‡∞Ç‡∞¶‡∞ø.


```r
set.seed(2056)
# Use elbow method to determine optimum number of clusters
kclusts %>% 
  ggplot(mapping = aes(x = k, y = tot.withinss)) +
  geom_line(size = 1.2, alpha = 0.8, color = "#FF7F0EFF") +
  geom_point(size = 2, color = "#FF7F0EFF")
```

‡∞™‡±ç‡∞≤‡∞æ‡∞ü‡±ç ‡∞í‡∞ï ‡∞®‡±Å‡∞Ç‡∞°‡∞ø ‡∞∞‡±Ü‡∞Ç‡∞°‡±Å ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç‡∞≤‡∞ï‡±Å ‡∞∏‡∞Ç‡∞ñ‡±ç‡∞Ø ‡∞™‡±Ü‡∞∞‡∞ø‡∞ó‡∞ø‡∞®‡∞™‡±ç‡∞™‡±Å‡∞°‡±Å WCSS‡∞≤‡±ã ‡∞™‡±Ü‡∞¶‡±ç‡∞¶ ‡∞§‡∞ó‡±ç‡∞ó‡±Å‡∞¶‡∞≤ (‡∞ï‡∞æ‡∞¨‡∞ü‡±ç‡∞ü‡∞ø ‡∞é‡∞ï‡±ç‡∞ï‡±Å‡∞µ *‡∞ü‡±à‡∞ü‡±ç‚Äå‡∞®‡±Ü‡∞∏‡±ç*) ‡∞ö‡±Ç‡∞™‡∞ø‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø, ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å ‡∞∞‡±Ü‡∞Ç‡∞°‡±Å ‡∞®‡±Å‡∞Ç‡∞°‡∞ø ‡∞Æ‡±Ç‡∞°‡±Å ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç‡∞≤‡∞ï‡±Å ‡∞Æ‡∞∞‡∞ø‡∞Ç‡∞§ ‡∞ó‡∞Æ‡∞®‡∞ø‡∞Ç‡∞ö‡∞¶‡∞ó‡∞ø‡∞® ‡∞§‡∞ó‡±ç‡∞ó‡±Å‡∞¶‡∞≤ ‡∞â‡∞Ç‡∞ü‡±Å‡∞Ç‡∞¶‡∞ø. ‡∞Ü ‡∞§‡∞∞‡±ç‡∞µ‡∞æ‡∞§, ‡∞§‡∞ó‡±ç‡∞ó‡±Å‡∞¶‡∞≤ ‡∞§‡∞ï‡±ç‡∞ï‡±Å‡∞µ‡∞ó‡∞æ ‡∞â‡∞Ç‡∞ü‡±Å‡∞Ç‡∞¶‡∞ø, ‡∞´‡∞≤‡∞ø‡∞§‡∞Ç‡∞ó‡∞æ ‡∞ö‡∞æ‡∞∞‡±ç‡∞ü‡±ç‚Äå‡∞≤‡±ã ‡∞∏‡±Å‡∞Æ‡∞æ‡∞∞‡±Å ‡∞Æ‡±Ç‡∞°‡±Å ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç‡∞≤ ‡∞µ‡∞¶‡±ç‡∞¶ ‡∞í‡∞ï `‡∞é‡∞≤‡±ç‡∞¨‡±ã` üí™ ‡∞â‡∞Ç‡∞ü‡±Å‡∞Ç‡∞¶‡∞ø. ‡∞á‡∞¶‡∞ø ‡∞∞‡±Ü‡∞Ç‡∞°‡±Å ‡∞®‡±Å‡∞Ç‡∞°‡∞ø ‡∞Æ‡±Ç‡∞°‡±Å ‡∞§‡∞ó‡∞ø‡∞®‡∞Ç‡∞§‡∞ó‡∞æ ‡∞µ‡±á‡∞∞‡±Å ‡∞ö‡±á‡∞∏‡∞ø‡∞® ‡∞°‡±á‡∞ü‡∞æ ‡∞™‡∞æ‡∞Ø‡∞ø‡∞Ç‡∞ü‡±ç‡∞≤ ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç‡∞≤‡±Å ‡∞â‡∞®‡±ç‡∞®‡∞æ‡∞Ø‡∞®‡∞ø ‡∞Æ‡∞Ç‡∞ö‡∞ø ‡∞∏‡±Ç‡∞ö‡∞®.

‡∞á‡∞™‡±ç‡∞™‡±Å‡∞°‡±Å ‡∞Æ‡∞®‡∞Ç `k = 3` ‡∞â‡∞®‡±ç‡∞® ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡∞ø‡∞Ç‡∞ó‡±ç ‡∞Æ‡±ã‡∞°‡∞≤‡±ç‚Äå‡∞®‡±Å ‡∞§‡±Ä‡∞∏‡±Å‡∞ï‡±ã‡∞µ‡∞ö‡±ç‡∞ö‡±Å:

> `pull()`: ‡∞í‡∞ï‡±á ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‚Äå‡∞®‡±Å ‡∞§‡±Ä‡∞∏‡±Å‡∞ï‡±ã‡∞µ‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞∏‡±ç‡∞§‡∞æ‡∞∞‡±Å
>
> `pluck()`: ‡∞ú‡∞æ‡∞¨‡∞ø‡∞§‡∞æ‡∞≤‡±Å ‡∞µ‡∞Ç‡∞ü‡∞ø ‡∞°‡±á‡∞ü‡∞æ ‡∞®‡∞ø‡∞∞‡±ç‡∞Æ‡∞æ‡∞£‡∞æ‡∞≤‡∞®‡±Å ‡∞∏‡±Ç‡∞ö‡∞ø‡∞Ç‡∞ö‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞∏‡±ç‡∞§‡∞æ‡∞∞‡±Å


```r
# Extract k = 3 clustering
final_kmeans <- kclusts %>% 
  filter(k == 3) %>% 
  pull(model) %>% 
  pluck(1)


final_kmeans
```

‡∞ö‡∞æ‡∞≤‡∞æ ‡∞¨‡∞æ‡∞ó‡±Å‡∞Ç‡∞¶‡∞ø! ‡∞Æ‡∞®‡∞Ç ‡∞™‡±ä‡∞Ç‡∞¶‡∞ø‡∞® ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç‡∞≤‡∞®‡±Å ‡∞¶‡±É‡∞∂‡±ç‡∞Ø‡±Ä‡∞ï‡∞∞‡∞ø‡∞Ç‡∞ö‡±Å‡∞ï‡±Å‡∞Ç‡∞¶‡∞æ‡∞Ç. `plotly` ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞ø ‡∞ï‡±ä‡∞Ç‡∞§ ‡∞á‡∞Ç‡∞ü‡∞∞‡∞æ‡∞ï‡±ç‡∞ü‡∞ø‡∞µ‡∞ø‡∞ü‡±Ä ‡∞ï‡∞æ‡∞µ‡∞æ‡∞≤‡∞æ?


```r
# Add predicted cluster assignment to data set
results <-  augment(final_kmeans, df_numeric_select) %>% 
  bind_cols(df_numeric %>% select(artist_top_genre)) 

# Plot cluster assignments
clust_plt <- results %>% 
  ggplot(mapping = aes(x = popularity, y = danceability, color = .cluster, shape = artist_top_genre)) +
  geom_point(size = 2, alpha = 0.8) +
  paletteer::scale_color_paletteer_d("ggthemes::Tableau_10")

ggplotly(clust_plt)
```

‡∞™‡±ç‡∞∞‡∞§‡∞ø ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç (‡∞µ‡∞ø‡∞µ‡∞ø‡∞ß ‡∞∞‡∞Ç‡∞ó‡±Å‡∞≤‡∞§‡±ã ‡∞™‡±ç‡∞∞‡∞æ‡∞§‡∞ø‡∞®‡∞ø‡∞ß‡±ç‡∞Ø‡∞Ç ‡∞µ‡∞π‡∞ø‡∞Ç‡∞ö‡∞¨‡∞°‡∞ø‡∞®) ‡∞µ‡±á‡∞∞‡±ç‡∞µ‡±á‡∞∞‡±Å ‡∞ú‡∞æ‡∞®‡∞∞‡±ç‡∞≤‡∞®‡±Å (‡∞µ‡∞ø‡∞µ‡∞ø‡∞ß ‡∞Ü‡∞ï‡∞æ‡∞∞‡∞æ‡∞≤‡∞§‡±ã ‡∞™‡±ç‡∞∞‡∞æ‡∞§‡∞ø‡∞®‡∞ø‡∞ß‡±ç‡∞Ø‡∞Ç ‡∞µ‡∞π‡∞ø‡∞Ç‡∞ö‡∞¨‡∞°‡∞ø‡∞®) ‡∞ï‡∞≤‡∞ø‡∞ó‡∞ø ‡∞â‡∞Ç‡∞ü‡±Å‡∞Ç‡∞¶‡∞®‡∞ø ‡∞Æ‡∞®‡∞Ç ‡∞Ü‡∞∂‡∞ø‡∞Ç‡∞ö‡∞µ‡∞ö‡±ç‡∞ö‡±Å.

‡∞Æ‡±ã‡∞°‡∞≤‡±ç ‡∞Ø‡±ä‡∞ï‡±ç‡∞ï ‡∞ñ‡∞ö‡±ç‡∞ö‡∞ø‡∞§‡∞§‡±ç‡∞µ‡∞æ‡∞®‡±ç‡∞®‡∞ø ‡∞ö‡±Ç‡∞¶‡±ç‡∞¶‡∞æ‡∞Ç.


```r
# Assign genres to predefined integers
label_count <- results %>% 
  group_by(artist_top_genre) %>% 
  mutate(id = cur_group_id()) %>% 
  ungroup() %>% 
  summarise(correct_labels = sum(.cluster == id))


# Print results  
cat("Result:", label_count$correct_labels, "out of", nrow(results), "samples were correctly labeled.")

cat("\nAccuracy score:", label_count$correct_labels/nrow(results))
```

‡∞à ‡∞Æ‡±ã‡∞°‡∞≤‡±ç ‡∞Ø‡±ä‡∞ï‡±ç‡∞ï ‡∞ñ‡∞ö‡±ç‡∞ö‡∞ø‡∞§‡∞§‡±ç‡∞µ‡∞Ç ‡∞ö‡±Ü‡∞°‡±Å‡∞ó‡∞æ ‡∞≤‡±á‡∞¶‡±Å, ‡∞ï‡∞æ‡∞®‡±Ä ‡∞ó‡±ä‡∞™‡±ç‡∞™‡∞¶‡∞ø‡∞ó‡∞æ ‡∞≤‡±á‡∞¶‡±Å. ‡∞°‡±á‡∞ü‡∞æ K-Means ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡∞ø‡∞Ç‡∞ó‡±ç‚Äå‡∞ï‡±Å ‡∞¨‡∞æ‡∞ó‡∞æ ‡∞∏‡∞∞‡∞ø‡∞™‡±ã‡∞µ‡∞ï‡∞™‡±ã‡∞µ‡∞ö‡±ç‡∞ö‡±Å. ‡∞à ‡∞°‡±á‡∞ü‡∞æ ‡∞ö‡∞æ‡∞≤‡∞æ ‡∞Ö‡∞∏‡∞Æ‡∞§‡±Å‡∞≤‡±ç‡∞Ø‡∞Ç‡∞ó‡∞æ ‡∞â‡∞Ç‡∞¶‡∞ø, ‡∞ö‡∞æ‡∞≤‡∞æ ‡∞§‡∞ï‡±ç‡∞ï‡±Å‡∞µ ‡∞∏‡∞Ç‡∞¨‡∞Ç‡∞ß‡∞Ç ‡∞â‡∞Ç‡∞¶‡∞ø ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç ‡∞µ‡∞ø‡∞≤‡±Å‡∞µ‡∞≤ ‡∞Æ‡∞ß‡±ç‡∞Ø ‡∞ö‡∞æ‡∞≤‡∞æ ‡∞µ‡±ç‡∞Ø‡∞§‡±ç‡∞Ø‡∞æ‡∞∏‡∞Ç ‡∞â‡∞Ç‡∞¶‡∞ø ‡∞ï‡∞æ‡∞¨‡∞ü‡±ç‡∞ü‡∞ø ‡∞¨‡∞æ‡∞ó‡∞æ ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç ‡∞ö‡±á‡∞Ø‡∞°‡∞Ç ‡∞ï‡∞∑‡±ç‡∞ü‡∞Ç. ‡∞µ‡∞æ‡∞∏‡±ç‡∞§‡∞µ‡∞æ‡∞®‡∞ø‡∞ï‡∞ø, ‡∞è‡∞∞‡±ç‡∞™‡∞°‡±á ‡∞ï‡±ç‡∞≤‡∞∏‡±ç‡∞ü‡∞∞‡±ç‡∞≤‡±Å ‡∞™‡±à‡∞ó‡∞æ ‡∞®‡∞ø‡∞∞‡±ç‡∞µ‡∞ö‡∞ø‡∞Ç‡∞ö‡∞ø‡∞® ‡∞Æ‡±Ç‡∞°‡±Å ‡∞ú‡∞æ‡∞®‡∞∞‡±ç ‡∞µ‡∞∞‡±ç‡∞ó‡∞æ‡∞≤ ‡∞¶‡±ç‡∞µ‡∞æ‡∞∞‡∞æ ‡∞¨‡∞≤‡∞Ç‡∞ó‡∞æ ‡∞™‡±ç‡∞∞‡∞≠‡∞æ‡∞µ‡∞ø‡∞§‡∞Ç ‡∞≤‡±á‡∞¶‡∞æ ‡∞µ‡∞ï‡±ç‡∞∞‡±Ä‡∞ï‡±É‡∞§‡∞Æ‡∞Ø‡±ç‡∞Ø‡±á ‡∞Ö‡∞µ‡∞ï‡∞æ‡∞∂‡∞Ç ‡∞â‡∞Ç‡∞¶‡∞ø.
