# ARIMA తో టైమ్ సిరీస్ ఫోర్కాస్టింగ్

ఈ నోట్‌బుక్‌లో, మేము ఎలా చేయాలో చూపిస్తాము:
- ARIMA టైమ్ సిరీస్ ఫోర్కాస్టింగ్ మోడల్ శిక్షణ కోసం టైమ్ సిరీస్ డేటాను సిద్ధం చేయడం
- టైమ్ సిరీస్‌లో తదుపరి HORIZON దశలను ముందుగా (సమయం *t+1* నుండి *t+HORIZON* వరకు) ఫోర్కాస్ట్ చేయడానికి ఒక సాదారణ ARIMA మోడల్‌ను అమలు చేయడం
- మోడల్‌ను మూల్యాంకనం చేయడం

ఈ ఉదాహరణలో డేటా GEFCom2014 ఫోర్కాస్టింగ్ పోటీ<sup>1</sup> నుండి తీసుకోబడింది. ఇది 2012 నుండి 2014 వరకు 3 సంవత్సరాల గంటల వారీ విద్యుత్ లోడ్ మరియు ఉష్ణోగ్రత విలువలను కలిగి ఉంది. పని భవిష్యత్తు విద్యుత్ లోడ్ విలువలను ఫోర్కాస్ట్ చేయడం. ఈ ఉదాహరణలో, మేము చారిత్రక లోడ్ డేటాను మాత్రమే ఉపయోగించి ఒక టైమ్ స్టెప్ ముందుకు ఎలా ఫోర్కాస్ట్ చేయాలో చూపిస్తాము.
