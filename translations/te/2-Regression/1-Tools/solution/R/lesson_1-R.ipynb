‡∞Ö‡∞≤‡∞æ‡∞ó‡±á, `glimpse()` ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞ø ‡∞Æ‡∞æ ‡∞°‡±á‡∞ü‡∞æ‡∞™‡±à ‡∞í‡∞ï ‡∞∏‡∞æ‡∞®‡∞ø‡∞ü‡±Ä ‡∞ö‡±Ü‡∞ï‡±ç ‡∞®‡∞ø‡∞∞‡±ç‡∞µ‡∞π‡∞ø‡∞Ç‡∞ö‡∞ø, `slice()` ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞ø ‡∞Æ‡±ä‡∞¶‡∞ü‡∞ø 5 ‡∞µ‡∞∞‡±Å‡∞∏‡∞≤‡∞®‡±Å ‡∞™‡±ç‡∞∞‡∞¶‡∞∞‡±ç‡∞∂‡∞ø‡∞∏‡±ç‡∞§‡∞æ‡∞Æ‡±Å.

‡∞Æ‡∞∞‡∞ø‡∞Ç‡∞§ ‡∞Æ‡±Å‡∞Ç‡∞¶‡±Å‡∞ï‡±Å ‡∞µ‡±Ü‡∞≥‡±ç‡∞≤‡±á‡∞Æ‡±Å‡∞Ç‡∞¶‡±Å, R ‡∞ï‡±ã‡∞°‡±ç‚Äå‡∞≤‡±ã ‡∞Æ‡±Ä‡∞∞‡±Å ‡∞§‡∞∞‡∞ö‡±Å‡∞ó‡∞æ ‡∞é‡∞¶‡±Å‡∞∞‡±ç‡∞ï‡±ä‡∞®‡±á ‡∞í‡∞ï ‡∞µ‡∞ø‡∞∑‡∞Ø‡∞Ç ‡∞™‡∞∞‡∞ø‡∞ö‡∞Ø‡∞Ç ‡∞ö‡±á‡∞¶‡±ç‡∞¶‡∞æ‡∞Ç ü•Åü•Å: ‡∞™‡±à‡∞™‡±ç ‡∞Ü‡∞™‡∞∞‡±á‡∞ü‡∞∞‡±ç `%>%`

‡∞™‡±à‡∞™‡±ç ‡∞Ü‡∞™‡∞∞‡±á‡∞ü‡∞∞‡±ç (`%>%`) ‡∞í‡∞ï ‡∞Ü‡∞¨‡±ç‡∞ú‡±Ü‡∞ï‡±ç‡∞ü‡±Å‡∞®‡±Å ‡∞´‡∞Ç‡∞ï‡±ç‡∞∑‡∞®‡±ç ‡∞≤‡±á‡∞¶‡∞æ ‡∞ï‡∞æ‡∞≤‡±ç ‡∞é‡∞ï‡±ç‡∞∏‡±ç‚Äå‡∞™‡±ç‡∞∞‡±Ü‡∞∑‡∞®‡±ç‚Äå‡∞ï‡±Å ‡∞Æ‡±Å‡∞Ç‡∞¶‡±Å‡∞ï‡±Å ‡∞™‡∞Ç‡∞™‡±Å‡∞§‡±Ç ‡∞≤‡∞æ‡∞ú‡∞ø‡∞ï‡∞≤‡±ç ‡∞ï‡±ç‡∞∞‡∞Æ‡∞Ç‡∞≤‡±ã ‡∞Ü‡∞™‡∞∞‡±á‡∞∑‡∞®‡±ç‡∞≤‡∞®‡±Å ‡∞®‡∞ø‡∞∞‡±ç‡∞µ‡∞π‡∞ø‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø. ‡∞Æ‡±Ä‡∞∞‡±Å ‡∞™‡±à‡∞™‡±ç ‡∞Ü‡∞™‡∞∞‡±á‡∞ü‡∞∞‡±ç‚Äå‡∞®‡±Å ‡∞Æ‡±Ä ‡∞ï‡±ã‡∞°‡±ç‚Äå‡∞≤‡±ã "‡∞Ö‡∞™‡±ç‡∞™‡±Å‡∞°‡±Å" ‡∞Ö‡∞®‡∞ø ‡∞ö‡±Ü‡∞™‡±ç‡∞™‡∞°‡∞Ç ‡∞≤‡∞æ‡∞ó‡∞æ ‡∞≠‡∞æ‡∞µ‡∞ø‡∞Ç‡∞ö‡∞µ‡∞ö‡±ç‡∞ö‡±Å.


```r
# Import the data set
diabetes <- read_table2(file = "https://www4.stat.ncsu.edu/~boos/var.select/diabetes.rwrite1.txt")


# Get a glimpse and dimensions of the data
glimpse(diabetes)


# Select the first 5 rows of the data
diabetes %>% 
  slice(1:5)
```

‡∞™‡∞∞‡±ç‡∞Ø‡∞µ‡±á‡∞ï‡±ç‡∞∑‡∞ø‡∞§ ‡∞Ö‡∞≠‡±ç‡∞Ø‡∞æ‡∞∏‡∞Ç‡∞≤‡±ã ‡∞°‡±á‡∞ü‡∞æ‡∞®‡±Å ‡∞∞‡±Ü‡∞Ç‡∞°‡±Å ‡∞â‡∞™‡∞∏‡∞Æ‡±Ç‡∞π‡∞æ‡∞≤‡±Å‡∞ó‡∞æ *‡∞µ‡∞ø‡∞≠‡∞ú‡∞ø‡∞Ç‡∞ö‡∞°‡∞Ç* ‡∞∏‡∞æ‡∞ß‡∞æ‡∞∞‡∞£ ‡∞Ü‡∞ö‡∞æ‡∞∞‡∞Ç; ‡∞í‡∞ï‡∞ü‡∞ø (‡∞∏‡∞æ‡∞ß‡∞æ‡∞∞‡∞£‡∞Ç‡∞ó‡∞æ ‡∞™‡±Ü‡∞¶‡±ç‡∞¶‡∞¶‡∞ø) ‡∞Æ‡±ã‡∞°‡∞≤‡±ç‚Äå‡∞®‡±Å ‡∞∂‡∞ø‡∞ï‡±ç‡∞∑‡∞£ ‡∞á‡∞µ‡±ç‡∞µ‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡±á ‡∞∏‡±Ü‡∞ü‡±ç, ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å ‡∞Æ‡∞∞‡±ä‡∞ï ‡∞ö‡∞ø‡∞®‡±ç‡∞® "‡∞π‡±ã‡∞≤‡±ç‡∞°‡±ç-‡∞¨‡±ç‡∞Ø‡∞æ‡∞ï‡±ç" ‡∞∏‡±Ü‡∞ü‡±ç ‡∞Æ‡±ã‡∞°‡∞≤‡±ç ‡∞é‡∞≤‡∞æ ‡∞™‡±ç‡∞∞‡∞¶‡∞∞‡±ç‡∞∂‡∞ø‡∞Ç‡∞ö‡∞ø‡∞Ç‡∞¶‡±ã ‡∞ö‡±Ç‡∞°‡∞ü‡∞æ‡∞®‡∞ø‡∞ï‡∞ø ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞∏‡±ç‡∞§‡∞æ‡∞∞‡±Å.

‡∞á‡∞™‡±ç‡∞™‡±Å‡∞°‡±Å ‡∞Æ‡∞®‡∞ï‡±Å ‡∞°‡±á‡∞ü‡∞æ ‡∞∏‡∞ø‡∞¶‡±ç‡∞ß‡∞Ç‡∞ó‡∞æ ‡∞â‡∞®‡±ç‡∞®‡∞Ç‡∞¶‡±Å‡∞®, ‡∞à ‡∞°‡±á‡∞ü‡∞æ‡∞∏‡±Ü‡∞ü‡±ç‚Äå‡∞≤‡±ã‡∞®‡∞ø ‡∞∏‡∞Ç‡∞ñ‡±ç‡∞Ø‡∞≤ ‡∞Æ‡∞ß‡±ç‡∞Ø ‡∞§‡∞æ‡∞∞‡±ç‡∞ï‡∞ø‡∞ï‡∞Æ‡±à‡∞® ‡∞µ‡∞ø‡∞≠‡∞ú‡∞®‡∞®‡±Å ‡∞Ø‡∞Ç‡∞§‡±ç‡∞∞‡∞Ç ‡∞∏‡∞π‡∞æ‡∞Ø‡∞Ç‡∞§‡±ã ‡∞®‡∞ø‡∞∞‡±ç‡∞£‡∞Ø‡∞ø‡∞Ç‡∞ö‡∞ó‡∞≤‡∞Æ‡∞æ ‡∞Ö‡∞®‡∞ø ‡∞ö‡±Ç‡∞°‡∞µ‡∞ö‡±ç‡∞ö‡±Å. ‡∞Æ‡∞®‡∞Ç Tidymodels ‡∞´‡±ç‡∞∞‡±á‡∞Æ‡±ç‚Äå‡∞µ‡∞∞‡±ç‡∞ï‡±ç‚Äå‡∞≤‡±ã ‡∞≠‡∞æ‡∞ó‡∞Æ‡±à‡∞® [rsample](https://tidymodels.github.io/rsample/) ‡∞™‡±ç‡∞Ø‡∞æ‡∞ï‡±á‡∞ú‡±Ä‡∞®‡∞ø ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞ø, ‡∞°‡±á‡∞ü‡∞æ‡∞®‡±Å *‡∞é‡∞≤‡∞æ* ‡∞µ‡∞ø‡∞≠‡∞ú‡∞ø‡∞Ç‡∞ö‡∞æ‡∞≤‡±ã ‡∞∏‡∞Æ‡∞æ‡∞ö‡∞æ‡∞∞‡∞Ç ‡∞ï‡∞≤‡∞ø‡∞ó‡∞ø‡∞® ‡∞í‡∞ï ‡∞Ü‡∞¨‡±ç‡∞ú‡±Ü‡∞ï‡±ç‡∞ü‡±ç‚Äå‡∞®‡±Å ‡∞∏‡±É‡∞∑‡±ç‡∞ü‡∞ø‡∞Ç‡∞ö‡∞µ‡∞ö‡±ç‡∞ö‡±Å, ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å ‡∞Ü‡∞™‡±à ‡∞∏‡±É‡∞∑‡±ç‡∞ü‡∞ø‡∞Ç‡∞ö‡∞ø‡∞® ‡∞∂‡∞ø‡∞ï‡±ç‡∞∑‡∞£ ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å ‡∞™‡∞∞‡±Ä‡∞ï‡±ç‡∞∑ ‡∞∏‡±Ü‡∞ü‡±ç‚Äå‡∞≤‡∞®‡±Å ‡∞§‡±Ä‡∞∏‡±Å‡∞ï‡±ã‡∞µ‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø ‡∞∞‡±Ü‡∞Ç‡∞°‡±Å ‡∞Æ‡∞∞‡∞ø‡∞®‡±ç‡∞®‡∞ø rsample ‡∞´‡∞Ç‡∞ï‡±ç‡∞∑‡∞®‡±ç‡∞≤‡∞®‡±Å ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞µ‡∞ö‡±ç‡∞ö‡±Å:


```r
set.seed(2056)
# Split 67% of the data for training and the rest for tesing
diabetes_split <- diabetes_select %>% 
  initial_split(prop = 0.67)

# Extract the resulting train and test sets
diabetes_train <- training(diabetes_split)
diabetes_test <- testing(diabetes_split)

# Print the first 3 rows of the training set
diabetes_train %>% 
  slice(1:10)
```

## 4. Tidymodels ‡∞§‡±ã ‡∞≤‡±Ä‡∞®‡∞ø‡∞Ø‡∞∞‡±ç ‡∞∞‡∞ø‡∞ó‡±ç‡∞∞‡±Ü‡∞∑‡∞®‡±ç ‡∞Æ‡±ã‡∞°‡∞≤‡±ç‚Äå‡∞®‡±Å ‡∞∂‡∞ø‡∞ï‡±ç‡∞∑‡∞£ ‡∞á‡∞µ‡±ç‡∞µ‡∞Ç‡∞°‡∞ø

‡∞á‡∞™‡±ç‡∞™‡±Å‡∞°‡±Å ‡∞Æ‡∞®‡∞Ç ‡∞Æ‡∞® ‡∞Æ‡±ã‡∞°‡∞≤‡±ç‚Äå‡∞®‡±Å ‡∞∂‡∞ø‡∞ï‡±ç‡∞∑‡∞£ ‡∞á‡∞µ‡±ç‡∞µ‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø ‡∞∏‡∞ø‡∞¶‡±ç‡∞ß‡∞Ç‡∞ó‡∞æ ‡∞â‡∞®‡±ç‡∞®‡∞æ‡∞Æ‡±Å!

Tidymodels ‡∞≤‡±ã, ‡∞Æ‡±Ä‡∞∞‡±Å `parsnip()` ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞ø ‡∞Æ‡±ã‡∞°‡∞≤‡±ç‡∞∏‡±ç‚Äå‡∞®‡±Å ‡∞Æ‡±Ç‡∞°‡±Å ‡∞≠‡∞æ‡∞µ‡∞®‡∞≤‡∞®‡±Å ‡∞™‡±á‡∞∞‡±ç‡∞ï‡±ä‡∞®‡∞°‡∞Ç ‡∞¶‡±ç‡∞µ‡∞æ‡∞∞‡∞æ ‡∞®‡∞ø‡∞∞‡±ç‡∞¶‡±á‡∞∂‡∞ø‡∞∏‡±ç‡∞§‡∞æ‡∞∞‡±Å:

- ‡∞Æ‡±ã‡∞°‡∞≤‡±ç **‡∞∞‡∞ï‡∞Ç** ‡∞≤‡±Ä‡∞®‡∞ø‡∞Ø‡∞∞‡±ç ‡∞∞‡∞ø‡∞ó‡±ç‡∞∞‡±Ü‡∞∑‡∞®‡±ç, ‡∞≤‡∞æ‡∞ú‡∞ø‡∞∏‡±ç‡∞ü‡∞ø‡∞ï‡±ç ‡∞∞‡∞ø‡∞ó‡±ç‡∞∞‡±Ü‡∞∑‡∞®‡±ç, ‡∞°‡±Ü‡∞∏‡∞ø‡∞∑‡∞®‡±ç ‡∞ü‡±ç‡∞∞‡±Ä ‡∞Æ‡±ã‡∞°‡∞≤‡±ç‡∞∏‡±ç ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å ‡∞á‡∞§‡∞∞ ‡∞Æ‡±ã‡∞°‡∞≤‡±ç‡∞∏‡±ç‚Äå‡∞®‡±Å ‡∞µ‡±á‡∞∞‡±Å‡∞ö‡±á‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø.

- ‡∞Æ‡±ã‡∞°‡∞≤‡±ç **‡∞Æ‡±ã‡∞°‡±ç** ‡∞∞‡∞ø‡∞ó‡±ç‡∞∞‡±Ü‡∞∑‡∞®‡±ç ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å ‡∞µ‡∞∞‡±ç‡∞ó‡±Ä‡∞ï‡∞∞‡∞£ ‡∞µ‡∞Ç‡∞ü‡∞ø ‡∞∏‡∞æ‡∞ß‡∞æ‡∞∞‡∞£ ‡∞é‡∞Ç‡∞™‡∞ø‡∞ï‡∞≤‡∞®‡±Å ‡∞ï‡∞≤‡∞ø‡∞ó‡∞ø ‡∞â‡∞Ç‡∞ü‡±Å‡∞Ç‡∞¶‡∞ø; ‡∞ï‡±ä‡∞®‡±ç‡∞®‡∞ø ‡∞Æ‡±ã‡∞°‡∞≤‡±ç ‡∞∞‡∞ï‡∞æ‡∞≤‡±Å ‡∞µ‡±Ä‡∞ü‡∞ø‡∞≤‡±ã ‡∞è‡∞¶‡±à‡∞®‡∞æ ‡∞Æ‡±ã‡∞°‡±ç‚Äå‡∞®‡±Å ‡∞Æ‡∞¶‡±ç‡∞¶‡∞§‡±Å ‡∞á‡∞∏‡±ç‡∞§‡∞æ‡∞Ø‡∞ø, ‡∞Æ‡∞∞‡∞ø‡∞ï‡±ä‡∞®‡±ç‡∞®‡∞ø ‡∞í‡∞ï‡±ç‡∞ï ‡∞Æ‡±ã‡∞°‡±ç ‡∞Æ‡∞æ‡∞§‡±ç‡∞∞‡∞Æ‡±á ‡∞ï‡∞≤‡∞ø‡∞ó‡∞ø ‡∞â‡∞Ç‡∞ü‡∞æ‡∞Ø‡∞ø.

- ‡∞Æ‡±ã‡∞°‡∞≤‡±ç **‡∞á‡∞Ç‡∞ú‡∞ø‡∞®‡±ç** ‡∞Ö‡∞®‡±á‡∞¶‡∞ø ‡∞Æ‡±ã‡∞°‡∞≤‡±ç‚Äå‡∞®‡±Å ‡∞´‡∞ø‡∞ü‡±ç ‡∞ö‡±á‡∞Ø‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡±á ‡∞ó‡∞£‡∞®‡∞æ‡∞§‡±ç‡∞Æ‡∞ï ‡∞∏‡∞æ‡∞ß‡∞®‡∞Ç. ‡∞á‡∞µ‡∞ø ‡∞§‡∞∞‡∞ö‡±Å‡∞ó‡∞æ R ‡∞™‡±ç‡∞Ø‡∞æ‡∞ï‡±á‡∞ú‡±Ä‡∞≤‡±Å, ‡∞â‡∞¶‡∞æ‡∞π‡∞∞‡∞£‡∞ï‡±Å **`"lm"`** ‡∞≤‡±á‡∞¶‡∞æ **`"ranger"`**

‡∞à ‡∞Æ‡±ã‡∞°‡∞≤‡∞ø‡∞Ç‡∞ó‡±ç ‡∞∏‡∞Æ‡∞æ‡∞ö‡∞æ‡∞∞‡∞Ç ‡∞Æ‡±ã‡∞°‡∞≤‡±ç ‡∞∏‡±ç‡∞™‡±Ü‡∞∏‡∞ø‡∞´‡∞ø‡∞ï‡±á‡∞∑‡∞®‡±ç‚Äå‡∞≤‡±ã ‡∞∏‡±á‡∞ï‡∞∞‡∞ø‡∞Ç‡∞ö‡∞¨‡∞°‡±Å‡∞§‡±Å‡∞Ç‡∞¶‡∞ø, ‡∞ï‡∞æ‡∞¨‡∞ü‡±ç‡∞ü‡∞ø ‡∞í‡∞ï‡∞¶‡∞æ‡∞®‡±ç‡∞®‡∞ø ‡∞®‡∞ø‡∞∞‡±ç‡∞Æ‡∞ø‡∞¶‡±ç‡∞¶‡∞æ‡∞Ç!


```r
# Build a linear model specification
lm_spec <- 
  # Type
  linear_reg() %>% 
  # Engine
  set_engine("lm") %>% 
  # Mode
  set_mode("regression")


# Print the model specification
lm_spec
```

‡∞Æ‡±ã‡∞°‡∞≤‡±ç‚Äå‡∞®‡±Å *‡∞∏‡±ç‡∞™‡±Ü‡∞∏‡∞ø‡∞´‡±à* ‡∞ö‡±á‡∞∏‡∞ø‡∞® ‡∞§‡∞∞‡±ç‡∞µ‡∞æ‡∞§, ‡∞Æ‡±ã‡∞°‡∞≤‡±ç‚Äå‡∞®‡±Å ‡∞∏‡∞æ‡∞ß‡∞æ‡∞∞‡∞£‡∞Ç‡∞ó‡∞æ ‡∞í‡∞ï ‡∞´‡∞æ‡∞∞‡±ç‡∞Æ‡±Å‡∞≤‡∞æ ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å ‡∞ï‡±ä‡∞Ç‡∞§ ‡∞°‡±á‡∞ü‡∞æ ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞ø [`fit()`](https://parsnip.tidymodels.org/reference/fit.html) ‡∞´‡∞Ç‡∞ï‡±ç‡∞∑‡∞®‡±ç ‡∞¶‡±ç‡∞µ‡∞æ‡∞∞‡∞æ `estimated` ‡∞≤‡±á‡∞¶‡∞æ `trained` ‡∞ö‡±á‡∞Ø‡∞µ‡∞ö‡±ç‡∞ö‡±Å.

`y ~ .` ‡∞Ö‡∞Ç‡∞ü‡±á ‡∞Æ‡∞®‡∞Ç `y` ‡∞®‡±Å ‡∞Ö‡∞Ç‡∞ö‡∞®‡∞æ ‡∞µ‡±á‡∞Ø‡∞¨‡∞°‡±á ‡∞™‡∞∞‡∞ø‡∞Æ‡∞æ‡∞£‡∞Ç/‡∞≤‡∞ï‡±ç‡∞∑‡±ç‡∞Ø‡∞Ç‡∞ó‡∞æ ‡∞´‡∞ø‡∞ü‡±ç ‡∞ö‡±á‡∞∏‡±ç‡∞§‡∞æ‡∞Æ‡±Å, ‡∞á‡∞¶‡∞ø ‡∞Ö‡∞®‡±ç‡∞®‡∞ø ‡∞™‡±ç‡∞∞‡∞ø‡∞°‡∞ø‡∞ï‡±ç‡∞ü‡∞∞‡±ç‡∞≤‡±Å/‡∞´‡±Ä‡∞ö‡∞∞‡±ç‡∞≤‡±Å ‡∞Ö‡∞Ç‡∞ü‡±á `.` ‡∞¶‡±ç‡∞µ‡∞æ‡∞∞‡∞æ ‡∞µ‡∞ø‡∞µ‡∞∞‡∞ø‡∞Ç‡∞ö‡∞¨‡∞°‡±Å‡∞§‡±Å‡∞Ç‡∞¶‡∞ø (‡∞à ‡∞∏‡∞Ç‡∞¶‡∞∞‡±ç‡∞≠‡∞Ç‡∞≤‡±ã, ‡∞Æ‡∞®‡∞ï‡±Å ‡∞í‡∞ï‡±á ‡∞í‡∞ï ‡∞™‡±ç‡∞∞‡∞ø‡∞°‡∞ø‡∞ï‡±ç‡∞ü‡∞∞‡±ç ‡∞â‡∞Ç‡∞¶‡∞ø: `bmi`)


```r
# Build a linear model specification
lm_spec <- linear_reg() %>% 
  set_engine("lm") %>%
  set_mode("regression")


# Train a linear regression model
lm_mod <- lm_spec %>% 
  fit(y ~ ., data = diabetes_train)

# Print the model
lm_mod
```

‡∞Æ‡±ã‡∞°‡∞≤‡±ç ‡∞Ö‡∞µ‡±Å‡∞ü‡±ç‡∞™‡±Å‡∞ü‡±ç ‡∞®‡±Å‡∞Ç‡∞°‡∞ø, ‡∞Æ‡∞®‡∞Ç ‡∞∂‡∞ø‡∞ï‡±ç‡∞∑‡∞£ ‡∞∏‡∞Æ‡∞Ø‡∞Ç‡∞≤‡±ã ‡∞®‡±á‡∞∞‡±ç‡∞ö‡±Å‡∞ï‡±Å‡∞®‡±ç‡∞® ‡∞ó‡±Å‡∞£‡∞ï‡∞æ‡∞≤‡∞®‡±Å ‡∞ö‡±Ç‡∞°‡∞µ‡∞ö‡±ç‡∞ö‡±Å. ‡∞Ö‡∞µ‡∞ø ‡∞®‡∞ø‡∞ú‡∞Æ‡±à‡∞® ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å ‡∞Ö‡∞Ç‡∞ö‡∞®‡∞æ ‡∞µ‡±á‡∞∞‡∞ø‡∞Ø‡∞¨‡±Å‡∞≤‡±ç ‡∞Æ‡∞ß‡±ç‡∞Ø ‡∞Æ‡±ä‡∞§‡±ç‡∞§‡∞Ç ‡∞≤‡±ã‡∞§‡±à‡∞® ‡∞™‡±ä‡∞∞‡∞™‡∞æ‡∞ü‡±Å ‡∞§‡∞ï‡±ç‡∞ï‡±Å‡∞µ‡∞ó‡∞æ ‡∞â‡∞Ç‡∞°‡±á ‡∞â‡∞§‡±ç‡∞§‡∞Æ ‡∞∏‡∞∞‡∞≥‡∞ø ‡∞Ø‡±ä‡∞ï‡±ç‡∞ï ‡∞ó‡±Å‡∞£‡∞ï‡∞æ‡∞≤‡∞®‡±Å ‡∞∏‡±Ç‡∞ö‡∞ø‡∞∏‡±ç‡∞§‡∞æ‡∞Ø‡∞ø.

## 5. ‡∞™‡∞∞‡±Ä‡∞ï‡±ç‡∞∑ ‡∞∏‡±Ü‡∞ü‡±ç ‡∞™‡±à ‡∞Ö‡∞Ç‡∞ö‡∞®‡∞æ‡∞≤‡±Å ‡∞ö‡±á‡∞Ø‡∞Ç‡∞°‡∞ø

‡∞á‡∞™‡±ç‡∞™‡±Å‡∞°‡±Å ‡∞Æ‡∞®‡∞Ç ‡∞í‡∞ï ‡∞Æ‡±ã‡∞°‡∞≤‡±ç ‡∞∂‡∞ø‡∞ï‡±ç‡∞∑‡∞£ ‡∞á‡∞ö‡±ç‡∞ö‡∞ø‡∞®‡∞Ç‡∞¶‡±Å‡∞®, ‡∞¶‡∞æ‡∞®‡±ç‡∞®‡∞ø ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞ø ‡∞™‡∞∞‡±Ä‡∞ï‡±ç‡∞∑ ‡∞°‡±á‡∞ü‡∞æ‡∞∏‡±Ü‡∞ü‡±ç ‡∞ï‡±ã‡∞∏‡∞Ç ‡∞µ‡±ç‡∞Ø‡∞æ‡∞ß‡∞ø ‡∞™‡±Å‡∞∞‡±ã‡∞ó‡∞§‡∞ø‡∞®‡∞ø y ‡∞Ö‡∞Ç‡∞ö‡∞®‡∞æ ‡∞µ‡±á‡∞Ø‡∞µ‡∞ö‡±ç‡∞ö‡±Å [parsnip::predict()](https://parsnip.tidymodels.org/reference/predict.model_fit.html) ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞ø. ‡∞á‡∞¶‡∞ø ‡∞°‡±á‡∞ü‡∞æ ‡∞∏‡∞Æ‡±Ç‡∞π‡∞æ‡∞≤ ‡∞Æ‡∞ß‡±ç‡∞Ø ‡∞∏‡∞∞‡∞≥‡∞ø ‡∞ó‡±Ä‡∞Ø‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞¨‡∞°‡±Å‡∞§‡±Å‡∞Ç‡∞¶‡∞ø.


```r
# Make predictions for the test set
predictions <- lm_mod %>% 
  predict(new_data = diabetes_test)

# Print out some of the predictions
predictions %>% 
  slice(1:5)
```

‡∞Ö‡∞Ç‡∞ö‡∞®‡∞æ‡∞≤‡±Å ‡∞ö‡±á‡∞Ø‡∞°‡∞Ç‡∞≤‡±ã, tidymodels ‡∞∏‡∞Ç‡∞™‡±ç‡∞∞‡∞¶‡∞æ‡∞Ø‡∞Ç ‡∞é‡∞™‡±ç‡∞™‡±Å‡∞°‡±Ç ‡∞´‡∞≤‡∞ø‡∞§‡∞æ‡∞≤ ‡∞Ø‡±ä‡∞ï‡±ç‡∞ï ‡∞ü‡∞ø‡∞¨‡±Å‡∞≤‡±ç/‡∞°‡±á‡∞ü‡∞æ ‡∞´‡±ç‡∞∞‡±á‡∞Æ‡±ç‚Äå‡∞®‡±Å ‡∞™‡±ç‡∞∞‡∞Æ‡∞æ‡∞£‡±Ä‡∞ï‡±É‡∞§ ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç ‡∞™‡±á‡∞∞‡±ç‡∞≤‡∞§‡±ã ‡∞â‡∞§‡±ç‡∞™‡∞§‡±ç‡∞§‡∞ø ‡∞ö‡±á‡∞Ø‡∞°‡∞Ç. ‡∞á‡∞¶‡∞ø ‡∞Ö‡∞∏‡∞≤‡±Å ‡∞°‡±á‡∞ü‡∞æ ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å ‡∞Ö‡∞Ç‡∞ö‡∞®‡∞æ‡∞≤‡∞®‡±Å ‡∞ï‡∞≤‡∞ø‡∞™‡∞ø ‡∞§‡∞∞‡±Å‡∞µ‡∞æ‡∞§‡∞ø ‡∞Ü‡∞™‡∞∞‡±á‡∞∑‡∞®‡±ç‡∞≤ ‡∞ï‡±ã‡∞∏‡∞Ç ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞¶‡∞ó‡∞ø‡∞® ‡∞´‡∞æ‡∞∞‡±ç‡∞Æ‡∞æ‡∞ü‡±ç‚Äå‡∞≤‡±ã ‡∞â‡∞Ç‡∞ö‡∞°‡∞Ç ‡∞∏‡±Å‡∞≤‡∞≠‡∞Ç ‡∞ö‡±á‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø, ‡∞â‡∞¶‡∞æ‡∞π‡∞∞‡∞£‡∞ï‡±Å ‡∞™‡±ç‡∞≤‡∞æ‡∞ü‡∞ø‡∞Ç‡∞ó‡±ç.

`dplyr::bind_cols()` ‡∞Ö‡∞®‡±á‡∞¶‡∞ø ‡∞¨‡∞π‡±Å‡∞≥ ‡∞°‡±á‡∞ü‡∞æ ‡∞´‡±ç‡∞∞‡±á‡∞Æ‡±ç‚Äå‡∞≤‡∞®‡±Å ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‚Äå‡∞≤‡±Å‡∞ó‡∞æ ‡∞∏‡∞Æ‡∞∞‡±ç‡∞•‡∞µ‡∞Ç‡∞§‡∞Ç‡∞ó‡∞æ ‡∞¨‡∞Ç‡∞ß‡∞ø‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø.


```r
# Combine the predictions and the original test set
results <- diabetes_test %>% 
  bind_cols(predictions)


results %>% 
  slice(1:5)
```
