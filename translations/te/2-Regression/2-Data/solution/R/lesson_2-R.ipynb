R ‡∞≤‡±ã ‡∞≤‡±á‡∞®‡∞ø ‡∞≤‡±á‡∞¶‡∞æ ‡∞§‡±Ü‡∞≤‡∞ø‡∞Ø‡∞®‡∞ø ‡∞µ‡∞ø‡∞≤‡±Å‡∞µ‡∞≤‡∞®‡±Å ‡∞™‡±ç‡∞∞‡∞§‡±ç‡∞Ø‡±á‡∞ï ‡∞∏‡±Ü‡∞Ç‡∞ü‡∞ø‡∞®‡±Ü‡∞≤‡±ç ‡∞µ‡∞ø‡∞≤‡±Å‡∞µ‡∞§‡±ã ‡∞∏‡±Ç‡∞ö‡∞ø‡∞∏‡±ç‡∞§‡∞æ‡∞∞‡±Å: `NA` (‡∞Ö‡∞Ç‡∞¶‡±Å‡∞¨‡∞æ‡∞ü‡±Å‡∞≤‡±ã ‡∞≤‡±á‡∞¶‡±Å).

‡∞ï‡∞æ‡∞¨‡∞ü‡±ç‡∞ü‡∞ø ‡∞°‡±á‡∞ü‡∞æ ‡∞´‡±ç‡∞∞‡±á‡∞Æ‡±ç ‡∞≤‡±ã ‡∞≤‡±á‡∞®‡∞ø ‡∞µ‡∞ø‡∞≤‡±Å‡∞µ‡∞≤‡±Å ‡∞â‡∞®‡±ç‡∞®‡∞æ‡∞Ø‡±ã ‡∞≤‡±á‡∞¶‡±ã ‡∞é‡∞≤‡∞æ ‡∞§‡±Ü‡∞≤‡±Å‡∞∏‡±Å‡∞ï‡±ã‡∞µ‡∞æ‡∞≤‡∞ø?
<br>
-   ‡∞í‡∞ï ‡∞∏‡∞∞‡∞≥‡∞Æ‡±à‡∞® ‡∞µ‡∞ø‡∞ß‡∞æ‡∞®‡∞Ç base R ‡∞´‡∞Ç‡∞ï‡±ç‡∞∑‡∞®‡±ç `anyNA` ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞°‡∞Ç, ‡∞á‡∞¶‡∞ø ‡∞≤‡∞æ‡∞ú‡∞ø‡∞ï‡∞≤‡±ç ‡∞Ü‡∞¨‡±ç‡∞ú‡±Ü‡∞ï‡±ç‡∞ü‡±Å‡∞≤‡±Å `TRUE` ‡∞≤‡±á‡∞¶‡∞æ `FALSE` ‡∞®‡±Å ‡∞§‡∞ø‡∞∞‡∞ø‡∞ó‡∞ø ‡∞á‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø.


```r
pumpkins %>% 
  anyNA()
```

‡∞ö‡∞æ‡∞≤‡∞æ ‡∞¨‡∞æ‡∞ó‡±Å‡∞Ç‡∞¶‡∞ø, ‡∞ï‡±ä‡∞Ç‡∞§ ‡∞°‡±á‡∞ü‡∞æ ‡∞≤‡±á‡∞Æ‡∞ø ‡∞â‡∞Ç‡∞¶‡∞ø ‡∞Ö‡∞®‡∞ø‡∞™‡∞ø‡∞∏‡±ç‡∞§‡±ã‡∞Ç‡∞¶‡∞ø! ‡∞Ö‡∞¶‡∞ø ‡∞™‡±ç‡∞∞‡∞æ‡∞∞‡∞Ç‡∞≠‡∞ø‡∞Ç‡∞ö‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø ‡∞Æ‡∞Ç‡∞ö‡∞ø ‡∞∏‡±ç‡∞•‡∞≤‡∞Ç.

-   ‡∞Æ‡∞∞‡±ã ‡∞Æ‡∞æ‡∞∞‡±ç‡∞ó‡∞Ç `is.na()` ‡∞´‡∞Ç‡∞ï‡±ç‡∞∑‡∞®‡±ç ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞°‡∞Ç, ‡∞á‡∞¶‡∞ø ‡∞è ‡∞µ‡±ç‡∞Ø‡∞ï‡±ç‡∞§‡∞ø‡∞ó‡∞§ ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç ‡∞Ö‡∞Ç‡∞∂‡∞æ‡∞≤‡±Å ‡∞≤‡±á‡∞Æ‡∞ø ‡∞â‡∞®‡±ç‡∞®‡∞æ‡∞Ø‡±ã ‡∞≤‡∞æ‡∞ú‡∞ø‡∞ï‡∞≤‡±ç `TRUE` ‡∞§‡±ã ‡∞∏‡±Ç‡∞ö‡∞ø‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø.


```r
pumpkins %>% 
  is.na() %>% 
  head(n = 7)
```

‡∞∏‡∞∞‡±á, ‡∞™‡∞®‡∞ø ‡∞™‡±Ç‡∞∞‡±ç‡∞§‡∞Ø‡∞ø‡∞Ç‡∞¶‡∞ø ‡∞ï‡∞æ‡∞®‡±Ä ‡∞á‡∞≤‡∞æ‡∞Ç‡∞ü‡∞ø ‡∞™‡±Ü‡∞¶‡±ç‡∞¶ ‡∞°‡±á‡∞ü‡∞æ ‡∞´‡±ç‡∞∞‡±á‡∞Æ‡±ç‚Äå‡∞§‡±ã, ‡∞™‡±ç‡∞∞‡∞§‡∞ø ‡∞µ‡∞∞‡±Å‡∞∏ ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‚Äå‡∞®‡±Å ‡∞µ‡±ç‡∞Ø‡∞ï‡±ç‡∞§‡∞ø‡∞ó‡∞§‡∞Ç‡∞ó‡∞æ ‡∞∏‡∞Æ‡±Ä‡∞ï‡±ç‡∞∑‡∞ø‡∞Ç‡∞ö‡∞°‡∞Ç ‡∞Ö‡∞®‡∞∞‡±ç‡∞•‡∞ï‡∞Ç‡∞ó‡∞æ ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å ‡∞™‡±ç‡∞∞‡∞æ‡∞Ø‡±ã‡∞ó‡∞ø‡∞ï‡∞Ç‡∞ó‡∞æ ‡∞Ö‡∞∏‡∞æ‡∞ß‡±ç‡∞Ø‡∞Ç ‡∞Ö‡∞µ‡±Å‡∞§‡±Å‡∞Ç‡∞¶‡∞øüò¥.

-   ‡∞Æ‡∞∞‡∞ø‡∞Ç‡∞§ ‡∞∏‡±Å‡∞≤‡∞≠‡∞Æ‡±à‡∞® ‡∞µ‡∞ø‡∞ß‡∞æ‡∞®‡∞Ç ‡∞™‡±ç‡∞∞‡∞§‡∞ø ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç ‡∞ï‡±ã‡∞∏‡∞Ç ‡∞≤‡±á‡∞®‡∞ø ‡∞µ‡∞ø‡∞≤‡±Å‡∞µ‡∞≤ ‡∞Æ‡±ä‡∞§‡±ç‡∞§‡∞Ç ‡∞≤‡±Ü‡∞ï‡±ç‡∞ï‡∞ø‡∞Ç‡∞ö‡∞°‡∞Ç ‡∞Ö‡∞µ‡±Å‡∞§‡±Å‡∞Ç‡∞¶‡∞ø:


```r
pumpkins %>% 
  is.na() %>% 
  colSums()
```

‡∞á‡∞Ç‡∞ï‡∞æ ‡∞¨‡∞æ‡∞ó‡±Å‡∞Ç‡∞¶‡∞ø! ‡∞ï‡±ä‡∞Ç‡∞§ ‡∞°‡±á‡∞ü‡∞æ ‡∞Æ‡∞ø‡∞∏‡±ç‡∞∏‡∞µ‡±Å‡∞Ç‡∞¶‡∞ø, ‡∞ï‡∞æ‡∞®‡±Ä ‡∞à ‡∞™‡∞®‡∞ø‡∞ï‡∞ø ‡∞Ö‡∞¶‡∞ø ‡∞™‡±Ü‡∞¶‡±ç‡∞¶ ‡∞∏‡∞Æ‡∞∏‡±ç‡∞Ø ‡∞ï‡∞æ‡∞ï‡∞™‡±ã‡∞µ‡∞ö‡±ç‡∞ö‡±Å. ‡∞Æ‡∞∞‡∞ø‡∞Ç‡∞§ ‡∞µ‡∞ø‡∞∂‡±ç‡∞≤‡±á‡∞∑‡∞£ ‡∞è‡∞Æ‡∞ø ‡∞§‡±Ü‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡±ã ‡∞ö‡±Ç‡∞¶‡±ç‡∞¶‡∞æ‡∞Ç.

#### dplyr::select()

`select()` ‡∞Ö‡∞®‡±á‡∞¶‡∞ø `dplyr` ‡∞™‡±ç‡∞Ø‡∞æ‡∞ï‡±á‡∞ú‡±Ä‡∞≤‡±ã ‡∞â‡∞®‡±ç‡∞® ‡∞í‡∞ï ‡∞´‡∞Ç‡∞ï‡±ç‡∞∑‡∞®‡±ç, ‡∞á‡∞¶‡∞ø ‡∞Æ‡±Ä‡∞∞‡±Å ‡∞®‡∞ø‡∞≤‡±Å‡∞™‡±Å‡∞ï‡±ã‡∞µ‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø ‡∞≤‡±á‡∞¶‡∞æ ‡∞§‡∞™‡±ç‡∞™‡∞ø‡∞Ç‡∞ö‡±Å‡∞ï‡±ã‡∞µ‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‡∞∏‡±ç‚Äå‡∞®‡±Å ‡∞é‡∞Ç‡∞ö‡±Å‡∞ï‡±ã‡∞µ‡∞°‡∞Ç‡∞≤‡±ã ‡∞∏‡∞π‡∞æ‡∞Ø‡∞™‡∞°‡±Å‡∞§‡±Å‡∞Ç‡∞¶‡∞ø.

‡∞Æ‡±Ä ‡∞°‡±á‡∞ü‡∞æ ‡∞´‡±ç‡∞∞‡±á‡∞Æ‡±ç‚Äå‡∞§‡±ã ‡∞™‡∞®‡∞ø ‡∞ö‡±á‡∞Ø‡∞°‡∞Ç ‡∞∏‡±Å‡∞≤‡∞≠‡∞Ç ‡∞ö‡±á‡∞Ø‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø, `select()` ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞ø ‡∞¶‡∞æ‡∞®‡∞ø ‡∞ï‡±ä‡∞®‡±ç‡∞®‡∞ø ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‡∞∏‡±ç‚Äå‡∞®‡±Å ‡∞§‡±ä‡∞≤‡∞ó‡∞ø‡∞Ç‡∞ö‡∞Ç‡∞°‡∞ø, ‡∞Æ‡±Ä‡∞∞‡±Å ‡∞Ö‡∞µ‡∞∏‡∞∞‡∞Ç ‡∞â‡∞®‡±ç‡∞® ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‡∞∏‡±ç ‡∞Æ‡∞æ‡∞§‡±ç‡∞∞‡∞Æ‡±á ‡∞®‡∞ø‡∞≤‡±Å‡∞™‡±Å‡∞ï‡±ã‡∞Ç‡∞°‡∞ø.

‡∞â‡∞¶‡∞æ‡∞π‡∞∞‡∞£‡∞ï‡±Å, ‡∞à ‡∞µ‡±ç‡∞Ø‡∞æ‡∞Ø‡∞æ‡∞Æ‡∞Ç‡∞≤‡±ã, ‡∞Æ‡∞æ ‡∞µ‡∞ø‡∞∂‡±ç‡∞≤‡±á‡∞∑‡∞£‡∞≤‡±ã `Package`, `Low Price`, `High Price` ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å `Date` ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‡∞∏‡±ç ‡∞â‡∞Ç‡∞ü‡∞æ‡∞Ø‡∞ø. ‡∞à ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‡∞∏‡±ç‚Äå‡∞®‡±Å ‡∞é‡∞Ç‡∞ö‡±Å‡∞ï‡±Å‡∞Ç‡∞¶‡∞æ‡∞Ç.


```r
# Select desired columns
pumpkins <- pumpkins %>% 
  select(Package, `Low Price`, `High Price`, Date)


# Print data set
pumpkins %>% 
  slice_head(n = 5)
```

#### dplyr::mutate()

`mutate()` ‡∞Ö‡∞®‡±á‡∞¶‡∞ø `dplyr` ‡∞™‡±ç‡∞Ø‡∞æ‡∞ï‡±á‡∞ú‡±Ä‡∞≤‡±ã ‡∞í‡∞ï ‡∞´‡∞Ç‡∞ï‡±ç‡∞∑‡∞®‡±ç, ‡∞á‡∞¶‡∞ø ‡∞Æ‡±Ä‡∞∞‡±Å ‡∞â‡∞®‡±ç‡∞® ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‡∞∏‡±ç‚Äå‡∞®‡±Å ‡∞â‡∞Ç‡∞ö‡±Å‡∞§‡±Ç ‡∞ï‡±ä‡∞§‡±ç‡∞§ ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‡∞∏‡±ç‚Äå‡∞®‡±Å ‡∞∏‡±É‡∞∑‡±ç‡∞ü‡∞ø‡∞Ç‡∞ö‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø ‡∞≤‡±á‡∞¶‡∞æ ‡∞Æ‡∞æ‡∞∞‡±ç‡∞ö‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø ‡∞∏‡∞π‡∞æ‡∞Ø‡∞™‡∞°‡±Å‡∞§‡±Å‡∞Ç‡∞¶‡∞ø.

mutate ‡∞Ø‡±ä‡∞ï‡±ç‡∞ï ‡∞∏‡∞æ‡∞ß‡∞æ‡∞∞‡∞£ ‡∞®‡∞ø‡∞∞‡±ç‡∞Æ‡∞æ‡∞£‡∞Ç:

`data %>%   mutate(new_column_name = what_it_contains)`

`Date` ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‚Äå‡∞®‡±Å ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞ø ‡∞ï‡±ç‡∞∞‡∞ø‡∞Ç‡∞¶‡∞ø ‡∞Ü‡∞™‡∞∞‡±á‡∞∑‡∞®‡±ç‡∞≤‡∞®‡±Å ‡∞ö‡±á‡∞Ø‡∞°‡∞Ç ‡∞¶‡±ç‡∞µ‡∞æ‡∞∞‡∞æ `mutate` ‡∞®‡±Å ‡∞™‡±ç‡∞∞‡∞Ø‡∞§‡±ç‡∞®‡∞ø‡∞¶‡±ç‡∞¶‡∞æ‡∞Ç:

1. ‡∞§‡±á‡∞¶‡±Ä‡∞≤‡∞®‡±Å (‡∞™‡±ç‡∞∞‡∞∏‡±ç‡∞§‡±Å‡∞§‡∞Ç character ‡∞∞‡∞ï‡∞Ç) ‡∞®‡±Ü‡∞≤ ‡∞´‡∞æ‡∞∞‡±ç‡∞Æ‡∞æ‡∞ü‡±ç‚Äå‡∞ï‡±Å ‡∞Æ‡∞æ‡∞∞‡±ç‡∞ö‡∞Ç‡∞°‡∞ø (‡∞á‡∞µ‡∞ø US ‡∞§‡±á‡∞¶‡±Ä‡∞≤‡±Å, ‡∞ï‡∞æ‡∞¨‡∞ü‡±ç‡∞ü‡∞ø ‡∞´‡∞æ‡∞∞‡±ç‡∞Æ‡∞æ‡∞ü‡±ç `MM/DD/YYYY`).

2. ‡∞§‡±á‡∞¶‡±Ä‡∞≤ ‡∞®‡±Å‡∞Ç‡∞°‡∞ø ‡∞®‡±Ü‡∞≤‡∞®‡±Å ‡∞ï‡±ä‡∞§‡±ç‡∞§ ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‚Äå‡∞ó‡∞æ ‡∞§‡±Ä‡∞∏‡±Å‡∞ï‡±ã‡∞Ç‡∞°‡∞ø.

R ‡∞≤‡±ã, [lubridate](https://lubridate.tidyverse.org/) ‡∞™‡±ç‡∞Ø‡∞æ‡∞ï‡±á‡∞ú‡±Ä Date-time ‡∞°‡±á‡∞ü‡∞æ‡∞§‡±ã ‡∞™‡∞®‡∞ø ‡∞ö‡±á‡∞Ø‡∞°‡∞Ç ‡∞∏‡±Å‡∞≤‡∞≠‡∞§‡∞∞‡∞Ç ‡∞ö‡±á‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø. ‡∞ï‡∞æ‡∞¨‡∞ü‡±ç‡∞ü‡∞ø, `dplyr::mutate()`, `lubridate::mdy()`, `lubridate::month()` ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞ø ‡∞™‡±à ‡∞≤‡∞ï‡±ç‡∞∑‡±ç‡∞Ø‡∞æ‡∞≤‡∞®‡±Å ‡∞é‡∞≤‡∞æ ‡∞∏‡∞æ‡∞ß‡∞ø‡∞Ç‡∞ö‡∞æ‡∞≤‡±ã ‡∞ö‡±Ç‡∞¶‡±ç‡∞¶‡∞æ‡∞Ç. ‡∞§‡∞∞‡±Å‡∞µ‡∞æ‡∞§‡∞ø ‡∞Ü‡∞™‡∞∞‡±á‡∞∑‡∞®‡±ç‡∞≤‡∞≤‡±ã ‡∞Æ‡∞≥‡±ç‡∞≤‡±Ä ‡∞Ö‡∞µ‡∞∏‡∞∞‡∞Ç ‡∞≤‡±á‡∞ï‡∞™‡±ã‡∞µ‡∞°‡∞Ç‡∞§‡±ã Date ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‚Äå‡∞®‡±Å ‡∞§‡±Ä‡∞∏‡∞ø‡∞µ‡±á‡∞Ø‡∞µ‡∞ö‡±ç‡∞ö‡±Å.


```r
# Load lubridate
library(lubridate)

pumpkins <- pumpkins %>% 
  # Convert the Date column to a date object
  mutate(Date = mdy(Date)) %>% 
  # Extract month from Date
  mutate(Month = month(Date)) %>% 
  # Drop Date column
  select(-Date)

# View the first few rows
pumpkins %>% 
  slice_head(n = 7)
```

```r
# Create a new column Price
pumpkins <- pumpkins %>% 
  mutate(Price = (`Low Price` + `High Price`)/2)

# View the first few rows of the data
pumpkins %>% 
  slice_head(n = 5)
```

‡∞¶‡±Ä‡∞®‡∞ø‡∞®‡∞ø ‡∞®‡∞ø‡∞∞‡±ç‡∞ß‡∞æ‡∞∞‡∞ø‡∞¶‡±ç‡∞¶‡∞æ‡∞Ç:

#### dplyr::filter() ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å stringr::str_detect()

[`dplyr::filter()`](https://dplyr.tidyverse.org/reference/filter.html): ‡∞Æ‡±Ä ‡∞∑‡∞∞‡∞§‡±Å‡∞≤‡∞®‡±Å ‡∞§‡±Ä‡∞∞‡±Å‡∞∏‡±ç‡∞§‡±Å‡∞®‡±ç‡∞® **‡∞™‡∞Ç‡∞ï‡±ç‡∞§‡±Å‡∞≤‡±Å** ‡∞Æ‡∞æ‡∞§‡±ç‡∞∞‡∞Æ‡±á ‡∞ï‡∞≤‡∞ø‡∞ó‡∞ø‡∞® ‡∞°‡±á‡∞ü‡∞æ ‡∞â‡∞™‡∞∏‡±Ü‡∞ü‡±ç‚Äå‡∞®‡±Å ‡∞∏‡±É‡∞∑‡±ç‡∞ü‡∞ø‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø, ‡∞à ‡∞∏‡∞Ç‡∞¶‡∞∞‡±ç‡∞≠‡∞Ç‡∞≤‡±ã, `Package` ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‚Äå‡∞≤‡±ã *bushel* ‡∞∏‡±ç‡∞ü‡±ç‡∞∞‡∞ø‡∞Ç‡∞ó‡±ç ‡∞â‡∞®‡±ç‡∞® ‡∞™‡∞Ç‡∞™‡±ç‡∞ï‡∞ø‡∞®‡±ç‡∞≤‡±Å.

[stringr::str_detect()](https://stringr.tidyverse.org/reference/str_detect.html): ‡∞∏‡±ç‡∞ü‡±ç‡∞∞‡∞ø‡∞Ç‡∞ó‡±ç‚Äå‡∞≤‡±ã ‡∞í‡∞ï ‡∞®‡∞Æ‡±Ç‡∞®‡∞æ ‡∞â‡∞®‡±ç‡∞®‡∞¶‡±ã ‡∞≤‡±á‡∞¶‡±ã ‡∞ó‡±Å‡∞∞‡±ç‡∞§‡∞ø‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø.

[`stringr`](https://github.com/tidyverse/stringr) ‡∞™‡±ç‡∞Ø‡∞æ‡∞ï‡±á‡∞ú‡±ç ‡∞∏‡∞æ‡∞ß‡∞æ‡∞∞‡∞£ ‡∞∏‡±ç‡∞ü‡±ç‡∞∞‡∞ø‡∞Ç‡∞ó‡±ç ‡∞Ü‡∞™‡∞∞‡±á‡∞∑‡∞®‡±ç‡∞≤ ‡∞ï‡±ã‡∞∏‡∞Ç ‡∞∏‡±Å‡∞≤‡∞≠‡∞Æ‡±à‡∞® ‡∞´‡∞Ç‡∞ï‡±ç‡∞∑‡∞®‡±ç‡∞≤‡∞®‡±Å ‡∞Ö‡∞Ç‡∞¶‡∞ø‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø.


```r
# Retain only pumpkins with "bushel"
new_pumpkins <- pumpkins %>% 
       filter(str_detect(Package, "bushel"))

# Get the dimensions of the new data
dim(new_pumpkins)

# View a few rows of the new data
new_pumpkins %>% 
  slice_head(n = 5)
```

#### dplyr::case_when()

**‡∞ï‡∞æ‡∞®‡±Ä ‡∞µ‡±á‡∞ö‡∞ø ‡∞â‡∞Ç‡∞°‡∞Ç‡∞°‡∞ø! ‡∞á‡∞Ç‡∞ï‡∞æ ‡∞í‡∞ï ‡∞µ‡∞ø‡∞∑‡∞Ø‡∞Ç ‡∞ö‡±á‡∞Ø‡∞æ‡∞≤‡±ç‡∞∏‡∞ø ‡∞â‡∞Ç‡∞¶‡∞ø**

‡∞Æ‡±Ä‡∞∞‡±Å ‡∞ó‡∞Æ‡∞®‡∞ø‡∞Ç‡∞ö‡∞æ‡∞∞‡∞æ, ‡∞¨‡∞∑‡±Ü‡∞≤‡±ç ‡∞™‡∞∞‡∞ø‡∞Æ‡∞æ‡∞£‡∞Ç ‡∞™‡±ç‡∞∞‡∞§‡∞ø ‡∞µ‡∞∞‡±Å‡∞∏‡∞ï‡±Å ‡∞Æ‡∞æ‡∞∞‡±Å‡∞§‡±Å‡∞Ç‡∞¶‡∞ø? ‡∞Æ‡±Ä‡∞∞‡±Å ‡∞ß‡∞∞‡∞≤‡∞®‡±Å ‡∞∏‡∞æ‡∞ß‡∞æ‡∞∞‡∞£‡±Ä‡∞ï‡∞∞‡∞ø‡∞Ç‡∞ö‡∞æ‡∞≤‡∞ø, ‡∞Ö‡∞Ç‡∞ü‡±á 1 1/9 ‡∞≤‡±á‡∞¶‡∞æ 1/2 ‡∞¨‡∞∑‡±Ü‡∞≤‡±ç‚Äå‡∞ï‡±Å ‡∞ï‡∞æ‡∞ï‡±Å‡∞Ç‡∞°‡∞æ ‡∞™‡±ç‡∞∞‡∞§‡∞ø ‡∞¨‡∞∑‡±Ü‡∞≤‡±ç‚Äå‡∞ï‡±Å ‡∞ß‡∞∞ ‡∞ö‡±Ç‡∞™‡∞ø‡∞Ç‡∞ö‡∞æ‡∞≤‡∞ø. ‡∞¶‡±Ä‡∞®‡∞ø ‡∞ï‡±ã‡∞∏‡∞Ç ‡∞ï‡±ä‡∞Ç‡∞§ ‡∞ó‡∞£‡∞ø‡∞§‡∞Ç ‡∞ö‡±á‡∞Ø‡∞æ‡∞≤‡±ç‡∞∏‡∞ø ‡∞â‡∞Ç‡∞¶‡∞ø.

‡∞Æ‡∞®‡∞Ç ‡∞ï‡±ä‡∞®‡±ç‡∞®‡∞ø ‡∞∑‡∞∞‡∞§‡±Å‡∞≤ ‡∞Ü‡∞ß‡∞æ‡∞∞‡∞Ç‡∞ó‡∞æ Price ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‚Äå‡∞®‡±Å *mutate* ‡∞ö‡±á‡∞Ø‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø [`case_when()`](https://dplyr.tidyverse.org/reference/case_when.html) ‡∞´‡∞Ç‡∞ï‡±ç‡∞∑‡∞®‡±ç‚Äå‡∞®‡±Å ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞∏‡±ç‡∞§‡∞æ‡∞Æ‡±Å. `case_when` ‡∞Ö‡∞®‡±á‡∞¶‡∞ø ‡∞Ö‡∞®‡±á‡∞ï `if_else()` ‡∞∏‡±ç‡∞ü‡±á‡∞ü‡±ç‡∞Æ‡±Ü‡∞Ç‡∞ü‡±ç‡∞≤‡∞®‡±Å ‡∞µ‡±Ü‡∞ï‡±ç‡∞ü‡∞∞‡±à‡∞ú‡±ç ‡∞ö‡±á‡∞Ø‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø ‡∞Ö‡∞®‡±Å‡∞Æ‡∞§‡∞ø‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø.


```r
# Convert the price if the Package contains fractional bushel values
new_pumpkins <- new_pumpkins %>% 
  mutate(Price = case_when(
    str_detect(Package, "1 1/9") ~ Price/(1 + 1/9),
    str_detect(Package, "1/2") ~ Price/(1/2),
    TRUE ~ Price))

# View the first few rows of the data
new_pumpkins %>% 
  slice_head(n = 30)
```

‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç ‡∞∏‡±ç‡∞•‡∞æ‡∞®‡∞æ‡∞≤‡∞®‡±Å ‡∞Æ‡∞æ‡∞∞‡±ç‡∞ö‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø `dplyr::relocate()` ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞∏‡±ç‡∞§‡∞æ‡∞∞‡±Å.

R ‡∞Ö‡∞®‡±á‡∞ï ‡∞ó‡±ç‡∞∞‡∞æ‡∞´‡±ç ‡∞§‡∞Ø‡∞æ‡∞∞‡±Ä ‡∞µ‡±ç‡∞Ø‡∞µ‡∞∏‡±ç‡∞•‡∞≤‡∞®‡±Å ‡∞Ö‡∞Ç‡∞¶‡∞ø‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø, ‡∞ï‡∞æ‡∞®‡±Ä [`ggplot2`](https://ggplot2.tidyverse.org/index.html) ‡∞Ö‡∞§‡±ç‡∞Ø‡∞Ç‡∞§ ‡∞Ö‡∞Ç‡∞¶‡∞Æ‡±à‡∞® ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å ‡∞¨‡∞π‡±Å‡∞Æ‡±Å‡∞ñ‡∞Æ‡±à‡∞® ‡∞µ‡∞æ‡∞ü‡∞ø‡∞≤‡±ã ‡∞í‡∞ï‡∞ü‡∞ø. `ggplot2` ‡∞Æ‡±Ä‡∞∞‡±Å **‡∞∏‡±ç‡∞µ‡∞§‡∞Ç‡∞§‡±ç‡∞∞ ‡∞≠‡∞æ‡∞ó‡∞æ‡∞≤‡∞®‡±Å ‡∞ï‡∞≤‡∞ø‡∞™‡∞ø** ‡∞ó‡±ç‡∞∞‡∞æ‡∞´‡±ç‚Äå‡∞≤‡∞®‡±Å ‡∞∞‡±Ç‡∞™‡±ä‡∞Ç‡∞¶‡∞ø‡∞Ç‡∞ö‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø ‡∞Ö‡∞®‡±Å‡∞Æ‡∞§‡∞ø‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø.

Price ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å Month ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‡∞∏‡±ç ‡∞ï‡±ã‡∞∏‡∞Ç ‡∞í‡∞ï ‡∞∏‡∞æ‡∞¶‡∞æ ‡∞∏‡±ç‡∞ï‡∞æ‡∞ü‡∞∞‡±ç ‡∞™‡±ç‡∞≤‡∞æ‡∞ü‡±ç ‡∞§‡±ã ‡∞™‡±ç‡∞∞‡∞æ‡∞∞‡∞Ç‡∞≠‡∞ø‡∞¶‡±ç‡∞¶‡∞æ‡∞Ç.

‡∞ö‡∞æ‡∞∞‡±ç‡∞ü‡±ç‡∞≤‡±Å ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ï‡∞∞‡∞Æ‡±à‡∞® ‡∞°‡±á‡∞ü‡∞æ‡∞®‡±Å ‡∞™‡±ç‡∞∞‡∞¶‡∞∞‡±ç‡∞∂‡∞ø‡∞Ç‡∞ö‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø, ‡∞Æ‡±Ä‡∞∞‡±Å ‡∞∏‡∞æ‡∞ß‡∞æ‡∞∞‡∞£‡∞Ç‡∞ó‡∞æ ‡∞°‡±á‡∞ü‡∞æ‡∞®‡±Å ‡∞è‡∞¶‡±ã ‡∞µ‡∞ø‡∞ß‡∞Ç‡∞ó‡∞æ ‡∞∏‡∞Æ‡±Ç‡∞π‡±Ä‡∞ï‡∞∞‡∞ø‡∞Ç‡∞ö‡∞æ‡∞≤‡∞ø. ‡∞â‡∞¶‡∞æ‡∞π‡∞∞‡∞£‡∞ï‡±Å, ‡∞Æ‡∞® ‡∞ï‡±á‡∞∏‡±Å‡∞≤‡±ã, ‡∞™‡±ç‡∞∞‡∞§‡∞ø ‡∞®‡±Ü‡∞≤‡∞ï‡±Å ‡∞ó‡±Å‡∞Æ‡±ç‡∞Æ‡∞°‡∞ø‡∞ï‡∞æ‡∞Ø‡∞≤ ‡∞∏‡∞ó‡∞ü‡±Å ‡∞ß‡∞∞‡∞®‡±Å ‡∞ï‡∞®‡±Å‡∞ó‡±ä‡∞®‡∞°‡∞Ç ‡∞Æ‡∞® ‡∞°‡±á‡∞ü‡∞æ‡∞≤‡±ã‡∞®‡∞ø ‡∞Ö‡∞Ç‡∞§‡∞∞‡±ç‡∞ó‡∞§ ‡∞®‡∞Æ‡±Ç‡∞®‡∞æ‡∞≤‡∞™‡±à ‡∞Æ‡∞∞‡∞ø‡∞Ç‡∞§ ‡∞Ö‡∞µ‡∞ó‡∞æ‡∞π‡∞®‡∞®‡±Å ‡∞Ö‡∞Ç‡∞¶‡∞ø‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø. ‡∞á‡∞¶‡∞ø ‡∞Æ‡∞æ‡∞ï‡±Å ‡∞Æ‡∞∞‡±ä‡∞ï **dplyr** ‡∞´‡±ç‡∞≤‡±à‡∞¨‡±à‡∞ï‡∞ø ‡∞¶‡∞æ‡∞∞‡∞ø‡∞§‡±Ä‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø:

#### `dplyr::group_by() %>% summarize()`

R ‡∞≤‡±ã ‡∞∏‡∞Æ‡±Ç‡∞π‡±Ä‡∞ï‡±É‡∞§ ‡∞∏‡∞Æ‡±ç‡∞Æ‡±á‡∞≥‡∞®‡∞Ç ‡∞∏‡±Å‡∞≤‡∞≠‡∞Ç‡∞ó‡∞æ ‡∞≤‡±Ü‡∞ï‡±ç‡∞ï‡∞ø‡∞Ç‡∞ö‡∞µ‡∞ö‡±ç‡∞ö‡±Å

`dplyr::group_by() %>% summarize()`

- `dplyr::group_by()` ‡∞µ‡∞ø‡∞∂‡±ç‡∞≤‡±á‡∞∑‡∞£ ‡∞Ø‡±Ç‡∞®‡∞ø‡∞ü‡±ç‚Äå‡∞®‡±Å ‡∞™‡±Ç‡∞∞‡±ç‡∞§‡∞ø ‡∞°‡±á‡∞ü‡∞æ‡∞∏‡±Ü‡∞ü‡±ç ‡∞®‡±Å‡∞Ç‡∞°‡∞ø ‡∞µ‡±ç‡∞Ø‡∞ï‡±ç‡∞§‡∞ø‡∞ó‡∞§ ‡∞∏‡∞Æ‡±Ç‡∞π‡∞æ‡∞≤‡∞ï‡±Å ‡∞Æ‡∞æ‡∞∞‡±ç‡∞ö‡±Å‡∞§‡±Å‡∞Ç‡∞¶‡∞ø, ‡∞â‡∞¶‡∞æ‡∞π‡∞∞‡∞£‡∞ï‡±Å ‡∞™‡±ç‡∞∞‡∞§‡∞ø ‡∞®‡±Ü‡∞≤‡∞ï‡±Å.

- `dplyr::summarize()` ‡∞Æ‡±Ä‡∞∞‡±Å ‡∞™‡±á‡∞∞‡±ç‡∞ï‡±ä‡∞®‡±ç‡∞® ‡∞∏‡∞Æ‡±ç‡∞Æ‡∞∞‡±Ä ‡∞ó‡∞£‡∞æ‡∞Ç‡∞ï‡∞æ‡∞≤ ‡∞ï‡±ã‡∞∏‡∞Ç ‡∞™‡±ç‡∞∞‡∞§‡∞ø ‡∞∏‡∞Æ‡±Ç‡∞π‡±Ä‡∞ï‡∞∞‡∞£ ‡∞µ‡±á‡∞∞‡∞ø‡∞Ø‡∞¨‡±Å‡∞≤‡±ç‚Äå‡∞ï‡±Å ‡∞í‡∞ï ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å ‡∞í‡∞ï ‡∞ï‡±ä‡∞§‡±ç‡∞§ ‡∞°‡±á‡∞ü‡∞æ ‡∞´‡±ç‡∞∞‡±á‡∞Æ‡±ç‚Äå‡∞®‡±Å ‡∞∏‡±É‡∞∑‡±ç‡∞ü‡∞ø‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø.

‡∞â‡∞¶‡∞æ‡∞π‡∞∞‡∞£‡∞ï‡±Å, ‡∞Æ‡∞®‡∞Ç `dplyr::group_by() %>% summarize()` ‡∞®‡±Å ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞ø ‡∞ó‡±Å‡∞Æ‡±ç‡∞Æ‡∞°‡∞ø‡∞ï‡∞æ‡∞Ø‡∞≤‡∞®‡±Å **Month** ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç ‡∞Ü‡∞ß‡∞æ‡∞∞‡∞Ç‡∞ó‡∞æ ‡∞∏‡∞Æ‡±Ç‡∞π‡∞æ‡∞≤‡±Å‡∞ó‡∞æ ‡∞µ‡∞ø‡∞≠‡∞ú‡∞ø‡∞Ç‡∞ö‡∞ø, ‡∞™‡±ç‡∞∞‡∞§‡∞ø ‡∞®‡±Ü‡∞≤‡∞ï‡±Å **‡∞∏‡∞ó‡∞ü‡±Å ‡∞ß‡∞∞** ‡∞ï‡∞®‡±Å‡∞ó‡±ä‡∞®‡∞µ‡∞ö‡±ç‡∞ö‡±Å.


```r
# Find the average price of pumpkins per month
new_pumpkins %>%
  group_by(Month) %>% 
  summarise(mean_price = mean(Price))
```

‡∞∏‡∞Ç‡∞ï‡±ç‡∞∑‡∞ø‡∞™‡±ç‡∞§‡∞Ç‡∞ó‡∞æ!‚ú®

‡∞®‡±Ü‡∞≤‡∞≤ ‡∞µ‡∞Ç‡∞ü‡∞ø ‡∞µ‡∞∞‡±ç‡∞ó‡±Ä‡∞ï‡∞∞‡∞£ ‡∞≤‡∞ï‡±ç‡∞∑‡∞£‡∞æ‡∞≤‡∞®‡±Å ‡∞¨‡∞æ‡∞∞‡±ç ‡∞™‡±ç‡∞≤‡∞æ‡∞ü‡±ç üìä ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞ø ‡∞Æ‡±Ü‡∞∞‡±Å‡∞ó‡±ç‡∞ó‡∞æ ‡∞™‡±ç‡∞∞‡∞¶‡∞∞‡±ç‡∞∂‡∞ø‡∞Ç‡∞ö‡∞µ‡∞ö‡±ç‡∞ö‡±Å. ‡∞¨‡∞æ‡∞∞‡±ç ‡∞ö‡∞æ‡∞∞‡±ç‡∞ü‡±ç‡∞∏‡±ç ‡∞ï‡±ã‡∞∏‡∞Ç ‡∞¨‡∞æ‡∞ß‡±ç‡∞Ø‡∞§ ‡∞µ‡∞π‡∞ø‡∞Ç‡∞ö‡±á ‡∞≤‡±á‡∞Ø‡∞∞‡±ç‡∞≤‡±Å `geom_bar()` ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å `geom_col()` ‡∞â‡∞®‡±ç‡∞®‡∞æ‡∞Ø‡∞ø. ‡∞Æ‡∞∞‡∞ø‡∞Ç‡∞§ ‡∞§‡±Ü‡∞≤‡±Å‡∞∏‡±Å‡∞ï‡±ã‡∞µ‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø `?geom_bar` ‡∞®‡±Å ‡∞ö‡±Ç‡∞°‡∞Ç‡∞°‡∞ø.

‡∞í‡∞ï‡∞ü‡∞ø ‡∞§‡∞Ø‡∞æ‡∞∞‡±Å ‡∞ö‡±á‡∞¶‡±ç‡∞¶‡∞æ‡∞Ç!


```r
# Find the average price of pumpkins per month then plot a bar chart
new_pumpkins %>%
  group_by(Month) %>% 
  summarise(mean_price = mean(Price)) %>% 
  ggplot(aes(x = Month, y = mean_price)) +
  geom_col(fill = "midnightblue", alpha = 0.7) +
  ylab("Pumpkin Price")
```
