‡∞Ü‡∞∏‡∞ï‡±ç‡∞§‡∞ø‡∞ï‡∞∞‡∞Ç‡∞ó‡∞æ ‡∞â‡∞Ç‡∞¶‡∞ø! ‡∞¶‡±É‡∞∑‡±ç‡∞ü‡∞ø‡∞≤‡±ã‡∞ï‡∞ø ‡∞§‡±Ä‡∞∏‡±Å‡∞ï‡±Å‡∞Ç‡∞ü‡±á, ‡∞Æ‡±ä‡∞¶‡∞ü‡∞ø ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç ‡∞í‡∞ï ‡∞∞‡∞ï‡∞Æ‡±à‡∞® `id` ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç ‡∞Ö‡∞®‡∞ø ‡∞Ö‡∞®‡∞ø‡∞™‡∞ø‡∞∏‡±ç‡∞§‡±ã‡∞Ç‡∞¶‡∞ø. ‡∞°‡±á‡∞ü‡∞æ ‡∞ó‡±Å‡∞∞‡∞ø‡∞Ç‡∞ö‡∞ø ‡∞Æ‡∞∞‡∞ø‡∞Ç‡∞§ ‡∞∏‡∞Æ‡∞æ‡∞ö‡∞æ‡∞∞‡∞Ç ‡∞™‡±ä‡∞Ç‡∞¶‡±Å‡∞ï‡±Å‡∞Ç‡∞¶‡∞æ‡∞Ç.

## ‡∞µ‡±ç‡∞Ø‡∞æ‡∞Ø‡∞æ‡∞Æ‡∞Ç - ‡∞µ‡∞Ç‡∞ü‡∞ï‡∞æ‡∞≤ ‡∞ó‡±Å‡∞∞‡∞ø‡∞Ç‡∞ö‡∞ø ‡∞§‡±Ü‡∞≤‡±Å‡∞∏‡±Å‡∞ï‡±ã‡∞µ‡∞°‡∞Ç

‡∞á‡∞™‡±ç‡∞™‡±Å‡∞°‡±Å ‡∞™‡∞®‡∞ø ‡∞Æ‡∞∞‡∞ø‡∞Ç‡∞§ ‡∞Ü‡∞∏‡∞ï‡±ç‡∞§‡∞ø‡∞ï‡∞∞‡∞Ç‡∞ó‡∞æ ‡∞Æ‡∞æ‡∞∞‡±Å‡∞§‡±Å‡∞Ç‡∞¶‡∞ø. ‡∞µ‡∞Ç‡∞ü‡∞ï‡∞æ‡∞≤ ‡∞™‡±ç‡∞∞‡∞ï‡∞æ‡∞∞‡∞Ç ‡∞°‡±á‡∞ü‡∞æ ‡∞™‡∞Ç‡∞™‡∞ø‡∞£‡±Ä‡∞®‡∞ø ‡∞ï‡∞®‡±Å‡∞ó‡±ä‡∞®‡±Å‡∞ï‡±Å‡∞Ç‡∞¶‡∞æ‡∞Ç.


```r
# Count observations per cuisine
df %>% 
  count(cuisine) %>% 
  arrange(n)

# Plot the distribution
theme_set(theme_light())
df %>% 
  count(cuisine) %>% 
  ggplot(mapping = aes(x = n, y = reorder(cuisine, -n))) +
  geom_col(fill = "midnightblue", alpha = 0.7) +
  ylab("cuisine")
```

‡∞§‡∞∞‡±ç‡∞µ‡∞æ‡∞§, ‡∞™‡±ç‡∞∞‡∞§‡∞ø ‡∞∞‡±Å‡∞ö‡∞ø‡∞®‡∞ø ‡∞¶‡∞æ‡∞®‡∞ø ‡∞µ‡±ç‡∞Ø‡∞ï‡±ç‡∞§‡∞ø‡∞ó‡∞§ ‡∞ü‡∞ø‡∞¨‡±Å‡∞≤‡±ç‚Äå‡∞≤‡±ã ‡∞ï‡±á‡∞ü‡∞æ‡∞Ø‡∞ø‡∞Ç‡∞ö‡∞ø, ‡∞™‡±ç‡∞∞‡∞§‡∞ø ‡∞∞‡±Å‡∞ö‡∞ø‡∞ï‡∞ø ‡∞é‡∞Ç‡∞§ ‡∞°‡±á‡∞ü‡∞æ (‡∞™‡∞Ç‡∞ï‡±ç‡∞§‡±Å‡∞≤‡±Å, ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‡∞∏‡±ç) ‡∞Ö‡∞Ç‡∞¶‡±Å‡∞¨‡∞æ‡∞ü‡±Å‡∞≤‡±ã ‡∞â‡∞Ç‡∞¶‡±ã ‡∞§‡±Ü‡∞≤‡±Å‡∞∏‡±Å‡∞ï‡±Å‡∞Ç‡∞¶‡∞æ‡∞Ç.

```r
# Create individual tibble for the cuisines
thai_df <- df %>% 
  filter(cuisine == "thai")
japanese_df <- df %>% 
  filter(cuisine == "japanese")
chinese_df <- df %>% 
  filter(cuisine == "chinese")
indian_df <- df %>% 
  filter(cuisine == "indian")
korean_df <- df %>% 
  filter(cuisine == "korean")


# Find out how much data is available per cuisine
cat(" thai df:", dim(thai_df), "\n",
    "japanese df:", dim(japanese_df), "\n",
    "chinese_df:", dim(chinese_df), "\n",
    "indian_df:", dim(indian_df), "\n",
    "korean_df:", dim(korean_df))
```

```r
# Creates a functions that returns the top ingredients by class

create_ingredient <- function(df){
  
  # Drop the id column which is the first colum
  ingredient_df = df %>% select(-1) %>% 
  # Transpose data to a long format
    pivot_longer(!cuisine, names_to = "ingredients", values_to = "count") %>% 
  # Find the top most ingredients for a particular cuisine
    group_by(ingredients) %>% 
    summarise(n_instances = sum(count)) %>% 
    filter(n_instances != 0) %>% 
  # Arrange by descending order
    arrange(desc(n_instances)) %>% 
    mutate(ingredients = factor(ingredients) %>% fct_inorder())
  
  
  return(ingredient_df)
} # End of function
```

‡∞á‡∞™‡±ç‡∞™‡±Å‡∞°‡±Å ‡∞Æ‡∞®‡∞Ç ‡∞Ü ‡∞´‡∞Ç‡∞ï‡±ç‡∞∑‡∞®‡±ç‚Äå‡∞®‡±Å ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞ø ‡∞µ‡∞Ç‡∞ü‡∞ï‡∞æ‡∞≤ ‡∞µ‡∞æ‡∞∞‡±Ä‡∞ó‡∞æ ‡∞ü‡∞æ‡∞™‡±ç ‡∞ü‡±Ü‡∞®‡±ç ‡∞Ö‡∞§‡±ç‡∞Ø‡∞Ç‡∞§ ‡∞™‡±ç‡∞∞‡∞æ‡∞ö‡±Å‡∞∞‡±ç‡∞Ø‡∞Ç ‡∞™‡±ä‡∞Ç‡∞¶‡∞ø‡∞® ‡∞™‡∞¶‡∞æ‡∞∞‡±ç‡∞•‡∞æ‡∞≤ ‡∞Ü‡∞≤‡±ã‡∞ö‡∞® ‡∞™‡±ä‡∞Ç‡∞¶‡∞µ‡∞ö‡±ç‡∞ö‡±Å. ‡∞¶‡±Ä‡∞®‡±ç‡∞®‡∞ø `thai_df` ‡∞§‡±ã ‡∞™‡∞∞‡±Ä‡∞ï‡±ç‡∞∑‡∞ø‡∞¶‡±ç‡∞¶‡∞æ‡∞Ç.


```r
# Call create_ingredient and display popular ingredients
thai_ingredient_df <- create_ingredient(df = thai_df)

thai_ingredient_df %>% 
  slice_head(n = 10)
```

‡∞Æ‡±Å‡∞®‡±Å‡∞™‡∞ü‡∞ø ‡∞µ‡∞ø‡∞≠‡∞æ‡∞ó‡∞Ç‡∞≤‡±ã, ‡∞Æ‡±á‡∞Æ‡±Å `geom_col()` ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞æ‡∞Æ‡±Å, ‡∞á‡∞™‡±ç‡∞™‡±Å‡∞°‡±Å ‡∞Æ‡±Ä‡∞∞‡±Å `geom_bar` ‡∞®‡±Å ‡∞ï‡±Ç‡∞°‡∞æ ‡∞é‡∞≤‡∞æ ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞µ‡∞ö‡±ç‡∞ö‡±ã ‡∞ö‡±Ç‡∞¶‡±ç‡∞¶‡∞æ‡∞Ç, ‡∞¨‡∞æ‡∞∞‡±ç ‡∞ö‡∞æ‡∞∞‡±ç‡∞ü‡±ç‡∞∏‡±ç ‡∞∏‡±É‡∞∑‡±ç‡∞ü‡∞ø‡∞Ç‡∞ö‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø. ‡∞Æ‡∞∞‡∞ø‡∞Ç‡∞§ ‡∞ö‡∞¶‡∞µ‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø `?geom_bar` ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞Ç‡∞°‡∞ø.

-   ‡∞Æ‡±Ü‡∞ú‡∞æ‡∞∞‡∞ø‡∞ü‡±Ä ‡∞µ‡∞∞‡±ç‡∞ó‡∞Ç ‡∞®‡±Å‡∞Ç‡∞°‡∞ø ‡∞ó‡∞Æ‡∞®‡∞ø‡∞ï‡∞≤‡∞®‡±Å ‡∞§‡±ä‡∞≤‡∞ó‡∞ø‡∞Ç‡∞ö‡∞°‡∞Ç: `‡∞Ö‡∞Ç‡∞°‡∞∞‡±ç-‡∞∏‡∞æ‡∞Ç‡∞™‡±ç‡∞≤‡∞ø‡∞Ç‡∞ó‡±ç`

‡∞á‡∞™‡±ç‡∞™‡±Å‡∞°‡±Å `recipe` ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞ø ‡∞Ö‡∞∏‡∞Æ‡∞§‡±Å‡∞≤‡±ç‡∞Ø ‡∞°‡±á‡∞ü‡∞æ ‡∞∏‡±Ü‡∞ü‡±ç‚Äå‡∞≤‡∞®‡±Å ‡∞é‡∞≤‡∞æ ‡∞®‡∞ø‡∞∞‡±ç‡∞µ‡∞π‡∞ø‡∞Ç‡∞ö‡∞æ‡∞≤‡±ã ‡∞ö‡±Ç‡∞™‡∞ø‡∞¶‡±ç‡∞¶‡∞æ‡∞Ç. ‡∞í‡∞ï recipe ‡∞Ö‡∞®‡±á‡∞¶‡∞ø ‡∞°‡±á‡∞ü‡∞æ ‡∞µ‡∞ø‡∞∂‡±ç‡∞≤‡±á‡∞∑‡∞£‡∞ï‡±Å ‡∞∏‡∞ø‡∞¶‡±ç‡∞ß‡∞Ç ‡∞ö‡±á‡∞Ø‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø ‡∞°‡±á‡∞ü‡∞æ ‡∞∏‡±Ü‡∞ü‡±ç‚Äå‡∞™‡±à ‡∞è ‡∞¶‡∞∂‡∞≤‡∞®‡±Å ‡∞µ‡∞∞‡±ç‡∞§‡∞ø‡∞Ç‡∞ö‡∞æ‡∞≤‡±ã ‡∞µ‡∞ø‡∞µ‡∞∞‡∞ø‡∞Ç‡∞ö‡±á ‡∞í‡∞ï ‡∞¨‡±ç‡∞≤‡±Ç‡∞™‡±ç‡∞∞‡∞ø‡∞Ç‡∞ü‡±ç‚Äå‡∞ó‡∞æ ‡∞≠‡∞æ‡∞µ‡∞ø‡∞Ç‡∞ö‡∞µ‡∞ö‡±ç‡∞ö‡±Å.


```r
# Load themis package for dealing with imbalanced data
library(themis)

# Create a recipe for preprocessing data
cuisines_recipe <- recipe(cuisine ~ ., data = df_select) %>% 
  step_smote(cuisine)

cuisines_recipe
```

‡∞Æ‡∞®‡∞Ç ‡∞Æ‡∞æ ‡∞™‡±ç‡∞∞‡±Ä‡∞™‡±ç‡∞∞‡∞æ‡∞∏‡±Ü‡∞∏‡∞ø‡∞Ç‡∞ó‡±ç ‡∞¶‡∞∂‡∞≤‡∞®‡±Å ‡∞µ‡∞ø‡∞°‡∞¶‡±Ä‡∞Ø‡∞Ç‡∞°‡∞ø.

-   ‡∞´‡∞æ‡∞∞‡±ç‡∞Æ‡±Å‡∞≤‡∞æ‡∞§‡±ã `recipe()` ‡∞ï‡∞æ‡∞≤‡±ç ‡∞µ‡±á‡∞∞‡∞ø‡∞Ø‡∞¨‡±Å‡∞≤‡±ç‡∞∏‡±ç ‡∞Ø‡±ä‡∞ï‡±ç‡∞ï *‡∞™‡∞æ‡∞§‡±ç‡∞∞‡∞≤‡∞®‡±Å* `df_select` ‡∞°‡±á‡∞ü‡∞æ‡∞®‡±Å ‡∞∏‡±Ç‡∞ö‡∞®‡∞ó‡∞æ ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞ø ‡∞∞‡∞ø‡∞∏‡∞ø‡∞™‡±Ä‡∞ï‡∞ø ‡∞ö‡±Ü‡∞¨‡±Å‡∞§‡±Å‡∞Ç‡∞¶‡∞ø. ‡∞â‡∞¶‡∞æ‡∞π‡∞∞‡∞£‡∞ï‡±Å `cuisine` ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‚Äå‡∞ï‡±Å `outcome` ‡∞™‡∞æ‡∞§‡±ç‡∞∞ ‡∞ï‡±á‡∞ü‡∞æ‡∞Ø‡∞ø‡∞Ç‡∞ö‡∞¨‡∞°‡∞ø‡∞Ç‡∞¶‡∞ø, ‡∞Æ‡∞ø‡∞ó‡∞§‡∞æ ‡∞ï‡∞æ‡∞≤‡∞Æ‡±ç‡∞∏‡±ç‚Äå‡∞ï‡±Å `predictor` ‡∞™‡∞æ‡∞§‡±ç‡∞∞ ‡∞ï‡±á‡∞ü‡∞æ‡∞Ø‡∞ø‡∞Ç‡∞ö‡∞¨‡∞°‡∞ø‡∞Ç‡∞¶‡∞ø.

-   [`step_smote(cuisine)`](https://themis.tidymodels.org/reference/step_smote.html) ‡∞í‡∞ï ‡∞∞‡∞ø‡∞∏‡∞ø‡∞™‡±Ä ‡∞¶‡∞∂ ‡∞Ø‡±ä‡∞ï‡±ç‡∞ï *‡∞µ‡∞ø‡∞µ‡∞∞‡∞£* ‡∞∏‡±É‡∞∑‡±ç‡∞ü‡∞ø‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø, ‡∞á‡∞¶‡∞ø ‡∞à ‡∞ï‡±á‡∞∏‡±Å‡∞≤ ‡∞∏‡∞Æ‡±Ä‡∞™ ‡∞™‡±ä‡∞∞‡±Å‡∞ó‡±Å‡∞µ‡∞æ‡∞∞‡∞ø‡∞®‡∞ø ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞ø ‡∞Æ‡±à‡∞®‡∞æ‡∞∞‡∞ø‡∞ü‡±Ä ‡∞§‡∞∞‡∞ó‡∞§‡∞ø ‡∞Ø‡±ä‡∞ï‡±ç‡∞ï ‡∞ï‡±ä‡∞§‡±ç‡∞§ ‡∞â‡∞¶‡∞æ‡∞π‡∞∞‡∞£‡∞≤‡∞®‡±Å ‡∞∏‡∞ø‡∞Ç‡∞•‡∞ü‡∞ø‡∞ï‡±ç‚Äå‡∞ó‡∞æ ‡∞â‡∞§‡±ç‡∞™‡∞§‡±ç‡∞§‡∞ø ‡∞ö‡±á‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø.

‡∞á‡∞™‡±ç‡∞™‡±Å‡∞°‡±Å, ‡∞Æ‡∞®‡∞Ç ‡∞™‡±ç‡∞∞‡±Ä‡∞™‡±ç‡∞∞‡∞æ‡∞∏‡±Ü‡∞∏‡±ç‡∞°‡±ç ‡∞°‡±á‡∞ü‡∞æ‡∞®‡±Å ‡∞ö‡±Ç‡∞°‡∞æ‡∞≤‡∞®‡±Å‡∞ï‡±Å‡∞Ç‡∞ü‡±á, ‡∞Æ‡∞®‡∞Ç [**`prep()`**](https://recipes.tidymodels.org/reference/prep.html) ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å [**`bake()`**](https://recipes.tidymodels.org/reference/bake.html) ‡∞Æ‡∞æ ‡∞∞‡∞ø‡∞∏‡∞ø‡∞™‡±Ä‡∞®‡∞ø ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞æ‡∞≤‡∞ø.

`prep()`: ‡∞∂‡∞ø‡∞ï‡±ç‡∞∑‡∞£ ‡∞∏‡±Ü‡∞ü‡±ç ‡∞®‡±Å‡∞Ç‡∞°‡∞ø ‡∞Ö‡∞µ‡∞∏‡∞∞‡∞Æ‡±à‡∞® ‡∞™‡∞æ‡∞∞‡∞æ‡∞Æ‡∞ø‡∞§‡±Å‡∞≤‡∞®‡±Å ‡∞Ö‡∞Ç‡∞ö‡∞®‡∞æ ‡∞µ‡±á‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø, ‡∞á‡∞µ‡∞ø ‡∞§‡∞∞‡±ç‡∞µ‡∞æ‡∞§ ‡∞á‡∞§‡∞∞ ‡∞°‡±á‡∞ü‡∞æ ‡∞∏‡±Ü‡∞ü‡±ç‚Äå‡∞≤‡∞ï‡±Å ‡∞µ‡∞∞‡±ç‡∞§‡∞ø‡∞Ç‡∞ö‡∞µ‡∞ö‡±ç‡∞ö‡±Å.

`bake()`: ‡∞™‡±ç‡∞∞‡∞ø‡∞™‡±ç ‡∞ö‡±á‡∞∏‡∞ø‡∞® ‡∞∞‡∞ø‡∞∏‡∞ø‡∞™‡±Ä‡∞®‡∞ø ‡∞§‡±Ä‡∞∏‡±Å‡∞ï‡±Å‡∞®‡∞ø ‡∞Ü‡∞™‡∞∞‡±á‡∞∑‡∞®‡±ç‡∞≤‡∞®‡±Å ‡∞è‡∞¶‡±à‡∞®‡∞æ ‡∞°‡±á‡∞ü‡∞æ ‡∞∏‡±Ü‡∞ü‡±ç‚Äå‡∞ï‡±Å ‡∞µ‡∞∞‡±ç‡∞§‡∞ø‡∞Ç‡∞™‡∞ú‡±á‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø.


```r
# Prep and bake the recipe
preprocessed_df <- cuisines_recipe %>% 
  prep() %>% 
  bake(new_data = NULL) %>% 
  relocate(cuisine)

# Display data
preprocessed_df %>% 
  slice_head(n = 5)

# Quick summary stats
preprocessed_df %>% 
  introduce()
```

‡∞á‡∞™‡±ç‡∞™‡±Å‡∞°‡±Å ‡∞Æ‡∞® ‡∞µ‡∞Ç‡∞ü‡∞ï‡∞æ‡∞≤ ‡∞™‡∞Ç‡∞™‡∞ø‡∞£‡±Ä‡∞®‡∞ø ‡∞™‡∞∞‡∞ø‡∞∂‡±Ä‡∞≤‡∞ø‡∞Ç‡∞ö‡∞ø, ‡∞Ö‡∞µ‡∞ø ‡∞Ö‡∞∏‡∞Æ‡∞§‡±Å‡∞≤‡±ç‡∞Ø ‡∞°‡±á‡∞ü‡∞æ‡∞§‡±ã ‡∞™‡±ã‡∞≤‡±ç‡∞ö‡±Å‡∞ï‡±Å‡∞Ç‡∞¶‡∞æ‡∞Ç.


```r
# Distribution of cuisines
new_label_count <- preprocessed_df %>% 
  count(cuisine) %>% 
  arrange(desc(n))

list(new_label_count = new_label_count,
     old_label_count = old_label_count)
```

>
> ‡∞Ö‡∞Ç‡∞¶‡±Å‡∞µ‡∞≤‡±ç‡∞≤, ‡∞Æ‡±Ä‡∞∞‡±Å ‡∞∏‡∞æ‡∞ß‡∞æ‡∞∞‡∞£‡∞Ç‡∞ó‡∞æ tidymodels ‡∞â‡∞™‡∞Ø‡±ã‡∞ó‡∞ø‡∞Ç‡∞ö‡∞ø‡∞®‡∞™‡±ç‡∞™‡±Å‡∞°‡±Å ‡∞∞‡±Ü‡∞∏‡∞ø‡∞™‡±Ä‡∞≤‡∞®‡±Å **`prep()`** ‡∞Æ‡∞∞‡∞ø‡∞Ø‡±Å **`bake()`** ‡∞ö‡±á‡∞Ø‡∞æ‡∞≤‡±ç‡∞∏‡∞ø‡∞® ‡∞Ö‡∞µ‡∞∏‡∞∞‡∞Ç ‡∞≤‡±á‡∞¶‡±Å, ‡∞ï‡∞æ‡∞®‡±Ä ‡∞Ö‡∞µ‡∞ø ‡∞Æ‡±Ä ‡∞ü‡±Ç‡∞≤‡±ç‚Äå‡∞ï‡∞ø‡∞ü‡±ç‚Äå‡∞≤‡±ã ‡∞â‡∞Ç‡∞°‡±á ‡∞∏‡∞π‡∞æ‡∞Ø‡∞ï ‡∞´‡∞Ç‡∞ï‡±ç‡∞∑‡∞®‡±ç‡∞≤‡±Å, ‡∞Æ‡∞æ ‡∞∏‡∞Ç‡∞¶‡∞∞‡±ç‡∞≠‡∞Ç‡∞≤‡±ã ‡∞â‡∞®‡±ç‡∞®‡∞ü‡±ç‡∞≤‡±Å‡∞ó‡∞æ ‡∞∞‡±Ü‡∞∏‡∞ø‡∞™‡±Ä‡∞≤‡±Å ‡∞Æ‡±Ä‡∞∞‡±Å ‡∞Ü‡∞∂‡∞ø‡∞Ç‡∞ö‡∞ø‡∞®‡∞ü‡±ç‡∞≤‡±Å‡∞ó‡∞æ ‡∞™‡∞®‡∞ø‡∞ö‡±á‡∞∏‡±ç‡∞§‡±Å‡∞®‡±ç‡∞®‡∞æ‡∞Ø‡∞æ ‡∞Ö‡∞®‡∞ø ‡∞®‡∞ø‡∞∞‡±ç‡∞ß‡∞æ‡∞∞‡∞ø‡∞Ç‡∞ö‡±Å‡∞ï‡±ã‡∞µ‡∞°‡∞æ‡∞®‡∞ø‡∞ï‡∞ø.
>
> ‡∞Æ‡±Ä‡∞∞‡±Å **`bake()`** ‡∞ö‡±á‡∞∏‡∞ø‡∞®‡∞™‡±ç‡∞™‡±Å‡∞°‡±Å ‡∞™‡±ç‡∞∞‡∞ø‡∞™‡±ç ‡∞ö‡±á‡∞∏‡∞ø‡∞® ‡∞∞‡±Ü‡∞∏‡∞ø‡∞™‡±Ä‡∞§‡±ã **`new_data = NULL`** ‡∞â‡∞Ç‡∞ü‡±á, ‡∞Æ‡±Ä‡∞∞‡±Å ‡∞∞‡±Ü‡∞∏‡∞ø‡∞™‡±Ä ‡∞®‡∞ø‡∞∞‡±ç‡∞µ‡∞ö‡∞ø‡∞Ç‡∞ö‡∞ø‡∞®‡∞™‡±ç‡∞™‡±Å‡∞°‡±Å ‡∞Ö‡∞Ç‡∞¶‡∞ø‡∞Ç‡∞ö‡∞ø‡∞® ‡∞°‡±á‡∞ü‡∞æ‡∞®‡±Å ‡∞§‡∞ø‡∞∞‡∞ø‡∞ó‡∞ø ‡∞™‡±ä‡∞Ç‡∞¶‡±Å‡∞§‡∞æ‡∞∞‡±Å, ‡∞ï‡∞æ‡∞®‡±Ä ‡∞™‡±ç‡∞∞‡±Ä‡∞™‡±ç‡∞∞‡∞æ‡∞∏‡±Ü‡∞∏‡∞ø‡∞Ç‡∞ó‡±ç ‡∞¶‡∞∂‡∞≤ ‡∞¶‡±ç‡∞µ‡∞æ‡∞∞‡∞æ ‡∞ó‡∞°‡∞ø‡∞ö‡∞ø‡∞®.
