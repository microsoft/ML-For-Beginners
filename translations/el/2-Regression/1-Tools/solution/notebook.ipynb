## Γραμμική Παλινδρόμηση για το σύνολο δεδομένων Διαβήτη - Μάθημα 1


Εισαγωγή των απαραίτητων βιβλιοθηκών


```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

Φορτώστε το σύνολο δεδομένων για τον διαβήτη, χωρισμένο σε δεδομένα `X` και χαρακτηριστικά `y`


```python
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
```

Επιλέξτε μόνο ένα χαρακτηριστικό για να στοχεύσετε σε αυτήν την άσκηση


```python
# Selecting the 3rd feature
X = X[:, 2]
print(X.shape)
```

```python
#Reshaping to get a 2D array
X = X.reshape(-1, 1)
print(X.shape)
print(X)
```

Διαχωρίστε τα δεδομένα εκπαίδευσης και δοκιμής για τα `X` και `y`


```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

Επιλέξτε το μοντέλο και προσαρμόστε το με τα δεδομένα εκπαίδευσης


```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

Χρησιμοποιήστε δεδομένα δοκιμής για να προβλέψετε μια γραμμή


```python
y_pred = model.predict(X_test)
```

Εμφανίστε τα αποτελέσματα σε ένα γράφημα


```python
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.show()
```
