## 1. èˆ‡æ•¸æ“šå…±èˆï¼šç¸®å°ç¯„åœè‡³ä¸‰å¤§æœ€å—æ­¡è¿çš„éŸ³æ¨‚é¡å‹

é€™æ˜¯æˆ‘å€‘åœ¨ä¸Šä¸€èª²ä¸­æ‰€åšå…§å®¹çš„å›é¡§ã€‚è®“æˆ‘å€‘ä¾†åˆ‡åˆ†å’Œåˆ†æä¸€äº›æ•¸æ“šå§ï¼


```r
# Load the core tidyverse and make it available in your current R session
library(tidyverse)

# Import the data into a tibble
df <- read_csv(file = "https://raw.githubusercontent.com/microsoft/ML-For-Beginners/main/5-Clustering/data/nigerian-songs.csv", show_col_types = FALSE)

# Narrow down to top 3 popular genres
nigerian_songs <- df %>% 
  # Concentrate on top 3 genres
  filter(artist_top_genre %in% c("afro dancehall", "afropop","nigerian pop")) %>% 
  # Remove unclassified observations
  filter(popularity != 0)



# Visualize popular genres using bar plots
theme_set(theme_light())
nigerian_songs %>%
  count(artist_top_genre) %>%
  ggplot(mapping = aes(x = artist_top_genre, y = n,
                       fill = artist_top_genre)) +
  geom_col(alpha = 0.8) +
  paletteer::scale_fill_paletteer_d("ggsci::category10_d3") +
  ggtitle("Top genres") +
  theme(plot.title = element_text(hjust = 0.5))
```

## 2. æ›´å¤šæ•¸æ“šæ¢ç´¢

é€™äº›æ•¸æ“šæœ‰å¤šä¹¾æ·¨ï¼Ÿè®“æˆ‘å€‘ä½¿ç”¨ç›’å½¢åœ–æª¢æŸ¥æ˜¯å¦æœ‰ç•°å¸¸å€¼ã€‚æˆ‘å€‘å°‡å°ˆæ³¨æ–¼ç•°å¸¸å€¼è¼ƒå°‘çš„æ•¸å€¼å‹æ¬„ä½ï¼ˆé›–ç„¶ä½ ä¹Ÿå¯ä»¥æ¸…ç†æ‰ç•°å¸¸å€¼ï¼‰ã€‚ç›’å½¢åœ–å¯ä»¥é¡¯ç¤ºæ•¸æ“šçš„ç¯„åœï¼Œä¸¦å¹«åŠ©é¸æ“‡è¦ä½¿ç”¨çš„æ¬„ä½ã€‚è«‹æ³¨æ„ï¼Œç›’å½¢åœ–ä¸é¡¯ç¤ºæ–¹å·®ï¼Œè€Œæ–¹å·®æ˜¯è‰¯å¥½å¯èšé¡æ•¸æ“šçš„é‡è¦å…ƒç´ ã€‚è«‹åƒé–±[é€™ç¯‡è¨è«–](https://stats.stackexchange.com/questions/91536/deduce-variance-from-boxplot)ä»¥äº†è§£æ›´å¤šã€‚

[ç›’å½¢åœ–](https://en.wikipedia.org/wiki/Box_plot)ç”¨æ–¼ä»¥åœ–å½¢æ–¹å¼æè¿°`æ•¸å€¼å‹`æ•¸æ“šçš„åˆ†ä½ˆï¼Œå› æ­¤è®“æˆ‘å€‘å¾*é¸æ“‡*æ‰€æœ‰æ•¸å€¼å‹æ¬„ä½ä»¥åŠæµè¡ŒéŸ³æ¨‚é¡å‹é–‹å§‹ã€‚


```r
# Select top genre column and all other numeric columns
df_numeric <- nigerian_songs %>% 
  select(artist_top_genre, where(is.numeric)) 

# Display the data
df_numeric %>% 
  slice_head(n = 5)
```

çœ‹çœ‹é¸æ“‡è¼”åŠ©å·¥å…· `where` æ˜¯å¦‚ä½•è®“é€™ä»¶äº‹è®Šå¾—ç°¡å–®çš„ ğŸ’ï¼Ÿå¯ä»¥åœ¨[é€™è£¡](https://tidyselect.r-lib.org/)æ¢ç´¢å…¶ä»–é¡ä¼¼çš„å‡½æ•¸ã€‚

ç”±æ–¼æˆ‘å€‘å°‡ç‚ºæ¯å€‹æ•¸å€¼ç‰¹å¾µè£½ä½œç®±å‹åœ–ï¼Œä¸¦ä¸”å¸Œæœ›é¿å…ä½¿ç”¨è¿´åœˆï¼Œå› æ­¤æˆ‘å€‘éœ€è¦å°‡æ•¸æ“šé‡æ–°æ ¼å¼åŒ–ç‚º*æ›´é•·çš„*æ ¼å¼ï¼Œé€™æ¨£å°±å¯ä»¥åˆ©ç”¨ `facets`â€”â€”æ¯å€‹å­åœ–éƒ½é¡¯ç¤ºæ•¸æ“šçš„ä¸€å€‹å­é›†ã€‚


```r
# Pivot data from wide to long
df_numeric_long <- df_numeric %>% 
  pivot_longer(!artist_top_genre, names_to = "feature_names", values_to = "values") 

# Print out data
df_numeric_long %>% 
  slice_head(n = 15)
```

æ›´é•·äº†ï¼ç¾åœ¨æ˜¯æ™‚å€™ä½¿ç”¨ä¸€äº› `ggplots` äº†ï¼é‚£éº¼æˆ‘å€‘æœƒä½¿ç”¨å“ªç¨® `geom` å‘¢ï¼Ÿ


```r
# Make a box plot
df_numeric_long %>% 
  ggplot(mapping = aes(x = feature_names, y = values, fill = feature_names)) +
  geom_boxplot() +
  facet_wrap(~ feature_names, ncol = 4, scales = "free") +
  theme(legend.position = "none")
```

ç¾åœ¨æˆ‘å€‘å¯ä»¥çœ‹åˆ°é€™äº›æ•¸æ“šæœ‰äº›é›œäº‚ï¼šé€éè§€å¯Ÿæ¯ä¸€åˆ—çš„ç›’ç‹€åœ–ï¼Œå¯ä»¥çœ‹åˆ°æœ‰ç•°å¸¸å€¼ã€‚ä½ å¯ä»¥é€ä¸€æª¢æŸ¥æ•¸æ“šé›†ä¸¦ç§»é™¤é€™äº›ç•°å¸¸å€¼ï¼Œä½†é€™æ¨£æœƒä½¿æ•¸æ“šè®Šå¾—éå¸¸æœ‰é™ã€‚

ç›®å‰ï¼Œæˆ‘å€‘ä¾†é¸æ“‡è¦ç”¨æ–¼èšé¡ç·´ç¿’çš„åˆ—ã€‚è®“æˆ‘å€‘æŒ‘é¸ç¯„åœç›¸ä¼¼çš„æ•¸å€¼å‹åˆ—ã€‚æˆ‘å€‘å¯ä»¥å°‡ `artist_top_genre` ç·¨ç¢¼ç‚ºæ•¸å€¼å‹ï¼Œä½†ç¾åœ¨å…ˆå°‡å…¶æ’é™¤ã€‚


```r
# Select variables with similar ranges
df_numeric_select <- df_numeric %>% 
  select(popularity, danceability, acousticness, loudness, energy) 

# Normalize data
# df_numeric_select <- scale(df_numeric_select)
```

## 3. åœ¨ R ä¸­è¨ˆç®— k-means åˆ†ç¾¤

æˆ‘å€‘å¯ä»¥ä½¿ç”¨ R ä¸­å…§å»ºçš„ `kmeans` å‡½æ•¸ä¾†è¨ˆç®— k-meansï¼Œè«‹åƒé–± `help("kmeans()")`ã€‚`kmeans()` å‡½æ•¸çš„ä¸»è¦åƒæ•¸æ˜¯ä¸€å€‹åŒ…å«æ‰€æœ‰æ•¸å€¼å‹æ¬„ä½çš„è³‡æ–™æ¡†ã€‚

ä½¿ç”¨ k-means åˆ†ç¾¤çš„ç¬¬ä¸€æ­¥æ˜¯æŒ‡å®šæœ€çµ‚è§£æ±ºæ–¹æ¡ˆä¸­è¦ç”Ÿæˆçš„ç¾¤æ•¸ï¼ˆkï¼‰ã€‚æˆ‘å€‘çŸ¥é“å¾è³‡æ–™é›†ä¸­åˆ†å‡ºäº† 3 ç¨®æ­Œæ›²é¡å‹ï¼Œå› æ­¤æˆ‘å€‘å˜—è©¦è¨­ç½®ç‚º 3ï¼š


```r
set.seed(2056)
# Kmeans clustering for 3 clusters
kclust <- kmeans(
  df_numeric_select,
  # Specify the number of clusters
  centers = 3,
  # How many random initial configurations
  nstart = 25
)

# Display clustering object
kclust
```

kmeans ç‰©ä»¶åŒ…å«äº†è¨±å¤šè³‡è¨Šï¼Œé€™äº›è³‡è¨Šåœ¨ `help("kmeans()")` ä¸­æœ‰è©³ç´°èªªæ˜ã€‚ç›®å‰ï¼Œæˆ‘å€‘å…ˆå°ˆæ³¨æ–¼å¹¾å€‹é‡é»ã€‚æˆ‘å€‘å¯ä»¥çœ‹åˆ°è³‡æ–™å·²è¢«åˆ†æˆä¸‰å€‹ç¾¤çµ„ï¼Œå¤§å°åˆ†åˆ¥ç‚º 65ã€110 å’Œ 111ã€‚è¼¸å‡ºä¸­é‚„åŒ…å«äº†ä¸‰å€‹ç¾¤çµ„åœ¨äº”å€‹è®Šæ•¸ä¸Šçš„ç¾¤ä¸­å¿ƒï¼ˆå¹³å‡å€¼ï¼‰ã€‚

èšé¡å‘é‡æ˜¯æ¯å€‹è§€æ¸¬å€¼çš„ç¾¤çµ„åˆ†é…ã€‚æˆ‘å€‘å¯ä»¥ä½¿ç”¨ `augment` å‡½æ•¸å°‡ç¾¤çµ„åˆ†é…åŠ å…¥åˆ°åŸå§‹è³‡æ–™é›†ä¸­ã€‚


```r
# Add predicted cluster assignment to data set
augment(kclust, df_numeric_select) %>% 
  relocate(.cluster) %>% 
  slice_head(n = 10)
```

å¤ªå¥½äº†ï¼Œæˆ‘å€‘å‰›å‰›å°‡æ•¸æ“šé›†åˆ†æˆäº†ä¸‰å€‹ç¾¤çµ„ã€‚é‚£éº¼ï¼Œæˆ‘å€‘çš„åˆ†ç¾¤æ•ˆæœå¦‚ä½•å‘¢ ğŸ¤·ï¼Ÿè®“æˆ‘å€‘ä¾†çœ‹çœ‹ `Silhouette score`ã€‚

### **Silhouette score**

[Silhouette åˆ†æ](https://en.wikipedia.org/wiki/Silhouette_(clustering)) å¯ä»¥ç”¨ä¾†ç ”ç©¶çµæœç¾¤çµ„ä¹‹é–“çš„åˆ†é›¢è·é›¢ã€‚é€™å€‹åˆ†æ•¸ç¯„åœå¾ -1 åˆ° 1ï¼Œå¦‚æœåˆ†æ•¸æ¥è¿‘ 1ï¼Œè¡¨ç¤ºç¾¤çµ„å¯†é›†ä¸”èˆ‡å…¶ä»–ç¾¤çµ„åˆ†é›¢è‰¯å¥½ã€‚æ¥è¿‘ 0 çš„å€¼å‰‡è¡¨ç¤ºç¾¤çµ„é‡ç–Šï¼Œæ¨£æœ¬éå¸¸æ¥è¿‘é„°è¿‘ç¾¤çµ„çš„æ±ºç­–é‚Šç•Œã€‚[ä¾†æº](https://dzone.com/articles/kmeans-silhouette-score-explained-with-python-exam)ã€‚

å¹³å‡ Silhouette æ–¹æ³•è¨ˆç®—ä¸åŒ *k* å€¼ä¸‹è§€æ¸¬å€¼çš„å¹³å‡ Silhouette åˆ†æ•¸ã€‚é«˜çš„å¹³å‡ Silhouette åˆ†æ•¸è¡¨ç¤ºåˆ†ç¾¤æ•ˆæœè‰¯å¥½ã€‚

ä½¿ç”¨ cluster å¥—ä»¶ä¸­çš„ `silhouette` å‡½æ•¸ä¾†è¨ˆç®—å¹³å‡ Silhouette å¯¬åº¦ã€‚

> Silhouette å¯ä»¥ä½¿ç”¨ä»»ä½• [è·é›¢](https://en.wikipedia.org/wiki/Distance "Distance") åº¦é‡ä¾†è¨ˆç®—ï¼Œä¾‹å¦‚ [æ­å¹¾é‡Œå¾—è·é›¢](https://en.wikipedia.org/wiki/Euclidean_distance "Euclidean distance") æˆ– [æ›¼å“ˆé “è·é›¢](https://en.wikipedia.org/wiki/Manhattan_distance "Manhattan distance")ï¼Œé€™äº›æˆ‘å€‘åœ¨[ä¸Šä¸€èª²](https://github.com/microsoft/ML-For-Beginners/blob/main/5-Clustering/1-Visualize/solution/R/lesson_14-R.ipynb)ä¸­å·²ç¶“è¨è«–éã€‚


```r
# Load cluster package
library(cluster)

# Compute average silhouette score
ss <- silhouette(kclust$cluster,
                 # Compute euclidean distance
                 dist = dist(df_numeric_select))
mean(ss[, 3])
```

æˆ‘å€‘çš„åˆ†æ•¸æ˜¯ **.549**ï¼Œæ­£å¥½ä½æ–¼ä¸­é–“ä½ç½®ã€‚é€™è¡¨æ˜æˆ‘å€‘çš„æ•¸æ“šä¸¦ä¸ç‰¹åˆ¥é©åˆé€™ç¨®é¡å‹çš„èšé¡ã€‚è®“æˆ‘å€‘çœ‹çœ‹æ˜¯å¦å¯ä»¥é€šéè¦–è¦ºåŒ–ä¾†ç¢ºèªé€™å€‹çŒœæ¸¬ã€‚[factoextra å¥—ä»¶](https://rpkgs.datanovia.com/factoextra/index.html) æä¾›äº†å‡½æ•¸ (`fviz_cluster()`) ç”¨æ–¼è¦–è¦ºåŒ–èšé¡ã€‚


```r
library(factoextra)

# Visualize clustering results
fviz_cluster(kclust, df_numeric_select)
```

## 4. ç¢ºå®šæœ€ä½³ç¾¤é›†æ•¸é‡

åœ¨ K-Means ç¾¤é›†åˆ†æä¸­ï¼Œç¶“å¸¸å‡ºç¾çš„ä¸€å€‹åŸºæœ¬å•é¡Œæ˜¯â€”â€”åœ¨æ²’æœ‰å·²çŸ¥é¡åˆ¥æ¨™ç±¤çš„æƒ…æ³ä¸‹ï¼Œå¦‚ä½•çŸ¥é“æ‡‰å°‡æ•¸æ“šåˆ†æˆå¤šå°‘å€‹ç¾¤é›†ï¼Ÿ

æˆ‘å€‘å¯ä»¥å˜—è©¦çš„ä¸€ç¨®æ–¹æ³•æ˜¯ä½¿ç”¨æ•¸æ“šæ¨£æœ¬ä¾†`å‰µå»ºä¸€ç³»åˆ—ç¾¤é›†æ¨¡å‹`ï¼Œä¸¦é€æ­¥å¢åŠ ç¾¤é›†çš„æ•¸é‡ï¼ˆä¾‹å¦‚å¾ 1 åˆ° 10ï¼‰ï¼Œç„¶å¾Œè©•ä¼°ç¾¤é›†æŒ‡æ¨™ï¼Œä¾‹å¦‚ **Silhouette åˆ†æ•¸**ã€‚

è®“æˆ‘å€‘é€šéè¨ˆç®—ä¸åŒ *k* å€¼çš„ç¾¤é›†ç®—æ³•ä¾†ç¢ºå®šæœ€ä½³ç¾¤é›†æ•¸é‡ï¼Œä¸¦è©•ä¼° **ç¾¤é›†å…§å¹³æ–¹å’Œ**ï¼ˆWCSSï¼‰ã€‚ç¾¤é›†å…§å¹³æ–¹å’Œï¼ˆWCSSï¼‰ç¸½é‡è¡¡é‡ç¾¤é›†çš„ç·Šå¯†æ€§ï¼Œæˆ‘å€‘å¸Œæœ›å®ƒè¶Šå°è¶Šå¥½ï¼Œè¼ƒä½çš„å€¼æ„å‘³è‘—æ•¸æ“šé»æ›´æ¥è¿‘ã€‚

è®“æˆ‘å€‘æ¢ç´¢ä¸åŒçš„ `k` é¸æ“‡ï¼ˆå¾ 1 åˆ° 10ï¼‰å°æ­¤ç¾¤é›†çš„å½±éŸ¿ã€‚


```r
# Create a series of clustering models
kclusts <- tibble(k = 1:10) %>% 
  # Perform kmeans clustering for 1,2,3 ... ,10 clusters
  mutate(model = map(k, ~ kmeans(df_numeric_select, centers = .x, nstart = 25)),
  # Farm out clustering metrics eg WCSS
         glanced = map(model, ~ glance(.x))) %>% 
  unnest(cols = glanced)
  

# View clustering rsulsts
kclusts
```

ç¾åœ¨æˆ‘å€‘å·²ç¶“ç²å¾—æ¯å€‹èšé¡ç®—æ³•åœ¨ä¸­å¿ƒ *k* ä¸‹çš„ç¸½é›†ç¾¤å…§å¹³æ–¹å’Œ (tot.withinss)ï¼Œæ¥ä¸‹ä¾†æˆ‘å€‘ä½¿ç”¨[è‚˜éƒ¨æ³•](https://en.wikipedia.org/wiki/Elbow_method_(clustering))ä¾†å°‹æ‰¾æœ€ä½³çš„èšé¡æ•¸é‡ã€‚æ­¤æ–¹æ³•åŒ…æ‹¬å°‡WCSSä½œç‚ºèšé¡æ•¸é‡çš„å‡½æ•¸é€²è¡Œç¹ªåœ–ï¼Œä¸¦é¸æ“‡[æ›²ç·šçš„è‚˜éƒ¨](https://en.wikipedia.org/wiki/Elbow_of_the_curve "æ›²ç·šçš„è‚˜éƒ¨")ä½œç‚ºè¦ä½¿ç”¨çš„èšé¡æ•¸é‡ã€‚


```r
set.seed(2056)
# Use elbow method to determine optimum number of clusters
kclusts %>% 
  ggplot(mapping = aes(x = k, y = tot.withinss)) +
  geom_line(size = 1.2, alpha = 0.8, color = "#FF7F0EFF") +
  geom_point(size = 2, color = "#FF7F0EFF")
```

è©²åœ–é¡¯ç¤ºï¼Œç•¶ç¾¤é›†æ•¸é‡å¾ä¸€å€‹å¢åŠ åˆ°å…©å€‹æ™‚ï¼ŒWCSSï¼ˆå› æ­¤*ç·Šå¯†åº¦*ï¼‰å¤§å¹…æ¸›å°‘ï¼Œå¾å…©å€‹å¢åŠ åˆ°ä¸‰å€‹ç¾¤é›†æ™‚ä¹Ÿæœ‰æ˜é¡¯çš„æ¸›å°‘ã€‚ä¹‹å¾Œï¼Œæ¸›å°‘çš„å¹…åº¦è®Šå¾—ä¸é‚£éº¼æ˜é¡¯ï¼Œå°è‡´åœ–è¡¨åœ¨å¤§ç´„ä¸‰å€‹ç¾¤é›†è™•å‡ºç¾ä¸€å€‹ã€Œè‚˜éƒ¨ã€ğŸ’ªã€‚é€™æ˜¯ä¸€å€‹å¾ˆå¥½çš„æŒ‡æ¨™ï¼Œè¡¨æ˜æ•¸æ“šé»å¯ä»¥åˆç†åœ°åˆ†ç‚ºå…©åˆ°ä¸‰å€‹æ˜é¡¯åˆ†é›¢çš„ç¾¤é›†ã€‚

ç¾åœ¨æˆ‘å€‘å¯ä»¥ç¹¼çºŒæå– `k = 3` çš„ç¾¤é›†æ¨¡å‹ï¼š

> `pull()`ï¼šç”¨æ–¼æå–å–®ä¸€åˆ—
>
> `pluck()`ï¼šç”¨æ–¼ç´¢å¼•åƒåˆ—è¡¨é€™æ¨£çš„æ•¸æ“šçµæ§‹


```r
# Extract k = 3 clustering
final_kmeans <- kclusts %>% 
  filter(k == 3) %>% 
  pull(model) %>% 
  pluck(1)


final_kmeans
```

å¤ªå¥½äº†ï¼è®“æˆ‘å€‘ä¾†çœ‹çœ‹ç²å¾—çš„ç¾¤é›†ã€‚æƒ³ç”¨ `plotly` å¢åŠ ä¸€äº›äº’å‹•æ€§å—ï¼Ÿ


```r
# Add predicted cluster assignment to data set
results <-  augment(final_kmeans, df_numeric_select) %>% 
  bind_cols(df_numeric %>% select(artist_top_genre)) 

# Plot cluster assignments
clust_plt <- results %>% 
  ggplot(mapping = aes(x = popularity, y = danceability, color = .cluster, shape = artist_top_genre)) +
  geom_point(size = 2, alpha = 0.8) +
  paletteer::scale_color_paletteer_d("ggthemes::Tableau_10")

ggplotly(clust_plt)
```

æˆ–è¨±æˆ‘å€‘åŸæœ¬é æœŸï¼Œæ¯å€‹ç¾¤é›†ï¼ˆä»¥ä¸åŒé¡è‰²è¡¨ç¤ºï¼‰éƒ½æœƒæœ‰æ˜é¡¯ä¸åŒçš„é¡å‹ï¼ˆä»¥ä¸åŒå½¢ç‹€è¡¨ç¤ºï¼‰ã€‚

è®“æˆ‘å€‘ä¾†çœ‹çœ‹æ¨¡å‹çš„æº–ç¢ºæ€§ã€‚


```r
# Assign genres to predefined integers
label_count <- results %>% 
  group_by(artist_top_genre) %>% 
  mutate(id = cur_group_id()) %>% 
  ungroup() %>% 
  summarise(correct_labels = sum(.cluster == id))


# Print results  
cat("Result:", label_count$correct_labels, "out of", nrow(results), "samples were correctly labeled.")

cat("\nAccuracy score:", label_count$correct_labels/nrow(results))
```
