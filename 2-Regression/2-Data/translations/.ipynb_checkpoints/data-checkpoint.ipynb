```python
import pandas as pd
import matplotlib.pyplot as plt
pumpkins = pd.read_csv("D:/Applications/Obsidian/ML-For-Beginners/2-Regression/data/US-pumpkins.csv")
```

```python
pumpkins.tail()
```

```python
pumpkins.isnull().sum()
```

```python
new_columns = ['Package', 'Month', 'Low Price', 'High Price', 'Date']
pumpkins.drop([c for c in pumpkins.columns if c not in new_columns],axis=1)
```

```python
price = (pumpkins['High Price'] + pumpkins['Low Price'])/2
month = pd.DatetimeIndex(pumpkins['Date']).month
```

```python
new_pumpkins = pd.DataFrame({'Month':month,'Package':pumpkins['Package'],'Low Price':pumpkins['Low Price'],'High Price':pumpkins['High Price'],'Price':price})
```

```python
pumpkins = pumpkins[pumpkins['Package'].str.contains('bushel',case=True,regex=True)]
```

```python
pumpkins.head()
```

```python
new_pumpkins.loc[new_pumpkins['Package'].str.contains('1 1/9'),'Price'] = price/(1+1/9)

price = new_pumpkins.Price
month = new_pumpkins.Month
```

```python
plt.scatter(price,month)
plt.show()
```

```python
new_pumpkins.groupby(['Month'])['Price'].mean().plot(kind='bar')
plt.ylabel = 'Pumpkin Price'
```
