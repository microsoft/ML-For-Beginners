# Your first scikit-learn notebook

```python
from matplotlib import pyplot as plt
import numpy as np
from sklearn import datasets,linear_model,model_selection
```

```python
# load datasets
X,y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
```

```python
# 划分测试集与训练集
X = X[:,np.newaxis,2]
X_train,X_test,y_train,y_test = model_selection.train_test_split(X,y,test_size=0.33)
```

```python
model = linear_model.LinearRegression()
# 训练模型
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test,y_test,color='blue')
plt.plot(X_test,y_pred,color='green',linewidth=3)
```

```python
X_1,y_1 = datasets.load_linnerud(return_X_y=True)
print(X_1.shape)
print(X_1[0:2])
print(y_1.shape)
print(y_1[0:2])
```

```python
# 创建一个可以描述腰围和完成仰卧起坐个数关系的回归模型
waist = X_1[:,0]
situps = y_1[:,1]
waist_train, waist_test, situps_train, situps_test = model_selection.train_test_split(waist, situps, test_size=0.33)
waist_situps_model = linear_model.LinearRegression()
waist_situps_model.fit(waist_train.reshape(-1,1), situps_train)
situps_pred = waist_situps_model.predict(waist_test.reshape(-1,1))
plt.scatter(waist_test, situps_test, color='blue')
plt.plot(waist_test, situps_pred, color='green', linewidth=3)
```
